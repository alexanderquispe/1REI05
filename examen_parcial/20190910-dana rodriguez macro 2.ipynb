<h1>Defina el multiplicador keynesiano y dé un ejemplo de cómo influye este en el modelo Ingreso-Gasto Keynesiano<h1>

* Ante un aumento o disminución de C0, I0, G0 o X0; el multiplicador keynesiano va a multiplicar su variación afectando así con mayor intensidad a la producción. En este sentido, El multiplicador keynesiano representa en cuánto se va incrementar el producto ante un incremento de cualquier componente autónomo de la Demanda Agregada.

<h1>Grafique y explique cuándo sucede un exceso y déficit de demanda en el modelo Ingreso-Gasto. Señale las áreas donde ocurre cada caso. Explique cómo se converge al equilibrio a partir de estos dos escenarios. <h1>

* Tras un aumento de la Demanda Agregada se produce un exceso de la misma, por lo que la producción anterior se hace insuficiente. para que pueda volver al equilibrio estable, los productores deben aumentar la producción para satisfacer la demanda.


* Por otro lado, tras una disminución de la Demanda Agregada se produce un déficit de la misma, por lo que la producción anterior se hace demasiado alta. Para que pueda volver al equilibrio estable, los productores deben disminuir la producción. 

```python
# Valores de Parametros
Y_size = 100 

Co = 2
Io = 21
Go = 23
Xo = 10
h = 0.8
b = 0.7
m = 0.3
t = 0.3
r = 8.2

Y = np.arange(Y_size)

# Ecuación de la curva del ingreso de equilibrio

def DA(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA= (Co+b*(1-t)*Y+Io-h*r+Go+Xo-m*(1-t)*Y)
    return DA

DA_0= DA(Co, Io, Go, Xo, h, r, b, m, t, Y)
```

```python
#--------------------------------------------------
# Recta de 45°

def L_45(Y):
    L_45 = Y
    return L_45

L_45 = L_45(Y)
```

```python
#--------------------------------------------------

# Parámetros

Y_size = 100 

Co = 15
Io = 18
Go = 28
Xo = 5
h = 0.8
b = 0.5
m = 0.3
t = 0.3

Y = np.arange(Y_size)


# Ecuación 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = ((Co+Io+Go+Xo)/h-Y*((1-(1-t)*(b-m))/h))
    return r_IS

IS_0 = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
```

```python
# Gráfico de la derivación de la curva IS a partir de la igualdad (DA = Y)

    # Dos gráficos en un solo cuadro (ax1 para el primero y ax2 para el segundo)
fig, ax = plt.subplots(figsize=(10, 8))
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())  

#---------------------------------
    # Gráfico 1: ingreso de Equilibrio

ax.plot(Y, DA_0, label = "DA_0", color = "red") 
ax.plot(Y, L_45, color = "black") 

ax.axvline(x = 90, ymin= 0, ymax = 0.87, linestyle = ":", color = "grey")
ax.axvline(x = 30, ymin= 0, ymax = 0.57, linestyle = ":", color = "grey")
ax.axvline(x = 59, ymin= 0.59, ymax = 0.65, linestyle = ":", color = "grey")
ax.axvline(x = 69.12,  ymin= 0, ymax = 0.68, linestyle = ":", color = "grey")


plt.axhline(y = 76, xmin= 0.75, xmax = 0.87, linestyle = ":", color = "grey")
plt.axhline(y = 58, xmin= 0.34, xmax = 0.59, linestyle = ":", color = "grey")
plt.axhline(y = 68, xmin= 0, xmax = 0.69, linestyle = ":", color = "grey")
plt.axhline(y = 66, xmin= 0.6, xmax = 0.66, linestyle = ":", color = "grey")

ax.text(4, 2, '$45°$', fontsize = 11.5, color = 'black')
ax.text(1, -3, '$◝$', fontsize = 30, color = 'black')
ax.text(69.12, 0, '$Y^e$', fontsize = 12, color = 'black')
ax.text(29, 59, 'A', fontsize = 12, color = 'black')
ax.text(69.12, 72, 'E', fontsize = 12, color = 'black')
ax.text(90,76, '$B$', fontsize = 11.5, color = 'black')
ax.text(5,53, 'Exceso de Demanda', fontsize = 11.5, color = 'black')
ax.text(76 ,72, 'Deficit de Demanda', fontsize = 11.5, color = 'black')

ax.set(title = "Derivación de la curva IS a partir del equilibrio $Y=DA$", xlabel = r'Y', ylabel = r'DA')
```
