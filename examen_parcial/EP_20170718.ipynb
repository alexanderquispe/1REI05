### Pregunta 1 (2 puntos) Defina el Multiplicador Keynesiano y d√© un ejemplo de c√≥mo influye este en el modelo Ingreso-Gasto Keynesiano.

El multiplicador keynesiano afirma que un cambio en el gasto aut√≥nomo provoca un cambio m√°s que proporcional en el PIB real. Es decir, un cambio en la magnitud de cualquiera de los componentes aut√≥nomos de la DA (que conforman el intercepto), genera un proceso multiplicador del ingreso hasta converger al nuevo ingreso y producto de equilibrio. 

Si $$ DA = Œ±_0  + Œ±_1Y $$ y la condici√≥n de equilibrio $$ Y= DA $$, se obtiene $$ (1 - Œ±_1)Y = Œ±_0$$

Por lo tanto, el multiplicador Keynesiano es: $\frac{1}{1 - (b - m)(1 - t)}$

Influye en el modelo Ingreso-Gasto Keynesiano de la siguiente manera:

Un cambio en Œ±0 provoca un cambio multiplicado en Y.

$$ {‚àÜY} = \frac{1}{1 - Œ±_1}{‚àÜŒ±_0} $$



### Pregunta 2 (2 puntos) Grafique y explique c√∫ando sucede un exceso y d√©ficit de demanda en el modelo Ingreso-Gasto. Se√±ale las √°reas donde ocurre cada caso. Explique c√≥mo se converge al equilibrio a partir de estos dos escenarios.

Todos los puntos de la Curva IS corresponden a pares ordenados (Y, r) que equilibran el mercado de bienes.  Por ejemplo, el punto A representa el equilibrio; el par ordenado (YA, rA). Sin embargo, los puntos fuera de esta curva son de desequilibrio en el mercado o representan una situaci√≥n en el que el ahorro y la inversi√≥n no son iguales. 

Contrariamente el punto B, que se encuentra a la derecha de la IS,correspondiente al par ordenado (YB, rA) representa un desequilibrio. Como la tasa de inter√©s (r)  y la inversi√≥n (I) permanecen constantes, pero los ingresos (Y) son mayores, el ahorro de la gente es mayor que en el punto A . Entonces, el ahorro es mayor a la inversi√≥n (IA<SA) ; por lo tanto, AB a la derecha de la curva IS representa un exceso de oferta.

Asimismo, el punto C, que se encuentra a la izquierda de la IS, representa un desequilibrio. Dado que la tasa de inter√©s (r)  y la inversi√≥n (I)  permanecen constantes, pero el ingreso (Y) es menor, el ahorro de la gente es menor que en el punto A. Entonces, el ahorro es menor a la inversi√≥n (SA<IA) ; por lo tanto,  a la derecha de la curva IS representa un exceso de demanda.

A partir de estos dos escenarios se converge al equilibrio de la siguiente manera:

Los cambios en el intercepto provocados por los cambios en los componentes aut√≥nomos de la demanda agregada (consumo aut√≥nomo, la confianza de los inversionistas, el gasto del gobierno y las exportaciones aut√≥nomas), son los que pueden dar lugar a desplazamientos paralelos de la curva IS, hacia a la derecha o hacia a la izquierda. 

Por ejemplo:

Si el gobierno aplica una pol√≠tica fiscal expansiva aumentado el gasto, G, la curva IS debe desplazarse hacia la derecha con una distancia equivalente a la magnitud de dicho gasto. Si se adopta una pol√≠tica fiscal restrictiva aumentando la tasa de tributaci√≥n, la pendiente de la IS aumenta, provocando un movimiento en el sentido de las agujas del reloj. Por el contrario, el aumento de la propensi√≥n marginal a consumir, ‚Äúb‚Äù, aumenta el tama√±o del multiplicador, aumenta la pendiente de la DA y reduce la pendiente de la curva IS.


```python
# Par√°metros

Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3

Y = np.arange(Y_size)


# Ecuaci√≥n 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)

# Gr√°fico de la curva IS

# Dimensiones del gr√°fico
y_max = np.max(r)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, r, label = "IS", color = "#8B0A50") #Demanda agregada

plt.axvline(x=20,  ymin= 0, ymax= 0.5, linestyle = ":", color = "grey")
plt.axvline(x=50,  ymin= 0, ymax= 0.5, linestyle = ":", color = "grey")
plt.axvline(x=80,  ymin= 0, ymax= 0.5, linestyle = ":", color = "grey")

plt.axhline(y=169, xmin= 0, xmax= 0.78, linestyle = ":", color = "grey")

# Eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

# Texto agregado
plt.text(-10, 170, '$r_A$', fontsize = 13.5, color = 'black')

plt.text(19, 171, '$C$', fontsize = 13.5, color = 'black')
plt.text(49, 171, '$A$', fontsize = 13.5, color = 'black')
plt.text(79, 171, '$B$', fontsize = 13.5, color = 'black')

plt.text(22, 125, '$Y_C$', fontsize = 13.5, color = 'black')
plt.text(52, 125, '$Y_A$', fontsize = 13.5, color = 'black')
plt.text(82, 125, '$Y_B$', fontsize = 13.5, color = 'black')

plt.text(24, 140, 'Exceso de demanda', fontsize = 11.5, color = 'black')
plt.text(69, 185, 'Exceso de oferta', fontsize = 11.5, color = 'black')

# T√≠tulo, ejes y leyenda
ax.set(title = "Equilibrio y Desequilibrio en el Mercado de Bienes", xlabel= 'Y', ylabel= 'r')
ax.legend()

plt.show()
```
