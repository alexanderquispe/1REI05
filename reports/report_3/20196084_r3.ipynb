1.  Explique cuáles son los intrumentos de política monetaria que puede utilizar el Banco Central.

La autoridad monetaria puede recurrir a políticas expansivas o contractivas. Una de ella es la tasa de interés, esta influye en cuánto van a desear prestarse los bancos, pues a mayor tasa de interés, podrán prestarse menos dinero con lo que disminuye la oferta monetaria, y viseversa. Por eso hay una relación inversa entre ambos factores. Otro instrumento es comprar o vender activos financieros o bonos. Al adquirir bonoos, se practica una política expansiva, pues se cuenta con mayor base monetaria; pero cuando se venden es una política de carácter contractivo. Por último, el Banco Central puede aumentar o disminuir la cantidad de dinero que pueden disponer los bancos para realizar préstamos. Esto se llama tasa de encaje, al aumentar esta tasa, se reduce la posibilidad de crear dinero porque habría más en el mercado monetario, y viceversa.

El equilibrio en el Mercado de Dinero se deriva del equilibrio entre la Oferta de Dinero $(M^s)$ y Demanda de Dinero $(M^d)$:

$$ M^s = M^d $$

$$ \frac{M^s}{P} = kY - ji $$

Se puede suponer que la inflación esperada es cero, ya que a corto plazo el nivel de precios es fijo y exógeno. Por lo tanto, no habría una gran diferencia entre la tasa de interés nominal $(i)$ y la real $(r)$.

Entonces la ecuación del equilibrio en el mercado monetario es:

$$ \frac{M^s}{P} = kY - jr $$


Y puede graficarse de la siguiente manera:

```python
# Parameters
r_size = 100

k = 0.5
j = 0.2                
P  = 10 
Y = 35
MS_0 = 500

r = np.arange(r_size)

# Necesitamos crear la funcion de demanda 

def MD(k, j, P, r, Y):
    MD_eq = (k*Y - j*r)
    return MD_eq
MD_0 = MD(k, j, P, r, Y)
# Necesitamos crear la oferta de dinero.
MS = MS_0 / P
MS
```

```python
# Equilibrio en el mercado de dinero

# Creamos el seteo para la figura 
fig, ax1 = plt.subplots(figsize=(10, 8))

# Agregamos titulo t el nombre de las coordenadas
ax1.set(title="Money Market Equilibrium", xlabel=r'M^s / P', ylabel=r'r')

# Ploteamos la demanda de dinero
ax1.plot(MD_0, label= '$L_0$', color = '#CD5C5C')

# Para plotear la oferta de dinero solo necesitamos crear una linea vertical
ax1.axvline(x = MS,  ymin= 0, ymax= 1, color = "grey")

# Creamos las lineas puntadas para el equilibrio
ax1.axhline(y=7.5, xmin= 0, xmax= 0.5, linestyle = ":", color = "black")

# Agregamos texto
ax1.text(0, 7.5, "$r_0$", fontsize = 12, color = 'black')

ax1.yaxis.set_major_locator(plt.NullLocator())   
ax1.xaxis.set_major_locator(plt.NullLocator())

ax1.legend()

plt.show()
```

Siguiendo la última ecuación de equilibrio, la curva LM se da en función de la tasa de interés:

$$ r = -\frac{1}{j}\frac{M_o^s}{P_o} + \frac{k}{j}Y $$

```python
# Parameters
Y_size = 100

k = 10
j = 10                
Ms = 200             
P  = 20               

Y = np.arange(Y_size)


# Ecuación

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)
```

```python
    # Gráfico de la curva LM

# Dimensiones del gráfico
y_max = np.max(i)
v = [0, Y_size, 0, y_max]   
fig, ax = plt.subplots(figsize=(10, 8))

# Eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

# Curvas a graficar
ax.plot(Y, i, label="LM", color = '#3D59AB')

# Título y leyenda
ax.set(title="LM SCHEDULE", xlabel=r'Y', ylabel=r'r')
ax.legend()


plt.show()
```

2. ¿Cuál es el efecto de una disminución en la Masa Monetaria? Explica usando la intuición y gráficos.

Cuando baja la masa monetaria, se genera un exceso de demanda en el mercado, por ello la curva LM avanza a la izquierda (se contrae).

$$ M^s↓ → M^s_0↓ → M^s < M^d → r↑ $$

```python
#--------------------------------------------------
    # Curva LM ORIGINAL

# Parámetros

Y_size = 100

k = 2
j = 1                
Ms = 200            
P  = 20               

Y = np.arange(Y_size)

# Ecuación

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)

#--------------------------------------------------
    # NUEVA curva LM

# Definir SOLO el parámetro cambiado
Ms = 100

# Generar la ecuación con el nuevo parámetro
def i_LM_Ms( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i_Ms = i_LM_Ms( k, j, Ms, P, Y)
```
