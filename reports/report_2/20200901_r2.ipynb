### a. El Modelo Ingreso-Gasto: la Curva IS
#### Utilizando las ecuaciones, derive paso a paso la curva IS matemáticmente (a partir de la condición de equilibrio Y = DA).
#### Lea la sección 4.4 del material de enseñanza y explique cómo se deriva la curva IS a partir del equilibrio Y = DA y grafique.
##### (Se procederá a convinar ambas pautas dado que la explicación de la derivación es facilitada por la ayuda visual del razonamiento matemático)

Se parte de la condición de que hay equilibrio cuando Y = DA...


$$ Y = DA $$

Si se separa DA en sus componentes...

$$ DA = C + I + G + X - M $$

Entonces equilibrio podría ser interpretado como...

$$ Y = C + I + G + X - M $$

Ahora, no se desea Y, sino el Ingreso disponible ($Y^d$), y para ello se remueve el componente T de Y, ergo se remueve lo mismo del DA

$$ Y^d = C + I + G + X - M - T $$

Como el modelo es de Ingreso - Gasto, se aísla I en un lado, por lo que queda...

$$ I = Y^d - C - G - X + M + T $$

Recordar que el equilibrio Ahorro - Inversión implicaría...

$$ I = S $$

Entonces, ya se sabría los componentes de S cuando está en equilibrio con I, formando así el equilibrio Ahorro - Inversión. No obstante, la Curva IS es en función a r, por lo que hay que seguir separando los elementos y continuar derivándolos recordando que:

$C=C_0+by^d$

$G=G_0$

$X=X_0$

$M=my^d$

$T=tY$

$I=I_0-hr$

$$ I_0 - hr = (y^d - (C_0 + by^d)) + (tY - G_0) + (my^d - X_0)) $$

$$ I_0 - hr = (1 - (b - m)(1 - t))Y - (C_0 + G_0 + X_0) $$

$$ hr = I_0 - (1 - (b - m)(1 - t)Y - (C_0 + G_0 + X_0)) $$

$$ hr = (I_0 + C_0 + G_0 + X_0) - (1 - (b - m)(1 - t))Y $$

Y finalmente la Curva IS que se mide en función de r sería...

$$ r = \frac{I_0 + C_0 + G_0 + X_0}{h} - \frac{(1 - (b - m)(1 - t))}{h} Y $$

```python
# DA

TamañoY = 100 

Co = 30
Io = 40
Go = 60
Xo = 3
h = 0.6
b = 0.7
m = 0.3
t = 0.3
r = 35 # Es verdad que debería estar entre 0 y 1, pero por razones visuales, se concederá un valor mayor a 1

Y = np.arange(TamañoY)

def DA(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA

DA = DA(Co, Io, Go, Xo, h, r, b, m, t, Y)
```

```python
# Recta de 45°

a = 2.5 

def G45(a, Y):
    G45 = a*Y
    return G45

G45 = G45(a, Y)
```

```python
# DA1

r = 1

def DA1(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA1 = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA1

DA1 = DA1(Co, Io, Go, Xo, h, r, b, m, t, Y)
```

```python
# IS

TamañoY = 100 

Co = 30
Io = 40
Go = 60
Xo = 3
h = 0.6
b = 0.7
m = 0.3
t = 0.3

Y = np.arange(TamañoY)


def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
```

```python
# Gráficos

fig, (ax1, ax2) = plt.subplots(2, figsize=(8, 16))


# Gráfico 1

ax1.yaxis.set_major_locator(plt.NullLocator())   
ax1.xaxis.set_major_locator(plt.NullLocator())

ax1.plot(Y, DA, label = "DA_0", color = "Black") 
ax1.plot(Y, DA1, label = "DA_1", linestyle = ":", color = "Black") 
ax1.plot(Y, G45, color = "Grey") 

ax1.axvline(x = 60, ymin= 0, ymax = 0.60, linestyle = ":", color = "Grey")
ax1.axvline(x = 50,  ymin= 0, ymax = 0.50, linestyle = ":", color = "Grey")

ax1.text(6, 4, '$45°$', fontsize = 11.5, color = 'Grey')
ax1.text(2.5, -3, '$◝$', fontsize = 30, color = 'Grey')
ax1.text(50, -30, '$Y_0$', fontsize = 12, color = 'Black')
ax1.text(60, -30, '$Y_1$', fontsize = 12, color = 'Black')
ax1.text(44, 133, '$E_0$', fontsize = 12, color = 'Black')
ax1.text(54, 155, '$E_1$', fontsize = 12, color = 'Black')

ax1.text(63, 142, '$DA(r_1)$; siendo $r_1$ < $r_0$', fontsize = 9, color = 'Black')
ax1.text(80, 123, '$DA(r_0)$', fontsize = 9, color = 'Black')

ax1.set(title = "Derivación de la curva IS a partir del equilibrio $Y=DA$", xlabel = r'Y', ylabel = r'DA')


# Gráfico 2

ax2.yaxis.set_major_locator(plt.NullLocator())   
ax2.xaxis.set_major_locator(plt.NullLocator())

ax2.plot(Y, r, label = "IS", color = "Black") 

ax2.axvline(x = 60, ymin= 0, ymax = 1, linestyle = ":", color = "Grey")
ax2.axvline(x = 50,  ymin= 0, ymax = 1, linestyle = ":", color = "Grey")
plt.axhline(y = 150, xmin= 0, xmax = 0.60, linestyle = ":", color = "Grey")
plt.axhline(y = 162, xmin= 0, xmax = 0.50, linestyle = ":", color = "Grey")

ax2.text(50, 88, '$Y_0$', fontsize = 12, color = 'Black')
ax2.text(60, 88, '$Y_1$', fontsize = 12, color = 'Black')
ax2.text(-14, 162, '$r_0$', fontsize = 12, color = 'Black')
ax2.text(-14, 150, '$r_1$', fontsize = 12, color = 'Black')
ax2.text(52, 162, '$E_0$', fontsize = 12, color = 'Black')
ax2.text(63, 150, '$E_1$', fontsize = 12, color = 'Black')

ax2.set(xlabel = r'Y', ylabel = r'r')

plt.show()
```

### d. Movimientos de la Curva IS

```python
# Parámetros

TamañoY = 100 

Co = 30
Io = 40
Go = 60
Xo = 3
h = 0.6
b = 0.7
m = 0.3
t = 0.3

Y = np.arange(TamañoY)


def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
```

```python
Go1 = 30

def r_IS(b, m, t, Co, Io, Go1, Xo, h, Y):
    r_IS = (Co + Io + Go1 + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

rGo1 = r_IS(b, m, t, Co, Io, Go1, Xo, h, Y)
```

```python
# Gráfico

y_max = np.max(r)
fig, ax = plt.subplots(figsize=(12, 9))

# Curvas a graficar
ax.plot(Y, r, label = "IS", color = "Orange")
ax.plot(Y, rGo1, label = "IS_G", color = "Gold", linestyle = 'dashed')

# Texto agregado
plt.text(47, 135, '∆Go', fontsize=20, color='Red')
plt.text(49, 130, '←', fontsize=20, color='Red')

# Título, ejes y leyenda
ax.set(title = "Caída en el Gasto del Gobierno", xlabel= 'Y', ylabel= 'r')
ax.legend()

plt.show()
```

Al decrecer el valor de Go, la nueva Curva IS es paralela a la original pero se desplaza a la iquierda. Esto se debe a que Go es parte de la sección de la fórmula que genera el intercepto [(Io + Co + Go + Xo)/h], y no a la sección que modifica la pendiente [- (1 -(b-m)(1-t)/h)Y]. Al simplemente reducir el valor de una suma, tan solo decrece el valor del intercepto.

```python
t1 = 0.1

def r_IS(b, m, t1, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t1)))/h
    return r_IS

rt1 = r_IS(b, m, t1, Co, Io, Go, Xo, h, Y)
```

```python
# Gráfico

y_max = np.max(r)
fig, ax = plt.subplots(figsize=(12, 9))

# Curvas a graficar
ax.plot(Y, r, label = "IS", color = "Blue")
ax.plot(Y, rt1, label = "IS_t", color = "LightBlue", linestyle = 'dashed')

# Texto agregado
plt.text(85, 125, '∆t', fontsize=13, color='Cyan')
plt.text(85, 123, '→', fontsize=13, color='Cyan')

# Título, ejes y leyenda
ax.set(title = "Caída en la Tasa de Impuestos", xlabel= 'Y', ylabel= 'r')
ax.legend()

plt.show()
```

El valor t forma parte de la sección de la ecuación que dicta la pendiente [- (1 -(b-m)(1-t)/h)Y]. t le es restado a 1, y el resultado de eso se le multiplica a b-m, y finalmente el resultado de eso se le resta a un valor 1 final para determinar la pendiente (divida entre h). Es decir, si decrece t, incrementará el valor que le será restado a ese último valor 1. Por lo tanto, si decrece t, decrecerá la pendiente, y con una pendiente menos marcada la Curva IS nueva se traslada a la derecha.

```python
b1 = 0.5

def r_IS(b1, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b1-m)*(1-t)))/h
    return r_IS

rb1 = r_IS(b1, m, t, Co, Io, Go, Xo, h, Y)
```

```python
# Gráfico

y_max = np.max(r)
fig, ax = plt.subplots(figsize=(12, 9))

# Curvas a graficar
ax.plot(Y, r, label = "IS", color = "Green")
ax.plot(Y, rb1, label = "IS_b", color = "LightGreen", linestyle = 'dashed')

# Texto agregado
plt.text(70, 128, '∆b', fontsize=15, color='LightBlue')
plt.text(70, 123, '←', fontsize=15, color='LightBlue')

# Título, ejes y leyenda
ax.set(title = "Caída en la Propensión Marginal a Consumir", xlabel= 'Y', ylabel= 'r')
ax.legend()

plt.show()
```

El motivo del desplazamiento a la izquierda ante un decrecimiento en b es parecida a la explicación del anterior gráfico si se tiene en cuenta, de nuevo, la sección de la ecuación que dicta la pendiente [- (1 -(b-m)(1-t)/h)Y]. Como se puede observar, si decrece b, el valor que le será restado al valor 1 final será menor, por lo que el valor de la pendiente será mayor. Esto hace que la pendiente sea más marcada, ergo la empinada en la nueva Curva IS.
