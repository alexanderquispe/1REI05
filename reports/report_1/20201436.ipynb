```python
# Parámetros

Y_size = 100 

Co = 30
b = 0.6 

Y = np.arange(Y_size)

#ecuación de la curva del ingreso de equilibrio

def DA_A(Co, b):
    DA_A = (Co) + (b*Y)
    return DA_A

DA_A_A = DA_A(Co, b)
```

pregunta 4:

DA=Co+bY^d+Io-hr+Go+Xo-mY^d

Y = C_0 + bY^d + I_0 -hr + G_0 + X_0 - mY^d

Y = \frac{1}{1 - (b - m)(1 - t)} (C_0 + I_0 + G_0 + X_0 - hr)

```python
# - Parámetros

Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8 # b > m
m = 0.2
t = 0.3
r = 0.9

Y = np.arange(Y_size)

# Ecuación de la curva del ingreso de equilibrio

def DA_C(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_C = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_C

DA_C_C = DA_C(Co, Io, Go, Xo, h, r, b, m, t, Y)
```

```python
# Recta de 45°

a = 2.5 

def L_45(a, Y):
    L_45 = a*Y
    return L_45

L_45 = L_45(a, Y)
```

```python
#Gráfico

# Dimensiones del gráfico
y_max = np.max(DA_C_C)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, DA_C_C, label = "DA", color = "orange") #Demanda agregada
ax.plot(Y, L_45, color = "black") #Línea de 45º

# Eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

# Líneas punteadas punto de equilibrio
plt.axvline(x=70.5,  ymin= 0, ymax= 0.69, linestyle = ":", color = "red")
plt.axhline(y=176, xmin= 0, xmax= 0.7, linestyle = ":", color = "red")

# Texto agregado
    # punto de equilibrio
plt.text(0, 220, '$DA=α_o+α_1$', fontsize = 11.5, color = 'purple')
plt.text(72, 0, '$Y^e$', fontsize = 12, color = 'purple')
plt.text(0, 180, '$α_o=Co+Io+Go+Xo-hr$', fontsize = 15, color = 'blue')
plt.text(72, 130, '$α_1=(b-m)(1-t)$', fontsize = 15, color = 'blue')
    # línea 45º
plt.text(6, 4, '$45°$', fontsize = 11.5, color = 'green')
plt.text(2.5, -3, '$◝$', fontsize = 30, color = 'green')

# Título y leyenda
ax.set(title="El ingreso de Equilibrio a Corto Plazo", xlabel= r'Y', ylabel= r'DA')
ax.legend() #mostrar leyenda

plt.show()
```
