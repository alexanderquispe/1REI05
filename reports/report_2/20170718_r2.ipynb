En el modelo Ingreso-Gasto el equilibrio entre el Ingreso y la DA representa el equilibrio en el mercado de bienes.

Tomando en cuenta las siguientes ecuaciones:

$$ C = C_0 + bY^d $$

$$ I = I_0 - hr $$

$$ G = G_0 $$

$$ T = tY $$

$$ X = X_0 $$

$$ M = mY^d $$


Haciendo reemplazos y algunas operaciones algebraicas en esta condición de quilibrio se obtiene: 

$$ Y = \frac{1}{1 - (b - m)(1 - t)} (C_0 + I_0 + G_0 + X_0 - hr) $$

O, expresado de otra manera:

$$ r = \frac{1}{h}(C_0 + G_0 + I_0 + X_0) - \frac{1 - (b - m)(1 - t)}{h}Y $$

Entonces, la curva IS puede ser simplificada de la siguiente manera:

$$ r = \frac{B_0}{h} - \frac{B_1}{h}Y $$

Donde $ B_0 = C_0 + G_0 + I_0 + X_0  $ y $  B_1 = 1 - (b - m)(1 - t) $

Procedmos a graficar, recordando la ecuación de la Demanda Agregada (DA):


#Ejercicio 2: La Curva IS de Equilibrio en el Mercado de Bienes


La Cueva IS o el equilibrio Ahorro-Inersión:

El equilibrio entre el ingreso y la demanda agregada es equivalente al equilibrio entre el ahorro total y la inversión. Por lo que partimos de la condición de equilibrio DA=Y:


$$ Y = C + I + G + X - M $$


Tomando en cuenta las siguientes ecuaciones:


$$ C = C_0 + bY^d $$

$$ I = I_0 - hr $$

$$ G = G_0 $$

$$ T = tY $$

$$ X = X_0 $$

$$ M = mY^d $$


Realizamos algunas operaciones y llegos al equilibrio Ahorro-Inversión. Para obtener la ecuación ingreso disponibles (Yd) restamos la Tributación (T) de ambos miembros de igualdad:


$$ Y - T = C + I - T + G + X - M $$

$$ Y^d = C + I - T + G + X - M $$


Esta igualdad puede reescribirse como:


$$ (Y^d - C) + (T - G) + (M - X) = I $$


El lado derecho constituye el ahorro total que es igual a la inversión. El ahorro toral (S) concstituye el Ahorro Privado (Sp), el Ahorro del Gobierno (Sg) y el Ahorro Externo (Se).


$$ S_p + S_g + S_e = I_0 - hr $$

$$ S = I_0 - hr $$


Dicha igualdad entre el ahorro y la inversión representa la curva IS:


$$ S(Y) = I(r) $$


Reemplazando, obtenemos:


$$ (Y^d - C_0 - bY^d) + (T - G_0) + (mY^d - X_0) = I_0 - hr $$


A partir de aquí, con las ecuaciones previamente mostradas, obtenemos la ecuacuión de la Curva IS:


$$ [1 - (b - m)(1 - t)]Y - (C_0 + G_0 + X_0) = I_0 - hr $$


La curva IS se puede expresar con una ecuación donde la tasa de interés es una función del ingreso:

$$ hr = (C_0 + G_0 + I_0 + X_0) - (1 - (b - m)(1 - t))Y $$

$$ r = \frac{1}{h}(C_0 + G_0 + I_0 + X_0) - \frac{1 - (b - m)(1 - t)}{h}Y $$

Se puede simplificar de esta manera:

$$ r = \frac{B_0}{h} - \frac{B_1}{h}Y $$

Donde $ B_0 = C_0 + G_0 + I_0 + X_0  $ es el intercepto y $  B_1 = 1 - (b - m)(1 - t) $ es la pendiente.

Procedemos a graficar la curva IS:

```python
# Parámetros

Y_size = 100 

Co = 30
Io = 35
Go = 65
Xo = 2.5
h = 0.75
b = 0.85
m = 0.25
t = 0.35

Y = np.arange(Y_size)


# Ecuación 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
```

```python
# Gráfico de la curva IS

# Dimensiones del gráfico
y_max = np.max(r)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, r, label = "IS", color = "#A52A2A") #Demanda agregada

# Eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

# Título, ejes y leyenda
ax.set(title = "Curva IS", xlabel= 'Y', ylabel= 'r')
ax.legend()

plt.show()
```
