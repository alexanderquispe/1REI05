**2. (3 puntos) Regression Discontinuity Design (RDD)**

Supongamos que estamos estudiando la eficacia de un programa de becas creado por el gobierno Peruano sobre las notas de los alumnos al final del primer semestre en la universidad. PRONABEC ha decidido que entregará becas a aquellos postulantes que tengan un promedio superior durante la secundaria igual a 16 .

- **Indetifique todos los elementos necesarios para implementar un RDD en este ejemplo.**

**Motivación (que nos lleva a usar un RDD):**

- Un alumno con puntaje 16.001 es prácticamente idéntico a un alumno con puntaje 15.999 , pero el primer alumno sí puede participar del programa mientras que el segundo no. 
- El segundo alumno se puede tomar como contrafactual del primero.

La idea es que si se encuentran a la izquierda o derecha del valor crítico (Z), participan o no

Entonces, la idea radica en que los individuos de la izquierda del umbral Z deben ser muy parecidos a los de la derecha del umbral. Entonces, la derecha podría ser un contrafactual válido de la izquierda.

Supuestos:
El umbral Z es abitriario
Los regresores X,  y el error ( u)  son muy parecidos a la izquierda y derecha del umbral.

Entonces:
La comparación de estos dos grupos (tratamiento y control) es un estimador del impacto del tratamiento alrededor del punto de corte.
El parámetro estimado por la Regresión Discontinua es un EFECTO DEL TRATAMIENTO (o Local Average Treatment Effect)
Se centra en los individuos alrededor del umbral


- **Explique el procedimiento para encontrar el efecto causal en este escenario. Puede utilizar ecuaciones.**

- Sabemos: la continuidad: la ausencia de tratamientos simultáneos en el corte

- La hipótesis nula siempre debe ser la continuidad y que cualquier discontinuidad implica necesariamente alguna causa, porque la tendencia a que las cosas cambien gradualmente es lo que hemos llegado a esperar en la naturaleza. 

- Entonces, el efecto causal promedio para esta subpoblación se identifica como $X \rightarrow Z_i$ en el limite Esto es posible porque el punto de corte es el único punto donde los sujetos de tratamiento y control se superponen en el límite.


- **¿El efecto de la politica sobre el rendimiento de los estudiantes sería positivo o negativo (solo necesitan usar su intuición)? Argumente si los resultados serian causales o solamente de correlación .**


En definitiva, la política será positivo para los alumnos que puedan acceder al beneficio, es decir, habrá una correlación positiva debido al efecto que genera dicha política. Respecto a ello, entonces, los resultados serán causales,debido a que existe un contrafactual para los alumnos que no reciben la beca y que determinan un cambio en la linea de continuidad. Es decir, existe un efecto causal en $X \rightarrow Z_i$, ya que sin la intervención no se daría de la misma forma, ya que la discontinuidad implica necesariamente una causa.
