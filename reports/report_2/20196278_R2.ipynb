$Caso 1$ ->Política Fiscal Contractiva con caída del Gasto del Gobierno, (ΔG<0)

$Analisis-intuitivo$
Considerando que el gasto de gobierno es menor a cero, decimos que Gasto de gobierno decae 
$$G↓$$
Por tanto decimos que como como G cae, etonces la Demanda a gregada tambien cae. 
$$G↓=DA↓$$
Así, como la demanda cae, también lo hace la producción
$$DA↓=Y↓$$
Finalmente, como la producción cae, podemos decir que la taza de interes ha caido, para ajustarce a la nueva DA.  
el la grafica se puede ver uns desplazamietno a la izquierda, en el que se encuentra un nuevo equilibrio
