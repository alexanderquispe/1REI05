#### 1. (2 puntos) Defina el Multiplicador Keynesiano y dé un ejemplo de cómo influye este en el modelo Ingreso-Gasto Keynesiano.

El multiplicador Keynesiano es aquel que, dentro de la ecuación para hallar $Y^{equilibrio}$ dentro modelo Ingreso-Gasto Keynesiano, modifica cuán angosta será la recta de la Demanda Agregada. Es decir, el multiplicador Keynesiano es la pendiente en el modelo Ingreso-Gasto Keynesiano.

$Modelo:$
$$ Y = \frac{1}{1 - (b - m)(1 - t)} (C_0 + I_0 + G_0 + X_0 - hr) $$

$Multiplicador:$
$$\frac{1}{1 - (b - m)(1 - t)}$$

$Ejemplo:$

```python
# Original

TamañoY = 100 

Co = 20
Io = 20
Go = 60
Xo = 10
h = 0.5
b = 0.6 # b > m
m = 0.3
t = 0.6
r = 0.9

Y = np.arange(TamañoY)

def DA(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA

DA = DA(Co, Io, Go, Xo, h, r, b, m, t, Y)

# Incremento del multiplicador Keynesiano

TamañoY = 100 

Co = 20
Io = 20
Go = 60
Xo = 10
h = 0.5
b1 = 0.9 # b > m
m1 = 0.1
t1 = 0.2
r = 0.9

Y = np.arange(TamañoY)

def DA1(Co, Io, Go, Xo, h, r, b1, m1, t1, Y):
    DA1 = (Co + Io + Go + Xo - h*r) + ((b1 - m1)*(1 - t1)*Y)
    return DA1

DA1 = DA1(Co, Io, Go, Xo, h, r, b1, m1, t1, Y)
```

```python
a = 2.5 

def L_45(a, Y):
    L_45 = a*Y
    return L_45

L_45 = L_45(a, Y)
```

```python
# Gráfico

y_max = np.max(DA)
fig, ax = plt.subplots(figsize=(12, 9))

ax.plot(DA, label = "Multiplicador Original", color = "Red")
ax.plot(DA1, label = "Incremento en el Multiplicador Keynesiano", linestyle = ":", color = "Red")
ax.plot(L_45, color = "Black") 

ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

plt.text(6, 4, '$45°$', fontsize = 11.5, color = 'black')
plt.text(2.5, -3, '$◝$', fontsize = 30, color = '#404040')

ax.set(title="Aumento en Multiplicador Keynesiano", xlabel= r'Y', ylabel= r'DA')
ax.legend()

plt.show()
```

$Considerando...$

$$ DA = C + I + G + X - M $$

Un exceso de demanda ocurre cuando la DA es mayor que los valores de la línea de equilibrio (de 45°), cuando es mayor que la producción. Para converger nuevamente al equilibrio en este escenario, la producción (Y) deberá aumentar.

Un déficit de demanda ocurre cuando la DA es menor aque los valores de la línea de equilibrio (de 45°), cuando es menor que la producción. Para converger nuevamente al equilibrio en este escenario, la producción (Y) deberá disminuir.

#### 2. (3 puntos) Analice los efectos sobre las variables endógenas Y, P ante una subida de la tasa de interés de referencia $r_0$. ($∆r_0>0$). El análisis debe ser intuitivo, matemático y gráfico.

$Intuición...$

$IS-LM:$
$$ r↑ → I↓ → DA↓ → DA < Y → Y↓ $$
$$ r↑ → M^d<M^s → M^s↓ $$


$DA-OA:$
$$ Y↓ → P↓ $$

Como un incremento en r reduce el valor del DA, se genera un desequilibrio en el modelo IS - LM. En respuesta, Y decrece para restaurar el balance DA = Y

En el modelo DA - OA, un incremento en Y siempre genera un incremento en el valor de P y viceversa. Ambas variables son directamente proporcionales.

$Matemático...$

$$∆Y < 0$$
$$∆P < 0$$

$Grafico...$

```python
# Gráfico

# DA

TamañoY = 100 

Co = 40
Io = 50
Go = 70
Xo = 3
h = 0.8
b = 0.8 # más que m
m = 0.4 # menos que b
t = 0.5

k = 4
j = 2                
Ms = 300             
P  = 7  

Y = np.arange(TamañoY)

# Ecuación DA

β_0 = (Co + Io + Go + Xo)
β_1 = (1 - (b-m)*(1-t))

def P_AD(h, Ms, j, β_0, β_1, k, Y):
    P_AD = ((h*Ms + j*β_0)/h) - (Y*(j*β_1 + h*k)/h)
    return P_AD

P_AD = P_AD(h, Ms, j, β_0, β_1, k, Y)


# OA

TamañoY = 100

Pe = 160 # Precio Esperado
θ = 5
_Y = 60 # Producto Potencial

Y = np.arange(TamañoY)


# Ecuación OA

def P_AS(Pe, _Y, Y, θ):
    P_AS = Pe + θ*(Y-_Y)
    return P_AS

P_AS = P_AS(Pe, _Y, Y, θ)


# IS 

def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r_IS = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)


# LM

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i_LM = i_LM( k, j, Ms, P, Y)


# Nuevo LM

Ms1 = 200

def i_LM2( k, j, Ms1, P, Y):
    i_LM2 = (-Ms1/P)/j + k/j*Y
    return i_LM2

i_LM2 = i_LM2( k, j, Ms1, P, Y)


# Nuevo DA

def P_AD2(h, Ms1, j, β_0, β_1, k, Y):
    P_AD2 = ((h*Ms1 + j*β_0)/h) - (Y*(j*β_1 + h*k)/h)
    return P_AD2

P_AD2 = P_AD2(h, Ms1, j, β_0, β_1, k, Y)
```

```python
fig, (ax1, ax2) = plt.subplots(2, figsize=(8, 16)) 

# DA-OA

ax1.plot(Y, P_AD, label = "DA", color = "Crimson") 
ax1.plot(Y, P_AS, label = "OA", color = "Gold")
ax1.plot(Y, P_AD2, label = "DA_2", linestyle = ":", color = "Crimson")



ax1.yaxis.set_major_locator(plt.NullLocator())   
ax1.xaxis.set_major_locator(plt.NullLocator())


ax1.set(title="DA-OA si $∆r_0>0$", xlabel= r'Y', ylabel= r'P')
ax1.legend()


# IS-LM

ax2.plot(Y, r_IS, label = "IS", color = "Green") 
ax2.plot(Y, i_LM, label = "LM", color = "Orange")
ax2.plot(Y, i_LM2, label = "LM_1", linestyle = ":", color = "Orange")

ax2.axhline(y=131, xmin= 0, xmax= 0.99, label = "r_1", linestyle = ":", color = "Black")
ax2.axhline(y=128, xmin= 0, xmax= 0.99, label = "r_0", color = "Black")

ax2.yaxis.set_major_locator(plt.NullLocator())   
ax2.xaxis.set_major_locator(plt.NullLocator())


ax2.set(title="IS-LM si $∆r_0>0$", xlabel= r'Y', ylabel= r'r')
ax2.legend()


plt.show()
```
