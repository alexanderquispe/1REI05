### Pregunta 1

#### Curva IS

La curva IS se deriva de la igualdad entre el ingreso $(Y)$ y la demanda agregada $(DA)$:

$$ Y = C + I + G + X - M $$

Debemos recordar las ecuaciones de cada componente:

$$ C = C_0 + bY^d $$

$$ I = I_0 - hr $$

$$ G = G_0 $$

$$ T = tY $$

$$ X = X_0 $$

$$ M = mY^d $$

Con esto, podemos reemplazar en la fórmula con la Tributación.

$$ Y - T = C + I - T + G + X - M $$

$$ Y^d = C + I - T + G + X - M $$

$$ (Y^d - C) + (T - G) + (M - X) = I $$

Ahora podemos reemplazar esta fórmula con las demás fórmulas de los componentes de la demanda.

$$ (Y^d - C_0 - bY^d) + (T - G_0) + (mY^d - X_0) = I_0 - hr $$

$$ [1 - (b - m)(1 - t)]Y - (C_0 + G_0 + X_0) = I_0 - hr $$

La parte de la izquierda sería el ahorro S, y la parte de la derecha la inversión I.

Por lo tanto, tenemos que el equilibrio es Ahorro = Ingreso, expresado así:

$$ S(Y) = I(r) $$

Ahora solo falta derivar un poco más para llegar a la ecuación en función con r en función de Y.

$$ hr = (C_0 + G_0 + I_0 + X_0) - (1 - (b - m)(1 - t))Y $$

$$ r = \frac{1}{h}(C_0 + G_0 + I_0 + X_0) - \frac{1 - (b - m)(1 - t)}{h}Y $$

Podemos simplificarlo ahora con betas, en vez de los alphas que usamos en la curva de demanda.

$$ r = \frac{B_0}{h} - \frac{B_1}{h}Y $$

Donde B0 es el intercepto y B1 es la pendiente.

#### Derivación curva IS

La curva IS representa el equilibrio del mercado de bienes, que es equivalente al Ingreso = Demanda Agregada.

Debemos recordar las ecuaciones de cada componente:

$$ C = C_0 + bY^d $$

$$ I = I_0 - hr $$

$$ G = G_0 $$

$$ T = tY $$

$$ X = X_0 $$

$$ M = mY^d $$

Asimismo, la Demanda agregada y la condición de equilibrio son las siguientes ecuaciones:

$$ DA = C + I + G + X - M $$

$$ Y = C + I + G + X - M $$

Con esto podemos restar la tributación "T" de cada componente y obtener la ecuación de Ingreso disponible 

$$ Y^d = C + I - T + G + X - M $$

$$ (Y^d - C) + (T - G) + (M - X) = I $$

En base a esto, podemos reemplazar los valores que se igualan a Y y obtenemos la siguiente ecuación en función de r:

$$ Y = \frac{1}{1 - (b - m)(1 - t)} (C_0 + I_0 + G_0 + X_0) - \frac{h}{1 - (b - m)(1 - t)}r $$

Que también se puede leer así en función de Y:

$$ r = \frac{1}{h}(C_0 + G_0 + I_0 + X_0) - \frac{1 - (b - m)(1 - t)}{h}Y $$

Esta ecuación indica que existe una tasa de interés que se equilibra con los cambios en los componentes autónomos de la demanda agregada y el ingreso disponible. Esta es una relación inversa.

### Pregunta 2

La función ahorro inversión es igual a S = I, siendo S el ahorro e I la inversión:

$$ S(Y) = I(r) $$

Esta función se puede derivar a cada uno de los tipos de ahorro, así como la ecuación de inversión:

$$ S_p + S_g + S_e = I_0 - hr $$

El ahorro total sería $ (S) $ que equivale a la suma del ahorro privado $ (S_p) $, el ahorro del gobierno $ (S_g) $ y el ahorro externo $ (S_e) $. Ahora podemos reemplazar estos ahorros con las funciones del modelo de la parte a.

$$ (Y^d - C) + (T - G) + (M - X) = I $$

$$ (Y^d - C_0 - bY^d) + (T - G_0) + (mY^d - X_0) = I_0 - hr $$

A partir de esto podemos encontrar la ecuación de la curva IS en base a r e Y.

$$ r = \frac{1}{h}(C_0 + G_0 + I_0 + X_0) - \frac{1 - (b - m)(1 - t)}{h}Y $$

```python
# Parámetros

Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3

Y = np.arange(Y_size)


# Ecuación 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
```

```python
#### Gráfico de la curva IS

# Dimensiones del gráfico
y_max = np.max(r)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, r, label = "IS", color = "#8B0A50") #Demanda agregada

# Eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

plt.text(98, 136, 'IS', fontsize = 12, color = 'black')

# Título, ejes y leyenda
ax.set(title = "Curva IS", xlabel= 'Y', ylabel= 'r')
ax.legend()


plt.show()
```

### Pregunta 4

#### Analice una Política Fiscal Contractiva con caída del Gasto del Gobierno.

Análisis intuitivo:

$$ ↓Go → ↓DA → ↑S → S > I → ↓r $$

Si disminuye el gasto de gobierno, va a aumentar el ahorro. Esto genera un exceso de oferta de bienes, ya que la demanda agregada disminuye. Entonces, va a haber más ahorro que inversión, rompiendo el equilibrio. Entonces, para que la inversión aumente y se regrese al equilibrio, debe bajar la tasa de interés. 

El gasto de gobierno es un componente del intercepto de la curva IS. Si se reduce el gasto del gobierno, al ser una suma en el intercepto, va a afectarlo negativamente por lo tanto, la curva va a moverse paralelamente para abajo. El locus del punto de equilibrio se mueve a la izquierda.

```python
#--------------------------------------------------
    # Curva IS ORIGINAL

# Parámetros

Y_size = 100 

Co = 35
Io = 40
Go = 80
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3

Y = np.arange(Y_size)


# Ecuación 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)


#--------------------------------------------------
    # NUEVA curva IS
    
# Definir SOLO el parámetro cambiado
Go = 50

# Generar la ecuación con el nuevo parámetro
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r_G = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
```

```python
    # Gráfico

# Dimensiones del gráfico
y_max = np.max(r)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, r, label = "IS", color = "black") #IS orginal
ax.plot(Y, r_G, label = "IS_G", color = "C1", linestyle = 'dashed') #Nueva IS

# Texto agregado
plt.text(47, 162, '∆Go', fontsize=12, color='black')
plt.text(49, 159, '←', fontsize=15, color='grey')

# Título, ejes y leyenda
ax.set(title = "Disminución en el Gasto de Gobierno $(G_0)$", xlabel= 'Y', ylabel= 'r')
ax.legend()

plt.show()
```

#### Analice una Política Fiscal Expansiva con una caída de la Tasa de Impuestos.

Análisis intuitivo:

$$ t↓ → ↑DA → S↓ → I > S → Y↑ $$

Si disminuye la tasa de tributación, esto genera un aumento del Consumo y de las Importaciones. Al aumentar el consumo y las importaciones disminuye el ahorro, lo que genera que las inversiones sean mayores. 

Por lo tanto, al disminuír la tasa de tributación, la pendiente de la curva IS disminuye, ya que el parámetro "t" disminuye el tamaño del multiplicador.

```python
#--------------------------------------------------
    # Curva IS ORIGINAL

# Parámetros

Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.6

Y = np.arange(Y_size)


# Ecuación 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)


#--------------------------------------------------
    # NUEVA curva IS
    
# Definir SOLO el parámetro cambiado
t = 0.3

# Generar la ecuación con el nuevo parámetro
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r_t = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
```

```python
    # Gráfico

# Dimensiones del gráfico
y_max = np.max(r)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, r, label = "IS", color = "black") #IS orginal
ax.plot(Y, r_t, label = "IS_t", color = "C1", linestyle = 'dashed') #Nueva IS

# Texto agregado
plt.text(50, 162, '∆t', fontsize=12, color='black')
plt.text(50, 158, '→', fontsize=15, color='grey')

# Título, ejes y leyenda
ax.set(title = "Caída en la Tasa de Tributación $(t)$", xlabel= 'Y', ylabel= 'r')
ax.legend()

plt.show()
```

#### Analice una caída de la Propensión Marginal a Consumir.

Análisis intuitivo:

$$ b↓ → DA↓ → S↑ → S > I → Y↓ $$

Si disminuye la propensión marginal a consumir, entonces disminuye la demanda agregada y aumenta el ahorro. Por lo tanto, el Ingreso debe disminuír para regresar al equilibrio de ahorro = inversión.

En el caso de la curva IS, "b" es parte de la pendiente, pero es un multiplicador negativo, por lo que al disminuír "b", aumenta el tamaño de la pendiente de la curva IS.

```python
#--------------------------------------------------
    # Curva IS ORIGINAL

# Parámetros

Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.6

Y = np.arange(Y_size)


# Ecuación 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)


#--------------------------------------------------
    # NUEVA curva IS
    
# Definir SOLO el parámetro cambiado
b = 0.4

# Generar la ecuación con el nuevo parámetro
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r_t = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
```

```python
    # Gráfico

# Dimensiones del gráfico
y_max = np.max(r)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, r, label = "IS", color = "black") #IS orginal
ax.plot(Y, r_t, label = "IS_t", color = "C1", linestyle = 'dashed') #Nueva IS

# Texto agregado
plt.text(80, 115, '∆b', fontsize=12, color='black')
plt.text(80, 110, '←', fontsize=15, color='grey')

# Título, ejes y leyenda
ax.set(title = "Caída en la Propensión Marginal a Consumir $(b)$", xlabel= 'Y', ylabel= 'r')
ax.legend()

plt.show()
```
