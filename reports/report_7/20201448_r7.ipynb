El artículo “The Effects of Exposure to Better Neighborhoods on Children”, se centra en explicar en base a grupos de tratamiento y grupos de control, si efectivamente hay un cambio sustancial en los ingresos que perciben los niños en el futuro dependiendo si viven en un barrio con un ratio menor de pobreza. En este sentido la pregunta de investigación es ¿Existe algún efecto en el crecimiento a largo plazo de los niños dependiendo si viven o no en un barrio menos pobre?
En base a ello, se puede ver que una de sus principales fortalezas es la delimitación de edad hasta la cual consideraban a un individuo como niño, que en este caso fue 13 años. Ello permitió a los investigadores darse cuenta que si bien las personas menores de 13 años obtenían mejoras sustanciales a largo plazo en el ambito laboral y amoroso, ello no sucedía de la misma forma para individuos mayores de 13 años. 

Por otro lado, se puede resaltar la metodología del experimento, pues este para presentar resultados fidedignos distinguió entre dos grupos de tratamiento con los resultados de uno de control. El primer grupo de tratamiento fue trasladado a distritos con un indice de pobreza menor a 10%, mientras que el otro grupo era libre de escoger a que barrio mudarse. Este tipo de metodología ayuda de cierta forma a suplantar el problema de temporalidad de este tipo de experimentos, el cual se evidencia en que a una misma unidad de analisis no se le puede designar diferentes estados en el mismo tiempo, por lo que no se puede ver realmente si existe un efecto causal al 100 por ciento. Por ejemplo, en este caso no se puede ver si el niño dentro del grupo de tratamiento iba a recibir el mismo tipo de ingresos si se encontraba en el grupo de control. No obtante, este método al tratar con varios casos (niños) y dar diferencias contundentes con respecto al salario y estado civil de los niños en el futuro genera que se puedan sacar conclusiones pese a este problema existente. 

Ahora bien, con respecto a las debilidades, se puede ver el no identificar otro tipo de variables como el efecto positivo o no con respecto a la pertenencia a grupos delictivos a largo plazo en los niños al mudarse a barrios más seguros, pues si bien en estudios anteriores sí aparecen, considero que este estudio al intentar solventar la problemática de si efectivamente existe o no un efecto sustancial al mudarse de barrio, debían de abordar esta problemática tan importante que afecta a una parte significativa de personas en situaciones de pobreza. 

Ahora bien, este artículo aporta a la academia ayudando a responder una problemática muy grande sobre si existen efectos economicos positivos en los niños si se mudan a barrios menos pobres. Sobre este punto se terminó concluyendo que si existe una diferencia significativa en el desarrollo del niño (menores de 13 años) si se mudan a barrios menos pobres, en comparación con los que no lo hacen. No obstante, también concluye que no existen beneficios en los mayores de 13 años y que mas bien podría afectarles de manera negativa, debido a que su ambiente de vivienda cambia drasticamente (de uno pobre a uno que no lo es). Estos resultados no solo permiten al Estado poder entender cuan importante es el lugar de vivienda en los efectos a largo plazo en el niño. Sino, también beneficiarse pues a largo plazo esos niños que sean apoyados pagarán más impuestos y con ello el ingreso fiscal del Estado crecerá. Además, que estos al ser profesionales más competentes, podrán aportar más no solo con respecto a ingresos sino  también mediante externalidades positivas de sus trabajos, por ejemplo, si es un médico aportara tratando pacientes.

Por ejemplo, en el caso de Medina (2002), explica que ello se debe a que un niño no tiene conocimientos anteriores, por lo que le es sumamente sencillo aprender diferentes materias o acoplarse a nuevos escenarios, en cambio un adulto al tener ya mucho conocimiento almacenado, le es dificil aprender cosas que se contrapongan a estos conocimientos ya adquiridos. En ese sentido es mucho más dificil para un adulto adaptarse a un barrio socioeconomicamente muy diferente que en el caso de un niño.
