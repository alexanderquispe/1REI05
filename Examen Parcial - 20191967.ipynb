1. Defina el Multiplicador Keynesiano y dé un ejemplo de cómo influye este en el modelo Ingreso-Gasto Keynesiano.

El concepto de Multiplicador Keynesiano implica que un cambio en la magnitud de cualquiera de los componentes autónomos de la demanda agregada (consumo, inversión, gasto, exportaciones) va a generar un proceso multiplicador del ingreso, hasta converger al nuevo ingreso y producto de equilibrio. De esta manera, Jiménez plantea que "el aumento multiplicado del ingreso es resultado de los efectos directos e indirectos ocasionados por el aumento de cualquiera de los componentes autónomos de la demanda agregada".  A su vez, Keynes desarrolla el concepto de multiplicador de inversión (k) basándose en una economía cerrada y sin gobierno, limitada a la producción de Bienes de Inversión y Bienes de Consumo. De este modo indica que, cuando existe un incremento en la inversión total, el ingreso aumentará en una cantidad que es "k" veces el incremento de la inversión. Lo anterior es, a su vez, una forma de ilustrar la influencia del Multiplicador Keynesiano sobre el modelo Ingreso-Gasto.

$$k = \frac{1}{1 - (b - m)(1 - t)}$$

2. Grafique y explique cúando sucede un exceso y déficit de demanda en el modelo Ingreso-Gasto. Señale las áreas donde ocurre cada caso. Explique cómo se converge al equilibrio a partir de estos dos escenarios.


```python
##Curva de ingreso 

# Parámetros
Y_size = 100 

Co = 35
Io = 40
Go = 40
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3
r = 0.9

Y = np.arange(Y_size)

# Ecuación 
def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)

# Recta 45°

a = 2.5 

def L_45(a, Y):
    L_45 = a*Y
    return L_45

L_45 = L_45(a, Y)
```

```python
# Gráfico
y_max = np.max(DA_IS_K)
fig, ax = plt.subplots(figsize=(10, 8))

# Graficando curvas
ax.plot(Y, DA_IS_K, label = "Demanda Agregada", color = "red") #curva ORIGINAL
ax.plot(Y, L_45, color = "#404040") #Recta 45º

# Lineas punteadas
plt.axvline(x = 75, ymin= 0, ymax = 0.75, linestyle = ":", color = "grey")
plt.axvline(x = 56,  ymin= 0, ymax = 0.56, linestyle = ":", color = "grey")
plt.axvline(x = 20,  ymin= 0, ymax = 0.50, linestyle = ":", color = "grey")
plt.axhline(y = 139, xmin= 0, xmax = 0.55, linestyle = ":", color = "grey")

# Texto
plt.text(0, 150, '$DA^e$', fontsize = 11.5, color = 'black')
plt.text(55, 145, '$E$', fontsize = 11.5, color = 'black')
plt.text(75, 150, '$B$', fontsize = 11.5, color = 'black')
plt.text(20, 130, '$A$', fontsize = 11.5, color = 'black')
plt.text(6, 4, '$45°$', fontsize = 11.5, color = 'black')
plt.text(2.5, -3, '$◝$', fontsize = 30, color = '#404040')
plt.text(57, 0, '$Y^e$', fontsize = 12, color = 'black')

# Título
ax.set(title = " Exceso y déficit de demanda en modelo Ingreso-Gasto", xlabel = r'Y', ylabel = r'DA')
ax.legend()

plt.show()
```

Para el modelo de Ingreso-Gasto, el punto A indica el momento en que la demanda agregada (DA) es mayor que el nivel de producción, lo cual genera un exceso de demanda que reduce el inventario del bien disponible hasta llegar al equilibrio (E). Posteriormente, el punto B señala el momento en que la DA es menor que el nivel de producción, lo cual genera un déficit de demanda que genera el aumento de los inventarios y, por tanto, una consecuente disminución de la producción hasta (otra vez) llegar al equilibrio. De este modo, ambos escenarios convergen al equilibrio ya que la producción se ajusta en ambos casos a los cambios de la DA.

## Ingreso de Equilibrio

Primero, demanda agregada:

$$DA=C+I+G+X-M$$

Para lo cual:

$ C= C_0+bY^d$

$ I= I_0-hr$

$ G= G_0$

$ X= x_1Y^*+x_2\epsilon$

$ M= m_1Y^d-m_2\epsilon$

$ T= tY$

$Y^d=Y-T$

$DA=C+I+G+(X-M)$

$Y=DA$

$\epsilon=\epsilon_0-p(r-r^*)$

Así:

$ Y= C_0+bY^d+I_0-hr+G_0+x_1Y^*+x_2\epsilon-m_1Y^d+m_2\epsilon$

$ Y-(b-m_1)Y^d = C_0 + I_0 + G_0 + x_1Y^* +(x_2+m_2)\epsilon-hr$

$ Y-(b-m_1)(1-t)Y=C_0 + I_0 + G_0  + x_1Y^* + (x_2+m_2)(\epsilon_0-pr+pr^*)-hr$

$ Y[1-(b-m_1)(1-t)]=C_0 + I_0 + G_0 + x_1Y^* + (x_2+m_2)(\epsilon_0+pr^*)-[(x_2+m_2)p+h]r$

$Y=\frac{1}{1-(b-m_1)(1-t)} [C_0 + I_0 + G_0 + x_1Y^* + (x_2+m_2)(\epsilon_0+pr^*)]-\frac{1}{1-(b-m_1)(1-t)}[(x_2+m_2)p+h]r$

Donde $\frac{1}{1-(b-m_1)(1-t)}$ es el multiplicador keynesiano.
