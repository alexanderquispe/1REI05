### 1. (3 puntos) Instrumental Variables (IV)
Suponga que estamos interesados en estudiar la relación entre educación(X) y el nivel de ingresos(Y).
Un investigador propone utilizar como variable instrumental(Z) el numero de horas que una persona utilizó para ver televisión cuando era niño.

- **Identifique todos los elementos necesarios para implementar un enfoque de IV en este ejemplo.**

Para implementar un enfoque de variables instrumentales en este ejemplo, es necesario identificar una variable de tratamiento ($X$), una variable de resultados ($Y$) y una variable instrumental ($Z$). Luego, se debe verificar que el instrumento a utilizar $(z)$, en este caso el número de horas que una persona utilizó para ver televisión cuando era niño, cumpla con dos supuestos: relevancia y validez. El supuesto de relevancia se refiere a que $Z$ (el instrumento) realmente debe estar relacionado con $X$ (el tratamiento). $Z$ debe estar relacionada al menos con una parte de $X$, para poder aislar los resultados obtenidos. En este caso, se evidencia que el instrumento propuesto $Z$, variable que mide el número de horas que una persona utilizó para ver televisión cuando era niño, no cumple del todo con el supuesto de relevancia, pues no explica causalmente el nivel de educación. Podría tratarse de una correlación, mas no de una relación de causalidad. 

El supuesto de validez, por otro lado, se refiere a el instrumento $Z$ no debe tener puertas traseras abiertas propias; es decir, cualquier path entre el instrumento y el resultado debe ser influenciado por el tratamiento y no por otras variables externas. En este caso, pueden haber otros factores que influencien la variable instrumental, como el tiempo disponible, la crianza de los padres, etc.

- **Explique el procedimiento para encontrar el efecto causal en este escenario. Puede utilizar ecuaciones**

En primer lugar, se debe utilizar el instrumento $(Z)$, es decir el número de horas usado para ver televisión cuando niños, para explicar el tratamiento, es decir la educación. La variable instrumental mejora la regresión lineal, puesto que remueve todas las partes del tratamiento que no son explicadas por el instrumento. Por lo tanto, el instrumento debe ser utilizado para explicar el resultado. El número de horas usadas para ver televisión debe ser el instrumento utilizado para determinar el efecto del nivel de educación que tiene la población. A su vez, este debe presentar una relación de causalidad hacia el nivel de ingresos. Para mejorar el análisis, se debe eliminar toda aquella parte del resultado que no es explicado por el instrumento.

Se debe aplicar la variable instrumental para aislar a una parte de la población, logrando establecer una relación causal en esta variable aislada. Entre la variable instrumental y $Y$ debe existir una correlación que refleje la causalidad. Luego, se deben descartar la existencia de puertas traseras. 

La verdadera relación de causalidad se analiza observando la parte restante del resultado explicada por el instrumento, que a su vez proviene de reducir la muestra tomando sólo aquellos casos en las que la variable $X$, el tratamiento no es explicada por el instrumento. En este caso, el número de horas en las que se vio televisión cuando niño debe explicar y mantener una relación causal tanto con la educación como con el nivel de ingresos.

- **¿El efecto de la educación sobre el nivel de ingresos es positivo o negativo? Argumente si los resultados serian causales o solamente de correlación**

El efecto de la educación sobre el nivel de ingresos es positivo. En este caso, se evidencia no sólo una correlación, sino también una causalidad, puesto que el acceso a una educación de calidad, especialmente en el nivel superior, condiciona mejores oportunidades laborales, con una mejor remuneración, lo cual aumenta el nivel de ingresos. En el caso planteado, no podría existir una relación causal entre $Z$ y $X$, pero sí una correlación. En ese sentido, una mejor variable instrumental sería la tasa de admisión a instituciones de educación superior en la muestra de un grupo de escolares de último año. Esta variable cumpliría con los supuestos de validez y relevancia.

### 2. (3 puntos) Regression Discontinuity Design (RDD)
Supongamos que estamos estudiando la eficacia de un programa de becas creado por el gobierno Peruano sobre las notas de los alumnos al final del primer semestre en la universidad. PRONABEC ha decidido que entregará becas a aquellos postulantes que tengan un promedio superior durante la secundaria igual a 16 . 

- **Identifique todos los elementos necesarios para implementar un RDD en este ejemplo.**

El diseño de regresión discontinua (Regression Discontinuity Design, RDD por sus siglas en inglés), tiene como objetivo determinar el efecto causal de la variable de ejecución $x$ en la variable de resultados $y$. Para ello, se hará un análisis mediante un punto de corte o umbral en la intervención. El primer elemento a tomar en cuenta, como se mencionó, es la variable de ejecución $x$, en este caso la asignación de becas, que busca condicionar o influir en las notas de los estudiantes para determinar si reciben el tratamiento o no. Luego, se encuentra el corte ($C_o$), referido al límite de la variable donde se determina la recepción del tratamiento. En este caso, el corte sería un promedio igual o superior a 16 durante la secundaria. Además, se encuentra el ancho de banda ($h$), que marca el puntaje seleccionado para medir el impacto. Por último, se encuentra el tratamiento $D$, que obtiene un valor binario (1 = sí se recibió y 0 = no se recibió)

- **Explique el procedimiento para encontrar el efecto causal en este escenario. Puede utilizar ecuaciones.**

El procedimiento para encontrar el efecto causal en este escenario consiste, inicialmente, en verificar el cumplimiento de 3 requisitos. En primer lugar, se debe verificar que el comportamiento de otras características sea continuo alrededor del punto de corte, es decir que hayan similitudes entre los casos observables y que la única diferencia sea $D$, es decir la posibildiad de recibir tratamiento o no. En segundo lugar, se debe mantener la regla asignada para separar a los dos grupos, en este caso un promedio en la secundaria mayor o igual a 16. Se debe evitar manipular la asignación en el tratamiento. Por último, debe haber un aumento significativo de la probabilidad de recibir el tratamiento en el punto de corte.

Una vez verificados estos requisitos, es necesario definir las variables del resultado $X$ (tratamiento, programa de asignación de becas del PRONABEC) y $Y$ (eficacia de $X$ medida mediante las notas en las notas de la universidad). Luego, se debe definir la ponderación de las observaciones, seleccionar el ancho de banda y revisar el impacto de la variable $X$ mediante la RDD verificando la diferencia de medias entre ambos grupos, o hacer dos regresiones que nos ayuden a medir el impacto sobre $Y$ y la diferencia entre los grupos de control y tratemiento. La ecuación que podría determinar esto es: $$Yi = a1 + β1Xi + β2(X1 -X*) Di + ui$$

- **¿El efecto de la politica sobre el rendimiento de los estudiantes sería positivo o negativo (solo necesitan usar su intuición)? Argumente si los resultados serian causales o solamente de correlación.**

Es posible afirmar que el efecto de la política de asignación de becas sería positivo sobre el rendimiento de los estudiantes. No sólo estamos hablando de una correlación, sino también de una relación causal, puesto que un incentivo sobre las buenas calificaciones condiciona mejores resultados en el futuro. En este caso, es posible afirmar que el modelo de RDD es útil, pues en este caso contamos con unn punto de corte determinado que nos permite medir la relación de causalidad entre dos grupos.


### 3. (3 puntos) Difference in Difference (DiD)
Supongamos que estamos interesados en estudiar el impacto de la migración en la tasa de empleo en dos regiones del Perú en 2017-2018. Tenemos datos anuales sobre la tasa de empleo en Lima, que experimentó una afluencia significativa de inmigrantes en 2018, y Puno que no experimentó el ingreso de migrantes (2017 ninguna de las regiones experimentó afluencia de migrantes). Podemos usar un análisis DID para comparar el cambio en las tasas de empleo entre las dos regiones a lo largo del tiempo para ver si la migración tuvo algún efecto.

- **Identifique todos los elementos necesarios para implementar un DiD en este ejemplo.**

Los elementos necesarios para implementar un enfoque Difference in Difference (DiD) son cuatro. En primer lugar, se encuentra el tratamiento ($D$), en este caso los niveles de migración en el Perú entre 2017 y 2018. En segundo lugar, se encuentra el resultado ($Y$) en este caso la tasa de empleo en Lima. Luego, los grupos de tratamiento ($k$), en este estudio se eligió la población de Lima por ser la que mayor número de migrantes recibió, y de control ($U$). La población de Puno fue elegida al ser una región que no recibió migrantes.

- **Explique el procedimiento para encontrar el efecto causal en este escenario. Puede utilizar ecuaciones.**

Las ecuaciones utilizadas para el procedimiento serían:
$$\widehat{\delta}^{2\times 2}_{kU} = \bigg ( \overline{y}_k^{-post(k)} - \overline{y}_k^{-pre(k)} \bigg ) - \bigg ( \overline{y}_U^{-post(k)} - \overline{y}_U^{-pre(k)} \bigg )$$

Luego de realizar las ecuaciones, se deben determinar la diferencia entre el primer momento, antes de aplicar el tratamiento, y el segundo, luego de hacerlo.

- **¿El efecto de la migración sobre la tasa de empleo sería positivo o negativo (solo necesitan usar su intuición)? Argumente si los resultados serian causales o solamente de correlación.**

En este caso, el efecto de la migración sobre la tasa de empleo sería negativo, puesto que a mayor número de migrantes habrán menos puestos disponibles de trabajo. No se trata sólo de una correlación, sino también de una causalidad.
