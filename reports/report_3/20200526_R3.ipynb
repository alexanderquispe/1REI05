## Ejercicio 1.1

El Banco central de reserva, tiene la facultad de controlar la a oferta monetaria (que se expanga o contraiga). Los instrumentos de la política monetaria son: en la oferta monetaría como instrumetno de política es la "Operación de Mercado Abierto" que usa la compra o venta de activos financieros; luego, el se utiliza el coeficiente legal de encaje tanto para la política monetaria expansiva y contractiva, ya que la disminución de la tasa de encaje aumenta la posibilidad de creación de dinero bancario; por último, la tasa de interés como instrumento, en la politica monetaria expansiva el Banco Central reduce su tasa de interés de referencia al aumentar la cantidad de dinero y, en la política monetaria contractiva al banco aumentar la tasa de interés de referente al reducir el crédito dado. 

## Ejercicio 1.2

La oferta real de Dinero se denota como:

$$ M^s= \frac{M_0^s}{P} $$

Donde:
$ M_0^s $: Masa Monetaria 

$P$:Nivel de Precio

## Ejercicio 1.3

La demanda real de dinero se denota  como:
$$ M^d= kY-ji $$

Donde:

$k$: propensión marginal a demandar dinero

$j$: constante de cambio en la demanda de dinero

$i$: tasa de interés nominal (o el precio del dinero)

La magnitud del parámetro “k” indica la sensibilidad de la demanda de dinero ante variaciones del ingreso “Y”, y la del parámetro “j” indica cuán sensible es la demanda de dinero ante las variaciones de la tasa de interés nominal de los bonos, “i”.

## Ejercicio 1.4

El Equilibrio de Mercado se deriva del equilibrio enre la Oferta de Dinero $M^s$ y Demanda de Dinero $M^d$
 
$$M^s=M^d$$

$$ \frac{M_0^s}{P}=kY-ji  $$

Si sopnermos que la inflación esperada es cero, ya que a corto plazo el nivel de precios es fijo y exógeno. Por lo tanto, no habría una gran diferencia entre la tasa de interés nominal  y la real .

Entonces la ecuación del equilibrio en el mercado monetario es:


$$ \frac{M_0^s}{P}=kY-jr  $$

## Ejercicio 1.5
Y puede graficarse de la siguiente manera:

```python
# Dando valores de parametros 
k = 0.5
j = 0.2                
P  = 10 
Y = 35
MS_0 = 500

r = np.arange(100)

# Necesitamos crear la funcion de demanda 

def MD(k, j, P, r, Y):
    MD_eq = (k*Y - j*r)
    return MD_eq
MD_0 = MD(k, j, P, r, Y)
# Necesitamos crear la oferta de dinero.
MS = MS_0 / P
# Equilibrio en el mercado de dinero

# Creamos el marco para la figura 
fig, ax1 = plt.subplots(figsize=(10, 8))

# Agregamos titulo t el nombre de las coordenadas

ax1.set(title="Equilibrio del Mercado Monetario")

# Ploteamos la demanda de dinero
ax1.plot(MD_0, label= '$L_0$', color = 'blue')

# Para plotear la oferta de dinero solo necesitamos crear una linea vertical
ax1.axvline(x = MS,  ymin= 0, ymax= 0.96, color = "grey") 

# Creamos las lineas puntadas para el equilibrio
ax1.axhline(y=7.5, xmin= 0, xmax= 0.5, linestyle = ":", color = "black")

# Agregamos texto
ax1.text(-10, 7.3, "$r_0^e$", fontsize = 20, color = 'black')
ax1.text(93, 0.6, "$L(Y_0)$", fontsize = 20, color = 'black')

ax1.text(-10, 17, "$r$", fontsize = 20, color = 'black')

ax1.text(45, -4.5, r'$M_0^s / P_0$', fontsize = 20, color = 'black')
ax1.text(95, -4.5, r'$M^s / P$', fontsize = 20, color = 'black')

ax1.yaxis.set_major_locator(plt.NullLocator())   
ax1.xaxis.set_major_locator(plt.NullLocator())

ax1.legend()

plt.show()
```

## Ejercicio 3.1 

Para la Curva LM, se obtiene del equilibrio del mercado del cual despejaremos la tasa de interes $r$ 

$$ \frac{M_0^s}{P_0}=kY-jr  $$

$$ \frac{M_0^s}{P_0}-kY=-jr  $$

$$ jr= -\frac{M_0^s}{P_0}+kY  $$

$$ r= -\frac{1}{j} \frac{M_0^s}{P_0}+ \frac{k}{j}Y  $$

Entonces:

$$ LM = -\frac{1}{j} \frac{M_0^s}{P_0}+ \frac{k}{j}Y  $$

del cual tiene una pendiente positiva 

```python
#Gráfico 

k = 0.5
j = 0.2                
P  = 10 
Y = 35
r = np.arange(100)#
Y_size =100
# Ecuación
def Ms_MD(k, j, P, r, Y):
    Ms_MD = P*(k*Y - j*r)
    return Ms_MD

Ms_MD = Ms_MD(k, j, P, r, Y)

# Nuevos valores de Y (45 y 25)
Y1 = 45
def Ms_MD_Y1(k, j, P, r, Y1):
    Ms_MD = P*(k*Y1 - j*r)
    return Ms_MD
Ms_Y1 = Ms_MD_Y1(k, j, P, r, Y1)

Y2 = 25
def Ms_MD_Y2(k, j, P, r, Y2):
    Ms_MD = P*(k*Y2 - j*r)
    return Ms_MD
Ms_Y2 = Ms_MD_Y2(k, j, P, r, Y2)

#2----------------------Curva LM
# Parameters 
k = 0.5
j = 0.2                
P  = 10               
Ms = 30            
Y = np.arange(100)#Y_size =100
# Ecuación
def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)
```

## Ejercicio 3.3 

Si aumenta k, entonces la pendiente de la curva LM va a aumentar.Esto se debe a que si k aumenta, la sensibilidad de los ingresos de las personas para tenerlos en su bolsillo es más alta. Esto genera que se requiera un aumento de la tasa de interés para regresar al equilibrio, para que así las personas decidan poner más de sus ingresos en los bancos.

$$ k↑ → M^d↑ → M^s < M^d → r↑ $$

```python
#Gráfico

 # Curva LM ORIGINAL

# Parámetros

Y_size = 100

k = 0.5
j = 1                
Ms = 200             
P  = 20               

Y = np.arange(Y_size)

# Ecuación

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)

#--------------------------------------------------
# NUEVA curva LM
k = 0.9 # un numero mayor en k
# Generar la ecuación con el nuevo parámetro
def i_LM_k( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i_k = i_LM_k( k, j, Ms, P, Y)

# Dimensiones del gráfico
y_max = np.max(i)
v = [0, Y_size, 0, y_max]   
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, i, label="LM", color = 'red')
ax.plot(Y, i_k, label="LM_k", color = 'blue', linestyle = 'dashed')

# Texto agregado
plt.text(80, 45, '∆$k$', fontsize=12, color='black')
plt.text(80, 50, '↑', fontsize=15, color='grey')

# Título y leyenda
ax.set(title = "Incremento en k", xlabel=r'Y', ylabel=r'r')
ax.legend()


plt.show()
```
