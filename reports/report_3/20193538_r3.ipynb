### Explique cuáles son los intrumentos de política monetaria que puede utilizar el Banco Central

Los instrumentos de la politica monetaria se pueden clasificar en cuatro.
En primer lugar, se encuentra **La operacion del mercado abierto**, en donde el BCR adquiere un importante papel en la compra-venta, tanto de activos financieros como de bonos. Sobre estos ultimos, podemos dividirlo en dos escenarios, en el de una politica monetaria contractiva, donde se retiran soles de la economia mediante la venta a bancos comerciales; y en el de una politica monetaria expansiva, donde a diferencia del primer escenario, se "inyectan" soles a la economia.

En segundo lugar, esta el **coeficiente legal de encaje**. Al igual que el primer instrumento, podemos situarnos en el caso de una politica monetaria contractiva, en la cual se aumenta el coeficiente legal de encaje, haciendo que los bancos obtengan mas depositos en reservas, reduciendo asi el multiplicador bancario y la oferta monetaria. Mientras que en la politica monetaria expansiva, el BCR reduce el encaje, lo que aumenta el dinero de los bancos para asi generar prestamos, lo que origina que estos tengan menos dinero en reservas.

El tercero es **la tasa de interes**, en una politica contractiva, como el BCR aumenta la tasa de interes, se disminuye el dinero prestado a los bancos, asi como la oferta y la base monetarias. En una politica monetaria expansiva, se reduce la tasa de interes, haciendo que el dinero prestado a los bancos aumente, junto a la oferta y base monetarias.

Finalmente, la cuarta es **la funcion de oferta real de dinero** la cual ademas es una variable exogena. Para hallarla, se tiene que dividir la oferta nominal del dinero entre el nivel general de precios de la economia $P$

### Derive la demanda real de dinero. Explique qué papel cumplen los parametros "k" y "j"
Hay que tener en cuenta que existen motivos por los cuales se demanda dinero, los cuales pueden ser.
- Motivo Transaccion (depende directamente del ingreso Y), es decir, para realizar transacciones de bienes y servicios, los cuales tienen una relacion directa con el ingreso o producto de la economia


- Motivo Precaucion (tambien depende positivamente del ingreso Y), que ocurre porque las familias o empresas se endeudan y deben de pagar en plazos determinados, en donde su pago depende de los ingresos

Estos dos primeros motivos se pueden expresar como:
$$L_1=kY$$


- Motivo Especulacion (depende negativamente de la tasa de interes de los bonos), donde "el activo financiero compite con el activo financiero no monetario bono en funcion de reserva de valor" (Jimenez, F.). Y cuando la tasa de interes o rendimiento disminuye, se preferira tener liquidez en forma de dinero y no de bonos. Esta expresado como:
$$L_2=-ji$$


Asumiendo que ambos estan en terminos reales, la demanda real de dinero sera:

$$L=L_1+L_2$$
$$L=kY - ji$$

Entonces, la magnitud de la sensibilidad de la demanda de dinero $(k)$ ante las variaciones de ingreso $(Y)$. Por otro lado, el parametro $(j)$, indicara la sensibilidad de la demanda de dinero ante las variaciones de la tasa de interes nominal de los bonos $(i)$

### Asumiendo que no hay inflación podemos asumir que *i=r* Escriba en terminos reales la ecuacion de equilibrio en el mercado de dinero
- $M^S$ = oferta de dinero
- $M^d$ = demanda de dinero
debe haber un equilibrio entre ambos para tener equilibrio en el mercado de dinero, entonces:
$$M^S=M^d$$


Para que la inflacion sea 0, el nivel de precios sera fijo y exogeno al corto plazo, entonces tampoco habria diferencias entre las tasas de interes real (r) o nominal (i), y la ecuacion de equilibrio en el mercado de dinero seria:



### Grafique el equilibrio en el mercado de dinero.

```python
# Parameters
r_size = 100

k = 0.5
j = 0.2                
P  = 10 
Y = 35
MS_0 = 500

r = np.arange(r_size)

# Necesitamos crear la funcion de demanda 

def MD(k, j, P, r, Y):
    MD_eq = (k*Y - j*r)
    return MD_eq
MD_0 = MD(k, j, P, r, Y)
# Necesitamos crear la oferta de dinero.
MS = MS_0 / P
MS
```

```python
# Equilibrio en el mercado de dinero

# Creamos el seteo para la figura 
fig, ax1 = plt.subplots(figsize=(10, 8))

# Agregamos titulo t el nombre de las coordenadas
ax1.set(title="Money Market Equilibrium", xlabel=r'M^s / P', ylabel=r'r')

# Ploteamos la demanda de dinero
ax1.plot(MD_0, label= '$L_0$', color = '#006400')

# Para plotear la oferta de dinero solo necesitamos crear una linea vertical
ax1.axvline(x = MS,  ymin= 0, ymax= 1, color = "royalblue")

# Creamos las lineas puntadas para el equilibrio
ax1.axhline(y=7.5, xmin= 0, xmax= 0.5, linestyle = ":", color = "black")

# Agregamos texto
ax1.text(0, 7.8, "$r_0$", fontsize = 12, color = 'black')
ax1.text(52, 0, "$(M_o^s) / P_0$", fontsize = 12, color = 'black')
ax1.text(85, 1, "$L(Y_0)$", fontsize = 12, color = 'black')

ax1.yaxis.set_major_locator(plt.NullLocator())   
ax1.xaxis.set_major_locator(plt.NullLocator())

ax1.legend()

plt.show()
```

### 2)¿Cuál es el efecto de una disminución en la Masa Monetaria Ms<0? Explica usando la intuición y gráficos.
Si hay una disminucion de la masa monetaria, entonces la oferta tambien disminuiria y se generaria un desiquilibrio


```python
Y_size = 100

k = 2
j = 1                
Ms = 200             
P  = 20               

Y = np.arange(Y_size)

# Ecuación

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)

#--------------------------------------------------
    # NUEVA curva LM

# Definir SOLO el parámetro cambiado
Ms = 20

# Generar la ecuación con el nuevo parámetro
def i_LM_Ms( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i_Ms = i_LM_Ms( k, j, Ms, P, Y)
```

### 3) ¿Cuál es el efecto de un aumento en k>0? Explica usando intuición y gráficos.
Si $k$ aumenta, entonces tambien lo hace la demanda monetaria, lo que significa que la curva $lm$ no esta en equilibrio. Para volver al equilibrio, la tasa de interes tiene que subir
$$↑k→↑kY→M^d>M^s→↑r$$

```python
#--------------------------------------------------
    # Curva LM ORIGINAL

# Parámetros

Y_size = 100

k = 2
j = 1                
Ms = 200             
P  = 20               

Y = np.arange(Y_size)

# Ecuación

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)

#--------------------------------------------------
    # NUEVA curva LM

# Definir SOLO el parámetro cambiado
k = 4

# Generar la ecuación con el nuevo parámetro
def i_LM_Ms( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i_Ms = i_LM_Ms( k, j, Ms, P, Y)
```

```python
# Dimensiones del gráfico
y_max = np.max(i)
v = [0, Y_size, 0, y_max]   
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, i, label="LM", color = 'dodgerblue')
ax.plot(Y, i_Ms, label="LM_Ms", color = '#3CB371', linestyle = 'dashed')

# Texto agregado
plt.text(47, 140, '∆$k$', fontsize=12, color='black')
plt.text(47, 120, '←', fontsize=15, color='black')

# Título y leyenda
ax.set(title = "Aumento en k $(k)$", xlabel=r'Y', ylabel=r'r')
ax.legend()


plt.show()
```
