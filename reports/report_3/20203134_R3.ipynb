1.Explique cu√°les son los intrumentos de pol√≠tica monetaria que puede utilizar el Banco Central 

Los instrumentos de la pol√≠tica monetaria son cuatro:
1. Operaci√≥n de Mercado Abierto: En este instrumento, el BCR realiza operaciones de √≠ndole compra y venta de activos financieros o bonos a los bancos comerciales y privados. De este modo, es mediante la pol√≠tica monetaria expansiva, en la cual el BCR compra bonos y logra inyectar dinero en la econom√≠a. Paralelamente, a trav√©s de la pol√≠tica monetaria contractiva se da con la venta de bonos al mercado nacional (bancos comerciales) retira dinero de la econom√≠a para la reducci√≥n de una inflaci√≥n.


2. El coeficiente legal de encaje como instrumento de pol√≠tica. En el caso de una pol√≠tica monetaria expansiva el BCR puede aumentar el coeficiente de encaje que poseen los bancos para generar pr√©stamos mediante la reducci√≥n de la tasa de inter√©s. A partir de lo cual, se espera que los bancos comerciales tengan menos dinero en reservas dentro del BCR. En el caso de la pol√≠tica monetaria contractiva, el BCR aumenta el coeficiente de encaje, es as√≠ como los bancos comerciales tienen una mayor proporci√≥n de dep√≥sitos en reservas y ello reduce el multiplicador bancario y origina una disminuci√≥n en la oferta monetaria.


3. La tasa de inter√©s como instrumento de pol√≠tica. Con la consideraci√≥n de la tasa de inter√©s como pol√≠tica, la oferta monetaria pas√≥ a ser una variable end√≥gena. Es as√≠ que se comprende que la tasa pol√≠tica es la misma que la tasa de inter√©s de referencia de la pol√≠tica monetaria. En el caso de la pol√≠tica monetaria expansiva, el BCR reduce la tasa de inter√©s de referencia, con lo cual aumenta el dinero en modalidad de pr√©stamo hacia los bancos comerciales y ello incrementa la base monetaria y consecuentemente forma la oferta monetaria. Mientras que en el caso de la pol√≠tica monetaria contractiva el BCR aumenta la tasa de inter√©s de referencia, con lo cual reduce el dinero prestado a los bancos comerciales y ello reduce la base monetaria y la oferta monetaria.


4. Funci√≥n de Oferta Real de Dinero. La oferta nominal de dinero (ùëÄùë†) al ser una variable ex√≥gena e instrumento de pol√≠tica monetaria. Seguidamente en t√©rminos reales o a precios constantes, la oferta nominal se divide entre en nivel general de precios de la econom√≠a, (ùëÉ).

5. Grafique el equilibrio en el mercado de dinero

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import sympy as sy
import pandas as pd
import numpy as np
import random
import math
import sklearn
import scipy as sp
import networkx
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.iolib.summary2 import summary_col
```

```python
# Parameters
r_size = 100

k = 0.5
j = 0.2                
P  = 10 
Y = 35
MS_0 = 500

r = np.arange(r_size)

# Necesitamos crear la funcion de demanda 

def MD(k, j, P, r, Y):
    MD_eq = (k*Y - j*r)
    return MD_eq
MD_0 = MD(k, j, P, r, Y)
# Necesitamos crear la oferta de dinero.
MS = MS_0 / P
MS
# Equilibrio en el mercado de dinero

# Creamos el seteo para la figura 
fig, ax1 = plt.subplots(figsize=(10, 8))

# Agregamos titulo t el nombre de las coordenadas
ax1.set(title="Money Market Equilibrium", xlabel=r'M^s / P', ylabel=r'b')

# Ploteamos la demanda de dinero
ax1.plot(MD_0, label= '$L_0$', color = 'green')

# Para plotear la oferta de dinero solo necesitamos crear una linea vertical
ax1.axvline(x = MS,  ymin= 0, ymax= 1, color = "grey")

# Creamos las lineas puntadas para el equilibrio
ax1.axhline(y=7.5, xmin= 0, xmax= 0.5, linestyle = ":", color = "black")

# Agregamos texto
ax1.text(0, 7.5, "$r_0$", fontsize = 12, color = 'black')

ax1.yaxis.set_major_locator(plt.NullLocator())   
ax1.xaxis.set_major_locator(plt.NullLocator())

ax1.legend()

plt.show()
```

Al intentar cumplir esta condici√≥n de equilibrio, esta disminuci√≥n de la elasticidad del ingreso genera un exceso de oferta en el mercado. Es as√≠ como que, al demandar menos dinero, tambi√©n provocar√° que disminuya el nivel de consumo por la carencia de efectivo circulatorio y la sobreoferta de este. Por tanto, el equilibrio debe restablecerse mediante una disminuci√≥n en la tasa de inter√©s. Este efecto, se reflejar√° en un desplazamiento de la curva de demanda de dinero hacia abajo o el interior.
