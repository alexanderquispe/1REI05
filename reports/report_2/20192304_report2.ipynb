Sabemos que en el equilibrio el ingreso agregado tiene que ser igual a la demanda agregada, por lo que nos quedar√≠a as√≠
Y=DA.

A partir de Y=DA, seguimos con la ecuaci√≥n del punto de equilibrio a corto plazo:
    ùëå =(1/(1 ‚àí (ùëè ‚àí ùëö)(1 ‚àí ùë°)(ùê∂0 + ùêº0 + ùê∫0 + ùëã0 ‚àí ‚Ñéùëü)

Continuamos con las ecuaciones, pero desde la tasa de inter√©s que es lo que va dentro del an√°lisis:
    
     r=(1/h*(ùê∂0 + ùêº0 + ùê∫0 + ùëã0) ‚àí((1-(b-m)*(1-t)/h)(Y)
        
        Lo reducimos a:
        
        r=L1/h-(L2/h)y

El intercepto para la curva IS: 
    L1= ùê∂0 + ùê∫0 + Io + ùëã0

```python
%matplotlib widget
%matplotlib inline
import ipywidgets as widgets
import matplotlib.pyplot as plt
import numpy as np
import sympy as sy
from sympy import *
import pandas as pd
```

```python

Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3

Y = np.arange(Y_size)


# Ecuaci√≥n 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
```

Pregunta c: Desequilibrios en el mercado de bienes

El desquilibrio en los mercados pueden ocurrir porque, por ejemplo, un precio real est√° por encima del precio de equilbrio,
la demanda de los productos aumenta, pero como los precios son altos puede que haya menos demanda y por consiguiente el 
llamado "exceso de oferta".Por otro lado, cuando real es inferior al precio de equilibrio, habr√° menos oferta de los productos y m√°s demandantes dispuestos a adquirir esos productos. 
Se producir√° por tanto un exceso de demanda, por lo tanto, el precio real tender√° a aumentar debido a la cantidad
de personas que quieren el producto.Es por ello que no estos casos no son considerados dentro del punto de equilibrio en el mercado de bienes.

Parte 1:

Si una pol√≠tica Fiscal Contractiva orgina una ca√≠da en el gasto del gobierno,

Aquello genera un desplazamiento de la curva IS hacia la izquierda porque tendr√° que disminuir la
producci√≥n y eso generar√≠a un exceso de demanda.

Cuando la tasa de inter√©s de tributaci√≥n, por consecuencia la demanda agregada sube porque las personas ahorran m√°s dinero
que es el porcentaje que le quita el Estado.Entonces, la demanda al subir tambi√©n tiene que subir la producci√≥n, generando as√≠

lo cual se genera un desplazamiento hacia la derecha.

```python
# Curva IS original
# Par√°metros

Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3

Y = np.arange(Y_size)


# Ecuaci√≥n 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)


#--------------------------------------------------
    # NUEVA curva IS
    
# Definir SOLO el par√°metro cambiado
t = 0.05

# Generar la ecuaci√≥n con el nuevo par√°metro
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r_t = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
```

Parte 3: 
    (t) ‚Üì ‚Üí Co ‚Üì‚Üí C ‚Üì‚Üí DA> Y  ‚Üí Y ‚Üë t ‚Üì ‚Üí M ‚Üì ‚Üí DA ‚Üì ‚Üí DA < Y  ‚Üí Y ‚Üì 
    
    La relaci√≥n entre la propensi√≥n marginal a consumir y la Demanda agregada es directa. Si la propensi√≥n
    marginal cae, por ende, caer√° el consumo y eso causar√≠a la ca√≠da de la demanda agregada y un exceso de demanda.
    "Cuando aumenta (disminuye) ‚Äúm‚Äù, la pendiente de la curva IS se incrementa
    (reduce). Por el contrario, el aumento de la propensi√≥n marginal a consumir, ‚Äúb‚Äù,
    aumenta el tama√±o del multiplicador, aumenta la pendiente de la DA y reduce la pendiente
    de la curva IS. Esta gira en sentido contrario a las agujas del reloj" (Jimenez,2020,pp.74)
   
    
  

```python
# Curva IS original
# Par√°metros

Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3

Y = np.arange(Y_size)


# Ecuaci√≥n 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)


#--------------------------------------------------
    # NUEVA curva IS
    
# Definir SOLO el par√°metro cambiado
b = 0.5

# Generar la ecuaci√≥n con el nuevo par√°metro
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r_b = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
```

```python
# Dimensiones del gr√°fico
y_max = np.max(r)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, r, label = "IS", color = "green") #IS orginal
ax.plot(Y, r_b, label = "IS_t", color = "#A52A2A", linestyle = 'dashed') #Nueva IS

# Texto agregado
plt.text(49, 162, '‚àÜt', fontsize=12, color='green')
plt.text(49, 158, '‚Üê', fontsize=15, color='black')

# T√≠tulo, ejes y leyenda
ax.set(title = "Disminuci√≥n en la PMgC $(b)$", xlabel= 'Y', ylabel= 'r')
ax.legend()

plt.show()
```
