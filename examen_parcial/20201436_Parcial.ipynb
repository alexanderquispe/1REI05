1. (2 puntos) Defina el Multiplicador Keynesiano y dé un ejemplo de cómo influye este en el modelo Ingreso-Gasto Keynesiano.
 Y: ingreso
DA: Demanda Agregada
Definición del multiplicador keynesiano:  es el multiplicador que provoca que la renta se multiplique hasta alcanzar al nuevo ingreso y producto de equilibrio. Así, un aumento múltiple del ingreso (Y) es la consecuencia de los efectos directos e indirectos provocados por un aumento de los componentes autónomos de la demanda agregada (DA). El tamaño del multiplicador obedece a: -la magnitud de los parámetros en la propensión marginal a ahorrar (b), -la propensión marginal a importar (m) y -la tasa promedio de impuestos (t).
