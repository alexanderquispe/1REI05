### A partir del siguiente sistema de ecuaciones que representan el modelo IS-LM 

La funci√≥n que representa la Curva IS es:

         r=ùú∑o/h - (ùú∑1/h)Y 
         
donde:ùú∑o=Co+Io+Go+Xo y ùú∑1=1-(b-m)(1-t)

La funci√≥n de la Curva LM es:

        r=-1/j*Mo^s/Po + k/j*Y
        
Todos los par√°metros son positivos. La propensi√≥n marginal a consumir est√° entre cero y uno: 0 < b < 1

1. Encuentre las ecuaciones de Ingreso (Y^e) y tasa de interes (r^e) de equilibrio(Escriba paso a paso la derivacion de estas ecuaciones).

Para hallar el Y^e igualamos 

        ùõΩo/h - (ùõΩ1/h)Y = -1*Mo^s/j*P + KY/j
        
        KY/j + (ùõΩ1/h)Y = ùõΩo/h + 1*Mo^s/j*P
        
   Factorizamos: Y(K/j + ùõΩ1/h) = ùõΩo/h + 1*Mo^s/j*P
   
                 Y ((Kh + ùõΩ1*j)/jh) = (ùõΩo*jP + Mo^s*h)/hjP
                 
                 Y = (ùõΩo*jP + Mo^s*h)/P(Kh + ùõΩ1*j)
                 
                 Y= ùõΩo*jP/P(Kh + ùõΩ1*j) + Mo^s*h/P(Kh + ùõΩ1*j)
   
   Por lo tanto la ecuaci√≥n del ingreso de equilibrio (Y^e) es:
   
   Y^e = ùõΩo*j/(Kh + ùõΩ1*j) + (h/Kh + ùõΩ1*j) * M^o/P
   
   Para hallar la tasa de inter√©s de equilibrio reemplazamos 1 en 2 y despejamos r.
   
  De 1  r = ùõΩo/h - ùõΩ1*Y/h     despejamos Y
  
        Y = ùõΩo/ùõΩ1 - r*h/ùõΩ1
        
   Luego reemplazamos
   
       r = -1/j*Mo^s/P + K/j*Y
       
       r = -1/j*Mo^s/P + K/j (ùõΩo/ùõΩ1 - rh/ùõΩ1)
       
       r = -1/j*Mo^s/P + K*ùõΩo/jùõΩ1 - Krh/jùõΩ1
       
       r + Krh/jùõΩ1 = KùõΩo/jùõΩ1 - 1/j*Mo^s/P
       
  Factorizamos r
    
     r(1 + Kh/jùõΩ1) = KùõΩo/jùõΩ1 - 1/j * Mo^s/P
     
     r = [jùõΩ1(KùõΩo/jùõΩ1 - 1/j*Mo^s/P)]/jùõΩ1 + Kh = [KùõΩo - (ùõΩ1*Mo^s/P)]/jùõΩ1 + Kh
     
   Por lo tanto la ecuaci√≥n de la tasa de inter√©s de equilibrio (Y^e) es:
   
     r^e = KùõΩo/Kh + jùõΩi - (ùõΩ1/Kh + jùõΩ1)*Mo^s/P
   
   
        

#### GRAFICO

```python
# Para la Curva IS

# Par√°metros

Y_size = 100 

Co = 35
Io = 40
Go = 50
Xo = 2
h = 0.8
b = 0.4
m = 0.5
t = 0.8

Y = np.arange(Y_size)


# Ecuaci√≥n 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo)/h - ( ( 1-(b-m)*(1-t) ) / h)*Y  
    return r_IS

r_is = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)

#--------------------------------------------------
# Para la Curva LM 

# Par√°metros

Y_size = 100

k = 2
j = 1                
Ms = 200             
P  = 20               

Y = np.arange(Y_size)

# Ecuaci√≥n

def r_LM(k, j, Ms, P, Y):
    r_LM = - (1/j)*(Ms/P) + (k/j)*Y
    return r_LM

r_lm = r_LM( k, j, Ms, P, Y)
```

```python
# Gr√°fico del modelo IS-LM

# Dimensiones del gr√°fico
y_max = np.max(r_lm)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
# IS
ax.plot(Y, r_is, label = "IS", color = "pink") #IS
# LM
ax.plot(Y, r_lm, label="LM", color = "green")  #LM

# Eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

# Texto y figuras agregadas
# Graficar la linea horizontal - r
plt.axvline(x=51.5,  ymin= 0, ymax= 0.52, linestyle = ":", color = "black")
# Grafica la linea vertical - Y
plt.axhline(y=93, xmin= 0, xmax= 0.52, linestyle = ":", color = "black")

# Plotear los textos 
plt.text(49,100, '$E_0$', fontsize = 14, color = 'black')
plt.text(0,100, '$r_0$', fontsize = 12, color = 'black')
plt.text(53,-10, '$Y_0$', fontsize = 12, color = 'black')

# T√≠tulo, ejes y leyenda
ax.set(title="Equilibrio simultaneo en los mercados de bienes y de dinero", xlabel= r'Y', ylabel= r'r')
ax.legend()

plt.show()
```

1. Analice los efectos sobre las variables end√≥genas Y, r de una disminuci√≥n del gasto fiscal.(ŒîGo < 0) . El an√°lisis debe ser intuitivo, matem√°tico y gr√°fico.

Intuici√≥n 

    ŒîGo < 0
    
    Analisis a partir del mercado de bienes 
    
    G‚Üì ‚ü∂ DAC‚Üì ‚ü∂ DA < Y ‚ü∂ Y‚Üì
    
    Si el gasto fisacl baja esto origina que la demanda agregada disminuya, si esto sucede hay un desequilibrio entre DA y el producto porque DA est√° por debajo del producto. Esto origina que el nivel del producto disminuya y entonces la demanda agregada va a ser igual al nivel del producto. 
    
    Analisis a partir del mercado de dinero 
    
    Y‚Üì ‚ü∂ M^d‚Üì ‚ü∂ M^d < M^s ‚ü∂ r‚Üì
    
    Si el nivel del producto disminuye esto origina que la demanda de dinero disminuya entonces la demanda de dinero es menor que la oferta de dinero, y si queremos equilibrar ese mercado, la tasa de interes debe disminuir 
    
Matem√°ticamente
        
        ŒîY^e = (j/Kh + jùú∑1) * ŒîGo   < 0
                   (+)       (-) 
           
        Œîr^e = (K/Kh + jùú∑1) * ŒîGo   < 0
                   (+)        (-)

Intuicion
 
     ŒîMo^s < 0

     Mo^s‚Üì ‚ü∂ M^s < M^d ‚ü∂ r‚Üë

     r‚Üë ‚ü∂ I‚Üì ‚ü∂ DA‚Üì ‚ü∂ DA < Y ‚ü∂ Y‚Üì

           (-)    (+)       (+)   (-)
