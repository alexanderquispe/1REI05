**2.1 Explique de manera muy breve y resumida en qué consiste una regresión discontinua.**

De forma muy general, en el caso de un evento o suceso, busca explicar qué hubiese pasado si la "naturaleza hubiera funcionado de forma diferente".


De forma resumida, es un método cuasiexperimental utilizado para identificar efectos causales de algun tratamiento (R). Este método se basa en cortes que surgen por ley o por diseño y que implican una discontinuidad en la implementación del tratamiento (R), mismo que viene definido a lo largo de alguna variable, G(normalmente llamada la variable definitoria). Este método hará posible determinar la relación causal de cambios en el estatus de tratamiento sobre alguna otra variable (la variable dependiente).

**2.2 Detalle un ejemplo en la vida real en donde podría emplear una regresión discontinua.**

Para poder buscar un ejemplo de la vida real, se hizo una pequeña búsqueda de aplicación de regresión discontinua para poder explicarla en el ejemplo requerido. 

Un ejemplo en la realidad actual es en el caso del Sistema de Identificación de Potenciales Beneficiarios de Programas Sociales o por sus siglas SISBEN. Este fue un programa desarrollado para focalizar programas de política pública hacia la población mas vulnerable. En dicho modelo, se tomada valores de 0 a 100 (en forma de índice). Entonces, las personas con puntaje más bajo son los que deberían recibir el programa, caso contrario para los que tienen puntaje muy elevado.

Así, por ejemplo, este se divide en corte de forma que si el puntaje es menor a 11, pertenecería al primer corte que es "Sisben 1" y accede a determinado beneficio. Entonces, la idea del programa es identificar también a las familias que se encuentran dentro de un rango cercano. Es decir, el rango está asignado para un rango menor a 11, pero el programa busca que también incluya a una persona con 11.01, puesto que se encuentra cerca del rango. 

Esta es la notación de la Regresión Discontinua (RD):


La idea es que si se encuentran a la izquierda o derecha del valor crítico (Z), participan o no

Entonces, la idea radica en que los individuos de la izquierda del umbral Z deben ser muy parecidos a los de la derecha del umbral. Entonces, la derecha podría ser un contrafactual válido de la izquierda. 

Supuestos: 

- El umbral Z es abitriario
- Los regresores X, , y el error, u, son muy parecidos a la izquierda y derecha del umbral. 


Entonces:
- la comparación de estos dos grupos (tratamiento y control) es un estimador del impacto del tratamiento alrededor del punto de corte. 
- El parámetro estimado por la Regresión Discontinua es un EFECTO DEL TRATAMIENTO (o Local Average Treatment Effect)

- Se centra en los individuos alrededor del umbral


¿Cómo se aplica la Regresión Discontinua (RD) en el caso explicado?

Ejemplo: se seleccionará a beneficiarios de Canastas de comida

Población elegible: Hogares Sisben 1 (ya definidos previamente)
- Población cuyo puntaje es igual o inferior al umbral Z 
- Un hogar con un puntaje Sisben de Z o menos participa en el programa 

- La probabilidad de participación cambia cuando el puntaje Sisben Z pasa de estar por arriba de Z a estar por Debajo de Z


Entonces, la aplicación de RD, se basa en los determinantes de la variable del resultado Y, exceptuando la participación en el programa, son similares para individuos con un puntaje Sisben por debajo y por encima de Z. 

**2.3 Explique el diseño Sharp Desing y comente un ejemplo en la vida real en donde podría aplicarlo (sea lo más detallado posible).**

Este tipo de discontinuidad se refiere al caso en el cual la probabilidad de formar parte del grupo de tratamiento pasa de 0 a 1 después del corte que determina la discontinuidad. 

En este caso  $Ti$ es definida por la variable definitoria $(Gi)$ y el punto de discontinuidad es  $k$:

$$Ti=\mathcal{1}\{Gi≥k\}$$  

Todos los individuos con $Gi≥k$ estarán en el grupo de tratamiento y todos con $Gi<k$ estarán en el grupo de control.

        -Ejemplo 1:
$$Y = 15 + 22D + 2.5X + \epsilon$$

```python
## Random data
dat = pd.DataFrame({'x': np.random.normal(50, 25, 1000)})
dat.loc[dat.x<0, 'x'] = 0
dat = dat[dat.x<100]
dat['D'] = 0
# Cutoff
dat.loc[dat.x>50, 'D'] = 1
# Simulated Y
dat['y'] = 15 + 22*dat.D + 2.5 * dat.x + np.random.normal(0, 20, dat.shape[0])
# Not Treatment
dat_0 = dat[dat.D == 0]
# Treatment
dat_1 = dat[dat.D == 1]
```

```python
# Colors 
treat_c = "#568A9B" # Treatment
n_treat = "#A10728" # Not treatment
# Plot
plt.scatter("x", "y", data = dat_0, color = n_treat, alpha= .7, s = 7, label = "Not treatment")
plt.scatter("x", "y", data = dat_1, color = treat_c, alpha= .7, s = 7, label = "Treatment")
plt.axvline(50, color = "black", linestyle = "-.", label = r"$c_0$: Cutoff")
plt.legend();
```

Ahora estimamos un modelo para cada lado, en este caso el modelo que mejor se ajustaría sería el modelo lineal, para ello generamos una función para incluir las predicciones en la base de datos.

```python
def y_hat_lin(da_ta, x = "x", y = "y"):
    yhat = smf.ols(f"{y} ~ {x}", data = da_ta).fit().predict()
    da_ta["yhat"] = yhat
    da_ta.sort_values('yhat', inplace = True)
    return da_ta

data_y0 = y_hat_lin(dat_0)
data_y1 = y_hat_lin(dat_1)
```

```python
plt.scatter("x", "y", data = dat_0, color = n_treat, alpha= .2, s = 7, label = "Not treatment")
plt.scatter("x", "y", data = dat_1, color = treat_c, alpha= .2, s = 7, label = "Treatment")
plt.axvline(50, color = "black", linestyle = "-.", label = r"$c_0$: Cutoff")

plt.plot("x", "yhat", data = data_y0, color = n_treat, lw = 3)
plt.plot("x", "yhat", data = data_y1, color = treat_c, lw = 3);
```

```python
eff = min(data_y1.yhat) - max(data_y0.yhat)
eff = round(eff,  2)
print("Effect:", f"{eff}")
```

```python
plt.scatter("x", "y", data = dat_0, color = n_treat, alpha= .03, s = 7, label = "Not treatment")
plt.scatter("x", "y", data = dat_1, color = treat_c, alpha= .03, s = 7, label = "Treatment")
plt.axvline(50, color = "black", linestyle = "-.", label = r"$c_0$: Cutoff")

plt.plot("x", "yhat", data = data_y0, color = n_treat, lw = 3)
plt.plot("x", "yhat", data = data_y1, color = treat_c, lw = 3)
plt.axvline(50, color = "black", linestyle = "-.", label = r"$c_0$: Cutoff", lw = 1)
plt.annotate(f"{eff}", xy = (51, eff + 76))
plt.annotate("", xy = (51, max(data_y0.yhat)), xytext = (51, min(data_y1.yhat)), 
            arrowprops=dict(arrowstyle="<->",
                            connectionstyle="arc3"));
```
