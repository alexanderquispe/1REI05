### 1.1 Derivaci√≥n paso a paso de la curva IS

Teniendo en cuenta que: 
 $Y = C + I + G + X - M$
 
 $Y= C_o + b * Y^d + I_o - h * r + G_o + X_o - m * Y^d$
 
 $Y^d = Y - T = (1-t) Y$
 
 Entonces Y pasando
 
 $Y = C_o + b(1-t) * Y + I_o - h*r + G_o + X_o - m(1-t) * Y$
 
 Se deja sola R, siendo $IS$
 
 $ \frac{C_o+I_o+G_o+X_o}{h}-{Y(1-(1-t)*(b-m))}{h}=r$

### 1.2 Encontrar $Œîr/ŒîY$

$Œîùëü=-ŒîY \frac{1-(1-t)(b-m)}{h}$
Entonces ŒîY pasando al otro lado:
$\frac{Œîùëü}{ŒîY}=-\frac{1-(1-t)(b-m)}{h}$

Esto nos indica, que el signo de la pendiente se quedar√≠a en negativo.

### 1.3 Gr√°fico de Derivaci√≥n de la Curva IS a partir del equilibrio 

```python
Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8 
m = 0.2
t = 0.3
r = 16

Y = np.arange(Y_size)


def DA(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA

DA_I = DA(Co, Io, Go, Xo, h, r, b, m, t, Y)

    #Definici√≥n de de la recta 45¬∞
a = 2.5 

def L_45(a, Y):
    L_45 = a*Y
    return L_45

L_45 = L_45(a, Y)


    # Definici√≥n del nuevo parametro 
r= 0.1

def DA(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA= (Co+b*(1-t)*Y+Io-h*r+Go+Xo-m*(1-t)*Y)
    return DA

DA_1= DA(Co, Io, Go, Xo, h, r, b, m, t, Y)
```
