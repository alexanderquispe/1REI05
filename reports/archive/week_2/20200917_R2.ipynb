3. Lea la secci√≥n 4.4 del material de ense√±anza y explique c√≥mo se deriva la curva IS a partir del equilibrio  y grafique lo siguiente:

Llegamos al equilibrio ahorro-inversion y restamos la tributaci√≥n, as√≠ obtenemos la ecuaci√≥n del ingreso disponible. Recordemos que el ahorro total ya incluye al privado, del dobierno y externo. Es as√≠ que obtenemos la curva IS (igualdad entre ahorro y la inversi√≥n).

```python
Y_size = 100 

Co = 36
Io = 41
Go = 71
Xo = 3
h = 0.7
b = 0.8
m = 0.2
t = 0.3

Y = np.arange(Y_size)


# Ecuaci√≥n 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
```

```python
y_max = np.max(r)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, r, label = "IS", color = "#66CDAA") #Demanda agregada

# Eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

# T√≠tulo, ejes y leyenda
ax.set(title = "Curva IS", xlabel= 'Y', ylabel= 'r')
ax.legend()

plt.show()
```
