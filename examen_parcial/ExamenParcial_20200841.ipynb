**2. (2 puntos) Grafique y explique cúando sucede un exceso y déficit de demanda en el modelo Ingreso-Gasto. Señale las áreas donde ocurre cada caso. Explique cómo se converge al equilibrio a partir de estos dos escenarios.**

Ante un aumento de la Demanda Agregada ($DA$), ocurre un exceso de DA. En consecuencia, la producción ($y$) se vuelve insuficiente. Para volver a alcanzar un equilibrio, $y$ debe subir. Para ello, habrá más demanda de empleo y aumentará $y_d$, el aumento de los ingresos de las familias.

Ante una caída de la Demanda Agregada, ocurre un déficit de DA. En consecuencia, la producción anterior se vuelve demasiado alta. Los productores por lo tanto, reducen la producción, lo cual afecta al empleo y el ingreso disponible de las familias. Esto se realiza con e fin de volver al equilibrio.

El siguiente gráfico muestra cada área en donde ocurre el aumento o el déficit de demanda, respectivamente.  El punto $E$ representa el punto de equilibrio, el punto "A" representa el exceso de demanda y el punto $B$ el déficit de demanda. La convergencia ocurre al partir del desequilibrio que causa el aumento o la reducción de la producción, respectivamente.

```python
def DA(α_0, α_1, Y):
    DA = α_0 + (α_1*Y)
    return DA

DA = DA(α_0, α_1, Y)

Y_size = 100

a = 1    

Y = np.arange(Y_size)

def Y_45(a, Y):
    Y_45 = a*Y
    return Y_45

Y_45 = Y_45(a, Y)
```

**5. (3 puntos) Cuáles son las conclusiones principales del paper de Dancurt "Inflation Targeting in Peru: The Reasons for the Success"**

En el artículo titulado “Inflation Targeting in Peru: The Reasons for the Success”, Óscar Dancourt concluye que la política monetaria emprendida por el Banco Central de Reserva del Perú generó un buen desempeño macroeconómico en nuestro país durante la década de 2002-2013. En términos generales, se mantuvo una baja inflación, alto crecimiento del PBI y se impidieron recesiones severas. Esto se deriva del análisis realizado en función de lo que el autor considera los dos canales más importantes e influyentes en la actividad económica y el nivel de precios: el canal de crédito y el canal de tipo de cambio. 

Con respecto al canal del crédito, Dancourt concluye que medidas como el coeficiente de encaje, el desarrollo del mercado local de bonos públicos y el ajuste de la tasa de interés de referencia impactaron en la oferta y demanda de crédito en la banca comercial privada. La principal consecuencia fue la expansión del crédito y su conversión a moneda nacional. Dancourt también toma en cuenta factores externos como la crisis financiera internacional de 2008-2009, la cual limitó el funcionamiento externo y como consecuencia, redujo el crédito en moneda extranjera y fomentó el crédito en moneda nacional. 

Respecto al canal del tipo de cambio, Dancourt concluye que acumular suficientes reservas de divisas preservó la estabilidad del tipo de cambio. No obstante, Dancourt sostiene que se pudo haber mejorado la política monetaria del BCR. La caída sistemática del tipo de cambio en 2007 demuestra que una intervención en el mercado de divisas puede ser el mejor mecanismo para prevenirlo. Desdolarizar el crédito, según Dancourt, es la única medida correcta para controlar. Esto se logra mediante la intervención esterilizada en el mercado de divisas y la limitación del endeudamiento de los bancos comerciales en moneda extranjera.

```python
## GRAFICO
    # Curva IS ORIGINAL

# Parámetros

Y_size = 100 

Co = 50
Io = 40
Go = 90
Xo = 1
h = 0.5
b = 0.8
m = 0.2
t = 0.4

Y = np.arange(Y_size)


# Ecuación 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)


#--------------------------------------------------
    # NUEVA curva IS
    
# Definir SOLO el parámetro cambiado
Go = 60

# Generar la ecuación con el nuevo parámetro
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r_G = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
```

```python
    # Gráfico

# Dimensiones del gráfico
y_max = np.max(r)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, r, label = "IS", color = "orange") #IS orginal
ax.plot(Y, r_G, label = "IS_G", color = "orange", linestyle = 'dashed') #Nueva IS

# Texto agregado
plt.text(47, 162, '∆Go', fontsize=12, color='black')
plt.text(49, 159, '←', fontsize=15, color='grey')

# Título, ejes y leyenda
ax.set(title = "Disminución en el Gasto Fiscal $(G_0)$", xlabel= 'Y', ylabel= 'r')
ax.legend()

plt.show()
```
