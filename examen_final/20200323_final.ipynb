### Pregunta 1: Suponga que estamos interesados en estudiar la relación entre educación(X) y el nivel de ingresos(Y). Un investigador propone utilizar como variable instrumental(Z) el numero de horas que una persona utilizó para ver televisión cuando era niño.

#### Indetifique todos los elementos necesarios para implementar un enfoque de IV en este ejemplo.

Lo primero que debemos tener es una relación que querramos estudiar, en este caso, entre educación(X) y el nivel de ingresos(Y).

Ahora debemos encontrar una variable instrumental, que es una herramienta estadística que nos permite estimar relaciones causales, limpiando la relación entre otras dos variables. En este caso, la educación está correlacionada con el Error de la ecuación, ya que hay variables que la afectan, como la raza o el sexo. Por lo tanto, se necesita una variable instrumental Z que limpie el efecto de la "educación" sobre el "nivel de ingresos", para poder encontrar el verdadero efecto causal.

Ahora debemos comprobar que nuestra variable instrumental es una buena variable Z. En este caso, debemos pasar dos criterios. El primero es la relevancia, que refiere a la condición indispensable de que Z efectivamente se relacione con X. Mientras más fuerte sea esta correlación, es decir mientras más se acerque a 1, entonces podemos tener más confianza de que esta relación permitirá que la variable instrumental funcione correctamente. En este caso, lógicamente podemos decir que las horas que uno vio televisión de niño tiene al menos una relación con la educación, ya que más horas de TV pueden significar menos horas estudiando. Por lo tanto se cumple este criterio.

El segundo criterio es la validez, es decir, que la variable instrumental Z debe ser una variable que no tenga puertas traseras propias. Esto quiere decir que no debe haber ningún otro camino que lleve a Z a afectar a Y, excepto el camino de X, que es nuestra variable de tratamiento. A la validez también podemos llamar la condición de "exogeneidad", en el sentido de que Z puede ser retirada del modelo de Y luego de haber construído el camino de Z a X. Esto también se cumple, a pesar de que sea difícil de comprobar, ya que las horas de TV no se ven afectadas por otros sesgos que sí afectan a la educación, como la zona donde vives, la raza, el sexo, etc. Por lo tanto, también se cumple este supuesto.


#### Explique el procedimiento para encontrar el efecto causal en este escenario. Puede utilizar ecuaciones
Para encontrar el efecto causal en este escenario, lo que debemos hacer es instrumentalizar a la variable independiente X, que en este caso es "educación", con la variable instrumental Z, que es "horas que una persona utilizó para ver televisión cuando era niño", para tener una relación limpia con la variable que buscamos explicar, que es "nivel de ingresos" Y. Esto nos permite eliminar cualquier sesgo o variable no observable que afecte a la variable X. Este sesgo se encuentra normalmente en el Error E de la siguiente ecuación de regresión:

Y = B0 + B1*X + E

Vemos que B1 está sesgado, ya que el error va a afectar a la educación. Por lo tanto, la variable instrumental Z debe limpiar la relación entre X e Y, eliminando cualquier variable que se encuentre detrás, como la raza, la edad, el sexo, etc.

Por lo tanto, la variable instrumental debe de poder afectar a X "educación", pero no verse afectada por otras variables no observables. Esto permitirá cerrar todas las "puertas traseras" que afectan la relación casual de X con Y.

Para ello, debemos comprobar que la covarianza entre Z y el Error sea 0, con lo que podremos comprobar que Z no se ve afectada por ningún sesgo.

A partir de esto, si se logra encontrar un efecto de Z sobre X, entonces a partir de ello podremos correr una regresión para encontrar la relación entre X e Y de manera limpia, sin el sesgo. En este caso, si un niño se pasa más horas viendo televisión, puede que se pase menos horas estudiando. Además, las horas que uno ve televisión de niño no tienen una correlación con otras variables que generan sesgo, como el sexo o la raza. Por lo tanto, estamos frente a una buena variable instrumental que nos permitirá limpiar la relación entre X e Y.

Ahora solo debemos correr una regresión, y encontrar el Average Treatment Effect, teniendo como variable instrumental a las "horas que vio televisión de niño".

Esto se puede hacer con un método conocido como el 2SLS, donde se realizan 2 regresiones para estimar el modelo de variable instrumental. En la primera ecuación se debe utilizar el instrumento para predecir la variable de tratamiento, es decir, utilizar las "horas de televisión" para predecir la "educación", con una regresión OLS. Y luego, si se encuentra un valor de predicción, se utilizan esos valores para predecir a la variable "nivel de ingresos".

Yo considero que el efecto de la educación sobre el nivel de ingresos es positivo. Es decir, que por cada año de educación más, los ingresos de las personas deberían aumentar. ### Pregunta 2: Supongamos que estamos estudiando la eficacia de un programa de becas creado por el gobierno Peruano sobre las notas de los alumnos al final del primer semestre en la universidad. #### Indetifique todos los elementos necesarios para implementar un RDD en este ejemplo.

Una regresión discontinua es un método de identificación causal sin aleatorización, que consiste en comparar los resultados de dos grupos a ambos lados de un umbral, es decir, comparar a dos grupos que se dividen por una discontinuidad en el curso natural de las cosas (el umbral) que sirve como tratamiento. El tratamiento en este caso es recibir la beca PRONABEC.

Existen 4 elementos necesarios para hacer una RDD:

Running variable: la variable que determina si recibes o no el tratamiento. En En este caso esta variable es el promedio de la secundaria.

Cutoff: el límite de la running variable que determina si recibes el tratamiento o no. Debe ser objetivo, es decir, no puede ser manipulado por los participantes. En este caso, ese límite es tener un promedio mayor a 16, lo que hace que recibas la beca, es decir, el tratamiento.

Bandwidth: el área límite comparable alrededor el cutoff. Esto se determina por el investigador, y puede ser comparar a aquellos que sacaron un poco más de 16 y un poco menos de 16.

Tratamiendo: representa con 1 si recibes el tratamiento, o con 0 si no lo recibes.

Además, existen algunos supuestos que deben cumplirse:

Primero, ambos grupos, a ambos lados del cutoff, deben ser muy parecidos.

Además, el cutoff no debe verse intervenido por otra variable, sino que debe haber una continuidad de otras caracteísticas diferentes de la running variable. La edad, la raza, u otras variables no deberán cambiar al pasar por el cutoff.

Asimismo, el tratamiento no debe ser asignado por manipulación de otras variables, sino que la única condición para conseguirlo debe ser pasar por el cutoff.

#### Explique el procedimiento para encontrar el efecto causal en este escenario. Puede utilizar ecuaciones.

Lo que queremos encontrar aquí es el efecto de tratamiento local, es decir, al diferencia entre los grupos a cada lado del cutoff.

Para medir la eficacia de este programa de becas, lo que debemos hacer entonces es comparar a aquel grupo con un promedio un poco menor a 16 que no recibió la beca, junto con el grupo con un promedio mayor a 16, cerca del cutoff, que sí recibió la beca, para luego comparar sus notas del primer semestre de la universidad.

En este caso, estamos frente a un "sharp design" ya que la probabilidad de obtener la beca es de 1 si el promedio es mayor a 16, y de 0 si es que no se llega a este promedio.

Por lo tanto, ya tenemos el cutoff establecido, que es un promedio mayor a 16.

Ya tenemos el modelo, el sharp design.

Entonces lo que debemos hacer ahora es encontrar el "salto" en el cutoff, que nos indicará el efecto de obtener la beca PRONABEC en las notas de la universidad.

Para ver este salto, debemos construir una base de datos con nuestras observaciones, y graficar a los dos grupos, a la izquierda quienes no recibieron la beca, y a la derecha los que sí, y comparar la diferencia en sus notas de universidad. 

También podemos usar un método conocido como Kernel Weighting, para evaluar el efecto de la beca, y dividirla en el bandwidth que hemos escogido.

$$
K(R, c, h) = \mathcal{1}\{|R-c| \leq h\} * \left(1-\frac{|R-c|}{h}\right)
$$

### Pregunta 3: Supongamos que estamos interesados en estudiar el impacto de la migración en la tasa de empleo en dos regiones del Perú en 2017-2018. Tenemos datos anuales sobre la tasa de empleo en Lima, que experimentó una afluencia significativa de inmigrantes en 2018, y Puno que no experimentó el ingreso de migrantes (2017 ninguna de las regiones experimentó afluencia de migrantes). Podemos usar un análisis DID para comparar el cambio en las tasas de empleo entre las dos regiones a lo largo del tiempo para ver si la migración tuvo algún efecto.

#### Indetifique todos los elementos necesarios para implementar un DiD en este ejemplo.

Para implementar un DiD, se necesitan dos grupos y dos etapas de tiempo. Primero debemos tener a un grupo de control, donde no hubo un cambio. En este caso, tenemos a Puno, donde no hubo migración. Y luego debemos tener un grupo de tratamiento, donde encontremos un estímulo. En este caso es Lima, que sí recibió migrantes. Además tenemos dos etapas de tiempo, 2017 y 2018, que podrán ser usadas para la comparación, ya que el estímulo de Lima se introdujo en 2018, por lo que su parecido con Puno en 2017 nos permite compararlas.

Lo que nos interesa es encontrar el efecto de un tratamiento, en este caso la migración, en un resultado, en este caso la tasa de empleo.

Además, se debe cumplir un supuesto fundamental: el supuesto de "tendencias paralelas", que implica que si no había tratamiento, el comportamiento de ambos grupos debería de mantenerse igual. Esto se puede verificar comparando la evolución de la tendencia previo al tratamiento en ambos grupos, donde esta tendencia debe ser paralela. En el caso de este ejemplo, este supuesto se cumple, ya que se nos indica que que en 2017, el año previo al tratamiento de migración, ni Puno ni Lima experimentaron afluencia de migrantes. Por lo tanto, su tendencia antes del tratamiento era prácticamente igual, por lo que pueden ser comparadas.

#### Explique el procedimiento para encontrar el efecto causal en este escenario. Puede utilizar ecuaciones.

Lo primero que debemos obtener son 4 promedios: el promedio de la tasa de empleo del grupo de control, Puno, antes y despúes del estímulo, es decir en 2017 y 2018, y lo mismo en el grupo de tratamiento Lima: promedio de tasa de empleo en 2017 y 2018.

Una vez que tengamos esto, debemos encontrar el cambio en el tiempo en ambos grupos:

- Control 2018 - Control 2017: Dif control
- Tratamiento 2018 - Tratamiento 2017: Dif Tratamiento

Una vez que tenemos estos dos resultados, que representan los cambios en el tiempo, podemos utilizar el método de la diferencia en diferencia.

Para ello, restamos la diferencia del grupo de tratamiento menos la diferencia del grupo de control.

- Dif Tratamiento - Dif control.

Este resultado representa el efecto causal de la migración sobre la tasa de empleo, ya que se está comparando la tasa de empleo en un lugar donde hubo tratamiendo con un lugar donde no lo hubo.

Esta sería la ecuación general: 

$$
\widehat{\delta}^{2\times 2}_{kU} = \bigg ( \overline{y}_k^{-post(k)} - \overline{y}_k^{-pre(k)} \bigg ) - \bigg ( \overline{y}_U^{-post(k)} - \overline{y}_U^{-pre(k)} \bigg )
$$


#### ¿El efecto de la migración sobre la tasa de empleo sería positivo o negativo (solo necesitan usar su intuición)? Argumente si los resultados serian causales o solamente de correlación.

En este caso, considero que el efecto de la migración sobre la tasa de empleo será positivo. Al haber más población dentro de una ciudad, sobre todo migrantes, que no están establecidos, la demanda de empleos aumentará considerablemente, ya que estas personas necesitan dinero urgentemente para mantenerse. Por lo tanto, considero que es un efecto causal, donde al haber mayor afluencia de migración en una ciudad, la tasa de empleo aumentará. Además, podemos estar seguros de que la relación no tiene sesgos, ya que al comparar el cambio en el tiempo, y restar el tratamiento menos el control, eliminamos los contrafactuales. Por lo tanto, la introducción del estímulo de migración, originará un incremento en la tasa de empleo, demostrando un efecto causal.
