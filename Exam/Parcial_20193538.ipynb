### 1. Defina el Multiplicador Keynesiano y d un ejemplo de cmo influye este en el modelo Ingreso-Gasto Keynesiano.
El multiplicador Keynesiano es un cambio en la magnitud de los componentes autnomos de la demanda agregada que conforman el intercepto, genera un proceso multiplicador del ingreso hasta converger al nuevo ingreso y producto de equilibrio.

As, el aumento multiplicado del ingreso es resultado de los efectos directos e indirectos ocasionados por el aumento de cualquiera de los componentes autnomos de la demanda agregada

Es representado con la formula de $\frac{1}{1-_1}=\frac{1}{1-(b-m)(1-t)}$

### 1. Encuentre el Ingreso de Equilibrio e identifique el multiplicador keynesiano.
Tengamos en cuenta que:
- Demanda agregada:$DA=C+I+G+(X-M)$

Donde:
-	Consumo: $C=C_0+bY^d$
-	Inversin: $I=I_0-hr$
-	Gobierno: $G=G_0$
-	Exportaciones: $X=X_0$
-	Importaciones: $M=mY^d$

**Entonces el ingreso, al estar en equilibrio, tiene que ser igual a la demanda agregada (DA=Y)
Primero analizemos la demanda agregada, la cual se desdoblaria de la siguiente manera:**

$$DA = C_0 + b(1  t)Y + I_0  hr +G_0 + x_1Y^ + x_2 m_1(1  t)Y + m2_$$


$$DA = C_0 + I_0 +G_0 + [b(1  t) m_1 (1 t)]Y  hr + x_1Y^ + (x_2 + m_2)$$


$$DA = w hr + x_1Y^ + (x_2 + m_2 )_0  (x_2 + m_2 )r + (x_2 + m_2 )r^ + [(b m_1)(1 t)]Y$$

**Donde:**

$$w = C_0 +I_0 + G_0$$


$$DA = [w + x_1Y^ + (x_2 + m_2 )0 + (x_2 + m_2 )r^  (h + (x_2 + m_2 ))r] + [(b m_1)(1  t)]Y$$

**Entonces si la Demanda Agregada es igual al Ingreso de Eequilibrio, las ecuaciones se igualan:**

$$Y = [w + x_1Y^ + (x_2 + m_2 )0 + (x_2 + m_2 )r^  (h + (x_2 + m_2 ))r] + [(b m_1)(1  t)]Y$$


$$Y = \frac{1} {1  (b m1)(1 t)} [w + (x_2 + m_2 )_0 + (x_2 + m_2 )r  + x_1Y^]  [ \frac{h + (x_2 + m_2 )} {1 (b m1)(1 t)} ]r$$


**En esta ecuacion, se puede identificar el Multiplicador Keynesiano, el cual es**

$$\frac{1}{1-(b-m)(1-t)}$$
