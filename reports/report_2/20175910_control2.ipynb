## a. El Modelo Ingreso-Gasto: la Curva IS

1. Utilizando las ecuaciones derive paso a paso la curva IS matemáticamente (a partir de la condición de equilibrio Y = DA).

𝑌 = 𝐶 + 𝐼 + 𝐺 + 𝑋 − M

𝑌 − 𝑇 = 𝐶 + 𝐼 − 𝑇 + 𝐺 + 𝑋 − M

𝑌𝑑 = 𝐶 + 𝐼 − 𝑇 + 𝐺 + 𝑋 − M

Esta igualdad se puede reescribir como sigue:

(𝑌𝑑 − 𝐶) + (𝑇 − 𝐺) + (𝑀 − 𝑋) = I

El lado derecho constituye el ahorro total que es igual a la inversión. El ahorro total
incluye el Ahorro Privado, el Ahorro del Gobierno y el Ahorro Externo.

𝑆𝑝 + 𝑆𝑔 + 𝑆𝑒 = 𝐼0 − ℎ𝑟 


𝑆 = 𝐼0 − ℎ𝑟


Esta igualdad entre el ahorro y la inversión representa precisamente la curva IS

𝑆(𝑌) = 𝐼(𝑟)

Haciendo reemplazos se obtiene que:


(𝑌𝑑 − 𝐶0 − 𝑏𝑌𝑑) + (𝑇 − 𝐺0) + (𝑚𝑌𝑑 − 𝑋0) = 𝐼0 − ℎr

A partir de aquí, mediante algunas operaciones algebraicas, se obtiene la ecuación de la
Curva IS.

[1 − (𝑏 − 𝑚)(1 − 𝑡)]𝑌 − (𝐶0 + 𝐺0 + 𝑋0) = 𝐼0 − ℎr

Haciendo algunas operaciones, la Curva IS puede expresarse con una ecuación donde la
tasa de interés se encuentra en función del ingreso:

𝑟 =1/ℎ*(𝐶0 + 𝐼0 + 𝐺0 + 𝑋0)−(1 − (𝑏 − 𝑚)(1 − 𝑡))*y/ℎ

```python
#--------------------------------------------------
    # ORIGINAL IS Curve

# Parameters

Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3

Y = np.arange(Y_size)


# Equation 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)


#--------------------------------------------------
    # New IS Curve

# Define ONLY the changed parameter
Go = 60

# Equation with the changed parameter
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r_G = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
    # Graph

# Dimensions
y_max = np.max(r)
fig, ax = plt.subplots(figsize=(10, 8))

# Curves to plot
ax.plot(Y, r, label = "IS", color = "black") #IS orginal
ax.plot(Y, r_G, label = "IS_G", color = "C1", linestyle = 'dashed') #New IS

# Text added
plt.text(47, 162, '∆Go', fontsize=12, color='black')
plt.text(49, 159, '←', fontsize=15, color='grey')

# Title and legend
ax.set(title = "Caída en el gasto de gobierno", xlabel= 'Y', ylabel= 'r')
ax.legend()

plt.show()
```

De forma intuitiva sabemos que: 

DA= C+I+G+X-M

Entonces: 

G↓→β_0↓

G↓→DA↓→DA<Y→Y↓

Por tanto, la caída del gasto del gobierno genera una caída en la demanda agregada. Ello genera un desplazamiento de la curva IS a la izquierda.
