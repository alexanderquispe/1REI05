### Pregunta 1

Primero tenemos la ecuación de la curva IS:

$$ r = \frac{B_o}{h} - \frac{B_1}{h}Y $$

Donde $ B_0 = C_o + I_o + G_o + X_o $ y $ B_1 = 1 - (b - m)(1 - t)$

Y tenemos la ecuación de la curva LM:;

$$  r = -\frac{1}{j}\frac{Mo^s}{P} + \frac{k}{j}Y $$


Eliminando r y despejando P, se obtiene:

$$  P = -\frac{h Mo^s}{-j B_o + (jB_1 + hk)Y} $$

O, en función del nivel de ingresos (Y):

$$  Y = \frac{jB_o}{jB_1 + hk} + (\frac{hMo^s}{jB_1 + hk})\frac{1}{P} $$


Ahora tenemos que utilizar la nueva ecuación de equilibrio en el mercado monetario:

$$ \frac{Mo^s}{P} ----> Mo^s - P $$

$$ Mo^s - P = kY - jr $$

Se reemplaza r, y se obtiene la ecuación de la demanda agregada DA:

$$  P = \frac{h Mo^s + jB_o}{h} - \frac{jB_1 + hk}{h}Y $$


El corto plazo es un periodo en el cual el producto Y se ubica por debajo o por encima de su nivel de largo plazo o Producto Potencial $(\bar{Y})$.

Entonces, curva de OA de corto plazo se puede representar con la siguiente ecuación:

$$ P = P^e + θ(Y - \bar{Y}) $$ 

Donde P is the nivel de precios, $P^e$ el precio esperado y $\bar{Y}$ el producto potencial.

Para hallar $Y^{eq}$ igualamos las ecuaciones de OA y DA:

$$ \frac{h Mo^s + jB_o}{h} - \frac{jB_1 + hk}{h}Y = P^e + θ(Y - \bar{Y}) $$

$$ Y^{eq} = [ \frac{1}{(θ + \frac{jB_1 + hk}{h})} ]*[(\frac{h Mo^s + jB_o}{h} - P^e + θ\bar{Y})]$$

Para encontrar $P^{eq}$ solamente reemplazamos $Y^{eq}$ en la ecuación de oferta agregada

$$ P^{eq} = P^e + θ( Y^{eq} - \bar{Y} ) $$ 

$$ P^{eq} = P^e + θ( [ \frac{1}{(θ + \frac{jB_1 + hk}{h})} ]*[(\frac{h Mo^s + jB_o}{h} - P^e + θ\bar{Y})] - \bar{Y} ) $$ 

Para encontrar $r^{eq}$ solamente reemplazamos $P^{eq}$ en la ecuación de tasa de interés de equilibrio del modelo IS-LM. 

- Tasa de interés de equilibrio:

$$ r^{eq\_is\_lm} = \frac{kB_o}{kh + jB_1} - (\frac{B_1}{kh + jB_1})*(Ms_o - P)$$

- Tasa de interés de equilibrio en DA-OA
$$ r^{eq\_is\_lm} = \frac{kB_o}{kh + jB_1} - (\frac{B_1}{kh + jB_1})*(Ms_o - P^{eq})$$

$$ r^{eq} = \frac{kB_o}{kh + jB_1} - (\frac{B_1}{kh + jB_1})*(Ms_o - P^e + θ( [ \frac{1}{(θ + \frac{jB_1 + hk}{h})} ]*[(\frac{h Mo^s + jB_o}{h} - P^e + θ\bar{Y})] - \bar{Y} ) ) $$


Y así logramos obtener las ecuaciones de equilibrio de nuestras 3 variables endógenas.

1. $$ Y^{eq\_da\_oa} = [ \frac{1}{(θ + \frac{jB_1 + hk}{h})} ]*[(\frac{h Mo^s + jB_o}{h} - P^e + θ\bar{Y})]$$

2. $$ r^{eq\_da\_oa} = \frac{kB_o}{kh + jB_1} - (\frac{B_1}{kh + jB_1})*(Ms_o - P^e + θ( [ \frac{1}{(θ + \frac{jB_1 + hk}{h})} ]*\\
[(\frac{h Mo^s + jB_o}{h} - P^e + θ\bar{Y})] - \bar{Y} ) ) $$

3. $$ P^{eq\_da\_oa} = P^e + θ( [ \frac{1}{(θ + \frac{jB_1 + hk}{h})} ]*[(\frac{h Mo^s + jB_o}{h} - P^e + θ\bar{Y})] - \bar{Y} ) $$ 


Ahora debemos graficar equilibrio del modelo DA-OA

```python
#1--------------------------
    # Demanda Agregada
    
# Parámetros

Y_size = 100 

Co = 35
Io = 40
Go = 50
Xo = 2
h = 0.8
b = 0.4
m = 0.5
t = 0.8

k = 2
j = 1                
Ms = 200             
P  = 8  

Y = np.arange(Y_size)


# Ecuación

B0 = Co + Io + Go + Xo
B1 = 1 - (b-m)*(1-t)

def P_AD(h, Ms, j, B0, B1, k, Y):
    P_AD = ((h*Ms + j*B0)/h) - (Y*(j*B1 + h*k)/h)
    return P_AD

P_AD = P_AD(h, Ms, j, B0, B1, k, Y)


#2--------------------------
    # Oferta Agregada
    
# Parámetros

Y_size = 100

Pe = 100 
θ = 3
_Y = 20   

Y = np.arange(Y_size)


# Ecuación

def P_AS(Pe, _Y, Y, θ):
    P_AS = Pe + θ*(Y-_Y)
    return P_AS

P_AS = P_AS(Pe, _Y, Y, θ)
```

```python
# líneas punteadas autómaticas

# definir la función line_intersection
def line_intersection(line1, line2):
    xdiff = (line1[0][0] - line1[1][0], line2[0][0] - line2[1][0])
    ydiff = (line1[0][1] - line1[1][1], line2[0][1] - line2[1][1])

    def det(a, b):
        return a[0] * b[1] - a[1] * b[0]

    div = det(xdiff, ydiff)
    if div == 0:
       raise Exception('lines do not intersect')

    d = (det(*line1), det(*line2))
    x = det(d, xdiff) / div
    y = det(d, ydiff) / div
    return x, y

    # coordenadas de las curvas (x,y)
A = [P_AD[0], Y[0]] # DA, coordenada inicio
B = [P_AD[-1], Y[-1]] # DA, coordenada fin

C = [P_AS[0], Y[0]] # L_45, coordenada inicio
D = [P_AS[-1], Y[-1]] # L_45, coordenada fin

    # creación de intersección

intersec = line_intersection((A, B), (C, D))
intersec # (y,x)
```

```python
    # Gráfico del modelo DA-OA

# Dimensiones del gráfico
y_max = np.max(P)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar 
ax.plot(Y, P_AD, label = "DA", color = "C4") #DA
ax.plot(Y, P_AS, label = "OA", color = "C8") #OA

# Líneas punteadas
plt.axhline(y=intersec[0], xmin= 0, xmax= 0.5, linestyle = ":", color = "grey")
plt.axvline(x=intersec[1],  ymin= 0, ymax= 0.49, linestyle = ":", color = "grey")

# Texto agregado
plt.text(0, 200, '$P_0$', fontsize = 12, color = 'black')
plt.text(53, 25, '$Y_0$', fontsize = 12, color = 'black')
plt.text(50, 202, '$E_0$', fontsize = 12, color = 'black')


# Eliminar valores de ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

# Título, ejes y leyenda
ax.set(title="Modelo DA-OA", xlabel= r'Y', ylabel= r'P')
ax.legend()

plt.show()
```

Intuición:

Modelo IS-LM: 
$$ Go↓ → DA↓ → Y > DA → Y↓ $$
$$ Y↓ → Md↓ → Ms > Md → r↓ $$

Modelo DA-OA: 
$$ Y↓ → θ(Y-\bar{Y})↓ → P↓$$

Si baja el gasto fiscal, entonces se recuce la demanda agregada. Esto lleva a que haya un exceso de oferta, y el Producto Y deba bajar. Si Y baja, entonces la demanda de dinero disminuye, generando un exceso de oferta. Para regresar al equilibrio deberá reducirse también la tasa de interés.

Además, al bajar el producto, los precios también van a disminuir.

```python
#Matemática:

# nombrar variables como símbolos de IS
Co, Io, Go, Xo, h, r, b, m, t, beta_0, beta_1  = symbols('Co Io Go Xo h r b m t beta_0, beta_1')

# nombrar variables como símbolos de LM 
k, j, Ms, P, Y = symbols('k j Ms P Y')

# nombrar variables como símbolos para curva de oferta agregada
Pe, _Y, Y, θ = symbols('Pe, _Y, Y, θ')

# Beta_0 y beta_1
beta_0 = (Co + Io + Go + Xo)
beta_1 = ( 1-(b-m)*(1-t) )

# Producto de equilibrio en el modelo DA-OA
Y_eq = ( (1)/(θ + ( (j*beta_1+h*k)/h) ) )*( ( (h*Ms+j*beta_0)/h ) - Pe + θ*_Y )

# Precio de equilibrio en el modelo DA-OA 
P_eq = Pe + θ*(Y_eq - _Y)

# Tasa de interés de equilibrio en el modelo DA-OA
r_eq = (j*beta_0)/(k*h + j*beta_1) + ( h / (k*h + j*beta_1) )*(Ms - P_eq)
#((h*Ms+j*beta_0)/h) - ((j*beta_1+h*r)/h)*((P-Pe-θ*_Y)/θ)
```

```python
# Efecto del cambio del Gasto Fiscal sobre el Producto
df_Y_eq_Go = diff(Y_eq, Go)
print("El Diferencial del Producto con respecto al diferencial de la masa monetaria = ", df_Y_eq_Go)
print("\n")

# Efecto del cambio del Gasto Fiscal sobre Tasa de Interés
df_r_eq_Go = diff(r_eq, Go)
print("El Diferencial de la tasa de interés con respecto al diferencial de la masa monetaria = ", df_r_eq_Go)
print("\n")

# Efecto del cambio del Gasto Fiscal sobre los Precios
df_P_eq_Go = diff(P_eq, Go)
print("El Diferencial del nivel de precios con respecto al diferencial de la masa monetaria = ", df_P_eq_Go)
```

Los diferenciales de Y, P y r son positivos. Por lo tanto, cuando hay una disminución del gasto fiscal, debemos multiplicar este cambio negativo por los diferenciales positivos.

$$ ∆Y^eq = (+)(-) $$
$$ ∆r^eq = (+)(-) $$

#### Analice los efectos sobre las variables endógenas Y, P y r de una disminución de la masa monetaria. El análisis debe ser intuitivo, matemático y gráfico. En una figura, se debe que usar los ejes r e Y (modelo IS-LM), y en la otra, los ejes P, Y (modelo DA-OA).

Intuición:

Modelo IS-LM: 
$$ Ms↓ → Ms↓ → Md > Ms → r↑ $$
$$ r↑ → I↓ → DA↓ → Y > DA → Y↓ $$

Modelo DA-OA: 
$$ Y↓ → θ(Y-\bar{Y})↓ → P↓$$

Si disminuye la masa monetaria, la oferta de dinero baja, lo que genera un exceso de demanda de dinero. Para regresar al equilibrio se debe entonces aumentar la tasa de interés. Esto genera entonces una disminución de las inversiones, lo que disminuye la demanda agregada. Para regresar al equilibrio se debe reducir el Producto (Y).

Además, al bajar el producto, los precios también van a disminuir.

```python
# nombrar variables como símbolos de IS
Co, Io, Go, Xo, h, r, b, m, t, beta_0, beta_1  = symbols('Co Io Go Xo h r b m t beta_0, beta_1')

# nombrar variables como símbolos de LM 
k, j, Ms, P, Y = symbols('k j Ms P Y')

# nombrar variables como símbolos para curva de oferta agregada
Pe, _Y, Y, θ = symbols('Pe, _Y, Y, θ')

# Beta_0 y beta_1
beta_0 = (Co + Io + Go + Xo)
beta_1 = ( 1-(b-m)*(1-t) )

# Producto de equilibrio en el modelo DA-OA
Y_eq = ( (1)/(θ + ( (j*beta_1+h*k)/h) ) )*( ( (h*Ms+j*beta_0)/h ) - Pe + θ*_Y )

# Precio de equilibrio en el modelo DA-OA 
P_eq = Pe + θ*(Y_eq - _Y)

# Tasa de interés de equilibrio en el modelo DA-OA
r_eq = (j*beta_0)/(k*h + j*beta_1) + ( h / (k*h + j*beta_1) )*(Ms - P_eq)
#((h*Ms+j*beta_0)/h) - ((j*beta_1+h*r)/h)*((P-Pe-θ*_Y)/θ)
```

```python
# Efecto del cambio de Precio esperado sobre el Producto 
df_Y_eq_Ms = diff(Y_eq, Ms)
print("El Diferencial del Producto con respecto al diferencial de la masa monetaria = ", df_Y_eq_Ms)
print("\n")

# Efecto del cambio de Precio esperado sobre Tasa de Interés
df_r_eq_Ms = diff(r_eq, Ms)
print("El Diferencial de la tasa de interés con respecto al diferencial de la masa monetaria = ", df_r_eq_Ms)
print("\n")

# Efecto del cambio de Precio esperado sobre los Precios
df_P_eq_Ms = diff(P_eq, Ms)
print("El Diferencial del nivel de precios con respecto al diferencial de la masa monetaria = ", df_P_eq_Ms)
```

Los diferenciales de Y, P son positivos. Por lo tanto, cuando hay una disminución de la masa monetaria, debemos multiplicar este cambio negativo por los diferenciales positivos. En cambio, el diferencial de r es negativo, por lo que su multiplicación dará un número positivo.

$$ ∆Y^eq = (+)(-) $$
$$ ∆r^eq = (-)(-) $$
$$ ∆P^eq = (+)(-) $$

Por lo tanto, Y y P van a disminuir, mientras que r va a aumentar.

#### Analice los efectos sobre las variables endógenas Y, P y r de un incremento de la tasa de impuestos. El análisis debe ser intuitivo, matemático y gráfico. En una figura, se debe que usar los ejes r e Y (modelo IS-LM), y en la otra, los ejes P, Y (modelo DA-OA).

Intuición:

Modelo IS-LM: 
$$ t↑ → DA↓ → Y > DA → Y↓ $$
$$ Y↓ → Md↓ → Ms>Md → r↓ $$ 

Modelo DA-OA: 
$$ Y↓ → θ(Y-\bar{Y})↓ → P↓$$

Si sube la tasa de impuestos, entonces la demanda agregada baja. Esto hace que deba bajar Y para regresar al equilibrio. Si baja el producto, entonces la demanda de dinero baja, y con ello surge un exceso de oferta de dinero. Para regresar al equilibrio se deberá reducir la tasa de interés.

Además, al bajar el producto, los precios también van a disminuir.

```python
# nombrar variables como símbolos de IS
Co, Io, Go, Xo, h, r, b, m, t, beta_0, beta_1  = symbols('Co Io Go Xo h r b m t beta_0, beta_1')

# nombrar variables como símbolos de LM 
k, j, Ms, P, Y = symbols('k j Ms P Y')

# nombrar variables como símbolos para curva de oferta agregada
Pe, _Y, Y, θ = symbols('Pe, _Y, Y, θ')

# Beta_0 y beta_1
beta_0 = (Co + Io + Go + Xo)
beta_1 = ( 1-(b-m)*(1-t) )

# Producto de equilibrio en el modelo DA-OA
Y_eq = ( (1)/(θ + ( (j*beta_1+h*k)/h) ) )*( ( (h*Ms+j*beta_0)/h ) - Pe + θ*_Y )

# Precio de equilibrio en el modelo DA-OA 
P_eq = Pe + θ*(Y_eq - _Y)

# Tasa de interés de equilibrio en el modelo DA-OA
r_eq = (j*beta_0)/(k*h + j*beta_1) + ( h / (k*h + j*beta_1) )*(Ms - P_eq)
#((h*Ms+j*beta_0)/h) - ((j*beta_1+h*r)/h)*((P-Pe-θ*_Y)/θ)
```

```python
# Efecto del cambio de Precio esperado sobre el Producto 
df_Y_eq_t = diff(Y_eq, t)
print("El Diferencial del Producto con respecto al diferencial de la masa monetaria = ", df_Y_eq_t)
print("\n")

# Efecto del cambio de Precio esperado sobre Tasa de Interés
df_r_eq_t = diff(r_eq, t)
print("El Diferencial de la tasa de interés con respecto al diferencial de la masa monetaria = ", df_r_eq_t)
print("\n")

# Efecto del cambio de Precio esperado sobre los Precios
df_P_eq_t = diff(P_eq, t)
print("El Diferencial del nivel de precios con respecto al diferencial de la masa monetaria = ", df_P_eq_t)
```

Los diferenciales de Y, P y r son negativos. Por lo tanto, cuando hay una aumento de la tasa de impuestos, debemos multiplicar este cambio positivo por los diferenciales negativos.

$$ ∆Y^eq = (-)(+) $$
$$ ∆r^eq = (-)(+) $$
$$ ∆P^eq = (-)(+) $$

Por lo tanto, Y, P y r van a disminuir.
