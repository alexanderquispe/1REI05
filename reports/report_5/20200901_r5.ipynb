En su artículo ‘Metas de Inflación en el Perú: Las razones del éxito’, Oscar Dancourt busca describir los dos principales canales mediante los cuales la política monetaria influye la actividad económica: el crédito y el tipo de cambio. Para responder a esta pregunta, el autor explica los dos canales utilizados: el canal de crédito y el canal de tipo de cambio, así como sus componentes más importantes. Primero, para el canal de crédito, se examina principalmente la tasa de referencia para el crédito local impuesta por el banco central, así como el ratio entre moneda local y extranjera, y el proceso de desdolarización ligado a esto último. ### A partir del siguiente sistema de ecuaciones que representan el modelo DA-OA

#### Encuentre las ecuaciones de Ingreso ($Y^e$), tasa de interés ($r^e$),  ($P^e$) de equilibrio (Escriba paso a paso la derivación de estas ecuaciones).

$DA:$
$$  P = \frac{h M_0^s + jB_0}{h} - \frac{jB_1 + hk}{h}Y $$

$OA:$
$$ P = P^{esperado} + θ(Y - \bar{Y}) $$ 

$Y^e:$
$$P^{esperado} + θ(Y - \bar{Y}) = \frac{h M_0^s + jB_0}{h} - \frac{jB_1 + hk}{h}Y $$
$$P^{esperado} + θY - θ\bar{Y} = \frac{h M_0^s + jB_0}{h} - \frac{jB_1 + hk}{h}Y $$
$$P^{esperado} - θ\bar{Y} - \frac{h M_0^s + jB_0}{h} =(- \frac{jB_1 + hk}{h} - θ) Y $$
$$ Y^e = [ \frac{1}{(θ + \frac{jB_1 + hk}{h})} ]*[(\frac{h M_0^s + jB_0}{h} - P^{esperado} + θ\bar{Y})]$$

$P^e:$
$$ P^e = P^{esperado} + θ(Y^e - \bar{Y}) $$
$$ P^e = P^{esperado} + θ([ \frac{1}{(θ + \frac{jB_1 + hk}{h})} ]*[(\frac{h M_0^s + jB_0}{h} - P^{esperado} + θ\bar{Y})])- θ\bar{Y} $$

$r^e:$
$$ r^e = \frac{kB_0}{kh + jB_1} - (\frac{B_1}{kh + jB_1})*(M_0^s - P^e)$$

$$ r^e = \frac{kB_0}{kh + jB_1} - (\frac{B_1}{kh + jB_1})*(M_0^s - P^e + θ( [ \frac{1}{(θ + \frac{jB_1 + hk}{h})} ]*[(\frac{h M_0^s + jB_0}{h} - P^{esperado} + θ\bar{Y})] - \bar{Y} ) ) $$

#### Grafique el equilibrio simultáneo en el modelo DA-OA

```python
# DA

TamañoY = 100 

Co = 40
Io = 50
Go = 70
Xo = 3
h = 0.8
b = 0.8 # más que m
m = 0.4 # menos que b
t = 0.5

k = 4
j = 2                
Ms = 300             
P  = 7  

Y = np.arange(TamañoY)

# Ecuación DA

β_0 = (Co + Io + Go + Xo)
β_1 = (1 - (b-m)*(1-t))

def P_AD(h, Ms, j, β_0, β_1, k, Y):
    P_AD = ((h*Ms + j*β_0)/h) - (Y*(j*β_1 + h*k)/h)
    return P_AD

P_AD = P_AD(h, Ms, j, β_0, β_1, k, Y)


# OA

TamañoY = 100

Pe = 160 # Precio Esperado
θ = 5
_Y = 60 # Producto Potencial

Y = np.arange(TamañoY)


# Ecuación OA

def P_AS(Pe, _Y, Y, θ):
    P_AS = Pe + θ*(Y-_Y)
    return P_AS

P_AS = P_AS(Pe, _Y, Y, θ)
```

```python
# Primer intento de hacer líneas punteadas autómaticas...

def line_intersection(line1, line2):
    xdiff = (line1[0][0] - line1[1][0], line2[0][0] - line2[1][0])
    ydiff = (line1[0][1] - line1[1][1], line2[0][1] - line2[1][1])

    def det(a, b):
        return a[0] * b[1] - a[1] * b[0]

    div = det(xdiff, ydiff)
    if div == 0:
       raise Exception('lines do not intersect')

    d = (det(*line1), det(*line2))
    x = det(d, xdiff) / div
    y = det(d, ydiff) / div
    return x, y

A = [P_AD[0], Y[0]] # DA, coordenada inicio
B = [P_AD[-1], Y[-1]] # DA, coordenada fin

C = [P_AS[0], Y[0]] # L_45, coordenada inicio
D = [P_AS[-1], Y[-1]] # L_45, coordenada fin


intersec = line_intersection((A, B), (C, D))
intersec # (y,x)
```

$IS-LM:$
$$ t↑ → y^d↓ → C↓ → DA↓ → DA < Y → Y↓ $$
$$ Y↓ → M^d↓ → M^d < M^s → r↓ $$
$DA-OA:$
$$ Y↓ → P↓ $$
