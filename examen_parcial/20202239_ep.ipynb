#### 1. Encuentre el Ingreso de Equilibrio $(Y^e_q)$ e identifique el multiplicador keynesiano.

- En el modelo ingreso-gasto el equilibrio está determinado por la igualdad entre el Ingreso $(Y)$ y la Demanda Agregada $(DA)$:

$$ Y = DA $$

- Sin embargo, primero, considerando los nuevos parámetros de $X$ y $M$, es necesario establecer la nueva ecuación de la Demanda Agregada:

$$ DA = C + I + G + (X-M) $$

$$ DA = C_o + bY^d + I_o - hr + G_o + x_1Y^* + x_2ε - m_1Y^d - m_2ε $$

- Se reemplaza $Y^d$:

$$ DA = C_o + b(1 - t)Y + I_o - hr + G_o + x_1Y^* + x_2ε - m_1(1 - t)Y - m_2ε $$

- Se reordenan los componenten, teniendo a los autónomos a un lado, y se factoriza:

$$ DA =  C_o + I_o + G_o - hr + x_1Y^* + b(1 - t)Y - m_1(1 - t)Y + x_2ε -  m_2ε $$

$$ DA = C_o + I_o + G_o - hr + x_1Y^* + (b – m_1)(1 – t)Y + (x_2 - m_2)ε $$
 
  
- En base a dicha ecuación, se reemplaza en la igualdad de equilibrio $Y = DA$, y se despeja en función de $Y$ para hallar el ingreso de equilibrio:

  
  $$ Y = C_o + I_o + G_o - hr + x_1Y^* + (b – m_1)(1 – t)Y + (x_2 - m_2)ε $$
  
   
  $$ Y - (b – m_1)(1 – t)Y = C_o + I_o + G_o - hr + x_1Y^* + (x_2 - m_2)ε $$
  
  
  
  $$ Y[1 - (b – m_1)(1 – t)] = C_o + I_o + G_o - hr + x_1Y^* + (x_2 - m_2)ε $$
  
  
  
  $$ Y^e_q = \frac{1}{1 - (b - m_1)(1 - t)} [C_0 + I_0 + G_0 - hr + x_1Y^* + (x_2 - m_2)ε] $$


- De esa manera se deriva el ingreso de equilibrio. Asimismo, dentro de dicha ecuación se identifica y concluye que el multiplicador keynesiano $(k)$ es:
    
$$ k = \frac{1}{1 - (b - m_1)(1 - t)} $$ 

#### 2. Analice los efectos sobre la variable endógena $Y$ de una disminución del gasto fiscal $(∆G_o < 0)$. El análisis debe ser intuitivo, matemático y gráfico.

##### - Análisis intuitivo

- En primer lugar, se debe tener en cuenta que el Producto o ingreso, se ajusta siempre a la demanda agregada. En ese sentido, tenienod en cuenta que se trata de una Política Fiscal Contractiva, con una disminución del gasto fiscal $(∆G_o < 0)$, se puede concluir que, al disminuir el gasto autónomo, el gasto del gobierno también lo hace:

$$ G_o ↓ → G ↓ $$

- Entonces, tendo en cuenta la ecuación de Demanda Agregada $(DA = C+I+G+X-M)$, se puede intuir que, considerando que el Gasto de gobierno baja, la demanda agregada también lo hara, pues el $G$ es uno de los componentes que determina la misma y tienen una relación directamente proporcional. Por lo tanto:

$$ G ↓ → DA ↓ $$

- Entonces, como la demanda agregada disminuye, esta resulta menor al nivel del producto, por lo que se produce una situación de desequilibrio, donde la demanda agregada es menor al producto:

$$ DA > Y $$

- En ese sentido, como se estableció anteriormente, el nivel del producto siempre se ajusta a la demanda agregada, para así mantener el equilibrio. Por esa razón, teniendo en cuenta que la demanda agregada disminuye, generando una situación de desequilibrio; entonces, el nivel del producto necesita disminuir también para que, de esa manera, se regrese al equilibrio:

$$ DA ↓ → Y ↓ $$

- Entonces, se tiene que: 

$$ G_o ↓ → G → ↓ DA ↓ → Y ↓ $$

- Por ende, se puede concluir de forma intuitiva que, una política fiscal contractiva con disminución del gasto de gobierno, genera una caída de la demanda agregada, y por ende, también hace caer el nivel del producto, lo cual genera una contracción en la economía.

##### - Análisis matemático
- Se sabe que $(∆G_0 < 0)$, entonces ¿$∆Y$?
- Se considera la función del ingreso de equilibrio: 


$$ Y = \frac{1}{1 - (b - m)(1 - t)} [C_0 + I_0 + G_0 - hr + x_1Y^* + (x_2 - m_2)ε] $$

- Como se sabe que no hay cambios en los parámetros del multiplicador keynesiado, ni en los otros componentes autónomos de la ecuación. Por lo tanto, se puede decir que sus diferenciales son iguales a 0. Entonces se tiene únicamente:

$$ ∆Y = ∆G_0 $$

- En ese sentido, como se sabe que el gasto autónomo disminuye, su diferencial es negativo:

$$ ∆Y = (-) $$

- Por lo tanto, el diferencial de $Y$ también resulta regativo:

$$ ∆Y = (-) $$

$$ ∆Y < 0 $$

- Entonces, matemáticamente también se concluye que la disminución del gasto de gobierno, genera una disminución en el nivel del producto, y por lo tanto, se contrae la economía.

##### - Análisis gráfico

#### 1. Encuentre las ecuaciones de equilibrio de $Y^e_q (DA-OA)$, $P^e_q (DA-OA)$

##### Ingreso de equilibrio:

- Modelo IS

$$ [1 - b(1-t)]Y = Co + Io + Go - hro $$

- Se despeja $Y$:


$$ Y = \frac{1}{1 - b(1 - t)}(Co + Io + Go) - \frac{1}{1 - b(1 - t)}(hro) $$

$$ Y = \frac{β_0}{β_1} - \frac{hro}{β_1} $$

- donde:

    - $β_0 = C_0 + I_0 + G_0$
    - $β_1 = 1 - b(1 - t)$

- Además, se halla la curva LM, sabiendo que la oferta monetaria está determinada por:

$$ M_s = M_o - P $$

- Entonces, el equilibrio en el mercado monetario estará dado por:

$$ M_o - P = kY - jr $$

$$ M_o = P[kY - jr] $$

- Luego se halla la masa monetaria de equilibrio:

$$ M^e_o = P[kY^e - jro] $$



$$ M^e_o = P[k(\frac{β_0}{β_1} - \frac{1}{β_1}hro) - jro] $$


$$ M^e_o = P[k(\frac{β_0}{β_1}) - (\frac{k}{β_1}h + j)ro] $$

- Entonces, el Ingreso de equilibrio:

$$ Y^e_q = \frac{β_0}{β_1} - \frac{hro}{β_1} $$

#### Precio de equilibrio

Para hallar el precio de equilibrio, solo es necesario reemplazar el ingreso de equilibrio $(Y^e)$ en la función de la oferta agregada:

$$ P = P^e + θ(Y - \bar{Y}) $$

$$ P_q^e = P^e + θ(Y - \bar{Y}) $$

$$ P_q^e = P^e + θ(Y_q^e - \bar{Y}) $$

$$ P_q^e = P^e + θ(\frac{β_0}{β_1} - \frac{hro}{β_1} - \bar{Y}) $$

donde:

- $P$: nivel de precios
- $P^e$: precio esperado
- $\bar{Y}$: producto potencial

```python
 def line_intersection(line1, line2):
    xdiff = (line1[0][0] - line1[1][0], line2[0][0] - line2[1][0])
    ydiff = (line1[0][1] - line1[1][1], line2[0][1] - line2[1][1])

    def det(a, b):
        return a[0] * b[1] - a[1] * b[0]

    div = det(xdiff, ydiff)
    if div == 0:
       raise Exception('lines do not intersect')

    d = (det(*line1), det(*line2))
    x = det(d, xdiff) / div
    y = det(d, ydiff) / div
    return x, y
```
