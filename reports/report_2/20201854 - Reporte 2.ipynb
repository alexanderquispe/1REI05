- A. EL MODELO DE INGRESO - GASTO: LA CURVA IS

C = C_0 + bY^d

I = I_0 - hr

G = G_0

T = tY

M = mY^d

1.1 Utilizando las ecuaciones derive paso a paso la curva IS matemáticamente (a partir de la condición de equilibrio Y = DA)

Y=DA

Y=C+I+G+X-M

Y=C0+b*Y^d+I0-h*r+G0+X0-m*Y^d

Y^d=Y-T=(1-t)*Y

Y=C0+b*(1-t)Y+I0-h*r+G0+X0-m(1-t)*Y

-Y*(1-(1-t)*(b-m)))=-C0-I0-G0-X0+h*r

IS:   \dfrac{-Y*(1-(1-t)*(b-m))}{h}+ \dfrac{(C0+I0+G0+X0)}{h}=r

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sy
import pandas as pd
from causalgraphicalmodels import CausalGraphicalModel
from sympy import *
```

```python
Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3
r = 0.9

Y = np.arange(Y_size)
```

```python
def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)
```

```python
a = 2.5

def L_45(a, Y):
    L_45 = a*Y
    return L_45

L_45 = L_45(a, Y)
```

```python
Y_size = 100 

Co = 15
Io = 18
Go = 28
Xo = 5
h = 0.8
b = 0.5
m = 0.3
t = 0.3

Y = np.arange(Y_size)
```

```python
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = ((Co+Io+Go+Xo)/h-Y*((1-(1-t)*(b-m))/h))
    return r_IS

IS = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
```

Estos puntos están fuera de los puntos de equilibrio en el mercado de bienes debido a que los puntos de equilibro estan ubicados dentro de la curva is.
