```python
# Parámetros

Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3

Y = np.arange(Y_size)


# Ecuación 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (1/h)*(Co + Io + Go + Xo) - (1/h)*(1-(b-m)*(1-t))*Y
    return r_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
```

```python
# Gráfico de la curva IS

# Dimensiones del gráfico
y_max = np.max(r)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, r, label = "IS", color = "red") #Demanda agregada

# Eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

plt.text(98, 136, 'IS', fontsize = 12, color = 'black')
plt.text(70, 180, 'Exceso de Oferta', fontsize = 12, color = 'black')
plt.text(15, 150, 'Exceso de Demanda', fontsize = 12, color = 'black')
plt.text(30, 187, 'A', fontsize = 12, color = 'black')
plt.text(48, 187, 'B', fontsize = 12, color = 'black')
plt.text(12, 187, 'C', fontsize = 12, color = 'black')

plt.axhline(y = 185, xmin= 0, xmax = 0.49, linestyle = ":", color = "grey")

ax.axvline(x = 12, ymin= 0, ymax = 0.68, linestyle = ":", color = "grey")
ax.axvline(x = 30,  ymin= 0, ymax = 0.68, linestyle = ":", color = "grey")
ax.axvline(x = 48,  ymin= 0, ymax = 0.68, linestyle = ":", color = "grey")

# Título, ejes y leyenda
ax.set(title = "Curva IS", xlabel= 'Y', ylabel= 'r')
ax.legend()


plt.show()
```

Respecto de la política monetaria:

Hay 3 posibles instrumentos de la política monetaria:
1) La Oferta Monetaria como instrumento de política

El primer instrumento de política monetaria que utiliza el Banco Central para aumentar o disminuir la oferta monetaria, es la Operación de Mercado Abierto. El Banco Central hace estas operaciones mediante la compra o venta de activos financieros o bonos a los bancos comerciales. En esta, la oferta monetaria es una variable endógena. 


2) El coeficiente legal de encaje como instrumento de política

El Banco Central puede incrementar la cantidad de dinero que tienen los bancos para realizar préstamos mediante la disminución de la tasa de encaje, porque ello implica que los bancos comerciales deben tener menos dinero en forma de reservas en el balance del Banco Central. En otras palabras, la disminución de la tasa de encaje aumenta laposibilidad de creación de dinero bancario porque aumenta el multiplicador de dinero bancario; y, este aumento del dinero bancario implica un aumento de la oferta monetaria.
Por otra parte, cuando el Banco Central aumenta el coeficiente de encaje, los bancos comerciales deben tener una mayor proporción de depósitos en forma de reservas, lo que reduce el multiplicador bancario y, por tanto, origina una disminución de la oferta monetaria.

3) La tasa de interés como instrumento de política

Desde los años 1990 los Bancos Centrales optaron por un nuevo esquema de política monetaria. La tasa de interés se convirtió en instrumento de política y la oferta monetaria pasó a ser una variable endógena. La tasa de política es la tasa de interés de referencia de la política monetaria.
El BCR, disminuyendo la tasa de interés de referencia, aumenta la cantidad de dinero prestada a los bancos comerciales, que aumenta la política monetaria y con ello la oferta monetaria. Por otra parte, si aumenta la tasa de interés de referencia, disminuye la cantidad de dinero prestada (el crédito) a los bancos comerciales, lo que reduce la base monetaria y con ello la oferta monetaria.
