## Causality and DAGS

- Explicar la diferencia entre correlación y causalidad

Lo que diferencia correlación de causalidad, es que la correlación refiere a una relación donde se percibe una afectación en una variable con el cambio de otra; mientras que en la causalidad Hay una relación estrecha y directa entre las variables donde un cambio en la variable afecta directamente la otra. Esta diferencia afecta la posibilidad de predecir: en una correlación solo se puede identificar que habrá mayor afectación en la segunda variable mientras más fuerte sea la correlación, pero no hay una relación lógica entre las variables o sucesos; mientras que en una relación causal sí se puede señalar que una variable cambie por consecuencia de otra.
Por ello, para distinguir estos casos es importante no solo considerar los datos sólidos sino que también la lógica. Así se puede evitar correlaciones espurias donde se confunde correlaciones con relaciones causales, señalando una causa y efecto entre dos variables que no es demostrable. Por ejemplo, una noticia publicada en LA Times sobre las causas del autismo tomaba la opinión de la autora principal del artículo «Este estudio no dice que la exposición a la contaminación del aire o la exposición al tráfico cause autismo… pero podría ser uno de los factores que están contribuyendo a su aumento” para poner en el titular que "Vivir al lado de las autopistas causa autismo". Así se presenta una correlación como una causa definitiva sin respaldo lógico o científico. (https://www.latimes.com/archives/la-xpm-2010-dec-16-la-he-autism-20101217-story.html)

- Qué es un DAG

Un DAG es un modelo de diagrama causal, o "gráfico acíclico dirigido" (DAG por sus siglas en inglés de ‘directed acyclic graph’). Esta es una herramienta útil para representar visualmente hipótesis y relaciones causales entre un conjunto de variables, lo que permite también reconocer algunas fuentes potenciales de sesgo que podrían alterar los resultados de un estudio. Así, este modelo es útil para identificar supuestos, minimizar sesgos, diferenciar efectos individuales de relaciones causales y determinar las covariables adecuadas para el análisis estadístico.

- Explicar qué es un Cofounder

Si bien hay discusiones sobre lo que significa un cofounder, proponemos la definición que considera ambas propiedades de discusión. Por un lado, se ha discutido si (i) bajo la definición candidata el control de todos los "factores de confusión" basta para controlar la confusión en el sesgo  que surge cuando el tratamiento y el resultado comparten una causa común y (ii) si cada factor de confusión en algún contexto ayuda a eliminar o reducir el sesgo de confusión. 

De manera que entendemos un cofounder o "factor de confusión" como una covariable previa a la exposición C para la que existe un conjunto de otras covariables X tal que el efecto de la exposición sobre el resultado no tiene factores de confusión condicional en (X, C) pero tal que para ningún subconjunto adecuado de (X, C) el efecto de la exposición sobre el resultado no tiene factores de confusión dado el subconjunto. O sea, es el sesgo en el tratamiento y resultado pero no son todas las variables previas a estos procesos ni es necesario controlar todas las variables para eliminar los factores de confusión o incluso podría sesgar los resultados.

El criterio de la puerta trasera o Backdoor criterion es que se han bloqueado todas las puertas traseras desde el tratamiento hasta el resultado y  se han controlado los descendientes del tratamiento para no incluirlos, esto ayuda a identificar fácilmente las variables sobre las que tenemos que queremos condicionar. 

## RCT

- ¿Qué es un RCT y cómo se diferencia de otros enfoques de investigación?

- En un RCT, ¿cómo se seleccionan los participantes y cómo se les asignan al grupo de intervención o control?

El RCT (Randomized controlled trials) son un método de control para probar nuevos tratamientos en cualquier área, por ejemplo sirven para calcular y demostrar el nivel de eficacia y seguridad de un fármaco experimental. Se diferencian de otros enfoques de investigación por su metodlogía donde toman 2 grupos para evitar la distorsión o manipulación deliberada de los resultados. 
> Aleatorizado: Los investigadores deciden al azar (se selecciona de manera aleatoria) qué participantes del ensayo reciben el nuevo tratamiento y cuáles reciben un placebo, o tratamiento falso.

> Controlado: El ensayo utiliza un grupo de control como comparación o referencia. En el grupo de control, los participantes no reciben el nuevo tratamiento, sino un placebo o tratamiento de referencia. 
Normalmente, los investigadores seleccionan las personas de un grupo de control por edad, sexo y etnia, junto con cualquier otro factor que pueda influir en el efecto de un fármaco o tratamiento, como el peso corporal, el hábito de fumar o las comorbilidades.

Una variable instrumental es una tercera variable introducida en el análisis de regresión que está correlacionada con la variable predictora, pero no correlacionada con la variable de respuesta. Al usar esta variable, es posible estimar el verdadero efecto causal que alguna variable predictora tiene sobre una variable de respuesta para hacer las mejores estimaciones del efecto de X sobre Y dejando de lado los efectos en que y causa x.


- ¿Cuáles son las características de una buena variable instrumental y cómo se verifican?

El supuesto de relevancia y validez son las dos condiciones mínimas que se deben satisfacer para que una variable instrumental funcione correctamente. El término "relevancia" significa que debe existir una relación mínima entre la variable de tratamiento X y la variable instrumental Z. Mientras que la "validez" quiere decir que la variable instrumental Z debe ser una variable que no tenga puertas traseras propias.

El "diseño de regresión-discontinuidad" (RDD)  refiere a un modelo que permite determinar los efectos causales de las intervenciones asignando un punto de corte o umbral por encima o por debajo del cual se asigna la intervención. Se diferencia de otros modelos de investigación porque toma un punto de corte o umbral que distingue los valores para así diferenciar las observaciones.

- ¿Cuáles son los supuestos del método RDD?

El principal supuesto de este diseño es la continuidad de las variables potencialmente relevantes además de la variable de tratamiento y la variable de resultado sean continuas (variable X), ya que cualquier discontinuidad implica necesariamente alguna causa. 

- ¿Qué es el punto de corte en un RDD y cómo se determina?

El punto de corte en un RDD es el límite de la variable donde se determina si recibe tratamiento o no, las normas detrás de cutoff son las que no recibieron tratamiento, y las de la derecha son las que si lo recibieron. Se determina  según un modelo probabilístico respecto al umbral y el de otros factores que el investigador puede considerar importantes.

- ¿Qué es el efecto de tratamiento en un RDD y cómo se estima?

El tratamiento representa si recibió el tratamiento (1) o no (0) se estima con el efecto causal medio.

## Diff-Diff

- ¿Qué es el método de diferencia en diferencia y cómo se diferencia de otros enfoques de investigación?

La diferencia en diferencias ($DiD$) es una herramienta para estimar los efectos del tratamiento comparando las diferencias antes y después del tratamiento en el resultado de un tratamiento y un grupo de control. Así, se puede observar y comparar los cambios en el tiempo en la variable de interés cuando no es posible aleatorizar. Se diferencia de otros enfoques de investigación porque estima el efecto de un tratamiento $D_i$ (por ejemplo, medicación) sobre un resultado $Y_i$ (por ejemplo, salud) antes y después en un determinado período en el tiempo.

- ¿Cuáles son los supuestos del método de diferencia en diferencia y cómo se verifican?

Mantiene todos los supuestos del modelo de mínimos cuadrados ordinarios. Además, el supuesto del método Diff-Diff es el de "tendencias paralelas" según el cual si el tratamiento no le hubiera ocurrido a nadie, la diferencia entre ambos se habría mantenido igual y los dos grupos hubiesen seguido la misma trayectoria en los resultados de interés. Se supone que la composición de los individuos de los dos grupos se mantenga sin cambios en el tiempo.

- ¿Qué es el efecto de tratamiento en un método de diferencia en diferencia y cómo se estima?


La estimación del efecto del tratamiento usando DID en el caso más simple, es decir, un grupo de control y un grupo de tratamiento observados durante dos períodos de tiempo, uno antes y otro después del tratamiento.

Se puede estimar utilizando la ecuación de cambio, y asumiendo un efecto fijo de estado y un efecto fijo de tiempo constantes, podemos escribir un modelo de regresión simple. 

$$
Y = \beta_0 + \beta_1GrupoTratado + \beta_2DespuésDelTratamiento + \beta_3GrupoTratado + \varepsilon
$$
