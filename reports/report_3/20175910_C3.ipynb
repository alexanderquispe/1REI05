## parte 3 Curva LM ##

1. Derive paso a paso la curva LM matemáticamente (a partir del equilibrio en el Mercado Monetario) y grafique. 

Sabemos: 

$$ M^d = kY - jr $$

$$ M^S = \frac{M_0^s}{P} $$

Por tanto: 

$$ M^d = M^S $$

$$ kY-jr=\frac{M_0^s}{P} $$

$$ kY - \frac{M_0^s}{P} = jr $$

$$ \frac{kY}{j}-\frac{M_0^s}{Pj} = r $$ 

Entonces, la tasa de interés que equilibra el mercado monetario: 

$$ r=-\frac{1}{j}\frac{M_0^s}{P_0}+\frac{k}{j}Y $$

```python

#1----------------------Equilibrio mercado monetario

    # Parameters
r_size = 100

k = 0.5
j = 0.2                
P  = 10 
Y = 35

r = np.arange(r_size)


    # Ecuación
def Ms_MD(k, j, P, r, Y):
    Ms_MD = P*(k*Y - j*r)
    return Ms_MD

Ms_MD = Ms_MD(k, j, P, r, Y)


    # Nuevos valores de Y
Y1 = 45

def Ms_MD_Y1(k, j, P, r, Y1):
    Ms_MD = P*(k*Y1 - j*r)
    return Ms_MD

Ms_Y1 = Ms_MD_Y1(k, j, P, r, Y1)


Y2 = 25

def Ms_MD_Y2(k, j, P, r, Y2):
    Ms_MD = P*(k*Y2 - j*r)
    return Ms_MD

Ms_Y2 = Ms_MD_Y2(k, j, P, r, Y2)

#2----------------------Curva LM

    # Parameters
Y_size = 100

k = 0.5
j = 0.2                
P  = 10               
Ms = 30            

Y = np.arange(Y_size)


# Ecuación

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)
```
