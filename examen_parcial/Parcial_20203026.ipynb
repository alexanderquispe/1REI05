### 1.1 

El multiplicador Keynesiano  contribuye a demostrar una variación en relación a la magnitud de los componentes de la demanda agregada, que forman  parte del intercepto de la ecuación. De esta manera se tiene que el aumento multiplicador del ingreso es consecuencia directa o indirecta por el aumento de alguno de los componentes autónomos de la demanda agregada. Matematicamente se expresa: $ \frac{1}{1-(b-m)(1-t)}  $

Un ejemplo que ayuda a visualizar su función en el modelo Ingreso-Gasto es si el Estado decide generar más gasto de gobierno $(Go)$, que es uno de los componentes autonomos, en la implementación de más infraestructura en el sector de salud pública; por lo que,  esto genera a su vez un incremento en el ingreso y se genera un nuevo equilibrio ante el proceso multiplicador.  


### 1.2

Un exceso y decifit de demanda en el modelo Ingreso-Gasto se da cuando no hay un eequilibrio, en el caso del exceso es porque hay una excases de propucto por lo que se debe incrementar del nivel de producción para contrarresta el exceso de la demanda, que a su vez tiene efectos en el empleo y su tasa . Por el lado del déficit, se da una disminución de producción, lo cual tiene efectos negativos en el empleo. Así, se generar un ajuste en relación a la demadna y la oferta

```python
Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8 # b > m
m = 0.2
t = 0.3
r = 0.9

Y = np.arange(Y_size)


def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)

a = 2.5 

def L_45(a, Y):
    L_45 = a*Y
    return L_45

L_45 = L_45(a, Y)
```

```python
y_max = np.max(DA_IS_K)
fig, ax = plt.subplots(figsize=(10, 8))


ax.plot(Y, DA_IS_K, label = "DA", color = "purple") #Demanda agregada
ax.plot(Y, L_45, color = "#404040") #Línea de 45º


# Líneas punteadas punto de equilibrio
plt.axvline(x=70.5,  ymin= 0, ymax= 0.69, linestyle = ":", color = "grey")
plt.axhline(y=176, xmin= 0, xmax= 0.7, linestyle = ":", color = "grey")
plt.axhline(y=146, xmin= 0, xmax= 0.8, linestyle = ":", color = "grey")

# Texto agregado
    # punto de equilibrio
plt.text(0, 180, '$DA^e$', fontsize = 11.5, color = 'black')
plt.text(72, 0, '$Y^e$', fontsize = 12, color = 'black')

    # línea 45º
plt.text(6, 4, '$45°$', fontsize = 11.5, color = 'black')
plt.text(2.5, -3, '$◝$', fontsize = 30, color = '#404040')
   
plt.text(72, 160, '$Disminución de producción$', fontsize = 12, color = "grey")
  
    
plt.text(82, 193, '$DA<Y$', fontsize = 11, color = "purple")
plt.text(80, 187, '$↖$', fontsize = 18, color = "purple")

ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

# Título y leyenda
ax.set(title="Exceso y Déficit de demanda en el modelo Ingreso-Gasto", xlabel= r'Y', ylabel= r'DA')
ax.legend() 

plt.show()
```
