1. Explique cuáles son los intrumentos de política monetaria que puede utilizar el Banco Central.
Hay tres posibles instrumentos de política monetaria:
- El primer instrumento de política monetaria que utiliza el Banco Central para aumentar o disminuir la oferta monetaria, es la Operación de Mercado Abierto. El Banco Central hace estas operaciones mediante la compra o venta de activos financieros o bonos a los bancos comerciales.
-  El banco central dispone como segundo instrumento el coeficiente legal de encaje. En la política expansiva, Banco Central puede incrementar la cantidad de dinero que tienen los bancos para realizar préstamos mediante la disminución de la tasa de encaje, porque ello implica que los bancos comerciales deben tener menos dinero en forma de reservas en el balance del Banco Central. Además, en la Política monetaria contractiva,Cuando el Banco Central aumenta el coeficiente de encaje, los bancos comerciales deben tener una mayor proporción de depósitos en forma de reservas, lo que reduce el multiplicador bancario y, por tanto, origina una disminución de la oferta monetaria.
- Desde 1990, La tasa de interés se convirtió en instrumento de política: la tasa de interés de referencia. El Banco Central hace política monetaria expansiva reduciendo su tasa de interés de referencia. Y hace política monetaria contractiva aumentando su tasa de interés de referencia.

3. Derive la demanda real de dinero. Explique qué papel cumplen los parametros "k" y "j".

- La demanda de dinero por los motivos transacción y precaución, será:
    $$ L1=k*Y$$
- Por motivos de especulación será:
    $$ L2= -j*i$$
- Si asumimos que ambos tipos de demanda están en términos reales, la función de demanda
real de dinero será:
    $$ L= L1+ L2 $$
    $$ L= k*Y- j*i$$
- La magnitud del parámetro “k” indica la sensibilidad de la demanda de dinero antevariaciones del ingreso “Y”, y la del parámetro “j” indica cuán sensible es la demanda de dinero ante las variaciones de la tasa de interés nominal de los bonos, “i”.


4. Asumiendo que no hay inflación podemos asumir que i=r. Escriba en terminos reales la eucacion de equilibrio en el mercado de dinero.

5. Grafique el equilibrio en el mercado de dinero.


```python
# Parameters
r_size = 100

k = 0.7
j = 0.6                
P  = 14
Y = 30
MS_0 = 500

r = np.arange(r_size)

# Necesitamos crear la funcion de demanda 

def MD(k, j, P, r, Y):
    MD_eq = (k*Y - j*r)
    return MD_eq
MD_0 = MD(k, j, P, r, Y)
# Necesitamos crear la oferta de dinero.
MS = MS_0 / P
MS
```

```python
# Equilibrio en el mercado de dinero

# Creamos el seteo para la figura 
fig, ax1 = plt.subplots(figsize=(10, 8))

# Agregamos titulo t el nombre de las coordenadas
ax1.set(title="Equilibrio en el mercado de dinero", xlabel=r'M^s / P', ylabel=r'r')

# Ploteamos la demanda de dinero
ax1.plot(MD_0, label= '$L_0$', color = "yellow")

# Para plotear la oferta de dinero solo necesitamos crear una linea vertical
ax1.axvline(x = MS,  ymin= 0, ymax= 1, color = "grey")

# Creamos las lineas puntadas para el equilibrio
ax1.axhline(y=0, xmin= 0, xmax= 0.37, linestyle = ":", color = "black")

# Agregamos texto
ax1.text(0, 7.5, "$r_0$", fontsize = 12, color = 'black')

ax1.yaxis.set_major_locator(plt.NullLocator())   
ax1.xaxis.set_major_locator(plt.NullLocator())

ax1.legend()

plt.show()
```

# Curva LM

- Derive paso a paso la curva LM matemáticamente (a partir del equilibrio en el Mercado Monetario) y grafique.
$$Md=Ms$$
$$k*Y- j*r= Mso/ P$$
$$k*Y-Mso/P= j*r$$
$$k*Y/j- Mso/P*j=r$$
$$r=Mso/P*j + k*Y/j$$

```python
# Parameters
Y_size = 100

k = 10
j = 10                
Ms = 200             
P  = 20               

Y = np.arange(Y_size)


# Ecuación

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM
i = i_LM( k, j, Ms, P, Y)
y_max = np.max(i)
v = [0, Y_size, 0, y_max]   
fig, ax = plt.subplots(figsize=(10, 8))

# Eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

# Curvas a graficar
ax.plot(Y, i, label="LM", color = '#3D59AB')
ax.set(title="CURVA LM", xlabel=r'Y', ylabel=r'r')
ax.legend()


plt.show()
```
