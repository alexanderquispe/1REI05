#  A partir del siguiente sistema de ecuaciones que representan el modelo IS-LM

* La funcion que representa la Curva IS (véase Capítulo 4) es
  
 $$ r = \frac{B_0}{h} - \frac{B_1}{h}Y $$
 
 Donde: 
 $ B_0 = C_0 + I_0 + G_0 +X_0 $ y $ B_1 = 1 - (b - m)(1 - t) $
 
 De otro lado la funcion que representa la Curva LM (véase Capítulo 5) es:    
 $$ r = -\frac{1}{j}\frac{M_O^S}{P_O} + \frac{k}{J}Y $$

 Son dos ecuaciones simultáneas. Todos los parámetros son positivos. La propensión marginal a consumir esta entre cero y uno:
 $$ 0 < b < 1 $$
 

# Encuentre las ecuaciones de ingreso $ (Y_e) $ y tasa de intéres $ (r_e) $ de equilibrio (Escriba paso a paso la derivacion de estas ecuaciones).

* Igualando $ r $ 

   $$ \frac{B_0}{h} - \frac{B_1}{h}Y  = -\frac{1}{j}\frac{M_O^S}{P_O} + \frac{k}{J}Y $$

* Obteniendo  $ Y_e $

   $$  \frac{B_0}{h} - \frac{B_1}{h}Y_e  = -\frac{1}{j}\frac{M_O^S}{P_O} + \frac{k}{J}Y_e $$

   $$ Y_e  = \frac{jB_0}{kh+jB_1} - \frac{h}{kh+jB_1} \frac{M_O^S}{P_O} $$

* Obteniendo  $ r_e $

 $$ rh = B_0 - B_1Y $$ donde  $$ Y = \frac{B_0 - rh}{B_1} $$
 
 $$ rj = -\frac{M_O^S}{P_O} + kY $$ donde $$Y = \frac{rj}{k} + \frac{M_O^S}{P_Ok}$$

 $$ \frac{B_0 - r_eh}{B_1} = \frac{r_ej}{k} + \frac{M_O^S}{P_Ok} $$

* Por lo tanto  $ r_e $

   $$ r_e  = \frac{kB_0}{kh+jB_1} - \frac{B_1}{kh+jP_1} \frac{M_O^S}{P_0} $$


## Grafique el equilibrio simultáneo en los mercados de bienes y de dinero

```python

Y_size = 100 
Co = 35
Io = 40
Go = 50
Xo = 2
h = 0.8
b = 0.4
m = 0.5
t = 0.8
Y = np.arange(Y_size)


def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo)/h - ( ( 1-(b-m)*(1-t) ) / h)*Y  
    return r_IS

r_is = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)

 



Y_size = 100
k = 2
j = 1                
Ms = 200             
P  = 20               
Y = np.arange(Y_size)



def r_LM(k, j, Ms, P, Y):
    r_LM = - (1/j)*(Ms/P) + (k/j)*Y
    return r_LM

r_lm = r_LM( k, j, Ms, P, Y)

y_max = np.max(r_lm)
fig, ax = plt.subplots(figsize=(10, 8))

ax.plot(Y, r_is, label = "IS", color = "C1") #IS
ax.plot(Y, r_lm, label="LM", color = "C0")  #LM


ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

plt.axvline(x=51.5,  ymin= 0, ymax= 0.52, linestyle = ":", color = "black")
plt.axhline(y=93, xmin= 0, xmax= 0.52, linestyle = ":", color = "black")


plt.text(49,100, '$E_0$', fontsize = 14, color = 'black')
plt.text(0,100, '$r_0$', fontsize = 12, color = 'black')
plt.text(53,-10, '$Y_0$', fontsize = 12, color = 'black')


ax.set(title="Equilibrio Simúltaneo en los mercados de bienes y de dinero", xlabel= r'Y', ylabel= r'r')
ax.legend()

plt.show()
```

* Funcion de consumo: $$ C = C_0 + bY_d $$

* Gasto del gobierno: $$ G = G_0 $$

* Tributacion: $$ T = tY $$

* Tasa de Interes real: $$ r = i - (pi)^e $$

* Identidad Ingreso-Gasto: $$ Y = C + I + G $$

* Mercado de dinero: la ecuacion LM : $$ \frac{M_O^S}{P_O} = kY - J(r + (pi)^e) $$


# Encuentre las ecuaciones de ingreso $ (Y^e) $ y tasa de intéres $ (r^e) $ de equilibrio (escriba paso a paso la derivación de estas ecuaciones).


* Funcion de Inversion:
$$ I = I_0 + hr $$
$$ hr = I - I_0 $$
$$ r = \frac{I}{h} - \frac{I_0}{h} $$

* donde: $$ I =  Y - C - G  $$

$$ r = \frac{Y - C - G}{h} - \frac{I_0}{h} $$

$$ r = \frac{Y}{h} -\frac{C}{h} -\frac{G}{h}  - \frac{I_0}{h} $$


* Mercado de dinero: la ecuacion LM : 
$$ \frac{M_O^S}{P_O} = kY - J(r + (pi)^e) $$
$$ J(r + (pi)^e) = kY - \frac{M_O^S}{P_O}  $$
$$ r + (pi)^e = \frac{1}{J}kY - \frac{1}{J} \frac{M_O^S}{P_O}  $$
$$ r  = \frac{1}{J}kY - \frac{1}{J} \frac{M_O^S}{P_O} - (pi)^e  $$

* Igualando r(Y) para hallar $Y^e$

$$\frac{Y^e}{h} -\frac{C}{h} -\frac{G}{h}  - \frac{I_0}{h} = \frac{1}{J}kY^e - \frac{1}{J} \frac{M_O^S}{P_O} - (pi)^e $$

$$\frac{Y^e}{h} - \frac{1}{J}kY^e +\frac{-C - G - I_0}{h} = - \frac{1}{J} \frac{M_O^S}{P_O} - (pi)^e $$

$$Y^e (\frac{1}{h} - \frac{1}{J}k)  = - \frac{1}{J} \frac{M_O^S}{P_O} + \frac{C + G + I_0}{h} - (pi)^e $$

$$Y^e (\frac{J -hk}{hJ})   = - \frac{1}{J} \frac{M_O^S}{P_O} + \frac{C + G + I_0}{h} - (pi)^e $$

$$Y^e   = - (\frac{1}{J -hk})(\frac{hM_O^S}{P_O}) + (\frac{1}{J -hk})J(C + G + I_0) - (\frac{1}{J -hk})hJ(pi)^e $$

* Hallando $r^e$
$$ r = \frac{Y}{h} -\frac{C}{h} -\frac{G}{h}  - \frac{I_0}{h} $$

$$  \frac{Y}{h} =   - \frac{C}{h} -\frac{G}{h}  - \frac{I_0}{h} - r $$

$$  Y =   - C - G - I_0 - hr $$

* Hallando $r^e$

$$  kY = \frac{M_O^S}{P_O} + J(r + (pi)^e) $$

$$  Y  = \frac{1}{k} \frac{M_O^S}{P_O} + \frac{1}{k}J(r^e + (pi)^e) $$

* Igualando Y(r) para $r^e$

$$  - C - G - I_0 - hr^e = \frac{1}{k} \frac{M_O^S}{P_O} + \frac{1}{k}J(r^e + (pi)^e) $$

$$  - C - G - I_0 - hr^e = \frac{1}{k} \frac{M_O^S}{P_O} + \frac{Jr^e}{k} + \frac{J(pi)^e}{k} $$

$$  - hr^e - \frac{Jr^e}{k} = \frac{1}{k} \frac{M_O^S}{P_O}  + \frac{J(pi)^e}{k} + C + G + I_0  $$ 

$$  -r^e(\frac{hk + J}{k}) = \frac{1}{k} \frac{M_O^S}{P_O}  + \frac{J(pi)^e}{k} + C + G + I_0  $$ 

$$  -r^e =(\frac{k}{hk + J}) (\frac{1}{k} \frac{M_O^S}{P_O})  + (\frac{k}{hk + J})(\frac{J(pi)^e}{k}) + (C + G + I_0)(\frac{k}{hk + J})  $$ 

$$  r^e = - (\frac{1}{hk + J})(\frac{M_O^S}{P_O}) - (\frac{1}{hk + J})(J(pi)^e) - (\frac{1}{hk + J})k(C + G + I_0)  $$ 


```python

Y_size = 100 
Co = 35
Io = 40
Go = 50
Xo = 2
h = 0.8
b = 0.4
m = 0.5
t = 0.8
Y = np.arange(Y_size)

def r_IS(Co, Io, Go, h, Y):
    r_IS = (Y - Co - Go - Io)/h   
    return r_IS

r_is = r_IS(Co, Io, Go, h, Y)

Y_size = 100
k = 2
j = 1                
Ms = 200             
P  = 20               
Y = np.arange(Y_size)

def r_LM(k, j, Ms, P, Y):
    r_LM =  ((1/j)*k*Y) - ((1/j)*(Ms/P))
    r_LM = r_LM
    return r_LM

r_lm = r_LM( k, j, Ms, P, Y)

y_max = np.max(r_lm)
fig, ax = plt.subplots(figsize=(10, 8))

ax.plot(Y, r_is, label = 'IS', color = "C1") #IS
ax.plot(Y, r_lm, label='LM ', color = "C0")  #LM

ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

ax.set(title="IS-LM Modelo", xlabel= r'Y', ylabel= r'r')
ax.legend()

plt.show()
```
