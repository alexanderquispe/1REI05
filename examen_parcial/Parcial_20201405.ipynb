Existen 3 instrumentos de política monetaria:

Oferta Monetaria: Esto lo utiliza el Banco Central para aumentar o disminuir la oferta monetaria. Se hacen estas operaciones mediante la compra o venta de activos financieros. Se divide en dos:

 - Política Monetaria Expansiva, la cual se realiza mediante bonos de mercado, y con la operación de compra de bonos el Banco Centra inyecta dinero a la economía.
 - Política Monetaria Contractiva, esta se realiza mediante la venta de bonos de mercado, con esta operación el Banco Central retira dinero de la economía. 
El coeficiente legal de encaje: tambien se divide en

 - Política Monetaria Expansiva, en donde la disminución de la tasa de encaje aumenta la posibilidad de creación de dinero bancario porque aumenta el multiplicador de dinero bancario, lo cual, este aumento del dinero bancario implica un aumento de la oferta monetaria.
 - Política Monetaria Contractiva, en donde los bancos comerciales deben tener una mayor proporción de depósitos en forma de reservas, lo que reduce el multiplicador bancario y, por tanto, origina una disminución de la oferta monetaria. 
La tasa de interés: se convirtió en instrumento de política y la oferta monetaria pasó a ser una variable endógena. Cuenta con:

 - Política Monetaria Expansiva, la cual reduciendo su tasa de interés de referencia, así aumenta la cantidad de dinero prestada a los bancos comerciales lo que aumenta la base monetaria y con ello la oferta monetaria. 
 - Política Monetaria Contractiva, la cual aumentando su tasa de interés de referencia, disminuye la cantidad de dinero prestada a los bancos comerciales, lo que reduce la base monetaria y con ello la oferta monetaria.
