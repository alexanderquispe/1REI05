#### 1. 

Existen 3 instrumentos de política monetaria:

1. Oferta Monetaria: Esto lo utiliza el Banco Central para aumentar o disminuir la oferta monetaria. Se hacen estas operaciones mediante la compra o venta de activos financieros. Se divide en dos:
		- Política Monetaria Expansiva, la cual se realiza mediante bonos de mercado, y con la operación de compra de bonos el Banco Centra inyecta dinero a la economía.
		- Política Monetaria Contractiva, esta se realiza mediante la venta de bonos de mercado, con esta operación el Banco Central retira dinero de la economía. 

2. El coeficiente legal de encaje: tambien se divide en 
		- Política Monetaria Expansiva, en donde la disminución de la tasa de encaje aumenta la posibilidad de creación de dinero bancario porque aumenta el multiplicador de dinero bancario, lo cual, este aumento del dinero bancario implica un aumento de la oferta monetaria.
		- Política Monetaria Contractiva, en donde los bancos comerciales deben tener una mayor proporción de depósitos en forma de reservas, lo que reduce el multiplicador bancario y, por tanto, origina una disminución de la oferta monetaria. 

3. La tasa de interés: se convirtió en instrumento de política y la oferta monetaria pasó a ser una variable endógena. Cuenta con: 
		- Política Monetaria Expansiva, la cual reduciendo su tasa de interés de referencia, así aumenta la cantidad de dinero prestada a los bancos comerciales lo que aumenta la base monetaria y con ello la oferta monetaria. 
		- Política Monetaria Contractiva, la cual aumentando su tasa de interés de referencia, disminuye la cantidad de dinero prestada a los bancos comerciales, lo que reduce la base monetaria y con ello la oferta monetaria.

#### 3. Derive la demanda real de dinero. Explique que papel cumplen los parámetros "k" y "j"
    
    La demanda de dinero por transacción y precaución es: 
        Li = KY
    donde : K = propensión marginal a demandar dinero 
    La demanda de dinero por especulación es:
        L2 = - ji
    donde : i = es la tasa de interés nominal
            j = constante de cambio en la demanda de dinero 
    La demanda de dinero total es :
            M^d = L1 + L2
            M^d = KY - ji 
            
    El parámetro "k" es la elasticidad total de los ingresos dentro de la economía, es la sensibilidad de la demanda de dinero, mientras más grande sea "k" mayor cantidad de dinero voy a demandar en cuanto se incremente el nivel de ingreso de la economía. 
    
    El parámetro "j" es la elasticidad de sustitución del dinero en el bolsillo con dinero en el banco, es decir, es la sensibilidad de la demanda de dinero frente a las variaciones de la tasa de interes nominal. Mientras el "j" sea más grande quiere decir que se va a querer mandar más dinero de mi bolsillo al banco pero si "j" es más pequeño en menos proporción voy a querer enviar dinero al banco.

#### 4. Asumiendo que no hay inflación podemos asumir que i=r
   Escriba en términos reales la ecuación de equilibrio en el mercado de dinero. 
   
   La tasa de interés real es:
   
       r = i - π
       
   donde: i -> tasa de interés nominal
          π -> inflación
          
   Como la inflación es muy pequeña asumimos que la tasa de interés nominal es igual a la tasa de interés real. 
    
        i = r
        
   Entonces la demanda de dinero será:
   
        M^d = KY - jr
        
   La oferta real de dinero es: 
   
       M^s = Mo^s/P
       
   El equilibrio en el mercado monetario es:
   
       M^s = M^d
       
       Mo^s/P = KY - jr
  
   En términos reales la ecuación de equilibrio en el mercado de dinero es:
   
       Mo^s = P ( KY - jr )

#### 5. Grafique el equilibrio en el mercado de dinero

```python
# Parameters
r_size = 100

k = 0.5
j = 0.2                
P  = 10 
Y = 35
MS_0 = 500

r = np.arange(r_size)

# Necesitamos crear la funcion de demanda 

def MD(k, j, P, r, Y):
    MD_eq = (k*Y - j*r)
    return MD_eq
MD_0 = MD(k, j, P, r, Y)
# Necesitamos crear la oferta de dinero.
MS = MS_0 / P
MS
```

```python
# Equilibrio en el mercado de dinero

# Creamos el seteo para la figura 
fig, ax1 = plt.subplots(figsize=(10, 8))

# Agregamos titulo t el nombre de las coordenadas
ax1.set(title="Money Market Equilibrium", xlabel=r'M^s / P', ylabel=r'r')

# Ploteamos la demanda de dinero
ax1.plot(MD_0, label= '$L_0$', color = '#CD5C5C')

# Para plotear la oferta de dinero solo necesitamos crear una linea vertical
ax1.axvline(x = MS,  ymin= 0, ymax= 1, color = "grey")

# Creamos las lineas puntadas para el equilibrio
ax1.axhline(y=7.5, xmin= 0, xmax= 0.5, linestyle = ":", color = "black")

# Agregamos texto
ax1.text(0, 7.5, "$r_0$", fontsize = 12, color = 'black')

ax1.yaxis.set_major_locator(plt.NullLocator())   
ax1.xaxis.set_major_locator(plt.NullLocator())

ax1.legend()

plt.show()
```

### Curva LM

#### 1. Derive paso a paso la curva LM matemáticamente (a partir del equilibrio en el Mercado Monetario) y grafique 

   Demanda de dinero: 
     
     M^d = Mo^s/P
     
   Oferta de dinero: 
    
     M^s = Mo^s/P
     
   El equilibrio: M^d = M^s
   Entonces:
   
       KY - jr = Mo^s/P
       KY - Mo^s/P = jr
       r = KY/j - Mo^s/Pj
       
   Por lo tanto la Curva LM es: 
   
       r = - Mo^s/Pj + KY/j
       
   en donde : K/j es la pendiente y Mo^s/Pj es el intercepto
   
     
     

```python
#1----------------------Equilibrio mercado monetario

    # Parameters
r_size = 100

k = 0.5
j = 0.2                
P  = 10 
Y = 35

r = np.arange(r_size)


    # Ecuación
def Ms_MD(k, j, P, r, Y):
    Ms_MD = P*(k*Y - j*r)
    return Ms_MD

Ms_MD = Ms_MD(k, j, P, r, Y)


    # Nuevos valores de Y
Y1 = 45

def Ms_MD_Y1(k, j, P, r, Y1):
    Ms_MD = P*(k*Y1 - j*r)
    return Ms_MD

Ms_Y1 = Ms_MD_Y1(k, j, P, r, Y1)


Y2 = 25

def Ms_MD_Y2(k, j, P, r, Y2):
    Ms_MD = P*(k*Y2 - j*r)
    return Ms_MD

Ms_Y2 = Ms_MD_Y2(k, j, P, r, Y2)

#2----------------------Curva LM

    # Parameters
Y_size = 100

k = 0.5
j = 0.2                
P  = 10               
Ms = 30            

Y = np.arange(Y_size)


# Ecuación

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)
```

```python
#--------------------------------------------------
    # Curva LM ORIGINAL

# Parámetros

Y_size = 100

k = 2
j = 1              
Ms = 200             
P  = 20               

Y = np.arange(Y_size)

# Ecuación

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)

#--------------------------------------------------
    # NUEVA curva LM

# Definir SOLO el parámetro cambiado
k = 4

# Generar la ecuación con el nuevo parámetro
def i_LM_Ms( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i_Ms = i_LM_Ms( k, j, Ms, P, Y)
```

```python
# Dimensiones del gráfico
y_max = np.max(i)
v = [0, Y_size, 0, y_max]   
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, i, label="LM", color = 'purple')
ax.plot(Y, i_Ms, label="LM_Ms", color = '#ea899a', linestyle = 'dashed')

# Texto agregado
plt.text(47, 140, '∆$k$', fontsize=12, color='black')
plt.text(47, 120, '←', fontsize=15, color='black')

# Título y leyenda
ax.set(title = "Aumento en k $(k)$", xlabel=r'Y', ylabel=r'r')
ax.legend()


plt.show()
```
