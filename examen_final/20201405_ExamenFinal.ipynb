### 1. (3 puntos) Instrumental Variables (IV)
Suponga que estamos interesados en estudiar la relación entre educación(X) y el nivel de ingresos(Y).
Un investigador propone utilizar como variable instrumental(Z) el numero de horas que una persona utilizó para ver televisión cuando era niño.

Indetifique todos los elementos necesarios para implementar un enfoque de IV en este ejemplo.
- Podemos identificar los siguientes elementos en el modelo presentado:
   - "Z" la cual no está correlacionada con el error del modelo: C(Z,e) = 0
   - "Z" que si está correlacionada con la variable endógena C(Z; X) ≠ 0
Podemos entender que aquí se debe satisfacer la relevancia y la exclusión. Por otro lado, se necesita una variable instrumental "Z" que limpie el efecto de la "educación" sobre el "nivel de ingresos". Además, podemos concluir que las horas que uno vio televisión de niño tiene al menos una relación con la educación, ya que más horas de TV pueden significar menos horas estudiando. Finalmente la variable instrumental "Z" debe ser una variable que no tenga puertas traseras propias. Es decir, que no debe haber ningún otro camino que lleve a "Z" a afectar a "Y", excepto el camino de "X", que es nuestra variable de tratamiento

Explique el procedimiento para encontrar el efecto causal en este escenario. Puede utilizar ecuaciones
- Debemos identificar si la variable instrumental debe de poder afectar a X "educación", pero esta a la vez no verse afectada por otras variables no observables. Entonces, si logramos encontrar un efecto de "Z" sobre "X", a partir de ello podremos correr una regresión para encontrar la relación entre "X" e "Y" de manera limpia, sin el sesgo.Finalmente, solo deberíamos correr una regresión, y encontrar el "Average Treatment Effect", teniendo como variable instrumental a las "horas que vio televisión de niño". 
¿El efecto de la educación sobre el nivel de ingresos es positivo o negativo? Argumente si los resultados serian causales o solamente de correlación
- El efecto de la educacion sobre el nivel de ingresos es positivo, ya que que por cada año de educación más, los ingresos de las personas tienden a aumentar. Lo que en este caso lo convertiria en una relación causal, ya que el instrumento que hemos utilizado cumple con los supuestos de relevancia y validez. 

### 2. (3 puntos) Regression Discontinuity Design (RDD)

Supongamos que estamos estudiando la eficacia de un programa de becas creado por el gobierno Peruano sobre las notas de los alumnos al final del primer semestre en la universidad. PRONABEC ha decidido que entregará becas a aquellos postulantes que tengan un promedio superior durante la secundaria igual a 16 .

Identifique todos los elementos necesarios para implementar un RDD en este ejemplo.
- El primer elemento que debemos tomar en cuenta para una regresión discontinua es el RUMNING VARIABLE, es decir, es aquella variable que va a determinar si somos tratado o no y en este caso sería promedio de notas. El segundo elemento es el CUT OFF que es una variable que determina si somos tratado o control que en este caso sería la nota superior a 16. El BANDWIDTH son las observaciones que vamos a incluir que tan lejos o cerca va a estar de la discontinuidad, sería la nota 16 más o menos 0.5.  Finalmente, el Tratamiento, el cual representa con 1 si recibes el tratamiento, o con 0 si no lo recibes.

Explique el procedimiento para encontrar el efecto causal en este escenario. Puede utilizar ecuaciones.
- En este ejemplo, podemos identificar el caso como un "sharp design" ya que la probabilidad de obtener la beca es de 1 si el promedio es mayor a 16, y de 0 si es que no se llega a este promedio. Como se mencionó anteriormente. contamos con el cutoff establecido, que es un promedio mayor a 16. Acontinuacion debemos identificar el "salto" en este cut off, ya que, este nos puede indicar el efecto de obtener la beca PRONABEC en las notas de la universidad. Ahora, ¿como encontramos e identificamos ese "salto"? Lo que debemos realizar, es construir una base de datos con nuestras observaciones, y crear un gráfico, a  la izquierda quienes no recibieron la beca, y a la derecha los que sí. Finalmente comparar estos resultados. 

¿El efecto de la politica sobre el rendimiento de los estudiantes sería positivo o negativo (solo necesitan usar su intuición)? Argumente si los resultados serian causales o solamente de correlación .
- El efecto sería positivo, ya que las notas de los estudiantes que alcancen una beca serán más altas. Sin embargo, puede que no exista una relación causal entre obtener una beca y tener mejor rendimiento. Pueden existir factores no observables que afectan el rendimiento de los alumnos, y que los hacen esforzarse más académicamente. Es decir, el estudiante que tiene mas interes en ir a la universidad y que sabe que podria obtener una beca por su nivel academico, se esforzará más por obtener tener notas y así poder conseguirla.
