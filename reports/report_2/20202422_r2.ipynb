### a. El modelo Ingreso-Gasto: la Curva IS

En el modelo Ingreso-Gasto el equilibrio entre el Ingreso y la DA representa el equilibrio en el mercado de bienes.
$$FunciónConsumo: C = C_o + bY^d$$
$$FunciónInversión: I = I_o + hr$$
$$Gasto del Gobierno: G = G_o$$
$$Tributación: T = tY$$
$$Exportación: X = X_o$$
$$Importaciones: M = mY^d$$
$$Gasto o Demanda Agregada: DA = C + I + G + X - M$$
$$Condición de equilibrio: Y = C + I + G + X - M$$

#### 1. Utilizando las ecuaciones, derive paso a paso la curva IS matemáticamente (a partir de la condición de equilibrio Y = DA)

#### Derivando
El ingreso está determinado por el consumo, la inversión, el gasto, las exportaciones netas y las importaciones netas:
$$Y= C + I + G + X - M . . . (1)$$

Si quiere llegar a un equilibrio Ahorro-inversión (Curva IS):

Si al nivel de ingreso se le quita la tasa de tributación
$$Y - T = C + I + G + X - M - T . . . (2)$$

Si el nivel de ingreso se le quita todo lo se tiene de tributación, queda el ingreso disponible Y^d. Entonces, Y - T = Y^d :
$$Y^d = C + I + G + X - M - T$$

Limpiamos a la inversión:
$$Y^d - C - G - X + M + T = I$$

Separamos la ecuación por conveniencia:
$$(Y^d - C) + (T - G) + (M - X) = I$$

Entonces, se obtiene 3 ahorros que representan el ahorro total (S) 
$$Y^d - C = ahorro privado$$
$$T - G = ahorro del gobierno$$
$$M - X = ahorro externo$$ 

Por lo que:
$$S = I . . . (3)$$

Entonces, con la ecuación por conveniencia, se reemplaza por la función de Ingreso-Gasto: 
$$(Y^d - C) + (T - G) + (M - X) = I$$

$$=$$

$$[Y^d - (C_o + bY^d)] + [tY - G_o] + mY^d - X_o] = I_o - hr$$

Se reemplaza la Y^d = (1 - t)Y

$$[(1 - t)Y - (C_o + b(1 - t)Y)] + [tY - G_o] + m(1 - t)Y - X_o] = I_o - hr$$

Entonces, se obtiene:
$$[1 - (b - m)(1 - t)]Y - (C_o + G_o + X_o) = I_o - hr$$
Asimismo, la parte izquierda de la ecuación representa a S y la parte de la derecha representa I

Finalmente, dejamos a r solo para que esté en función de Y (para que se pueda lograr el mercado de bienes):
$$hr = I_o - [1 - (b - m)(1 - t)]Y - (C_o + G_o + X_o)]$$

$$hr = [I_o + C_o + G_o + X_o] - [1 - (b - m)(1 - t)]Y]$$

$$r = \frac{1}{h} (C_o + I_o + G_o + X_o) - \frac{1 - (b - m)(1 - t)}{h} Y$$

La última ecuación encontrada se puede poner como:

$$r = \frac{beta_0}{h} - \frac{beta_1}{h} Y$$

donde,
$$beta_0 = C_o + I_o + G_o + X_o$$
$$beta_1 = 1 - (b - m)(1 - t)$$

#### 2. Utilizando la función de equilibrio IS donde r está en función de Y, encuentre △r/△Y ¿qué representa la pendiente y qué signo tiene?

#### Derivando

Si, por el ejercicio anterior, ya se tenía que:

$$r = -\frac{1}{h} (C_o + I_o + G_o + X_o) - \frac{1 - (b - m)(1 - t)}{h} Y$$

Entonces se reemplaza:

$$△r = -\frac{1- ((b - m)(1 - t}{h} △Y$$

△Y pasa al otro lado de la ecuación:

$$\frac{△r}{△Y} = -\frac{1- ((b - m)(1 - t)}{h} < 0$$

La pendiente es representa por $$\frac{1- ((b - m)(1 - t)}{h}$$ y, si antes era mayor a cero, con △ se ha vuelto negativa (<0). Por otro lado, la curva IS es negativa puesto que en cuanto r aumente y disminuye.

### b. La curva IS o el equilibrio Ahorro-Inversión

#### 1. Derive la ecuación de IS a partir de la igualdad Ahorro-Inversión y grafique la curva IS de equilibrio en el Mercado de Bienes.

#### Derivando

Si la función de ahorro es:
$$S = −(C_o + G_o + X_o) + [1 − (b − m)(1 − t)]Y$$

Y la función de inversión es:
$$I = I_o − hr$$

Entonces, se plantea y se mueve a conveniencia:
$$- (C_o + G_o + X_o) + [1 - (b - m)(1 + t)] Y = I_o - hr$$

$$hr = I_o - [(1 - (b - m)(1 - t)Y - (C_o + G_o + X_o)]$$

$$hr = (I_o + C_o + G_o + X_o) - [(1 - (b - m)(1 - t)Y]

Separamos r:
$$r = -\frac{1}{h} (C_o + I_o + G_o + X_o) - \frac{1 - (b - m)(1 - t)}{h} Y$$





#### Graficando

```python
# Curvas
# Parámetros

Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3

Y = np.arange(Y_size)


# Ecuación 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (1/h)*(Co + Io + Go + Xo) - (1/h)*(1-(b-m)*(1-t))*Y
    return r_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
```

### d. Movimientos en la curva IS

#### 1. Analice una Política Fiscal Contractiva con caída del Gasto del Gobierno (△G < 0). El análisis debe ser intuitivo y gráfico.

#### Analizando

Si la expresión de la Curva IS es:

$$r = \frac{beta_0}{h} - \frac{beta_1}{h} Y$$

donde,
$$beta_0 = C_o + I_o + G_o + X_o$$
$$beta_1 = 1 - (b - m)(1 - t)$$

Jimenez menciona que si hay cambios en los componente autónomos de la demanda agregada, estos afectarán también al intercepto.
En este sentido, cuando el gobierno aplique una Política Fiscal Contractiva y se disminuya el gasto del gobierno (G), la curva IS caerá.

#### Graficando

```python
Y_size = 100 

Co = 35
Io = 40
Go = 60
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3

Y = np.arange(Y_size)
 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)


Go = 25

def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r_G = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
```

```python
# Dimensions
y_max = np.max(r)
fig, ax = plt.subplots(figsize=(10, 8))

# Curves to plot
ax.plot(Y, r, label = "IS", color = "green") 
ax.plot(Y, r_G, label = "IS_G", color = "turquoise", linestyle = 'dashed')

# Text added
plt.text(47, 135, '∆Go', fontsize=12, color='blue')
plt.text(48, 130, '←', fontsize=15, color='blue')

# Title and legend
ax.set(title = "Política Fiscal contractiva con caída del gasto del gobierno", xlabel= 'Y', ylabel= 'r')
ax.legend()

plt.show()
```
