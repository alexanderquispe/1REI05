#### 1. Defina el Multiplicador Keynesiano y dé un ejemplo de cómo influye este en el modelo Ingreso-Gasto Keynesiano.


Un cambio en la magnitud de cualquiera de los componentes autónomos de la Demanda Agregada, genera un proceso multiplicador del ingreso hasta converger al nuevo ingreso y producto de equilibrio. El multiplicador keynesiano representa en cuánto se va incrementar o disminuir el producto ante shocks en cualquiera de los componente autónomos de la $DA$, ya sea el Gasto de Gobierno $(G)$, el Consumo $(C)$, la Inversión $(I)$ o las Exportaciones Netas $(X)$. Además, el tamaño del multiplicador $k$ depende de la magnitud de los parámetros $b$, $m$ y $t$.

En este sentido, los excesos de demanda se enfrentan con aumentos de la producción. Cada vez que aumenta la producción, aumenta el empleo, disminuye la tasa de desempleo y aumenta el ingreso disponible de las familias. Las familias distribuyen el incremento de su ingreso disponible entre consumo de bienes y servicios, y ahorro.

Por otro lado, cuando se produce un déficit de demanda o exceso de producción, contrariamente, aumentan los inventarios y las empresas deciden disminuir su producción y el empleo de trabajadores. 

No obstante, en ambos casos, la oferta de producción se ajusta a los cambios en la demanda agregada. Por su parte, en el punto $A$, la demanda agregada es mayor que el nivel de producción. Hay exceso de demanda que provoca una disminución de los inventarios y aumentos de la producción hasta llegar al equilibrio. Mientras que en el punto $B$ hay un déficit de demanda o un exceso de producción que da lugar a un aumento de los inventarios y a disminuciones de la producción hasta llegar al equilibrio.

### Parte 2: El modelo ingreso-gasto en una economía abierta

#### 1. Encuentre el Ingreso de Equilibrio  e identifique el multiplicador keynesiano.

Reemplazando los parámetros se obtiene: 

$$ DA=C_0 +b(1−t)Y+I_0 −hr+G_0 +x_1Y∗ +x_2ε−m_1(1−t)Y+m_2ε$$

$$ DA=C_0 +I_0 +G_0 +[b(1−t)−m_1(1−t)]Y−hr+x_1Y∗ +(x_2 +m_2)ε $$

Se agrupa los componentes autónomos bajo el parámetro $w$:

$$ DA=w−hr+x_1Y∗ +(x_2 +m_2)ε_0 −(x_2 +m_2)ρr+(x_2 +m_2)ρr∗ +[(b−m_1)(1−t)]Y $$


La ecuación de la Demanda Agregada $(DA)$ es:

$$ DA=[𝑤+x_1Y∗ +(x_2 +m_2)ε_0 +(x_2 +m_2)ρr∗ −(h+(x_2 +m_2)ρ)r]+[(b−m_1)(1−t)]Y $$

$$ DA = α_0 + α_1Y $$


El ingreso de equilibrio se obtiene cuando se utiliza la igualdad $Y = DA$:

$$ Y=w−[h+(x_2 +m_2)ρ]r+x_1Y∗ +(x_2 +m_2)ε_0 +(x_2 +m_2)ρr∗ +[(b−m_1)(1−t)]Y $$

$$ Y = \frac{1}{1−(b−m1)(1−t)} [w+(x_2 +m_2)ε_0 +(x_2 +m_2)ρr∗ +x_1Y∗] - [\frac{h+(x_2 +m_2)ρ}{1−(b−m1)(1−t)}]r $$

Y el multiplicador está representado por:

$$ \frac{1}{1−(b−m1)(1−t)} $$


```python
# Parámetros

Y_size = 100 

Co = 35
Io = 40
Go = 50
h = 0.8
b = 0.9
m = 0.5
t = 0.1

x1 = 0.3
Yº = 12
x2 = 0.6
m2 = 0.4
Eo = 9
p = 0.8
rº = 8
r = 12
m1 = 0.1

Y = np.arange(Y_size)


# Ecuación

def DA(Co, Io, Go, h, b, m, t, x1, Yº, x2, m2, Eo, p, rº, r, m1):
    DA = [Co + Io + Go + x1 *Yº +(x2 + m2)* Eo + (x2 + m2)* p*rº -(h +(x2 +m2)*p)*r]+[(b-m1)*(1-t)]*Y
    return DA

DA = DA(Co, Io, Go, h, b, m, t, x1, Yº, x2, m2, Eo, p, rº, r, m1)
```

```python
# Recta de 45°

a = 2.5 

def L_45(a, Y):
    L_45 = a*Y
    return L_45

L_45 = L_45(a, Y)
```

```python
# Gráfico

# Dimensiones del gráfico
y_max = np.max(Y)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, DA, label = "DA", color = "#3D59AB") #Demanda agregada
ax.plot(Y, L_45, color = "#404040") #Línea de 45º

# Eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

# Líneas punteadas
plt.axvline(x=70.5,  ymin= 0, ymax= 0.69, linestyle = ":", color = "grey")
plt.axhline(y=176, xmin= 0, xmax= 0.7, linestyle = ":", color = "grey")

# Texto agregado
plt.text(0, 180, '$DA^e$', fontsize = 11.5, color = 'black')
plt.text(0, 130, '$α_o$', fontsize = 15, color = 'black')
plt.text(6, 4, '$45°$', fontsize = 11.5, color = 'black')
plt.text(2.5, -3, '$◝$', fontsize = 30, color = '#404040')
plt.text(72, 0, '$Y^e$', fontsize = 12, color = 'black')

# Título y leyenda
ax.set(title="El modelo ingreso-gasto en una economía abierta", xlabel= r'Y', ylabel= r'DA')
ax.legend()

plt.show()
```

### Parte 3: Modelo de OA-DA y el nuevo esquema institucional de Política Monetaria, donde la tasa de interés de referencia $r_0$ es usada como instrumento.

#### 1. Encuentre las ecuaciones de equilibrio de $Y^{eq\_da\_oa}$, $P^{eq\_da\_oa}$

- **Demanda Agregada**:

Considerando la ecuación de la curva IS: 

$$ Y =  \frac{1}{B_1} (B_0) $$

Donde $B_0 = C_0 + I_0 + G_0 + X_0 - hr$ y $B_1 = 1 - (b-m)(1-t)$

Y la ecuación de la LM:

$$ M^s = kY - jr - P_0$$


Reemplazando en función de $P$, se obtiene:

$$ P = M^s - k(\frac{B_1}{B_0}) + jr $$

Y, en función de $M^s$:

$$ M^s = k(\frac{B_1}{B_0}) - jr + P $$

- **Oferta Agregada**:

$$ P = P_e + θ(Y - \bar{Y}) $$

- **Ecuaciones de equilibrio**:

$$ Y^{eq\_da\_oa} = \frac{1}{1 - (b-m)(1-t)} (C_0 + I_0 + G_0 + X_0) $$



$$ P^{eq\_da\_oa} = P_e + θ(\frac{1}{B_1} B_0 - \bar{Y}) $$


```python
#1--------------------------
    # Demanda Agregada
    
# Parámetros

Y_size = 100 

Co = 32
Io = 20
Go = 40
Xo = 2
h = 0.2
b = 0.5
m = 0.4
t = 0.8
r = 5

Ms = 70
k = 0.5
j = 1.2                
P  = 8  

Y = np.arange(Y_size)


# Ecuación

B0 = Co + Io + Go + Xo - h*r
B1 = 1 - (b-m)*(1-t)

def P_AD(h, P, j, B0, B1, k, r):
    P_AD = k*(B0/B1) - j*r + P
    return P_AD

P_AD = P_AD(h, P, j, B0, B1, k, r)


#2--------------------------
    # Oferta Agregada
    
# Parámetros

Y_size = 100

Pe = 100 
θ = 3
_Y = 20   

Y = np.arange(Y_size)


# Ecuación

def P_AS(Pe, _Y, Y, θ):
    P_AS = Pe + θ*(Y-_Y)
    return P_AS

P_AS = P_AS(Pe, _Y, Y, θ)
```

```python
# líneas punteadas autómaticas

    # definir la función line_intersection
def line_intersection(line1, line2):
    xdiff = (line1[0][0] - line1[1][0], line2[0][0] - line2[1][0])
    ydiff = (line1[0][1] - line1[1][1], line2[0][1] - line2[1][1])

    def det(a, b):
        return a[0] * b[1] - a[1] * b[0]

    div = det(xdiff, ydiff)
    if div == 0:
       raise Exception('lines do not intersect')

    d = (det(*line1), det(*line2))
    x = det(d, xdiff) / div
    y = det(d, ydiff) / div
    return x, y
```
