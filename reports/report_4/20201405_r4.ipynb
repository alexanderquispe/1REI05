### A partir del siguiente sistema de ecuaciones que representan el modelo IS-LM 

La función que representa la Curva IS es:

         r=𝜷o/h - (𝜷1/h)Y 
         
donde:𝜷o=Co+Io+Go+Xo y 𝜷1=1-(b-m)(1-t)

La función de la Curva LM es:

        r=-1/j*Mo^s/Po + k/j*Y
        
Todos los parámetros son positivos. La propensión marginal a consumir está entre cero y uno: 0 < b < 1

1. Encuentre las ecuaciones de Ingreso (Y^e) y tasa de interes (r^e) de equilibrio(Escriba paso a paso la derivacion de estas ecuaciones).

Para hallar el Y^e igualamos 

        𝛽o/h - (𝛽1/h)Y = -1*Mo^s/j*P + KY/j
        
        KY/j + (𝛽1/h)Y = 𝛽o/h + 1*Mo^s/j*P
        
   Factorizamos: Y(K/j + 𝛽1/h) = 𝛽o/h + 1*Mo^s/j*P
   
                 Y ((Kh + 𝛽1*j)/jh) = (𝛽o*jP + Mo^s*h)/hjP
                 
                 Y = (𝛽o*jP + Mo^s*h)/P(Kh + 𝛽1*j)
                 
                 Y= 𝛽o*jP/P(Kh + 𝛽1*j) + Mo^s*h/P(Kh + 𝛽1*j)
   
   Por lo tanto la ecuación del ingreso de equilibrio (Y^e) es:
   
   Y^e = 𝛽o*j/(Kh + 𝛽1*j) + (h/Kh + 𝛽1*j) * M^o/P
   
   Para hallar la tasa de interés de equilibrio reemplazamos 1 en 2 y despejamos r.
   
  De 1  r = 𝛽o/h - 𝛽1*Y/h     despejamos Y
  
        Y = 𝛽o/𝛽1 - r*h/𝛽1
        
   Luego reemplazamos
   
       r = -1/j*Mo^s/P + K/j*Y
       
       r = -1/j*Mo^s/P + K/j (𝛽o/𝛽1 - rh/𝛽1)
       
       r = -1/j*Mo^s/P + K*𝛽o/j𝛽1 - Krh/j𝛽1
       
       r + Krh/j𝛽1 = K𝛽o/j𝛽1 - 1/j*Mo^s/P
       
  Factorizamos r
    
     r(1 + Kh/j𝛽1) = K𝛽o/j𝛽1 - 1/j * Mo^s/P
     
     r = [j𝛽1(K𝛽o/j𝛽1 - 1/j*Mo^s/P)]/j𝛽1 + Kh = [K𝛽o - (𝛽1*Mo^s/P)]/j𝛽1 + Kh
     
   Por lo tanto la ecuación de la tasa de interés de equilibrio (Y^e) es:
   
     r^e = K𝛽o/Kh + j𝛽i - (𝛽1/Kh + j𝛽1)*Mo^s/P
   
   
        

#### GRAFICO

```python
# Para la Curva IS

# Parámetros

Y_size = 100 

Co = 35
Io = 40
Go = 50
Xo = 2
h = 0.8
b = 0.4
m = 0.5
t = 0.8

Y = np.arange(Y_size)


# Ecuación 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo)/h - ( ( 1-(b-m)*(1-t) ) / h)*Y  
    return r_IS

r_is = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)

#--------------------------------------------------
# Para la Curva LM 

# Parámetros

Y_size = 100

k = 2
j = 1                
Ms = 200             
P  = 20               

Y = np.arange(Y_size)

# Ecuación

def r_LM(k, j, Ms, P, Y):
    r_LM = - (1/j)*(Ms/P) + (k/j)*Y
    return r_LM

r_lm = r_LM( k, j, Ms, P, Y)
```

```python
# Gráfico del modelo IS-LM

# Dimensiones del gráfico
y_max = np.max(r_lm)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
# IS
ax.plot(Y, r_is, label = "IS", color = "pink") #IS
# LM
ax.plot(Y, r_lm, label="LM", color = "green")  #LM

# Eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

# Texto y figuras agregadas
# Graficar la linea horizontal - r
plt.axvline(x=51.5,  ymin= 0, ymax= 0.52, linestyle = ":", color = "black")
# Grafica la linea vertical - Y
plt.axhline(y=93, xmin= 0, xmax= 0.52, linestyle = ":", color = "black")

# Plotear los textos 
plt.text(49,100, '$E_0$', fontsize = 14, color = 'black')
plt.text(0,100, '$r_0$', fontsize = 12, color = 'black')
plt.text(53,-10, '$Y_0$', fontsize = 12, color = 'black')

# Título, ejes y leyenda
ax.set(title="Equilibrio simultaneo en los mercados de bienes y de dinero", xlabel= r'Y', ylabel= r'r')
ax.legend()

plt.show()
```

1. Analice los efectos sobre las variables endógenas Y, r de una disminución del gasto fiscal.(ΔGo < 0) . El análisis debe ser intuitivo, matemático y gráfico.

Intuición 

    ΔGo < 0
    
    Analisis a partir del mercado de bienes 
    
    G↓ ⟶ DAC↓ ⟶ DA < Y ⟶ Y↓
    
    Si el gasto fisacl baja esto origina que la demanda agregada disminuya, si esto sucede hay un desequilibrio entre DA y el producto porque DA está por debajo del producto. Esto origina que el nivel del producto disminuya y entonces la demanda agregada va a ser igual al nivel del producto. 
    
    Analisis a partir del mercado de dinero 
    
    Y↓ ⟶ M^d↓ ⟶ M^d < M^s ⟶ r↓
    
    Si el nivel del producto disminuye esto origina que la demanda de dinero disminuya entonces la demanda de dinero es menor que la oferta de dinero, y si queremos equilibrar ese mercado, la tasa de interes debe disminuir 
    
Matemáticamente
        
        ΔY^e = (j/Kh + j𝜷1) * ΔGo   < 0
                   (+)       (-) 
           
        Δr^e = (K/Kh + j𝜷1) * ΔGo   < 0
                   (+)        (-)

2. Analice los efectos sobre las variables endógenas Y, r de una disminución de la masa monetaria.(ΔM^s < 0) . El análisis debe ser intuitivo, matemático y gráfico.

Intuicion
 
     ΔMo^s < 0
     
 En el mercado monetario 
 
     Mo^s↓ ⟶ M^s < M^d ⟶ r↑
     
     Si la oferta monetaria disminuye va a originar un desbalance en el mrecado monetario porque la oferta de dinero va a ser menor a la cantidad de dinero demandada. 
     
 En el mercado de bienes
 
     r↑ ⟶ I↓ ⟶ DA↓ ⟶ DA < Y ⟶ Y↓
     
 Matematicamente
 
     ΔY^e= (h/Kh + j𝛃1)1/P * ΔM^s    < 0
              (+)      (+)   (-)
              
     Δr^e = - (𝛃1/ Kh + j𝛃1)1/P * ΔMo^s   > 0
           (-)    (+)       (+)   (-)

#### GRAFICO

```python
#1--------------------------------------------------
    # Curva IS ORIGINAL

# Parámetros

Y_size = 100 

Co = 35
Io = 40
Go = 50
Xo = 2
h = 0.8
b = 0.4
m = 0.5
t = 0.8

Y = np.arange(Y_size)


# Ecuación 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r_0 = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)


#2--------------------------------------------------
    # Curva LM ORIGINAL
# Parámetros

Y_size = 100

k = 2
j = 1                
Ms = 700             
P  = 20               

Y = np.arange(Y_size)

# Ecuación

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i_0 = i_LM( k, j, Ms, P, Y)

#-------------------------------------------------
    # Curva LM NUEVA
    
#Parámetro que cambia: Masa monetaria (Ms)

Ms = 200

# Nueva curva LM con la variacion del Ms
def i_LM_Ms( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i_Ms = i_LM_Ms( k, j, Ms, P, Y)
```
