El artículo tiene como objetivo modelar la política monetaria no convencional que llevó a cabo la FED e incorporó innovaciones de política monetaria al modelo estándar de demanda y oferta agregada de economía cerrada, a fin de hacer frente a la crisis internacional de 2008-2009 en un contexto donde los instrumentos tradicionales de política monetaria resultaban insuficientes. En ese sentido, Mendoza busca responder si los viejos modelos y los viejos métodos son todavía útiles para tratar los problemas macroeconómicos actuales.

La fortaleza del artículo es que introduce dos diferencias al modelo IS-LM tradicional para efectos de su pregunta de investigación: 1. la FED administra la tasa de interés de corto plazo, no la oferta monetaria, que considera endógena; y 2. adiciona un mercado de bonos a largo plazo a los mercados de bienes, dinero y bonos a corto plazo para analizar el impacto de la flexibilización cuantiativa. De esa forma, logra proponer un modelo IS-LM útil y de fácil entendimiento para explicar la política monetaria de la FED: el modelo IS-LM-BB.

El artículo recupera el conocimiento tradicional de economistas keynesianos como el propio Keynes, Hicks, Tobin, Brainard y Branson, quienes reflexionaron en su momento sobre los problemas generados por el exceso de liquidez y la necesidad de intervenir en los mercados de deuda a largo plazo para hacerles frente. El modelo propuesto por Mendoza, inspirado en el keynesianismo clásico, logra tres cosas:

1. Determinar los valores de equilibrio de producción, la tasa de interés de largo plazo y el nivel de precios. Para este modelo, las variables exógenas son la producción, el stock de bonos de corto plazo u oferta monetaria y la tasa de interés a largo plazo. Al respecto, el equilibrio general se alcanza cuando hay equilibrio simultáneo en los tres mercados analizados: bienes (donde se determina la producción), dinero (donde se determina el stock de bonos de corto plazo) y bonos de largo plazo (donde se determina la tasa de interés de largo plazo). Si estos están en equilibrio, por la Ley de Walras, el mercado de bonos de corto plazo también lo está.  
2. Discutir los efectos de las políticas convencionales y no convencionales (como la de compra de bonos a largo plazo). Para ello, conjuga la demanda y la oferta agregada en el corto plazo.
3. Discutir los efectos dinámicos de la política fiscal y de los choques adversos de oferta sobre el conjunto de variables endógenas del modelo a lo largo del tiempo. Para ello, el autor utiliza el método de simulaciones y evalúa numéricamente las proposiciones formuladas en el artículo. 

Sin embargo, el trabajo tiene la debilidad de ser muy descriptivo y usar modelos abstractos sin aterrizar en la evaluación concreta de situaciones reales mediante dichos modelos. 

Además, si bien Mendoza menciona que el FMI realizó una evaluación sobre la efectividad de las políticas convencionales, la omisión de hacer referencia al éxito o fracaso de estas en el artículo genera dudas acerca de la relevancia del trabajo de Mendoza en términos prácticos. 

Asimismo, otro problema con el artículo de Mendoza es buscar modelar las políticas de la FED, en lugar de enfocarse en el BCRP. Las políticas de la FED no solo se orientan a mantener un nivel determinado de inflación, sino también a lograr el pleno empleo. Asimismo, otro aspecto a tener en cuenta es el estatus del dólar como divisa mundial y el enorme poder de la FED por tener el derecho de señoraje de esa moneda, lo que le permite realizar políticas monetarias no convencionales imposibles de llevar a cabo para el resto de bancos centrales. Por tanto, otra falencia de Mendoza es no señalar si lo explicado en su artículo tiene alguna aplicación práctica a nivel nacional, tomando en cuenta nuestras propias instituciones, y el porqué. 

Un paso para avanzar en la pregunta es aplicar el modelo IS-LM-BB frente a los problemas económicos actuales y así evaluar su capacidad de explicar las crisis internacionales. Hoy existe un aumento generalizado de la inflación a nivel global y los bancos centrales de todo el mundo han decidido subir sus tasas de interés para hacer frente a ese aumento (McKinsey & Company, 2022). En este escenario, cabe preguntarnos si el keynesianismo clásico es realmente útil para explicar una coyuntura tan complicada. 

Otro paso para avanzar en la pregunta es innovar en la ecuación del mercado de bienes del modelo IS-LM-BB. En ese sentido, Listokin sugiere introducir cambios en las normas legales a fin de aumentar la demanda agregada. Por ejemplo, cambiando las normas sobre zonificación para promover la construcción. ## A partir del siguiente sistema de ecuaciones que representan el modelo IS-LM:

- **Encuentre las ecuaciones de Ingreso ($Y^e$) y tasa de interes ($r^e$) de equilibrio (escriba paso a paso la derivación de estas ecuaciones).**

Resolviendo por cualquiera de los 3 métodos (reducción, igualación o sustitución), se obtiene los valores del ingreso y de la tasa de interés que equilibran ambos mercados.

Igualando las 2 ecuaciones:

$$-\frac{1}{j}\frac{M_0^s}{P_0} + \frac{k}{j}Y = \frac{B_0}{h} - \frac{B_1}{h}Y$$


Para hallar $Y^e$ debemos poner las ecuaciones en función a $Y$.

$Y^e$:

$$-\frac{1}{j}\frac{M_0^s}{P_0} + \frac{k}{j}Y = \frac{B_0}{h} - \frac{B_1}{h}Y$$

$$\frac{k}{j}Y + \frac{B_1}{h}Y = \frac{B_0}{h}+\frac{1}{j}\frac{M_0^s}{P_0}$$

$$Y(\frac{k}{j} + \frac{B_1}{h}) = \frac{B_0}{h}+\frac{1}{j}\frac{M_0^s}{P_0}$$

$$Y(P_0kh+jB_1P_0) = B_ojP_o+M_0^sh$$

$$Y^e = \frac{jB_0P_0+M_0^sh}{P_0kh+B_1jP_0}$$

$$Y^e = \frac{jB_0P_0}{P_0kh+B_1jP_0}+\frac{M_0^sh}{P_0kh+B_1jP_0}$$

$$Y^e = \frac{P_0(jB_0)}{P_0(kh+B_1j)}+\frac{M_0^sh}{P_0(kh+B_1j)}$$

$$Y^e = \frac{jB_0}{kh+jB_1}+(\frac{h}{kh+jB_1})\frac{M_0^s}{P_0}$$

Para hallar $r^e$, debemos reemplazar $Y$ por $Y^e$ en la ecuación de la Curva IS-LM:

$r^e$:

$$ r = \frac{B_0}{h} - \frac{B_1}{h}Y$$

$$ r^e = \frac{B_0}{h} - \frac{B_1}{h}Y^e$$

$$ r^e = \frac{B_0}{h} - \frac{B_1}{h} [\frac{jB_0}{kh+jB_1}+(\frac{h}{kh+jB_1})\frac{M_0^s}{P_0}]$$

$$ r^e = \frac{B_0}{h} - \frac{jB_0B_1}{h(kh+jB_1)} + \frac{B_1}{h}(\frac{h}{kh+jB_1})\frac{M_0^s}{P_0}$$

$$ r^e = \frac{B_0P_0(kh+jB_1)-jB_0B_1P_0+hB_1M_0^s}{hP_o(kh+jB_1)}$$

$$ r^e = \frac{khB_0P_0+jB_0B_1P_0-jB_0B_1P_0+hB_1M_0^s}{hP_0(kh+jB_1)}$$

$$ r^e = \frac{khB_0P_0+hB_1M_0^s}{hP_0(kh+jB_1)}$$

$$ r^e = \frac{khB_0P_0}{hP_0(kh+jB_1)}+\frac{hB_1M_0^s}{hP_0(kh+jB_1)}$$

$$r^e = \frac{kB_0}{kh + jB_1} - (\frac{B_1}{kh + jB_1})\frac{M^s_0}{P_0}$$

- **Grafique el equilibrio simultáneo en los mercados de bienes y de dinero.**

```python
#Curva IS

Y_size = 100 

Co = 35
Io = 40
Go = 50
Xo = 2
h = 0.8
b = 0.4
m = 0.5
t = 0.8

Y = np.arange(Y_size)
 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo)/h - ( ( 1-(b-m)*(1-t) ) / h)*Y  
    return r_IS

r_is = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)

#Curva LM

Y_size = 100

k = 2
j = 1                
Ms = 200             
P  = 20               

Y = np.arange(Y_size)

def r_LM(k, j, Ms, P, Y):
    r_LM = - (1/j)*(Ms/P) + (k/j)*Y
    return r_LM

r_lm = r_LM( k, j, Ms, P, Y)
```

```python
#Modelo IS-LM

y_max = np.max(r_lm)
fig, ax = plt.subplots(figsize=(10, 8))

ax.plot(Y, r_is, label = "IS", color = "blue") 
ax.plot(Y, r_lm, label="LM", color = "red")  

ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

plt.axvline(x=51.5,  ymin= 0, ymax= 0.52, linestyle = ":", color = "grey")

plt.axhline(y=93, xmin= 0, xmax= 0.52, linestyle = ":", color = "grey")


plt.text(50,100, '$E$', fontsize = 14, color = 'black')
plt.text(-10,90, '$r_0$', fontsize = 12, color = 'black')
plt.text(50,-30, '$Y_0$', fontsize = 12, color = 'black')
plt.text(50,150, '$I$', fontsize = 14, color = 'black')
plt.text(50,50, '$III$', fontsize = 14, color = 'black')
plt.text(25,100, '$IV$', fontsize = 14, color = 'black')
plt.text(75,100, '$II$', fontsize = 14, color = 'black')

ax.set(title="Gráfico 6.1\nEquilibrio simultáneo en los mercados de bienes y dinero", xlabel= r'Y', ylabel= r'r')
ax.legend()
ax.xaxis.set_label_coords(1, -0.05)
ax.yaxis.set_label_coords(-0.05, 0.95)

plt.show()
```

- **Analice los efectos sobre las variables endógenas Y, r de una disminución del gasto fiscal. ($\Delta G_o$ < 0). El análisis debe ser intuitivo, matemático y gráfico.**

### Análisis intuitivo 

- Mercado de Bienes:

$$Go↓ → DA↓ → DA < Y → Y↓$$

- Mercado de Dinero:

$$Y↓ → M^d↓ → M^d < M^s → r↓$$

### Análisis matemático

```python
Co, Io, Go, Xo, h, r, b, m, t, beta_0, beta_1  = symbols('Co Io Go Xo h r b m t beta_0 beta_1')

k, j, Ms, P, Y = symbols('k j Ms P Y')

beta_0 = (Co + Io + Go + Xo)
beta_1 = 1-(b-m)*(1-t) 

r_eq = (k*beta_0)/(k*h + j*beta_1) - ( beta_1 / (k*h + j*beta_1) )*(Ms/P)
Y_eq = (j*beta_0)/(k*h + j*beta_1) + ( h / (k*h + j*beta_1) )*(Ms/P)
```

```python
df_r_eq_Ms = diff(r_eq, Go)
print("El diferencial de la tasa de interes con respecto al diferencial del gasto autónomo es", df_r_eq_Ms) 
```

El signo del diferencial de $r$ con respecto al diferencial de $G_0$ es negativo (-).

```python
df_Y_eq_Ms = diff(Y_eq, Go)
print("El diferencial del producto con respecto al diferencial del gasto autónomo es", df_Y_eq_Ms)
```

El signo del diferencial de $Y$ con respecto al diferencial de $G_0$ es negativo (-).
