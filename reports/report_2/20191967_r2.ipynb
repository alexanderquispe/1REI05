## **a. El Modelo Ingreso-Gasto: la Curva IS**

**En el modelo Ingreso-Gasto el equilibrio entre el Ingreso y la DA representa el equilibrio en el mercado de bienes.**

**Función Consumo: $C = C_0 + bY^d$**

**Función Inversión: $I = I_0 - hr$**

**Gasto del Gobierno: $G = G_0$**

**Tributación: $T = tY$**

**Exportaciones: $X = X_0$**

**Importaciones: $M = mY^d$**

**Gasto o Demanda Agregada: $DA = C + I + G + X - M$**

**Condición de equilibrio: $Y = C + I + G + X - M$**

- **Utilizando las ecuaciones derive paso a paso la curva IS matemáticamente (a partir de la condición de equilibrio $Y = DA$).**

Si:

$$Y = DA$$

Y:

$$Y = C + I + G + X - M$$

$$DA = C + I + G + X - M$$

Igualando:

$$Y = C + I + G + X - M$$

Restando $T$ a ambos lados de la ecuación:

$$Y - T = C + I + G + X - M - T$$

Si:

$$Y^d = Y - T$$

Reemplazando:

$$Y^d = C + I + G + X - M - T$$

$$Y^d - C - G - X + M + T = I$$

$$(Y^d - C) - (T - G) + (M -X) = I$$

Si:

$$C = C_0 - bY^d$$

$$I = I_0 - hr$$

$$G = G_0$$

$$T = tY$$

$$X = X_0$$

$$M = mY^d$$

Reemplazando:

$$[Y^d - (C_0 + bY^d)] + [tY - G_0] + [mY^d - X_0] = I_0 - hr$$

Si: 

$$Y^d = (1 - t)Y$$

Reemplazando:

$$\{(1 - t)Y - [C_0 + b(1 - t)Y]\} + [tY - G_0] + [m(1-t)Y - X_0] = I_0 - hr$$

$$[(1 - (b - m)(1 - t)]Y - (C_0 + G_0 + X_0) = I_0 - hr$$

$$hr = I_0 - [(1 - (b - m)(1 - t)Y - (C_0 + G_0 + X_0)]$$

$$hr = (I_0 + C_0 + G_0 + X_0) - [(1 - (b - m)(1 - t)Y]$$

$$r = \frac{1}{h}(I_0 + C_0 + G_0 + X_0) - \frac{[(1 - (b - m)(1 - t)]}{h}Y$$


- **Utilizando la función de equilibrio $IS$ donde $r$ está en función de $Y$, encuentre $\frac{\Delta r}{\Delta Y}$ que representa la pendiente y qué signo tiene?**

La función de equilibrio $IS$ donde $r$ está en función de $Y$ es:

$$r = \frac{1}{h}(I_0 + C_0+ G_0 + X_0) - \frac{[(1 - (b - m)(1 - t)]}{h}Y$$

Esta ecuación indica que existen pares de valores ordenados de ingreso $Y$ y de tasa de interés $r$ que equilibran el mercado de bienes y que estos valores tienen una relación inversa.

Tomemos un ingreso de equilibrio $Y_0$ que corresponde a una tasa de interés real igual a $r_0$. Ahora, supongamos que se produce una reducción de la tasa de interés igual a $r_1$. Entonces, $r_0 < r_1$. Esta reducción de la tasa de interés eleva la magnitud del intercepto de la función $DA$, desplazándola hacia arriba, y da lugar a un ingreso de equilibrio de mayor magnitud e igual a $Y_1$. Entonces, $Y_0 > Y_1$. 

Ambas desigualdades que equilibran el mercado de bienes se pueden representar en el plano $(Y, r)$ como una **recta con pendiente negativa** que pasa por los puntos $(Y_0, r_0)$ y $(Y_1,  r_1)$. 

Entonces:

$$\frac{\Delta r}{\Delta Y} = \frac{r_0 - r_1}{Y_0 - Y_1} $$

Toda vez que se cumple que:

$$r_0 < r_1$$ 
$$Y_0 > Y_1$$

Tenemos que:

$$\frac{r_0 - r_1}{Y_0 - Y_1} = \frac{(-)}{(+)} = (-) $$

Por tanto:

$$\frac{\Delta r}{\Delta Y} < 0$$

Podemos concluir que **la pendiente es negativa**.

- **Lea la sección 4.4 del material de enseñanza y explique cómo se deriva la curva $IS$ a partir del equilibrio $Y = DA$ y grafique lo siguiente:**

El equilibrio entre el Ingreso y la Demanda Agregada representa el equilibrio en el mercado de bienes. Para derivar la curva $IS$, se debe partir de la condición $DA = Y$. Sabemos que $Y$ = $C$ + $I$ + $G$ + $X$ - $M$. Luego, restamos $T$ de ambos miembros de la igualdad. De esa forma, obtenemos la ecuación de $Y^d$. Luego, ponemos la ecuación en función de $I$. Después, agrupamos los términos de tal forma que el lado derecho corresponda al Ahorro Total, que es igual a la Inversión. Al terminar, reemplazamos los componentes por sus funciones. Finalmente, realizando algunas operaciones algebraicas, obtenemos la ecuación de la Curva $IS$. Esta también puede expresarse con una ecuación donde la tasa de interés se encuentra en función del Ingreso.

## **b. La Curva IS o el equilibrio Ahorro-Inversión**

$$S = I$$

$$S = -(C_0 + G_0 + X_o) + [1 - (b-m)(1-t)]Y$$
    
$$I = I_0 - hr$$

Entonces:

$$-(C_0 + G_0 + X_o) + [1 - (b-m)(1-t)]Y = I_0 - hr$$

$$hr = I_o - [(1 - (b - m)(1 - t)Y - (C_o + G_o + X_o)]$$

$$hr = (I_o + C_o + G_o + X_o) - [(1 - (b - m)(1 - t)Y]$$

$$r = \frac{1}{h}(I_o + C_o + G_o + X_o) - \frac{[(1 - (b - m)(1 - t)]}{h}Y$$

Los puntos de la Curva $IS$ corresponden a pares ordenados ($Y$, $r$) que equilibran el mercado de bienes. Por tanto, los puntos fuera de esa curva son de desequilibrio en el mercado. Existen dos tipos de desequilibrio de mercado: exceso de oferta y exceso de demanda (a la derecha y a la izquierda de la curva $IS$ respectivamente).

Si observamos el punto "A" del gráfico de arriba es de equilibrio. El par ordenado ($Y_A$, $r_A$) equilibra el ahorro con la inversión $I_A = S_A$. El punto "B", que se encuentra a la derecha de la $IS$, corresponde al par ordenado ($Y_B$, $r_A$). Como se mantiene la tasa de interés, la inversión permanece constante, pero el ahorro es mayor que el que corresponde al punto "A" porque el ingreso en el punto "B" es más alto. En el punto "B", $I_A$ < $S_B$; por lo tanto, a la derecha de la $IS$ el desequilibrio del mercado de bienes es exceso de oferta. Por otro lado, a la izquierda de la curva $IS$ el desequilibrio de mercado es de exceso de demanda.

- **Analice una Política Fiscal Contractiva con caída del Gasto del Gobierno ($\Delta G < 0$). El análisis debe ser intuitivo y gráfico.**

Los cambios en el intercepto producidos por los cambios en los componentes autónomos de la Demanda Agregada pueden dar lugar a desplazamientos paralelos de la curva IS, hacia la derecha o hacia la izquierda. 

La disminución de cualquier gasto autónomo aumenta la Demanda Agregada de bienes generando una disminución de demanda que da lugar a un desplazamiento de la curva IS hacia la izquierda, donde hay una disminución de la oferta, con una distancia equivalente a la magnitud de dicho gasto. 

Por lo tanto, si el gobierno aplica una Política Fiscal Contractiva disminuyendo el gasto, la curva IS debe desplazarse hacia la izquierda con una distancia equivalente a dicho gasto.

$$↓G_0 → ↓DA → DA < Y → ↓Y$$
