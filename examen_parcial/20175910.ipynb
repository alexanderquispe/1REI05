(2 puntos) Defina el Multiplicador Keynesiano y dé un ejemplo de cómo influye este en el modelo Ingreso-Gasto Keynesiano.

El multiplicador keynesiano es el “multiplicador” "k" que genera un proceso multiplicador en el ingreso hasta converger al nuevo ingreso y producto de equilibrio. Por ello, el aumento multiplicado del ingreso (Y) es resultado de los efectos directos e indirectos ocasionados por el aumento de los componentes autónomos de la demanda agregada (DA). 
El tamaño de este multiplicador depende de la magnitud de los parámetros en la propensión marginal a ahorrar (b), la propensión marginal a importar (m) y la tasa promedio de impuestos (t). Asimismo, y en relación con el párrafo anterior, se comprende que las filtraciones de demanda son las que reducen su tamaño.

Un ejemplo de cómo puede influir en el modelo Ingreso Gasto keynesaiano es, por ejemplo, si aumenta k, entonces la pendiente de la curva LM va a aumentar. Esto se debe a que si k aumenta, la sensibilidad de los ingresos de las personas para tenerlos en su bolsillo es más alta. Esto genera que se requiera un aumento de la tasa de interés para regresar al equilibrio, para que así las personas decidan poner más de sus ingresos en los bancos. :

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sy
import pandas as pd
import numpy as np
import random
import math
import sklearn
import scipy as sp
import networkx
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.iolib.summary2 import summary_col

#--------------------------------------------------
    # Curva LM ORIGINAL

# Parámetros

Y_size = 100

k = 0.5
j = 1                
Ms = 200             
P  = 20               

Y = np.arange(Y_size)

# Ecuación

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)

#--------------------------------------------------
    # NUEVA curva LM

# Definir SOLO el parámetro cambiado
k = 0.9

# Generar la ecuación con el nuevo parámetro
def i_LM_k( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i_k = i_LM_k( k, j, Ms, P, Y)
```

3. (2 puntos) ¿Cuáles son las herramientas de política fiscal y política monetaria? Dea un ejemplo para cada una de ellas dentro del contexto peruano. Coloque su fuente en caso sea necesario. (Solo necesita 1 para pol. Fiscal y 1 para Monetaria).

Respecto de la política monetaria:

Hay 3 posibles instrumentos de la política monetaria:
1) La Oferta Monetaria como instrumento de política

El primer instrumento de política monetaria que utiliza el Banco Central para aumentar o disminuir la oferta monetaria, es la Operación de Mercado Abierto. El Banco Central hace estas operaciones mediante la compra o venta de activos financieros o bonos a los bancos comerciales. En esta, la oferta monetaria es una variable endógena. 


2) El coeficiente legal de encaje como instrumento de política

El Banco Central puede incrementar la cantidad de dinero que tienen los bancos para realizar préstamos mediante la disminución de la tasa de encaje, porque ello implica que los bancos comerciales deben tener menos dinero en forma de reservas en el balance del Banco Central. En otras palabras, la disminución de la tasa de encaje aumenta laposibilidad de creación de dinero bancario porque aumenta el multiplicador de dinero bancario; y, este aumento del dinero bancario implica un aumento de la oferta monetaria.
Por otra parte, cuando el Banco Central aumenta el coeficiente de encaje, los bancos comerciales deben tener una mayor proporción de depósitos en forma de reservas, lo que reduce el multiplicador bancario y, por tanto, origina una disminución de la oferta monetaria.

3) La tasa de interés como instrumento de política

Desde los años 1990 los Bancos Centrales optaron por un nuevo esquema de política monetaria. La tasa de interés se convirtió en instrumento de política y la oferta monetaria pasó a ser una variable endógena. La tasa de política es la tasa de interés de referencia de la política monetaria.
El BCR, disminuyendo la tasa de interés de referencia, aumenta la cantidad de dinero prestada a los bancos comerciales, que aumenta la política monetaria y con ello la oferta monetaria. Por otra parte, si aumenta la tasa de interés de referencia, disminuye la cantidad de dinero prestada (el crédito) a los bancos comerciales, lo que reduce la base monetaria y con ello la oferta monetaria.

Curva IS

```python
#--------------------------------------------------

# Curva IS

r_size = 100 

Co = 13
b = 0.7
t = 0.25
Io = 6.4
Go = 5
h = 0.3
Xo = 2
m = 0.4

r = np.arange(r_size)

# Ecuación

def IS_CB(Co, b, t, Io, h, Xo, m, Go):
    IS_CB = (1/(1 - (b - m)*(1 - t)))*(Co + Io + Go + Xo - h*r)
    return IS_CB

IS_CB = IS_CB(Co, b, t, Io, h, Xo, m, Go)
```

Curva LM

```python
#--------------------------------------------------
# Curva LM
    
# Parámetros

Y_size = 100

P = 3.5
k = 1.76
j = 3
r = 15

Y = np.arange(Y_size)

# Ecuación

def LM_CB(Y, P, k, j, r):
    LM_CB = P*(k*Y - j*r)
    return LM_CB

LM_CB = LM_CB(Y, P, k, j, r)
```

```python
# líneas punteadas autómaticas

    # definir la función line_intersection
def line_intersection(line1, line2):
    xdiff = (line1[0][0] - line1[1][0], line2[0][0] - line2[1][0])
    ydiff = (line1[0][1] - line1[1][1], line2[0][1] - line2[1][1])

    def det(a, b):
        return a[0] * b[1] - a[1] * b[0]

    div = det(xdiff, ydiff)
    if div == 0:
       raise Exception('lines do not intersect')

    d = (det(*line1), det(*line2))
    x = det(d, xdiff) / div
    y = det(d, ydiff) / div
    return x, y
```

```python
r_ec = np.full((100), 15)

# coordenadas de las curvas (x,y)
A = [IS_CB[0], Y[0]] # DA, coordenada inicio
B = [IS_CB[-1], Y[-1]] # DA, coordenada fin

C = [r_ec[0], Y[0]] # L_45, coordenada inicio
D = [r_ec[-1], Y[-1]] # L_45, coordenada fin

# creación de intersección

intersec_1 = line_intersection((A, B), (C, D))
intersec_1 # (y,x)
```
