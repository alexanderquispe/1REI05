## a. El Modelo Ingreso-Gasto: la Curva IS

###### Utilizando las ecuaciones derive paso a paso la curva IS matemáticamente (a partir de la condición de equilibrio Y = DA). En el modelo Ingreso-Gasto el equilibrio entre el Ingreso y la DA representa el equilibrio en el mercado de bienes.

En primer lugar, establecer el equilibrio entre el ingreso (Y) y la demanda agregada (DA)

$$ Y = C + I + G + (X - M) $$

Para poder obtener un real equilibrio entre ahorro e inversión, se debe restar la tributación en ambos lados de la ecuación. Además, se puede reemplazar la diferencia del ingreso y la tributación (Y-T) por el ingreso disponible (Yd)

$$ Y - T = C + I + G + X - M - T $$

$$ Y^d = C + I + G + X - M - T $$

El siguiente paso es despejar la variable inversión y reagrupar la ecuación. 

$$ Y^d - C - G - X + M = I $$

$$ (Y^d - C) + (T - G) + (M - X) = I $$

De esta manera, en la ecuación es posible visualizar:

* $(Y^d - C)$ como el ahorro privado
* $(T - G)$ como el ahorro del gobierno
* $(M - X)$ como el ahorro externo

###### Utilizando la función de equilibrio IS donde r está en función de Y, encuentre $ \frac{∆r}{∆y} $  que representa la pendiente y qué signo tiene?

Para obtener la ecuación en función a la tasa de interés (r), el siguiente paso es reemplazar los valores.

$$ [Y^d - (C_0 + bY^d)] + [tY - G_0] + [mY^d - X_0] = I_0 - hr $$

Tomando en cuenta que:

$$ Y^d = (1 - t)Y $$

Se reemplaza y se obtiene la siguiente ecuación:

$$ [1 - (b - m)(1 - t)]Y - (C_0 + G_0 + X_0) = I_0 - hr $$

Con el objetivo de expresar la ecuación de manera que r se encuentre en función de Y, se reordena

$$ hr = (C_0 + G_0 + I_0 + X_0) - [1 - (b - m)(1 - t)]Y $$

$$ r = \frac{1}{h}(C_0 + G_0 + I_0 + X_0) - [\frac{1 - (b - m)(1 - t)}{h}]Y $$

Simplificando:

  $$ r = \frac{B_0}{h} - \frac{B_1}{h} Y $$

Donde:
$$ B_0 = (C_0 + G_0 + I_0 + X_0) $$
$$ B_1 = [1 - (b - m)(1 - t)] $$

Entonces:

$$ \frac{∆r}{∆y} = - \frac{B_1}{h} $$

$$ \frac{∆r}{∆y} = - \frac{[1 - (b - m)(1 - t)]}{h} $$

Por lo tanto, se trata de una pendiente de signo negativo.

###### Lea la sección 4.4 del material de enseñanza y explique cómo se deriva la curva IS a partir del equilibrio $ Y = DA $  y grafique:

El equilibrio Y = DA es un equilibrio en el mercado de bienes, lo que indica que todo lo producido en la economía de la comunidad es consumido por la misma. La curva IS representa la relación entre la tasa de interés referencial y el ingreso en una situación de equilibrio, y se obtiene derivando todas las partes de la ecuación de Demanda Agregada, incluyendo la inversión que contempla la sensibilidad a la tasa de interés.

```python
#Parámetros derivación Y = DA

Y_size = 100 

Co = 30
Io = 20
Go = 50
Xo = 4
h = 0.3
b = 0.5
m = 0.4
t = 0.6

Y = np.arange(Y_size)

def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)

a = 2.5 

def L_45(a, Y):
    L_45 = a*Y
    return L_45

L_45 = L_45(a, Y)

Go = 70

def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_G = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)
```

```python
#Parámetros Curva IS
Y_size = 100 

Co = 30
Io = 20
Go = 50
Xo = 4
h = 0.3
b = 0.5
m = 0.4
t = 0.6

Y = np.arange(Y_size)

def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
```

## b. La Curva IS o el equilibrio Ahorro- Inversión

En primer lugar, asumir la igualdad ahorro-inversión y derivarla a partir de sus componentes: ahorro privado ($S_p$), ahorro de gobierno ($S_g$) y ahorro externo ($S_e$). Luego, reemplazar los valores:

$$ S = I $$

$$ S_p + S_g + S_e = I $$

$$ (Y^d - C) + (T - G) + (M - X) = I $$

Luego, reemplazar los valores con las ecuaciones que representan a cada componente, incluida la inversión.

$$ [Y^d - (C_0 + bY^d)] + [tY - G_0] + [mY^d - X_0] = I_0 - hr $$

Factorizar:

$$ [1 - (b - m)(1 - t)]Y - (C_0 + G_0 + X_0) = I_0 - hr $$

Reordenar la ecuación de manera que todo se encuentre en función a la tasa de interés (r):

$$ hr = (C_0 + G_0 + I_0 + X_0) - [1 - (b - m)(1 - t)]Y $$

$$ r = \frac{1}{h}(C_0 + G_0 + I_0 + X_0) - [\frac{1 - (b - m)(1 - t)}{h}]Y $$

Simplificar:

  $$ r = \frac{B_0}{h} - \frac{B_1}{h} Y $$

Donde:
$$ B_0 = (C_0 + G_0 + I_0 + X_0) $$
$$ B_1 = [1 - (b - m)(1 - t)] $$

```python
# Parámetros

Y_size = 100 

Co = 30
Io = 20
Go = 50
Xo = 4
h = 0.3
b = 0.5
m = 0.4
t = 0.6

Y = np.arange(Y_size)


# Ecuación 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
```

```python
# Curva IS

y_max = np.max(r)
fig, ax = plt.subplots(figsize=(10, 8))

ax.plot(Y, r, label = "IS", color = "magenta")

ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

ax.set(title = "Curva IS en el Equilibrio en el Mercado de Bienes", xlabel= 'Y', ylabel= 'r')
ax.legend()

plt.show()
```

En la gráfica, se observan tres interceptos entre el eje $r$ (tasa de interés de referencia) y el eje $Y$ (ingresos). 

El primero es el punto $A$, que representa la intersección $(Y_A, r_A)$. Este es el único de los tres pares que es atravesado por la curva IS en equilibrio del mercado de bienes. En esta situación, hay un equilibrio entre el ahorro y la inversión $(I_A = S_A)$.

El segundo es el punto $B$, que es la intersección $(Y_B, r_A)$. Este es una situación de exceso de oferta. El ingreso aumenta, pese a que se mantenga la tasa de interés de referencia (r) y con ello la inversión. Por lo tanto, el ahorro también aumenta, lo cual rompe con la situación de equilibrio $(Y_B > Y_A)$. En este caso, la demanda es insuficiente para satisfacer la creciente oferta debido al aumento de los ingresos.

En el punto $C$, conformado por la intersección $(Y_C, r_C)$, estamos frente a una situación de exceso de demanda. No hay cambios en la tasa de interés, por lo que la inversión se mantiene. No obstante, una disminución en el ingreso generará que disminuya también el ahorro y se rompa nuevamente con el equilibrio $(Y_C < Y_A)$. En este caso, el ingreso es insuficiente para satisfacer la demanda.

Analice una Política Fiscal Contractiva con caída del Gasto del Gobierno $∆G < 0$. El análisis debe ser intuitivo y gráfico.

* Análisis intuitivo
$$G↓ → DA↓ → DA<Y → Y↓$$
Una caída en el gasto de gobierno generará una caída de la demanda agregada, teniendo en cuenta que $ Y = DA =  C + I + G + (X – M)$. Si el ingreso ($Y$) se mantiene igual, se generará inevitablemente un desequilibrio en el mercado de bienes $DA<Y$, por lo tanto, el ingreso se disminuirá también y se volverá al equilibrio. Por lo cual, se da un desplazamiento de la curva a la izquierda.

* Análisis gráfico

```python
#IS
Y_size = 100 

Co = 30
Io = 20
Go = 50
Xo = 4
h = 0.3
b = 0.5
m = 0.4
t = 0.6

Y = np.arange(Y_size)

def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)

#IS_G
Go = 30

def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r_G = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
```

```python
#Gráfico
y_max = np.max(r)
fig, ax = plt.subplots(figsize=(10, 8))

ax.plot(Y, r, label = "IS", color = "red") #IS
ax.plot(Y, r_G, label = "IS_G", color = "pink", linestyle = 'dashed') #IS_G

plt.text(47, 162, '∆Go', fontsize=12, color='black')
plt.text(49, 159, '←', fontsize=15, color='grey')

ax.set(title = "Reducción en el Gasto de Gobierno $(G_0)$", xlabel= 'Y', ylabel= 'r')
ax.legend()

plt.show()
```

Analice una caída de la Propensión Marginal a Consumir $∆b < 0$ . El análisis debe ser intuitivo y gráfico.
