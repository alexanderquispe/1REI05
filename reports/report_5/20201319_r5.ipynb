### EL MODELO DA - OA
**1. Encuentre las ecuaciones de Ingreso $(Y^e)$, tasa de interés $(r^e)$, $(P^e)$ de equilibrio (Escriba paso a paso la derivación de estas ecuaciones).**

Recordamos la ecuación de la demanda agregada $(DA)$: 

$$ P = \frac {hM^s_o + jBo}{h}- \frac{jB_1 + hk}{h}Y $$

Y, al mismo tiempo, la ecuación de la oferta agregada $(OA)$:

$$P = P^e + θ(Y - \bar{Y})$$

Hallamos $(Y^e)$ igualando ambas ecuaciones: 

$$\frac {hM^s_o + jBo}{h}- \frac{jB_1 + hk}{h}Y = P^e + θ(Y - \bar{Y}) $$

$$Y^e = [\frac{1}{θ + (\frac {jB1 +hk}{h})}]*[(\frac{hM^s_o + jB_o}{h} - P^e + θ\bar{Y})]$$

Hallamos $(P^e)$ reemplazando en la ecuación de la $OA$:

$$P^e = P^e + θ(Y^e - \bar{Y})$$

$$P^e_q = P^e + θ([\frac{1}{θ + (\frac {jB1 +hk}{h})}]*[(\frac{hM^s_o + jB_o}{h} - P^e + θ\bar{Y})]- \bar{Y})$$


Finalmente, para hallar la ecuación de la tasa de interés de equilibrio $(r)$ se reemplaza $P^e_q$:

$$r = \frac{kB_o}{kh+jB_1} - (\frac{B_1}{kh + jB_1}) * (M^s_o -P^e_q)$$

$$r^e = \frac{kB_o}{kh+jB_1} - (\frac{B_1}{kh + jB_1}) * (M^s_o - P^e + θ([\frac{1}{θ + (\frac {jB1 +hk}{h})}]*[(\frac{hM^s_o + jB_o}{h} - P^e + θ\bar{Y})]- \bar{Y}) )$$

**2. Grafique el equilibrio simultáneo en el modelo DA-OA.**

```python
#1--------------------------
    # Demanda Agregada
    
# Parámetros

Y_size = 100 

Co = 20
Io = 35
Go = 40
Xo = 3
h = 0.6
b = 0.3
m = 0.4
t = 0.8

k = 3
j = 2                
Ms = 300             
P  = 7  

Y = np.arange(Y_size)


# Ecuación

B0 = Co + Io + Go + Xo
B1 = 1 - (b-m)*(1-t)

def P_AD(h, Ms, j, B0, B1, k, Y):
    P_AD = ((h*Ms + j*B0)/h) - (Y*(j*B1 + h*k)/h)
    return P_AD

P_AD = P_AD(h, Ms, j, B0, B1, k, Y)

#2--------------------------
    # Oferta Agregada
    
# Parámetros

Y_size = 100

Pe = 200 
θ = 5
_Y = 30   

Y = np.arange(Y_size)


# Ecuación

def P_AS(Pe, _Y, Y, θ):
    P_AS = Pe + θ*(Y-_Y)
    return P_AS

P_AS = P_AS(Pe, _Y, Y, θ)
```

```python
# Gráfico del modelo DA-OA

# Dimensiones del gráfico
y_max = np.max(P)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar 
ax.plot(Y, P_AD, label = "DA", color = "#e63946") #DA
ax.plot(Y, P_AS, label = "OA", color = "#457b9d") #OA

# Líneas punteadas
plt.axhline(y=300.5, xmin= 0, xmax= 0.52, linestyle = ":", color = "grey")
plt.axvline(x=50.6, ymin= 0, ymax= 0.49, linestyle = ":", color = "grey")

# Texto agregado
plt.text(0, 320, '$P_0$', fontsize = 12, color = 'black')
plt.text(53, 25, '$Y_0$', fontsize = 12, color = 'black')
plt.text(50, 320, '$E_0$', fontsize = 12, color = 'black')


# Eliminar valores de ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

# Título, ejes y leyenda
ax.set(title="DA-OA", xlabel= r'Y', ylabel= r'P')
ax.legend()

plt.show()
```

**Matemática**

```python
# nombrar variables como símbolos de IS
Co, Io, Go, Xo, h, r, b, m, t, beta_0, beta_1  = symbols('Co Io Go Xo h r b m t beta_0, beta_1')

# nombrar variables como símbolos de LM 
k, j, Ms, P, Y = symbols('k j Ms P Y')

# nombrar variables como símbolos para curva de oferta agregada
Pe, _Y, Y, θ = symbols('Pe, _Y, Y, θ')

# Beta_0 y beta_1
beta_0 = (Co + Io + Go + Xo)
beta_1 = (1-(b-m)*(1-t))

# Producto de equilibrio en el modelo DA-OA
Y_eq = ( (1)/(θ + ( (j*beta_1+h*k)/h) ) )*( ( (h*Ms+j*beta_0)/h ) - Pe + θ*_Y )

# Precio de equilibrio en el modelo DA-OA 
P_eq = Pe + θ*(Y_eq - _Y)

# Tasa de interés de equilibrio en el modelo DA-OA
r_eq = (j*beta_0)/(k*h + j*beta_1) + ( h / (k*h + j*beta_1) )*(Ms - P_eq)
#((h*Ms+j*beta_0)/h) - ((j*beta_1+h*r)/h)*((P-Pe-θ*_Y)/θ)
```

```python
# nombrar variables como símbolos de IS
Co, Io, Go, Xo, h, r, b, m, t, beta_0, beta_1  = symbols('Co Io Go Xo h r b m t beta_0, beta_1')

# nombrar variables como símbolos de LM 
k, j, Ms, P, Y = symbols('k j Ms P Y')

# nombrar variables como símbolos para curva de oferta agregada
Pe, _Y, Y, θ = symbols('Pe, _Y, Y, θ')

# Beta_0 y beta_1
beta_0 = (Co + Io + Go + Xo)
beta_1 = ( 1-(b-m)*(1-t) )

# Producto de equilibrio en el modelo DA-OA
Y_eq = ( (1)/(θ + ( (j*beta_1+h*k)/h) ) )*( ( (h*Ms+j*beta_0)/h ) - Pe + θ*_Y )

# Precio de equilibrio en el modelo DA-OA 
P_eq = Pe + θ*(Y_eq - _Y)

# Tasa de interés de equilibrio en el modelo DA-OA
r_eq = (j*beta_0)/(k*h + j*beta_1) + ( h / (k*h + j*beta_1) )*(Ms - P_eq)
#((h*Ms+j*beta_0)/h) - ((j*beta_1+h*r)/h)*((P-Pe-θ*_Y)/θ)
```

```python
# Efecto del cambio de Precio esperado sobre la tasa de impuestos en el modelo DA-OA
df_Y_eq_t = diff(Y_eq, t)
print("El Diferencial del Producto con respecto al diferencial de la tasa de impuestos = ", df_Y_eq_t)
print("\n")

# Efecto del cambio de Precio esperado sobre la tasa de impuestos en el modelo DA-OA
df_r_eq_t = diff(r_eq, t)
print("El Diferencial de la tasa de interés con respecto al diferencial de la tasa de impuestos = ", df_r_eq_t)
print("\n")

# Efecto del cambio de Precio esperado sobre la tasa de impuestos en el modelo DA-OA
df_P_eq_t = diff(P_eq, t)
print("El Diferencial del nivel de precios con respecto al diferencial de la tasa de impuestos = ", df_P_eq_t)
```

```python
# IS-LM

#1--------------------------------------------------
    # Curva IS ORIGINAL

# Parámetros

Y_size = 100 

Co = 35
Io = 40
Go = 50
Xo = 2
h = 0.8
b = 0.7
m = 0.3
t = 0.1

Y = np.arange(Y_size)


# Ecuación 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)


#2--------------------------------------------------
    # Curva LM ORIGINAL

# Parámetros

Y_size = 100

k = 2
j = 1                
Ms = 200             
P  = 8           

Y = np.arange(Y_size)

# Ecuación

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)

#--------------------------------------------------
    # NUEVA curva IS: incremento en la tasa de impuestos (t)

# Definir SOLO el parámetro cambiado
t = 0.8

# Generar la ecuación con el nuevo parámetro
def r_IS_t(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r_t = r_IS_t(b, m, t, Co, Io, Go, Xo, h, Y)


#DA-OA

#1--------------------------
    # Demanda Agregada ORGINAL
    
# Parámetros

Y_size = 100 

Co = 35
Io = 40
Go = 50
Xo = 2
h = 0.8
b = 0.7
m = 0.2
t = 0.1

k = 2
j = 1                
Ms = 200             
P  = 8  

Y = np.arange(Y_size)


# Ecuación

B0 = Co + Io + Go + Xo
B1 = 1 - (b-m)*(1-t)

def P_AD(h, Ms, j, B0, B1, k, Y):
    P_AD = ((h*Ms + j*B0)/h) - (Y*(j*B1 + h*k)/h)
    return P_AD

P_AD = P_AD(h, Ms, j, B0, B1, k, Y)

#--------------------------------------------------
    # NUEVA Demanda Agregada

# Definir SOLO el parámetro cambiado

t = 0.8

# Generar la ecuación con el nuevo parámetro

B0 = Co + Io + Go + Xo
B1 = 1 - (b-m)*(1-t)

def P_AD_t(h, Ms, j, B0, B1, k, Y):
    P_AD = ((h*Ms + j*B0)/h) - (Y*(j*B1 + h*k)/h)
    return P_AD

P_t = P_AD_t(h, Ms, j, B0, B1, k, Y)


#2--------------------------
    # Oferta Agregada ORIGINAL
    
# Parámetros

Y_size = 100

Pe = 70
θ = 3
_Y = 20  

Y = np.arange(Y_size)

# Ecuación

def P_AS(Pe, _Y, Y, θ):
    P_AS = Pe + θ*(Y-_Y)
    return P_AS

P_AS = P_AS(Pe, _Y, Y, θ)

# Dos gráficos en un solo cuadro
fig, (ax1, ax2) = plt.subplots(2, figsize=(8, 16)) 

#---------------------------------
    # Gráfico 1: IS-LM

ax1.plot(Y, i, label="LM", color = "#e9c46a")  #LM 
ax1.plot(Y, r, label = "IS", color = "#2a9d8f") #IS
ax1.plot(Y, r_t, label="IS_t", color = "#2a9d8f", linestyle ='dashed')  #IS

ax1.axvline(x=66,  ymin= 0, ymax= 0.65, linestyle = ":", color = "grey")
ax1.axvline(x=58,  ymin= 0, ymax= 0.58, linestyle = ":", color = "grey")
ax1.axhline(y=105,  xmin= 0, xmax= 0.65, linestyle = ":", color = "grey")
ax1.axhline(y=92,  xmin= 0, xmax= 0.58, linestyle = ":", color = "grey")

ax1.text(75, 82, '∆$t$', fontsize=12, color='black')
ax1.text(79, 82, '↓', fontsize=15, color='grey')
ax1.text(60, 0, '←', fontsize=15, color='grey')
ax1.text(0, 95, '↓', fontsize=15, color='grey')
ax1.text(50, 0, '$Y_1$', fontsize=12, color='black')
ax1.text(70, 0, '$Y_0$', fontsize=12, color='black')
ax1.text(0, 82, '$r_1$', fontsize=12, color='black')
ax1.text(0, 110, '$r_0$', fontsize=12, color='black')


ax1.set(title="Efectos de un aumento en la tasa de impuestos", xlabel= r'Y', ylabel= r'r')
ax1.legend()


#---------------------------------
    # Gráfico 2: DA-OA

ax2.plot(Y, P_AD, label = "DA", color = "#e76f51") #DA
ax2.plot(Y, P_t, label = "DA_t", color = "#e76f51", linestyle = 'dashed') #DA_t
ax2.plot(Y, P_AS, label = "OA", color = "#264653") #OA

ax2.axvline(x=61,  ymin= 0, ymax= 0.53, linestyle = ":", color = "grey")
ax2.axvline(x=56,  ymin= 0, ymax= 0.49, linestyle = ":", color = "grey")
ax2.axhline(y=195,  xmin= 0, xmax= 0.6, linestyle = ":", color = "grey")
ax2.axhline(y=180,  xmin= 0, xmax= 0.56, linestyle = ":", color = "grey")

ax2.text(57, 30, '←', fontsize=15, color='grey')
ax2.text(86, 105, '↓', fontsize=15, color='grey')
ax2.text(83, 105, '∆$t$', fontsize=12, color='black')
ax2.text(0, 183, '↓', fontsize=15, color='grey')

ax2.text(50, 0, '$Y_1$', fontsize=12, color='black')
ax2.text(63, 0, '$Y_0$', fontsize=12, color='black')
ax2.text(0, 160, '$P_1$', fontsize=12, color='black')
ax2.text(0, 205, '$P_0$', fontsize=12, color='black')

ax2.set(xlabel= r'Y', ylabel= r'P')
ax2.set(title="DA-OA", xlabel= r'Y', ylabel= r'P')
ax2.legend()

plt.show
```
