# 1.1:
Existen tres posibles instrumentos de la política monetaria que puede usar el Banco Central: 


* La Oferta Monetaria: para aumentar o disminuir la oferta monetaria se realiza la Operación de Mercado Abierto. El Banco Central, hace estas operaciones mediante la compra o venta de activos financieros o bonos a los bancos comerciales. 
   * Politica monetaria expansiva: registro de un incremento de los activos del Banco Central, lo cual origina un aumento de los depósitos de los Bancos Comerciales hacia el Banco Central.
   * Politica monetaria contractiva: registro de una disimnución de los activos del Banco Central, lo cual origina una disminución de los depósitos de los Bancos Comerciales hacia el Banco Central. 
   

* El coeficiente legal de encaje:
   * Politica monetaria expansiva: el Banco Central puede incrementar la cantidad de dinero que tienen los bancos para realizar préstamos mediante una reducción de la tasa de encaje impuesta por el Banco Central. Es decir, la disminución de la tasa de encaje aumenta la posibilidad de creación de dinero bancario y esto, produce un aumento de la oferta monetaria.
   * Politica monetaria contractiva: el Banco Central aumenta el coeficiente de encaje, por ende, los bancos comerciales deben tener una mayor proporción de depósitos en forma de reservas,y esto pruduce una disminución de la oferta monetaria.
   

* La tasa de interés: 
   * Politica monetaria expansiva: El Banco Central hace política monetaria expansiva reduciendo su tasa de interés de referencia, por tanto, aumenta la cantidad de dinero prestada a los bancos comerciales lo que aumenta la base monetaria y a su vez la oferta monetaria.
   * Politica monetaria contractiva: El Banco Central hace política monetaria contractiva aumentando su tasa de interés de referencia, por tanto, disminuye la cantidad de dinero prestada (el crédito) a los bancos comerciales lo que reduce la base monetaria y a su vez la oferta monetaria.


# 1.2:

Componentes: 
* $ M^s $ : Oferta de Dinero (variable exógena y nominal)
* $ P $ : precio

Entonces: 
* Para expresar la variable nominal en términos reales o a precios constantes, se debe dividir la oferta nominal de dinero entre el nivel general de precios de la economía. 

Por tanto, tendremos lo siguiente: 

$$ \frac{M^s}{P}= \frac{M_0^s}{P} $$





# 1.3:

Para poder llegar a la ecuación de la demanda real de dinero, se debe tener presente 3 motivos: 
* Transacción: se demanda dinero por su función de medio de intercambio, para realizar transacciones. Por esta razón, la demanda de dinero por el motivo transacción depende directamente del Ingreso (Y).
* Precaución: se demanda dinero por el motivo precaución para honrar estos servicios de la deuda. Por esta razón, la demanda de dinero por el motivo transacción depende directamente del Ingreso (Y).

De estos dos motivos tenemos: $$ L_1 = kY $$ 

* Especulación: el activo financiero dinero compite con el activo financiero no monetario bono en la función de reserva de valor.Por lo tanto, esta demanda de dinero dependerá negativa o inversamente de la tasa de interés de los bonos.

De este tercer motivo tenemos: $$ L_2 = -ji $$


Por tanto, tendremos lo siguiente: 
$$ L=L_1+L_2 $$

Reemplazando tenemos: 

$$ L = kY - ji $$

$$ M^d = kY - ji $$


El parámetro $ “k” $ indica la sensibilidad de la demanda de dinero ante variaciones del ingreso $“Y”$, y el parámetro $“j”$ indica cuán sensible es la demanda de dinero ante las variaciones de la tasa de interés nominal de los bonos, $“i”$. 


# 1.4:

Asumiendo que no hay inflación, se puede asumir que $i=r$
- Sabemos que $i$ es la tasa de interés nominal
- Sabemos que $r=i-π$
Asumiendo lo anterior, podemos decir que la tasa de interés nominal es igual a la real, por tanto se tiene, reemplazando $i$ por $r$: 
$$ M^d=kY-jr $$
Despejando r:
$$ jr =kY-M^d $$

$$ r= \frac{kY}{j}- \frac{M^d}{j} $$

# 2.3:

- Intuición: Si la oferta monetaria disminuye, el déficit de oferta que se produce en el mercado, genera que haya un aumento de la tasa de interés y a su vez, disminuye la demanda, para con ello, se restablezca el equilibrio en el mercado. Al disminuir la cantidad de dinero disminuye de $M_0^s$ a $M_1^s$. Es por ello, que la recta de la oferta real de dinero se desplaza a la izquierda ($\frac{M_1^s}{P_0}$). Adicional a ello, la tasa de interés de equilibrio aumenta de $r_0^e$ a $r_1^e$.

$$ ↓M_0^s → ↓\frac{M^s}{P}$$

Esto produce que haya un desequilibrio porque la cantidad de dinero es menor a la cantidad de demanda del mercado monetario:
$\frac{M^s}{P} < M^d $

Sin embargo, no debe suceder, por tanto, para que vuelva a exister equilibrio, la tasa de interés debe subir: 
$\frac{M^s}{P} < M^d → ↑r$

Lo cual, da como resultado que: $\frac{M^s}{P} =M^d $

```python
# Parametros

r_size = 100

k = 0.6
j = 0.3                
P  = 10 
Y = 35

r = np.arange(r_size)

    # Ecuación

def Ms_MD(k, j, P, r, Y):
    Ms_MD = P*(k*Y - j*r)
    return Ms_MD

Ms_MD = Ms_MD(k, j, P, r, Y)


  # Equilibrio en el mercado de dinero

fig, ax1 = plt.subplots(figsize=(10, 8))

ax1.set(title="Equilibrio de mercado de Dinero")
ax1.plot(Ms_MD, label= '$L_0$', color = '#CD5C5C')

ax1.axvline(x = 45,  ymin= 0, ymax= 1, color = "green")
ax1.axvline(x = 30,  ymin= 0, ymax= 1, color = "green",linestyle = 'dashed')
ax1.axhline(y=74, xmin= 0, xmax= 0.46, linestyle = ":", color = "yellow")
ax1.axhline(y=120, xmin= 0, xmax= 0.32, linestyle = ":", color = "yellow")
ax1.text(-10, 200, "$r$", fontsize = 14, color = 'black')
ax1.text(100, -120, "$M^s/M^d$", fontsize = 14, color = 'black')

ax1.text(-8, 75, "$r_0^e$", fontsize = 12, color = 'black')
ax1.text(-8, 120, "$r_1^e$", fontsize = 12, color = 'black')
ax1.text(41,-116, "$M_0^s/P_0$", fontsize = 12, color = 'black')
ax1.text(46,74, "$E_0$", fontsize = 10, color = 'black')
ax1.text(28, -116, "$M_1^s/P_0$", fontsize = 12, color = 'black')
ax1.text(31,120, "$E_1$", fontsize = 10, color = 'black')
ax1.text(99, -90, "$L(Y_0)$", fontsize = 10, color = 'black')

ax1.yaxis.set_major_locator(plt.NullLocator())   
ax1.xaxis.set_major_locator(plt.NullLocator())

ax1.legend()

plt.show()
```

Tenemos:

* $ M^d = kY - jr $
* $ M^S = \frac{M_0^s}{P} $

Entonces:

 $$ M^d = M^S $$
 
$$ kY-jr=\frac{M_0^s}{P} $$ 

$$ kY - \frac{M_0^s}{P} = jr $$ 

$$ \frac{kY}{j}-\frac{M_0^s}{Pj} = r $$





Por tanto, la tasa de interés que equilibra el mercado monetario será igual a: 
$ r=-\frac{1}{j}\frac{M_0^s}{P_0}+\frac{k}{j}Y $

- Intuición: Si $k$ que es la sensibilidad aumenta, nos da una certeza de cuanto será la variación del ingreso. 

```python
# Parámetros

Y_size = 100
k = 2
j = 1                
Ms = 200            
P  = 20               
Y = np.arange(Y_size)
# Ecuación

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i_LMk0 = i_LM( k, j, Ms, P, Y)


# NUEVA curva LM

# Definir SOLO el parámetro cambiado
k = 5

i_LMk1 = i_LM( k, j, Ms, P, Y)


# Dimensiones del gráfico
y_max = np.max(i_LMk1)
v = [0, Y_size, 0, y_max]   
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, i_LMk0, label="$LM(k=2)$", color = 'black')
ax.plot(Y, i_LMk1, label="$LM(k=5)$", color = 'green', linestyle = 'dashed')

# Texto agregado
plt.text(47, 105, '∆$K$', fontsize=15, color='black')
plt.text(47, 150, '↑', fontsize=20, color='black')

# Título y leyenda
ax.set(title = "Aumento en la sensibilidad de la demanda", xlabel=r'Y', ylabel=r'r')
ax.legend()


plt.show()
```
