Una política fiscal de contracción del gasto fiscal, una variable endógnea afecta de la siguiente manera

$$Go↓=DA↓ → DA<Y→Y↓$$
$$Y↓=M^d↓ → M^d↓→r↑$$
                     
Por lo tanto en nuestro modelo DA-OA:

$$P= Pe + θ (Y-\bar{Y})$$
$$Y↓ → θ (Y-\bar{Y})↓  →   Pe + θ (Y-\bar{Y})< P → P↓$$

    c)Analice los efectos sobre las variables endógenas Y, P y r de un incremento de la tasa de impuestos. . El análisis debe ser intuitivo, matemático y gráfico. En una figura, se debe que usar los ejes r e Y (modelo IS-LM), y en la otra, los ejes P, Y (modelo DA-OA).

Una política fiscal con incremento de la tasa de impuesto,  como una variable endógnea, afecta de la siguiente manera

A el modelo de IS-LM

$$ t↑ = Co↓ = DA↓ → DA<Y = Y↓  → DA=Y $$
$$ Y↓ →Md↓Md<Ms→r↓ $$

Y en nuestro modelo DA-OA 
$$ Y↓ → θ(Y-\bar{Y})↓ → P↓$$

Partiendo desde nuestro análisis matemático

Modelo IS-LM:

```python
Co, Io, Go, Xo, h, r, b, m, t, beta_0, beta_1  = symbols('Co Io Go Xo h r b m t beta_0, beta_1')

k, j, Ms, P, Y = symbols('k j Ms P Y')

beta_0 = (Co + Io + Go + Xo)
beta_1 = ( 1-(b-m)*(1-t) )

Y_eq = (k*beta_0)/(k*h + j*beta_1) - ( beta_1 / (k*h + j*beta_1) )*(Ms/P)
r_eq = (j*beta_0)/(k*h + j*beta_1) + ( h / (k*h + j*beta_1) )*(Ms/P)
```

```python
df_Y_eq_t = diff(Y_eq, t)
print("El Diferencial del Producto con respecto al diferencial de la tasa de interés = ", df_Y_eq_t)
```

La pendiente es representada por el diferencial de $ ∆t/∆Y $: $\frac{Ms*j*(b - m)*(-(1 - t)*(b - m) + 1)/(P*(h*k + j*(-(1 - t)*(b - m) + 1))**2) - Ms*(b - m)/(P*(h*k + j*(-(1 - t)*(b - m) + 1))) - j*k*(b - m)*(Co + Go + Io + Xo)}{(h*k + j*(-(1 - t)*(b - m) + 1))**2} $ y es negativo

```python
df_r_eq_t = diff(r_eq, t)
print("El Diferencial de la tasa de interes con respecto al diferencial de precios = ", df_r_eq_t)
```

La pendiente es representada por el diferencial de $ ∆t/∆r $: $\frac{-Ms*h*j*(b - m)/(P*(h*k + j*(-(1 - t)*(b - m) + 1))**2) - j**2*(b - m)*(Co + Go + Io + Xo)}{(h*k + j*(-(1 - t)*(b - m) + 1))**2} $ y es positivo

Modelo de DA-OA

```python
h, Ms, j, beta_0, beta_1, k, Y, Pe, _Y, θ  = symbols('h, Ms, j, beta_0, beta_1, k, Y, Pe, _Y, θ ')

# Beta_0 y beta_1
beta_0 = (Co + Io + Go + Xo)
beta_1 = ( 1-(b-m)*(1-t) )

# Producto de equilibrio y la tasa de interes de equilibrio en el modelo IS-LM
Y_eq = (((h*Ms+j*beta_0)/h) - (Pe + θ*_Y))/(θ - ((j*beta_1+h*r)/h))
P_eq = ((h*Ms+j*beta_0)/h) - (((j*beta_1+h*r)/h)*((P-Pe-θ*_Y)/θ))
```

```python
df_Y_eq_t = diff(Y_eq, t)
print("El Diferencial del Producto con respecto al diferencial de la masa monetaria = ", df_Y_eq_t)
```

La pendiente es representada por el diferencial de $ ∆t/∆Y $: $ j*(b - m)*(-Pe - _Y*θ + \frac{(Ms*h + j*(Co + Go + Io + Xo))/h)/(h*(θ - (h*r + j*(-(1 - t)*(b - m) + 1))}{h})**2) $ y es negativo

```python
df_P_eq_t = diff(P_eq, t)
print("El Diferencial del nivel de precios con respecto al diferencial del precio esperado = ", df_P_eq_t)
```

La pendiente es representada por el diferencial de $ ∆Go/∆P $: $\frac{-j*(b - m)*(P - Pe - _Y*θ)}{(h*θ)} $ y es negativo
