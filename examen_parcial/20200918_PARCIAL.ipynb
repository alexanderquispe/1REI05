(2 puntos) Defina el Multiplicador Keynesiano y dé un ejemplo de cómo influye este en el modelo Ingreso-Gasto Keynesiano

El multiplicador postulado por keynes asevera que un aumento del gasto en inversión genera un aumento multiplicado de la producción y de la renta, ello indica que la inversión puede aumentar sin que el ahorro lo haga previamente. Esto sucede esencialmente porque un cambio en la magnitud de cualquiera de los componentes autónomos de la demanda agregada , generan un proceso multiplicador del ingreso hasta converger al nuevo ingreso y producto de equilibrio. Entonces el aumento multiplicador del ingreso resulta de los efectos ocasionados por los cambios de los componentes autónomos de la demanda agregada. También es importante mencionar que el tamaño del multiplicador depende de la magnitud de los parámetros de propensión marginal a ahorrar (b), la propensión marginal a importar (m) y la tasa promedio de impuestos (t). Toda esta idea surge de la idea de una economía cerrada y sin gobierno, con dos sectores de producción (bienes de inversión y bienes de consumo). El proceso empieza con el aumento de la demanda de inversión lo que hace que aumente la producción e ingreso en el sector que produce los bienes de inversión, así aumenta la demanda de consumo y de la producción e ingreso en el sector que produce bienes de consumo. El segundo incremento del ingreso aumenta la demanda de bienes de consumo y por tanto también la producción e ingreso del sector dedicado a los bienes de consumo. Con todo ello podemos evidenciar como influye este multiplicador en el modelo ingreso- gasto, pues, cuando existe un incremento en  la inversión total, el ingreso aumentará en una cantidad que es k veces el incremento de la inversión

(2 puntos) Grafique y explique cúando sucede un exceso y déficit de demanda en el modelo Ingreso-Gasto. Señale las áreas donde ocurre cada caso. Explique cómo se converge al equilibrio a partir de estos dos escenarios.

```python
 #Curva de ingreso 

    # Parámetros
Y_size = 100 

Co = 35
Io = 40
Go = 40
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3
r = 0.9

Y = np.arange(Y_size)

    # Ecuación 
def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)

#--------------------------------------------------
# 45°

a = 2.5 

def L_45(a, Y):
    L_45 = a*Y
    return L_45

L_45 = L_45(a, Y)
```

```python
# Gráfico
y_max = np.max(DA_IS_K)
fig, ax = plt.subplots(figsize=(10, 8))


# Curvas a graficar
ax.plot(Y, DA_IS_K, label = "DA", color = "#3D59AB") #curva ORIGINAL
ax.plot(Y, L_45, color = "#404040") #línea de 45º

# Lineas punteadas
plt.axvline(x = 75, ymin= 0, ymax = 0.75, linestyle = ":", color = "grey")
plt.axvline(x = 56,  ymin= 0, ymax = 0.56, linestyle = ":", color = "grey")
plt.axvline(x = 20,  ymin= 0, ymax = 0.50, linestyle = ":", color = "grey")
plt.axhline(y = 139, xmin= 0, xmax = 0.55, linestyle = ":", color = "grey")

# Texto agregado
plt.text(0, 150, '$DA^e$', fontsize = 11.5, color = 'black')
plt.text(55, 145, '$E$', fontsize = 11.5, color = 'black')
plt.text(75, 150, '$B$', fontsize = 11.5, color = 'black')
plt.text(20, 130, '$A$', fontsize = 11.5, color = 'black')
plt.text(6, 4, '$45°$', fontsize = 11.5, color = 'black')
plt.text(2.5, -3, '$◝$', fontsize = 30, color = '#404040')
plt.text(57, 0, '$Y^e$', fontsize = 12, color = 'black')

# Título y leyenda
ax.set(title = " exceso y déficit de demanda en el modelo Ingreso-Gasto.", xlabel = r'Y', ylabel = r'DA')
ax.legend()

plt.show()
```

El punto A refleja exceso de demanda, donde la demanda agregada es mayor que el nivel de producción. Esto genera una disminución del inventario y aumento de la producción, lo que hace que también aumente el empleo y el ingreso de las familias, distribuyendo los ingresos extras en el consumo de bienes, servicios y ahorro, lo cual lleva nuevamente al equilibrio .

Por otro lado el punto B representa un déficit de demanda que es igual a un exceso de producción, allí aumenta el inventario y las empresas disminuyen su producción y por tanto requieren de menos trabajadores, aumentando el desempleo. Estas disminuciones de producción llevarán de nuevo al equilibrio.

Lo que se evidencia en ambos casos es que la oferta de producción se ajusta a los cambios en la demanda agregada  


(2 puntos) ¿Cuáles son las herramientas de política fiscal y política monetaria? Dea un ejemplo para cada una de ellas dentro del contexto peruano. Coloque su fuente en caso sea necesario. (Solo necesita 1 para pol. Fiscal y 1 para Monetaria).

Las herramientas de política fiscal son el conjunto de acciones gubernamentales en donde, mediante instrumentos discrecionales, se busca modificar los ingresos y gastos, estos serían el gasto público y la tasa impositiva. Estás son aquellas variables exógenas que puede usar el gobierno para aumentar (política fiscal expansiva) o disminuir (política fiscal contractiva) la demanda agregada. Un ejemplo de política fiscal pueden ser los bonos que el estado ha entregado durante la pandemia para reactivar la economía. Estos generan un aumento en el gasto público lo que a su vez genera un aumento en la demanda y producción. Estos bonos, a consecuencia de esto último, generan un aumento en el consumo de los individuos lo que nuevamente aumenta la demanda agregada, cumpliendo así su objetivo.

Mientras tanto, las herramientas de política monetaria son la Operación de Mercado Abierto, el coeficiente legal de encaje y la tasa de interés. Por un lado, la Operación de Mercado Abierto, se refiere a la compra o venta de activos financieros o bonos por parte del Banco Central a los bancos comerciales. Por otro lado, referente al coeficiente legal de encaje, el Banco central puede aumentar o disminuir la cantidad de dinero que tienen los bancos para realizar préstamos mediante el aumento o disminución de la tasa de encaje, pues ello implica que los bancos comerciales deben tener menos o más dinero en forma de reservas en el balance del Banco Central. Finalmente, la tasa de interés es un instrumento usado por los Bancos Centrales gracias al cual la oferta monetaria pasó a ser una variable endógena. Estás son en términos generales  variables exógenas que puede usar el Banco Central para aumentar (política monetaria expansiva) o disminuir (política monetaria contractiva) la oferta o demanda monetaria y que el mercado mantenga su equilibrio. Un ejemplo de esto último es cuando el BCRP aumenta su tasa de interés de referencia en respuesta a los altos niveles de inflación. Esto conlleva a un aumento de las tasas de interés en préstamos que ofrecen los bancos, cajas y financieras. De esta forma, el Banco Central consigue desmotivar a las personas y reducir la demanda monetaria.

Dancourt en "Inflation Targeting in Peru: The Reasons for the Success" concluye que la política macroeconómica del BCR fue exitosa en términos de estabilidad monetaria en diferentes contextos debido a las principales decisiones que tomó en 2002 y 2013: implementar un sistema de metas con una tasa de interés de referencia, como principal instrumento de política monetaria; y mantener suficientes reservas de moneda extranjera, divisas.

Esclarece las medidas a tomar en determinados contextos, en uno favorable se debe subir las tasas de interés y comprar moneda extranjera, mientras que en uno desfavorable debe bajar las tasas de interés y vender moneda extranjera. Asimismo dice que el BCR debe limitar los préstamos de moneda extranjera de los bancos comerciales, estableciendo mayores requerimientos de reserva para la deuda con el fin de desdolarizar las deudas y, en consecuencia, la economía.  Así  limita el endeudamiento externo de bancos comerciales y previene crisis bancarias.

Finalmente, Dancourt concluye que a través de una regla de Taylor para gestionar la tasa de interés con una regla de intervención en el tipo de cambio, dirigida al equilibrio interno; junto con una regla de intervención del tipo de cambio que va contra el viento, dirigida al equilibrio externo, es posible estabilizar los niveles de precios y la actividad económica sin liquidar la diversificación productiva en el largo plazo.
