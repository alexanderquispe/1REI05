1. Defina el Multiplicador Keynesiano y d√© un ejemplo de c√≥mo influye este en el modelo Ingreso-Gasto Keynesiano.

Frente a un cambio en la magnitud de alg√∫n componente aut√≥nomo de la Demanda Agregada (intercepto), el multiplicador Keynesiano representa el proceso multiplicador del ingreso generado hasta converger al nuevo ingreso y producto de equilibrio. As√≠, el aumento multiplicado del ingreso es el resultado de los efectos directos e indirectos ocasionados por el aumento de cualquiera de los componentes aut√≥nomos de la demanda agregada (Jimenez, 2020). Finalmente, este multiplicador es mayor a 0. 

Un ejemplo de c√≥mo este multiplicador puede influir en el modelo Ingreso-Gasto keynesiano a trav√©s de una reducci√≥n en la tasa impositiva (t). En este caso, se estar√≠a reduciendo la magnitud del denominador del multiplicador keynesiano y, por lo tanto, el multiplicador aumenta su tama√±o. En consecuencia, la producci√≥n ser√≠a menor a la demanda agregada (Y<DA), por lo que para volver al equilibrio Y debe aumentar. Por otro lado, este aumento en la tasa impositiva reduce la pendiente de la curva de Demanda Agregada (DA), girando entonces en el sentido de las agujas de un reloj. 


2. Grafique y explique c√∫ando sucede un exceso y d√©ficit de demanda en el modelo Ingreso-Gasto. Se√±ale las √°reas donde ocurre cada caso. Explique c√≥mo se converge al equilibrio a partir de estos dos escenarios.




```python
z_size = 100          
z = np.arange(z_size) 
z 


#Parameters
Y_size = 100 
b = 0.2                 # propensi√≥n marginal a consumir
m = 0.2                 # propensi√≥n a importar
t = 1                   # tasa impositiva
Co = 20                 # consumo aut√≥nomo
Io = 10                 # inversi√≥n aut√≥noma
Go = 8                  # gasto gobierno 
Xo = 2                  # exportaci√≥n 
h =  5                  # constante de decisi√≥n a invertir

Y = np.arange(Y_size)


def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS


def Y_IS(b, m, t, Co, Io, Go, Xo, h, r):
    Y_IS = (Co + Io + Go + Xo - h*r)/(1-(b-m)*(1-t))
    return Y_IS


r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
r




#Graficando: 

y_max = np.max(r)
x_max = Y_IS(b, m, t, Co, Io, Go, Xo, h, 0)


v = [0, x_max, 0, y_max]                        
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="IS SCHEDULE", xlabel=r'Y', ylabel=r'r')
ax.plot(Y, r, "k-")

plt.axvline(x = 8, ymin= 0, ymax = 0.60, linestyle = ":", color = "grey")
plt.axvline(x = 16,  ymin= 0, ymax = 0.60, linestyle = ":", color = "grey")
plt.axvline(x = 24, ymin= 0, ymax = 0.60, linestyle = ":", color = "grey")
plt.axhline(y = 4.8, xmin= 0, xmax= 0.60, linestyle = ":", color = "grey")


plt.text(-2, 4.7, '$r_A$', fontsize = 12, color = 'black')
plt.text(7.2, -0.3, '$Y_C$', fontsize = 15, color = 'black')
plt.text(15.5, -0.3, '$Y_A$', fontsize = 15, color = 'black')  
plt.text(23.3, -0.3, '$Y_B$', fontsize = 15, color = 'black') 
plt.text(7.2, 4.9, '$C$', fontsize = 15, color = 'black')
plt.text(15.5, 4.9, '$A$', fontsize = 15, color = 'black')  
plt.text(23.7, 4.9, '$B$', fontsize = 15, color = 'black') 
plt.text(25, 5.5, 'Deficit  de  demanda', fontsize = 15, color = 'black')  
plt.text(7, 2.5, 'Exceso de demanda', fontsize = 15, color = 'black') 

ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())
    
plt.axis(v)                                    
plt.show()
```

Entonces, podemos observar que todos los puntos de la Curva IS corresponden a pares ordenados (Y, r) que equilibran el mercado de bienes (I=S). En este sentido, el punto A representa el equilibrio. Sin embargo, los puntos fuera de esta curva son de desequilibrio en el mercado (es decir, donde el ahorro y la inversi√≥n no son iguales).

Por otro lado, en el punto B representa un desequilibrio (se encuentra a la derecha de la IS). Dado que la tasa de inter√©s ‚Äúr‚Äù permaneci√≥ constante pero el ingreso aument√≥, este ahora es mayor a la inversi√≥n. Por tanto,  estamos ante un d√©ficit de demanda. En este caso, se tendr√° que disminuir el ingreso para que se nivele con la inversi√≥n y as√≠ volver al equilibrio. 

Finalmente, el punto C representa tambi√©n un desequilibrio (en este caso por ubicarse a la izquierda de la IS). Dado que la tasa de inter√©s se mantuvo constante y el ingreso disminuy√≥, la inversi√≥n ahora es mayor que el nivel de ingreso. Por tanto, estamos ante un exceso de demanda. En este sentido, para volver al equilibrio se deber√° aumentar el ingreso.
