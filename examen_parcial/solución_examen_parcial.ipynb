#### 1. Defina el Multiplicador Keynesiano y d√© un ejemplo de c√≥mo influye este en el modelo Ingreso-Gasto Keynesiano.


Un cambio en la magnitud de cualquiera de los componentes aut√≥nomos de la Demanda Agregada, genera un proceso multiplicador del ingreso hasta converger al nuevo ingreso y producto de equilibrio. El multiplicador keynesiano representa en cu√°nto se va incrementar o disminuir el producto ante shocks en cualquiera de los componente aut√≥nomos de la $DA$, ya sea el Gasto de Gobierno $(G)$, el Consumo $(C)$, la Inversi√≥n $(I)$ o las Exportaciones Netas $(X)$. Adem√°s, el tama√±o del multiplicador $k$ depende de la magnitud de los par√°metros $b$, $m$ y $t$.


#### 2. Grafique y explique c√∫ando sucede un exceso y d√©ficit de demanda en el modelo Ingreso-Gasto. Se√±ale las √°reas donde ocurre cada caso. Explique c√≥mo se converge al equilibrio a partir de estos dos escenarios.

En el siguiente gr√°fico se muestra el punto de equilibrio $E$ del modelo Ingreso-Gasto y tambi√©n un punto que representa el exceso de demanda $(A)$ y otro del d√©ficit de demanda $(B)$ 

Por un lado, aumentos de la demanda agregada $(DA)$ que dan lugar a excesos de demanda, provocan la ca√≠da de las existencias de las empresas y, por lo tanto, la decisi√≥n de las empresas de aumentar su producci√≥n en un monto similar al aumento de la demanda. En este sentido, los excesos de demanda se enfrentan con aumentos de la producci√≥n. Cada vez que aumenta la producci√≥n, aumenta el empleo, disminuye la tasa de desempleo y aumenta el ingreso disponible de las familias. Las familias distribuyen el incremento de su ingreso disponible entre consumo de bienes y servicios, y ahorro.

Por otro lado, cuando se produce un d√©ficit de demanda o exceso de producci√≥n, contrariamente, aumentan los inventarios y las empresas deciden disminuir su producci√≥n y el empleo de trabajadores. 

No obstante, en ambos casos, la oferta de producci√≥n se ajusta a los cambios en la demanda agregada. Por su parte, en el punto $A$, la demanda agregada es mayor que el nivel de producci√≥n. Hay exceso de demanda que provoca una disminuci√≥n de los inventarios y aumentos de la producci√≥n hasta llegar al equilibrio. Mientras que en el punto $B$ hay un d√©ficit de demanda o un exceso de producci√≥n que da lugar a un aumento de los inventarios y a disminuciones de la producci√≥n hasta llegar al equilibrio.



```python
# Par√°metros

Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8 # b > m
m = 0.2
t = 0.3
r = 0.9

Y = np.arange(Y_size)

# Ecuaci√≥n de la curva del ingreso de equilibrio

def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)
```

### Parte 2: El modelo ingreso-gasto en una econom√≠a abierta

#### 1. Encuentre el Ingreso de Equilibrio  e identifique el multiplicador keynesiano.

Reemplazando los par√°metros se obtiene: 

$$ DA=C_0 +b(1‚àít)Y+I_0 ‚àíhr+G_0 +x_1Y‚àó +x_2Œµ‚àím_1(1‚àít)Y+m_2Œµ$$

$$ DA=C_0 +I_0 +G_0 +[b(1‚àít)‚àím_1(1‚àít)]Y‚àíhr+x_1Y‚àó +(x_2 +m_2)Œµ $$

Se agrupa los componentes aut√≥nomos bajo el par√°metro $w$:

$$ DA=w‚àíhr+x_1Y‚àó +(x_2 +m_2)Œµ_0 ‚àí(x_2 +m_2)œÅr+(x_2 +m_2)œÅr‚àó +[(b‚àím_1)(1‚àít)]Y $$


La ecuaci√≥n de la Demanda Agregada $(DA)$ es:

$$ DA=[ùë§+x_1Y‚àó +(x_2 +m_2)Œµ_0 +(x_2 +m_2)œÅr‚àó ‚àí(h+(x_2 +m_2)œÅ)r]+[(b‚àím_1)(1‚àít)]Y $$

$$ DA = Œ±_0 + Œ±_1Y $$


El ingreso de equilibrio se obtiene cuando se utiliza la igualdad $Y = DA$:

$$ Y=w‚àí[h+(x_2 +m_2)œÅ]r+x_1Y‚àó +(x_2 +m_2)Œµ_0 +(x_2 +m_2)œÅr‚àó +[(b‚àím_1)(1‚àít)]Y $$

$$ Y = \frac{1}{1‚àí(b‚àím1)(1‚àít)} [w+(x_2 +m_2)Œµ_0 +(x_2 +m_2)œÅr‚àó +x_1Y‚àó] - [\frac{h+(x_2 +m_2)œÅ}{1‚àí(b‚àím1)(1‚àít)}]r $$

Y el multiplicador est√° representado por:

$$ \frac{1}{1‚àí(b‚àím1)(1‚àít)} $$


```python
# Par√°metros

Y_size = 100 

Co = 35
Io = 40
Go = 50
h = 0.8
b = 0.9
m = 0.5
t = 0.1

x1 = 0.3
Y¬∫ = 12
x2 = 0.6
m2 = 0.4
Eo = 9
p = 0.8
r¬∫ = 8
r = 12
m1 = 0.1

Y = np.arange(Y_size)


# Ecuaci√≥n

def DA(Co, Io, Go, h, b, m, t, x1, Y¬∫, x2, m2, Eo, p, r¬∫, r, m1):
    DA = [Co + Io + Go + x1 *Y¬∫ +(x2 + m2)* Eo + (x2 + m2)* p*r¬∫ -(h +(x2 +m2)*p)*r]+[(b-m1)*(1-t)]*Y
    return DA

DA = DA(Co, Io, Go, h, b, m, t, x1, Y¬∫, x2, m2, Eo, p, r¬∫, r, m1)
```

```python
# Recta de 45¬∞

a = 2.5 

def L_45(a, Y):
    L_45 = a*Y
    return L_45

L_45 = L_45(a, Y)
```

```python
# Gr√°fico

# Dimensiones del gr√°fico
y_max = np.max(Y)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, DA, label = "DA", color = "#3D59AB") #Demanda agregada
ax.plot(Y, L_45, color = "#404040") #L√≠nea de 45¬∫

# Eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

# L√≠neas punteadas
plt.axvline(x=70.5,  ymin= 0, ymax= 0.69, linestyle = ":", color = "grey")
plt.axhline(y=176, xmin= 0, xmax= 0.7, linestyle = ":", color = "grey")

# Texto agregado
plt.text(0, 180, '$DA^e$', fontsize = 11.5, color = 'black')
plt.text(0, 130, '$Œ±_o$', fontsize = 15, color = 'black')
plt.text(6, 4, '$45¬∞$', fontsize = 11.5, color = 'black')
plt.text(2.5, -3, '$‚óù$', fontsize = 30, color = '#404040')
plt.text(72, 0, '$Y^e$', fontsize = 12, color = 'black')

# T√≠tulo y leyenda
ax.set(title="El modelo ingreso-gasto en una econom√≠a abierta", xlabel= r'Y', ylabel= r'DA')
ax.legend()

plt.show()
```

### Parte 3: Modelo de OA-DA y el nuevo esquema institucional de Pol√≠tica Monetaria, donde la tasa de inter√©s de referencia $r_0$ es usada como instrumento.

#### 1. Encuentre las ecuaciones de equilibrio de $Y^{eq\_da\_oa}$, $P^{eq\_da\_oa}$

- **Demanda Agregada**:

Considerando la ecuaci√≥n de la curva IS: 

$$ Y =  \frac{1}{B_1} (B_0) $$

Donde $B_0 = C_0 + I_0 + G_0 + X_0 - hr$ y $B_1 = 1 - (b-m)(1-t)$

Y la ecuaci√≥n de la LM:

$$ M^s = kY - jr - P_0$$


Reemplazando en funci√≥n de $P$, se obtiene:

$$ P = M^s - k(\frac{B_1}{B_0}) + jr $$

Y, en funci√≥n de $M^s$:

$$ M^s = k(\frac{B_1}{B_0}) - jr + P $$

- **Oferta Agregada**:

$$ P = P_e + Œ∏(Y - \bar{Y}) $$

- **Ecuaciones de equilibrio**:

$$ Y^{eq\_da\_oa} = \frac{1}{1 - (b-m)(1-t)} (C_0 + I_0 + G_0 + X_0) $$



$$ P^{eq\_da\_oa} = P_e + Œ∏(\frac{1}{B_1} B_0 - \bar{Y}) $$


```python
#1--------------------------
    # Demanda Agregada
    
# Par√°metros

Y_size = 100 

Co = 32
Io = 20
Go = 40
Xo = 2
h = 0.2
b = 0.5
m = 0.4
t = 0.8
r = 5

Ms = 70
k = 0.5
j = 1.2                
P  = 8  

Y = np.arange(Y_size)


# Ecuaci√≥n

B0 = Co + Io + Go + Xo - h*r
B1 = 1 - (b-m)*(1-t)

def P_AD(h, P, j, B0, B1, k, r):
    P_AD = k*(B0/B1) - j*r + P
    return P_AD

P_AD = P_AD(h, P, j, B0, B1, k, r)


#2--------------------------
    # Oferta Agregada
    
# Par√°metros

Y_size = 100

Pe = 100 
Œ∏ = 3
_Y = 20   

Y = np.arange(Y_size)


# Ecuaci√≥n

def P_AS(Pe, _Y, Y, Œ∏):
    P_AS = Pe + Œ∏*(Y-_Y)
    return P_AS

P_AS = P_AS(Pe, _Y, Y, Œ∏)
```

```python
    # Gr√°fico del modelo DA-OA

# Dimensiones del gr√°fico
y_max = np.max(P)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar 
plt.axvline(P_AD, label = "DA", color = "C4") #DA

ax.plot(Y, P_AS, label = "OA", color = "C8") #OA

# L√≠neas punteadas
plt.axhline(y = 190, color = "grey", linestyle = ':', xmax= 0.5)

# Texto agregado
plt.text(0, 200, '$P_0$', fontsize = 12, color = 'black')
plt.text(53, 30, '$Y_0$', fontsize = 12, color = 'black')
plt.text(50, 202, '$E_0$', fontsize = 12, color = 'black')


# Eliminar valores de ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

# T√≠tulo, ejes y leyenda
ax.set(title="DA-OA", xlabel= r'Y', ylabel= r'P')
ax.legend()

plt.show()
```

#### 2. Analice los efectos sobre las variables end√≥genas Y, P ante una subida de la tasa de inter√©s de referencia  $(Œîr_0 > 0)$. El an√°lisis debe ser intuitivo, matem√°tico y gr√°fico.

- Intuici√≥n:

$$ r_0‚Üë ‚Üí I‚Üì ‚Üí DA‚Üì ‚Üí DA < Y ‚Üí Y‚Üì $$

$$ Y‚Üì ‚Üí (Y - \bar{Y})‚Üì ‚Üí P‚Üì $$

- Matem√°tica:

$$ ŒîY < 0 $$

```python
#------------------------------------- 
    # Curva IS

r_size = 100 

Co = 13
b = 0.7
t = 0.25
Io = 6.4
Go = 5
h = 0.3
Xo = 2
m = 0.4

r = np.arange(r_size)

# Ecuaci√≥n

def IS_CB(Co, b, t, Io, h, Xo, m, Go):
    IS_CB = (1/(1 - (b - m)*(1 - t)))*(Co + Io + Go + Xo - h*r)
    return IS_CB

IS_CB = IS_CB(Co, b, t, Io, h, Xo, m, Go)


#--------------------------------------------------
    # Curva LM
    
# Par√°metros

Y_size = 100

P = 3.5
k = 1.76
j = 3
r = 15

Y = np.arange(Y_size)

# Ecuaci√≥n

def LM_CB(Y, P, k, j, r):
    LM_CB = P*(k*Y - j*r)
    return LM_CB

LM_CB = LM_CB(Y, P, k, j, r)

#---------------------------------------------------
    # Nueva LM

r = 20 

def LM_CB_2(Y, P, k, j, r):
    LM_CB_2 = P*(k*Y - j*r)
    return LM_CB_2

LM_CB_2 = LM_CB_2(Y, P, k, j, r)
```

```python
# l√≠neas punteadas aut√≥maticas

    # definir la funci√≥n line_intersection
def line_intersection(line1, line2):
    xdiff = (line1[0][0] - line1[1][0], line2[0][0] - line2[1][0])
    ydiff = (line1[0][1] - line1[1][1], line2[0][1] - line2[1][1])

    def det(a, b):
        return a[0] * b[1] - a[1] * b[0]

    div = det(xdiff, ydiff)
    if div == 0:
       raise Exception('lines do not intersect')

    d = (det(*line1), det(*line2))
    x = det(d, xdiff) / div
    y = det(d, ydiff) / div
    return x, y
```

```python
r_ec = np.full((100), 15)

    # coordenadas de las curvas (x,y)
A = [IS_CB[0], Y[0]] # DA, coordenada inicio
B = [IS_CB[-1], Y[-1]] # DA, coordenada fin

C = [r_ec[0], Y[0]] # L_45, coordenada inicio
D = [r_ec[-1], Y[-1]] # L_45, coordenada fin

    # creaci√≥n de intersecci√≥n

intersec_1 = line_intersection((A, B), (C, D))
intersec_1 # (y,x)
```
