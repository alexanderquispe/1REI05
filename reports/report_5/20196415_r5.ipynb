Para realizar el analisis, el autor se centra en dos de los canales a través de los cuales la política monetaria influye en la actividad económica y el nivel de precios: el canal del crédito y el canal del tipo de cambio, que parecen ser los más importantes en la economía peruana. La hipótesis del autor es que el principal canal a través del cual la política monetaria impacta en el sector real de la economía peruana es el canal del crédito o préstamo bancario.
