# Ejercicio 4: 

Para calcular el "Ingreso de Equilibrio"(Y), primero debemos comenzar por hallar la ecuación de la "Demanda agregada" (DA): 

Datos: 
$$ C = C_0 + bY^d $$
$$ I = I_0 - hr $$
$$ G = G_0 $$
$$ T = tY $$
$$ X = X_0 $$
$$ M = mY^d $$
----------------------------
Resultado de la suma de "Datos" que es igual a "DA":
$ DA = C + I + G + X -M $

Reemplazamos en "DA" los "Datos" previos:
- $ DA = C_0 + bY^d + I_0 -hr + G_0 + X_0 - mY^d $
- $ DA = C_0 + I_0 + G_0 + X_0 - hr + Y(b - m)(1 - t) $

En donde:
* Intercepto: $ α_0 = C_0 + I_0 + G_0 + X_0 - hr $ 
* Pendiente: $ α_1 = (b - m)(1 - t) $

De la forma más corta tenemos: $ DA = α_0 + α_1Y $

Continuando con el procedimiento, reemplazamos en la ecuación del "Ingreso de equilibrio" en donde  $Y = DA$
$$ Y = C_0 + bY^d + I_0 -hr + G_0 + X_0 - mY^d $$
$$ Y = \frac{1}{1 - (b - m)(1 - t)} (C_0 + I_0 + G_0 + X_0 - hr) $$

$$ k=\frac{1}{1 - (b - m)(1 - t)} $$

De otra forma: 
$$ Y = k (C_0 + I_0 + G_0 + X_0 - hr) $$

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sy
import pandas as pd
from causalgraphicalmodels import CausalGraphicalModel
from sympy import *
```

```python
# Parámetros

Y_size = 100 

Co = 36
Io = 41
Go = 71
Xo = 3
h = 0.6
b = 0.7 # b > m
m = 0.1
t = 0.2
r = 0.8

Y = np.arange(Y_size)

# Ecuación de la curva del ingreso de equilibrio

def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)
```

```python
# Gráfico

# Dimensiones del gráfico
y_max = np.max(DA_IS_K)
fig, ax = plt.subplots(figsize=(15, 13))

# Curvas a graficar
ax.plot(Y, DA_IS_K, label = "DA", color = "#3D59AB") #Demanda agregada
ax.plot(Y, L_45, color = "#404040") #Línea de 45º

# Eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

# Líneas punteadas punto de equilibrio
plt.axvline(x=74.5,  ymin= 0, ymax= 0.73, linestyle = ":", color = "green")
plt.axhline(y=186, xmin= 0, xmax= 0.8, linestyle = ":", color = "green")
plt.axhline(y=150, xmin= 0, xmax= 0.8, linestyle = ":", color = "green")

# Texto agregado
    # punto de equilibrio
plt.text(0, 180, '$DA^e$', fontsize = 11.5, color = 'black')
plt.text(72, 0, '$Y^e$', fontsize = 12, color = 'black')
plt.text(0, 152, '$α_o$', fontsize = 15, color = 'black')
plt.text(100, 200, '$α_1 = (b-m)(1-t)Y$', fontsize = 15, color = 'black')
plt.text(78, 155, '$α_0 = (C_0 + I_0 + G_0 + X_0 - hr)$', fontsize = 15, color = 'black')
plt.text(90, 215, '$DA = α_0 + α_1Y$', fontsize = 15, color = 'black')
plt.text(92, 205, '$↓$', fontsize = 18, color = 'grey')
    # línea 45º
plt.text(6, 4, '$45°$', fontsize = 11.5, color = 'red')
plt.text(2.5, -3, '$◝$', fontsize = 30, color = '#404040')

# Título y leyenda
ax.set(title="El ingreso de Equilibrio a Corto Plazo", xlabel= r'Y', ylabel= r'DA')
plt.show()

# definir la función line_intersection
def line_intersection(line1, line2):
    xdiff = (line1[0][0] - line1[1][0], line2[0][0] - line2[1][0])
    ydiff = (line1[0][1] - line1[1][1], line2[0][1] - line2[1][1])

    def det(a, b):
        return a[0] * b[1] - a[1] * b[0]

    div = det(xdiff, ydiff)
    if div == 0:
       raise Exception('lines do not intersect')

    d = (det(*line1), det(*line2))
    x = det(d, xdiff) / div
    y = det(d, ydiff) / div
    return x, y

    # coordenadas de las curvas (x,y)
A = [DA_IS_K[0], Y[0]] # DA, coordenada inicio
B = [DA_IS_K[-1], Y[-1]] # DA, coordenada fin

C = [L_45[0], Y[0]] # L_45, coordenada inicio
D = [L_45[-1], Y[-1]] # L_45, coordenada fin

    # creación de intersección

intersec = line_intersection((A, B), (C, D))
intersec
```

Asimismo, el aumento del gasto del gobierno hace que se eleve el intercepto de la función de la Demanda Agregada para arriba. Entonces, si el Gasto(Go) aumenta, la Demanda Agregada (DA) también lo hará. Por ende, el Ingreso de Equilibrio también aumentará. 
$$ ↑Go → ↑DA → DA > Y → ↑Y $$

$$ Y_e = \frac{1}{1 - (b - m)(1 - t)} (C_0 + I_0 + G_0 + X_0 - hr) $$

Si : $ k=\frac{1}{1 - (b - m)(1 - t)} $ , en donde $ k > 0 $ , tenemos lo siguiente:

$$ Y = k (C_0 + I_0 + G_0 + X_0 - hr) $$


$$ ∆Y = k (∆C_0 + ∆I_0 + ∆G_0 + ∆X_0 - ∆hr) $$


Como no ha habido cambios en $C_0$, $I_0$, $X_0$, $h$ ni $r$, entonces: 

$$∆C_0 = ∆I_0 = ∆X_0 = ∆h = ∆r = 0$$

$$ ∆Y = k (∆G_0) $$


Sabiendo que $∆G_0 > 0 $ y que $k > 0$, la multiplicación de un número positivo con otro positivo dará un resultado positivo:

$$ ∆Y = (+)(+) $$
$$ ∆Y > 0 $$

# 5.2:
- Intuición: se dan dos situaciones contradictorias, teniendo en cuenta el Consumo autónomo (Co) y las Importaciones (M).

 * Si la tasa de impuestos baja, aumenta el Ingreso de Equilibrio. Por tanto, aumenta el ingreso disponible, la demanda de consumo y la demanda agregada: 
 $$ t↓ → Co↑ → DA↑ → DA > Y → Y↑ $$
 * También, aumentaría las importaciones, lo que ocasiona que se reduzca la pendiente de la Demanda Agregada y el valor del multiplicador. Lo cual, conlleva a la disminución del ingreso de equilibrio: 
 $$ t↓ → M↑ → DA↓ → DA < Y → Y↓ $$

- Matemáticamente: $∆t < 0  →  ¿∆Y?$


```python
Co, Io, Go, Xo, h, r, b, m, t = symbols('Co Io Go Xo h r b m t')

f = (Co + Io + Go + Xo - h*r)/(1-(b-m)*(1-t))


df_t = diff(f, t)
df_t #∆Y/∆t
```

Considernado el diferencial de $∆t$:

 * $∆Y$ =  $ (m-b)(Co + Go + Io + Xo - hr) $
 * $∆t$ = $ (1-(1-t)(b-m)+1)^2 $
 
Entonces: 
$$ \frac{∆Y}{∆t} = \frac{(m-b)(Co + Go + Io + Xo - hr)}{(1-(1-t)(b-m)+1)^2} $$

- Sabemos que b > m, entonces $(m-b) < 0$
- Los componentes autónomos no cambian: $∆C_0 = ∆I_0 = ∆X_0 = ∆h = ∆r = 0$
- Cualquier número elevado al cuadrado será positivo: $ (1-(1-t)(b-m)+1)^2 > 0 $

Entonces:

$$ \frac{∆Y}{∆t} = \frac{(-)}{(+)} $$

Dado que $∆t < 0$,  la multiplicación de dos negativos, da como resultado uno positivo. Por tanto quedaría en que la división de un positivo y otro positivo dará un resultado positivo:

$$ \frac{∆Y}{(-)} = \frac{(-)}{(+)} $$

$$ ∆Y = \frac{(-)(-)}{(+)} $$

$$ ∆Y = \frac{(+)}{(+)} $$

$$ ∆Y > 0 $$

# 6.2:
Para calcular el "Ingreso de Equilibrio"(Y), primero debemos comenzar por hallar la ecuación de la "Demanda agregada" (DA): 

Datos: 
$$ C = C_0 + bY^d $$
$$ I = I_0 - hr $$
$$ G = G_0 - gY $$
$$ T = tY $$
$$ X = X_0 $$
$$ M = mY^d $$
----------------------------
Resultado de la suma de "Datos" que es igual a "DA":
$ DA = C + I + G + X -M $

Reemplazamos en "DA" los "Datos" previos:
- $ DA = C_0 + bY^d + I_0 -hr + G_0 - gY + X_0 - mY^d $
- $ DA = C_0 + I_0 + G_0 - gY + X_0 - hr + Y(b - m)(1 - t) $

En donde:
* Intercepto: $ α_0 = C_0 + I_0 + G_0 + X_0 - hr $ 
* Pendiente: $ α_1 = ((b - m)(1 - t)-g)Y $

De la forma más corta tenemos: $ DA = α_0 + α_1Y $

Continuando con el procedimiento, reemplazamos en la ecuación del "Ingreso de equilibrio" en donde  $Y = DA$
$$ Y = C_0 + bY^d + I_0 -hr + G_0 - gY + X_0 - mY^d $$
$$ Y = \frac{1}{1 - (b - m)(1 - t) + g} (C_0 + I_0 + G_0 + X_0 - hr) $$

$$ k=\frac{1}{1 - (b - m)(1 - t) + g} $$

De otra forma: 
$$ Y = k (C_0 + I_0 + G_0 + X_0 - hr) $$

# 6.3:
- Intuición: Si el gobierno decide aumentar el gasto autónomo (∆𝐺 > 0) en un periódo de recesión donde disminuye la producción y aumenta el desempleo. Ante esta situación, el gobierno, deberá incrementar su gasto para recuperar la producción, o al menos estimularla. Esto podría dar como resultado que aumente el empleo en la economía. 
$$ ↑Go → ↑DA → DA > Y → ↑Y $$


- Matemáticamente: $∆G_0 > 0  →  ¿∆Y?$

$$ Y = \frac{1}{1 - (b - m)(1 - t) + g} (C_0 + I_0 + G_0 + X_0 - hr) $$

Si : $ k=\frac{1}{1 - (b - m)(1 - t)} $ , en donde $ k > 0 $ , tenemos lo siguiente:

$$ Y = k (C_0 + I_0 + G_0 + X_0 - hr) $$


$$ ∆Y = k (∆C_0 + ∆I_0 + ∆G_0 + ∆X_0 - ∆hr) $$


Como no ha habido cambios en $C_0$, $I_0$, $X_0$, $h$ ni $r$, entonces: 

$$∆C_0 = ∆I_0 = ∆X_0 = ∆h = ∆r = 0$$

$$ ∆Y = k (∆G_0) $$


Sabiendo que $∆G_0 > 0 $ y que $k > 0$, la multiplicación de un número positivo con otro positivo dará un resultado positivo:

$$ ∆Y = (+)(+) $$
$$ ∆Y > 0 $$
