**1. Explique cuáles son los intrumentos de política monetaria que puede utilizar el Banco Central.**

El Banco Central tiene como propósito regular la oferta monetaria de un país. Con el objetivo de cumplir con sus funciones, puede emplear tres instrumentos por medio de políticas monetarias expansivas o contractivas: 

- **La oferta monetaria:**
    - **Política monetaria expansiva:** el banco comienza a a comprar bonos. Por medio de esta operación, se incrementa la cantidad de dinero nominal dentro de la economía dentro de bancos comerciales y, a su vez, la oferta monetaria. 
    - **Política monetaria contractiva:** Por el contrario, el banco central comienza a vender bonos en el mercado. Al realizar tal acción, la entidad comienza a retirar dinero de la economía, reduciendo la oferta monetaria. 


- **El coeficiente legal de encaje**
    - **Política monetaria expansiva:** al disminuir la tasa de encaje, los bancos comerciales tienen la posibilidad de mantener menos dinero en reserva.
    - **Política monetaria contractiva:** Asimismo, el aumento de la tasa de encaje permite que los bancos puedan tener una mayor cantidad de depósitos en reserva. En conscuencia, se reduce la oferta monetaria.
   
   
- **La tasa de interés**  
    - **Política monetaria expansiva:** los bancos pueden optar por reducir la tasa de interés de referencia, aumentando de esta manera la oferta monetaria.  
    - **Política monetaria contractiva:** ante un incremento de la tasa de interés de referencia, disminuye la cantidad de créditos que los bancos comerciales que pueden obtener. 

**2. Derive la oferta real de dinero y explique cada uno de sus componentes.**

Para comenzar, debemos tener en cuenta que nos encontramos dentro de un modelo de corto plazo y, debido a ello, la oferta monetaria se trata de una variable exógena. 

La función de la oferta real de dinero se representa de la siguiente manera: 

$$ M^s = M^s_0/P $$

Y se compone de: 

$ M^s_0$ = masa monetaria o la oferta nominal de dinero

Se trata de la cantidad de dinero que se encuentra en circulación dentro de la economía. Asimismo, está es determinada por el banco central. 

$ P$ = nivel general de precios

De igual manera, el nivel general de precios de una economía es un valor que se encuentra dado por el modelo. 



**3. Derive la demanda real de dinero. Explique qué papel cumplen los parametros "k" y "j"**

En primer lugar, se debe tener en cuenta que hay motivos por los que el dinero es demandado. De acuerdo a sus funciones pueden ser divididos de la siguiente manera: 

**Transacción y precaucación:**

La **transacción** consiste en la demanda de dinero para realizar transacciones de productos dentro de la economía. 

Por su parte, la **precaución** consiste en la demanda de dinero para tener la capacidad de cumplir con los pagos de una deuda. 

Finalmente, esta demanda por estos motivos se puede expresar de la siguiente manera:

$$ L_1 = kY$$

**Especulación:**

La demanda de dinero dependerá inversamente de la tasa de interés de los bonos. Es decir, será preferible mantener el dinero en físico de acuerdo al nivel de la tasa de interés mencionada. 

Esta demanda se representa: 

$$ L_2 = -ji$$

Finalmente, la demanda real de dinero se expresa de esta manera: 

$$ L^d = L_1 + L_2$$

Reemplazando:

$$ M^d  = kY - ji$$

No obstante, para efectos del modelo podemos considerar que la tasa de interés nominal es igual a la tasa de interés real $i = r$. Por lo tanto: 

$$ M^d  = kY - jr$$


**El papel de los parametros "k" y "j"**

La propensión marginal a demandar dinero o **el parámetro $k$** indica la variación que se puede ver en la demanda de dinero frente a los cambios en el nivel de ingresos ($Y$). Por lo tanto, depende positivamente de este. 

Finalmente, **el parámetro $j$** o la elasticidad de sustitución del dinero permite determinar los cambios que se pueden dar en la demanda ante la variación de la tasa de interés nominal ($i$). Es decir, si es mejor tener en dinero dentro del banco o en la mano.

**4. Asumiendo que no hay inflación podemos asumir que $ i = r $. Escriba en terminos reales la ecuación de equilibrio en el mercado de dinero.**

Considerando que la función de la demanda real de dinero se expresa de la siguiente manera: 

$$ M^d  = kY - jr$$


Y, por su parte, la función de la oferta real monetaria: 

$$ M^s = M^s_o/P $$

El equilibrio en el mercado de dinero se expresa mediante la igualdad entre la oferta y la demanda. Por lo tanto:

$$M^s = M^d$$

$$ M^s_o/P = kY - jr$$

**5. Grafique el equilibrio en el mercado de dinero.**

```python
# Parameters
r_size = 100

k = 0.7
j = 0.3                
P = 15 
Y = 40
MS_0 = 700

r = np.arange(r_size)

# Función de demanda 

def MD(k, j, r, Y):
    MD_eq = (k*Y - j*r)
    return MD_eq
MD_0 = MD(k, j, r, Y)

# Oferta de dinero.
MS = MS_0 / P
MS
```

```python
# Creamos el seteo para la figura 
fig, ax1 = plt.subplots(figsize=(10, 8))

# Agregamos titulo t el nombre de las coordenadas
ax1.set(title="Equilibrio en el mercado de dinero", xlabel=r'M^s / P', ylabel=r'r')

# Ploteamos la demanda de dinero
ax1.plot(MD_0, label= '$L_0$', color = '#5FBABF')

# Para plotear la oferta de dinero solo necesitamos crear una linea vertical
ax1.axvline(x = MS,  ymin= 0, ymax= 1, color = "grey")

# Creamos las lineas puntadas para el equilibrio
ax1.axhline(y=14, xmin= 0, xmax= 0.475, linestyle = ":", color = "black")


# Agregamos texto
ax1.text(0,14.5, "$r_0$", fontsize = 12, color = 'black')
ax1.text(90, 2.5, "$L(Y_0)$", fontsize = 12, color = 'black')


ax1.yaxis.set_major_locator(plt.NullLocator())   
ax1.xaxis.set_major_locator(plt.NullLocator())

ax1.legend()

plt.show()
```

$$ \frac {M^s_o} {P} = kY - jr$$

$$ M^s = \frac {M^s_o} {P} $$

Por lo tanto, se observa una disminución en la oferta monetaria: 

$$M^s↓ → \frac {M^s_o} {P}↓  $$

Asimismo, ante la disminución de la oferta monetaria, la tasa de interés aumenta para que la demanda de dinero disminuya y se recupere el equilibrio: 

$$ M^s↓ → r↑  $$

### Curva LM

**1. Derive paso a paso la curva LM matemáticamente (a partir del equilibrio en el Mercado Monetario) y grafique.**

La ecuación del equilibrio en el mercado monetario se expresa de la siguiente manera:

$$ \frac {M^s_o} {P} = kY - jr$$

Se despeja de la tasa de interés ($r$): 

$$ jr = kY - \frac {M^s_o} {P} $$

Y se obtiene la ecuación de la curva LM: 

$$ r = - \frac {M^s_o} {Pj} + \frac {k} {j} Y  $$

Teniendo un intercepto negativo $- \frac {M^s_o} {Pj}$y una pendiente positiva $\frac {k} {j}$

**Gráfico:**

```python
# Parameters
Y_size = 100

k = 1
j = 2                
Ms = 700             
P  = 25               

Y = np.arange(Y_size)


# Ecuación

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)
```

```python
    # Gráfico de la curva LM

# Dimensiones del gráfico
y_max = np.max(i)
v = [0, Y_size, 0, y_max]   
fig, ax = plt.subplots(figsize=(10, 8))

# Eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

# Curvas a graficar
ax.plot(Y, i, label="LM", color = '#FF5733')

# Título y leyenda
ax.set(title="Curva LM", xlabel=r'Y', ylabel=r'r')
ax.legend()


plt.show()
```

**2. ¿Cuál es el efecto de una disminución en la Masa Monetaria $∆M_s<0$ ? Explica usando la intuición y gráficos.**

**Intuición:**

La curva LM es expresada de la siguiente manera: 

$$ r = - \frac {M^s_o} {Pj} + \frac {k} {j} Y  $$

Sabiendo que $∆M^s<0$:

$$ M^s_o↓ → M^s↓ → M^s<M^d → r↑ → M^d↓  $$

**Gráfico:**

```python
# Parámetros

Y_size = 100

k = 3
j = 2               
Ms = 800             
P  = 20               

Y = np.arange(Y_size)

# Ecuación

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)

#--------------------------------------------------

# Definir SOLO el parámetro cambiado

Ms = 200

def i_LM_Ms( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i_Ms = i_LM_Ms( k, j, Ms, P, Y)
```

```python
# Dimensiones del gráfico
y_max = np.max(i)
v = [0, Y_size, 0, y_max]   
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, i, label="LM", color = 'black')
ax.plot(Y, i_Ms, label="LM_Ms", color = '#C70039', linestyle = 'dashed')

# Texto agregado
plt.text(58, 76, '∆$M^s$', fontsize=12, color='black')
plt.text(54, 69, '←', fontsize=15, color='grey')

# Título y leyenda
ax.set(title = "Disminución en la Masa Monetaria $(M^s)$", xlabel=r'Y', ylabel=r'r')
ax.legend()


plt.show()
```

**3.¿Cuál es el efecto de un aumento en k $∆k>0$? Explica usando intuición y gráficos.**

**Intuición:**

La curva LM: 

$$ r = - \frac {M^s_o} {Pj} + \frac {k} {j} Y  $$

Teniendo en cuenta que $∆k>0$: 

$$k↑→ M^d↑ → M^d>M^s → r ↑ $$

$$k↑ → \frac {k}{j} ↑ $$

**Gráfico:**

```python
# Parámetros

Y_size = 100

k = 3
j = 2               
Ms = 800             
P  = 20               

Y = np.arange(Y_size)

# Ecuación

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)

#--------------------------------------------------

# Definir SOLO el parámetro cambiado

k = 6

def i_LM_k( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i_k = i_LM_k ( k, j, Ms, P, Y)
```

```python
# Dimensiones del gráfico
y_max = np.max(i)
v = [0, Y_size, 0, y_max]   
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, i, label="LM", color = 'black')
ax.plot(Y, i_k, label="LM_k", color = '#C70039', linestyle = 'dashed')

# Texto agregado
plt.text(55, 90, '∆$k$', fontsize=12, color='black')
plt.text(50, 90, '↑', fontsize=20, color='grey')

# Título y leyenda
ax.set(title = "Incremento en la elasticidad del ingreso $(k)$", xlabel=r'Y', ylabel=r'r')
ax.legend()


plt.show()
```
