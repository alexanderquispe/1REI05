1.Derive y explique la función de demanda de Consumo. Replique el gráfico siguiente:

Las familias toman decisiones de gasto y ahorro. Sabemos que la suma de salarios de los trabajadores y de los beneficios de los propietarios del stock de capital es igual al Ingreso Agregado (Y). Parte de este ingreso (Y) se entrega al Estado como tributación (T), y con lo que queda se decide consumir y ahorrar.
La tributación total, es:

$$𝑇 = 𝑡𝑌$$

Donde “t” es la tasa de impuestos o de tributación promedio. El ingreso neto de tributación es el Ingreso Disponible, pues se está suponiendo que las transferencias (TR) son iguales a cero:

$$𝑌^𝑑 =𝑌−𝑇$$
$$𝑌^𝑑 =(1−𝑡)𝑌$$

Parte de este ingreso disponible se utiliza para los gastos en consumo:

$$𝐶 = 𝐶_0 + 𝑏𝑌^𝑑$$

Donde “C0”es el nivel básico de consumo, conocido como el consumo autónomo, y “b” representa el cambio en el consumo por unidad de cambio en el ingreso disponible. Es conocida como la Propensión Marginal a Consumir, (PMgC). Esta propensión es mayor que cero y menor que uno, 0 < 𝑏 < 1.

```python

#Parámetros

Y_size=100

b=0.5
Co=10

Yd=np.arange(Y_size)
Yd
```

```python
#Ecuación

def C (Co, b, Yd):
    C = Co + b*Yd
    return C
C = C(Co, b, Yd)
C
```

```python
#Gráfico:

plt.plot(Yd, C, color='blue')
plt.title("Gráfico 4.2 - Función de Demanda de Consumo")
plt.xlabel("C")
plt.ylabel("Yd")
plt.grid()

texto1 = plt.text(30,55, r'$PMgC=\frac{\Delta  C}{\Delta  Y}=b=0.5$', fontsize=15)
texto1 = plt.text(30,15, r'$\Delta  Y^d$', fontsize=15)
texto1 = plt.text(42,25, r'$\Delta  C$', fontsize=15)

texto2 = plt.text(20,19, r'$------$', fontsize=16)
texto2 = plt.text(39,30, r'$|$', fontsize=16)
texto2 = plt.text(39,21, r'$|$', fontsize=16)
texto2 = plt.text(39,26, r'$|$', fontsize=16)

```

2.Derive y explique la función de Inversión. Replique el gráfico siguiente:

El gasto de inversión se puede representar con una función que depende inversamente de la tasa de interés real, “r”, y donde “I0” es el componente autónomo que depende básicamente de las expectativas de los inversionistas.

$$I = I_0 − hr$$

Donde:
- Io: Inversión autónoma (inversión fija).
    Si sube la inversión autónoma, sube la inversión general en el país.
- r: Tasa de interés real pasiva.
    Indica la tasa de interés que recibe el banco al realizar préstamos.
- h: Sensibilidad de los inversionistas ante la subida o bajada del interés. Es fija para efectos del modelo.
    Se encuentra entre 0 y 1.


```python

#Parámetros

Io=100
h=0.5
R_size=50

r=np.arange(R_size)
r
```

```python
#Ecuación

def I (Io, h, r):
    I = Io - h*r
    return I
I = I(Io, h, r)
I
```

```python
#Gráfico

plt.plot(r, I, color="blue")
plt.title("I = Io - hr \n \n Gráfico 4.3 \n Función de Demanda de Inversión")
plt.xlabel("r")
plt.ylabel("I")
plt.xticks([])
plt.yticks([])
```

```python
# 4.Encuentre el nivel de Ingreso de equilibrio Ye. 
#Replique el siguiente gráfico con todas descripciones encima de las rectas.

# Parámetros

Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8 # b > m
m = 0.2
t = 0.3
r = 0.9

Y = np.arange(Y_size)

# Ecuación de la curva del ingreso de equilibrio

def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)
```

```python
# Recta de 45°

a = 2.5 

def L_45(a, Y):
    L_45 = a*Y
    return L_45

L_45 = L_45(a, Y)
```

```python
# Gráfico

# Dimensiones del gráfico
y_max = np.max(DA_IS_K)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, DA_IS_K, label = "DA", color = "#3D59AB") #Demanda agregada
ax.plot(Y, L_45, color = "#404040") #Línea de 45º

# Eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

# Líneas punteadas punto de equilibrio
plt.axvline(x=70.5,  ymin= 0, ymax= 0.69, linestyle = ":", color = "grey")
plt.axhline(y=176, xmin= 0, xmax= 0.7, linestyle = ":", color = "grey")

# Texto agregado
    # punto de equilibrio
plt.text(0, 180, '$DA^e$', fontsize = 11.5, color = 'black')
plt.text(72, 0, '$Y^e$', fontsize = 12, color = 'black')
plt.text(0, 152, '$α_o$', fontsize = 15, color = 'black')
    # línea 45º
plt.text(6, 4, '$45°$', fontsize = 11.5, color = 'black')
plt.text(2.5, -3, '$◝$', fontsize = 30, color = '#404040')

# Título y leyenda
ax.set(title="El ingreso de Equilibrio a Corto Plazo", xlabel= r'Y', ylabel= r'DA')
ax.legend() #mostrar leyenda

plt.show()
```
