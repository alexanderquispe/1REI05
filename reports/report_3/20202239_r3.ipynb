##### 1. Explique cuáles son los instrumentos de política monetaria que puede utilizar el Banco Central. 

La oferta de dinero es uno de los instrumentos de política monetaria que el Banco Central puede utilizar. En este caso, la institución establece la cantidad de masa monetaria que va a circular en la economía del país. Por lo tanto, al establecer ajustes en esta variable, ya sean de carácter expansivo o contractivo, va a estar generando cambios en el mercado monetario.

##### 2. Derive la oferta real de dinero y explique cada uno de sus componentes. 

La oferta real de dinero está determinada por la ecuación: 

$$ M_s = \frac{M_0}{P} $$

donde: 
- $M_0$: masa monetaria 
- $P$: nivel de precios

En este caso, la masa monetaria es la cantidad de dinero que poseemos en términos nominales. Sin embargo, esta variable, por sí sola, no determina la oferta real (cantidad real) de dinero que existe en el mercado. En ese sentido, el nivel de precios es lo que permite establecer la oferta de dinero en términos reales, pues es esta variable la que determina la cantidad de dinero real que se tiene en un periodo de tiempo determinado, en la moneda del país. En nuestro caso, en soles.  

##### 3. Derive la demanda real de dinero. Explique qué papel cumplen los parámetro "k" y "j"

La demanda de dinero está determinada por los motivos que motivan a los agentes económicos (en este caso, familiar y productores/empresas) a demandar dinero. 

Un primer bloque $(L_1)$ está determinado por la transacción y la precaución. Por un lado, la transacción es el dinero que se demanda para realizar operaciones, que como su nombre lo indica, son de carácter transactorio. Por otro lado, la precaución se refiere al dinero demandado por los agentes para poder cubrir el pago de deudas en las que se incurre. 

Esta sección de la demanda agregada de dinero está determinada por la sensibilidad o elasticidad del ingreso $(k)$, que va a indicar la cantidad de dinero demandado en relación al ingreso de la economía $(Y)$. De esa manera, mientras más grande sea el valor de $k$, mayor será la cantidad demandada de dinero en el mercado. Asimismo, una mayor demanda de dinero va a ser resultado de un incremento en el ingreso de la economía $(Y)$. 

En ese sentido, esta primera sección está determinada por la siguiente ecuación:

$$ L_1 = kY $$

De igual forma, una segunda sección de la demanda $(L_2)$ va a estar determinada por la especulación, la cual estará relacionada al sector financiero. Según la especulación, habrán momentos en los que se demandará más o menor dinero, dependiendo de la tasa de interés nominal $(i)$ que se ofresca por mantener nuestro dinero en el banco. En ese sentido, mientras mayor sea la tasa de interés $(i)$, mayores serán las motivaciones para poner el dinero en el banco, y por ende, la cantidad de dinero demandada en el mercado será menor; y viceversa. 

Asimismo, esta sección de la demanda agregada de dinero, estará determinada por la elasticidad de sustitución $(j)$. Esta elasticidad va a indincar la propensión que van a tener los agentes a ingresar su dinero al banco. En ese sentido, mientras más grande sea su valor, implicará una mayor elasticidad, indicando que hay más propensión a ingresar nuestro dinero al banco y por ende, una menor demanda de dinero en el mercado. Esto se explica por la tasa de interés, como ya se mencionó anteriormente, pues mientras esta aumente, la elasticidad también lo hará, generando una reducción de la demanda de dinero en el mercado. Por lo tanto, se puede hablar de la existencia de una relación inversa entre la elasticidad de sustitución y la demanda de agregada de dinero. 

En este caso, la ecuación de dicha sección de la demanda agregada de dinero será: 

$$ L_2 = -ji $$

Entonces, en un primer momento tenemos que la demanda de dinero estará determinada por:

$$ M_d = L_1 + L_2 $$

Reemplazando se tiene que:

$$ M_d = kY - ji $$

Ahora, en este modelo se trabaja con la suposición de que la tasa de interés nominal $(i)$ es igual a la tasa real de interés $(r)$, debido a que se asume que no hay inflación. De esa manera, se puede hablar de la demanda real de dinero, donde su función es la siguiente:

$$ M_d = kY - jr $$



##### 4. Asumiendo que no hay inflación podemos asumir que $i=r$. Escriba en términos reales la ecuación de equilibrio en el mercado de dinero.

En este caso, sabiendo que:

$$ M_d = kY - ji $$

$$ M_s = \frac{M_0}{P} $$

donde: 

$$ i = r $$

Además, conociendo que en el equilibrio, la demanda de dinero es igual a la oferta de dinero:

$$ M_d = M_s $$

Podemos decir que:

$$ kY - jr = \frac{M_0}{P} $$

Se ordena la ecuación y se puede concluir que la ecuación de equilibrio, en términos reales, en el mercado de dinero es:

$$ M_0 = P(kY - jr) $$

##### 5. Grafique el equilibrio en el mercado de dinero. 

```python
r_size = 100

#Parámetros

k = 3
j = 0.2                
Y = 35

r = np.arange(r_size)

    # Ecuación

def Ms_MD(k, j, r, Y):
    Ms_MD = k*Y - j*r
    return Ms_MD

Ms_MD = Ms_MD(k, j, r, Y)
```

##### 1. Explique y grafique qué sucede en el mercado de dinero si $∆Y<0$

En este caso, cuando ocurre una reducción en el ingreso de la economía, la demanda real de dinero en el mercado monetario se reduce. Esto sucede porque, al haber menos ingreso, entonces existe una propensión o sensibilidad menor en los agentes económicos a demandar dinero. Por esa razón, ocurre un desplazamiento de la curva de demanda real de dinero, hacia la izquierda. 

En ese sentido, teniendo en cuenta el equilibrio en el mercado de dinero, donde la oferta de dinero es igual a la demanda:

$$ M_s = M_d $$

Se puede decir que al ocurrir a disminución de la Demanda real de dinero $(M_d↓)$, se genera un desequilibrio en el mercado pues la demanda actual $(Md_1)$ resulta menor a la oferta: 

$$ Md_1 < M_s $$

Por esa razón, para volver al equilibrio, debe ocurrir un ajusto en la tasa de interés real. Esto porque no se plantea, en este caso, ninguna variación dentro de la oferta de dinero, por lo tanto, se supone que esta se mantiene. De esa manera, el ajuste de la tasa de interés es lo que permitirá regresar al equilibrio. 

En este caso, ante una reducción de la demanda de dinero, la tasa de interés también deberá reducirse, para que de esa manera se regrese al equilibrio. Por lo tanto se puede decir que: 

Si $Md↓$ entonces $r↓$

Por lo que, se genera un nuevo punto de equilibrio, donde la demanda real de dinero y la tasa real de interés resultan menores a los valores iniciales, antes de que ocurriera una disminución en el ingreso.

###### Gráfico

```python
r_size = 100

#Parámetros

k = 2
j = 0.2                
Y = 35

r = np.arange(r_size)

# Ecuación

def Ms_MD(k, j, r, Y):
    Ms_MD = k*Y - j*r
    return Ms_MD

Ms_Y = Ms_MD(k, j, r, Y)

#Parámetro nuevo

Y = 20

#Nueva ecuación

def Ms_MD(k, j, r, Y):
    Ms_MD = k*Y - j*r
    return Ms_MD

Ms_IS_Y = Ms_MD(k, j, r, Y)
```

##### - Curva LM 
###### 1. Derive paso a paso la curva LM matemáticamente (apartir del equilibrio en el Mercado) y gráfique.

En primer lugar, se necesita establecer las ecuaciones que determinan el equilibrio de Mercado.
    
Por un lado, se tiene la ecuación que determina la Demanda de Dinero: 

$$ M_d = kY - ji $$

donde:
    
   - Md: Demanda de dinero
   - k: propensión marginal a demandar dinero
   - j: constancia de cambio en la demanda de dinero
   - i: tasa de interés nominal o precio del dinero
    
Por otro lado, se tiene la ecuación que determina la oferta de dinero: 

$$ M_s = \frac{M_o}{P} $$
    
donde:
   - Ms: Oferta de dinero
   - Mo: Masa monetaria nominal
   - P: nivel de precios
    
En ese sentido, se sabe que la curva LM deriva del equilibrio en el Mercado Monetario, dado por:

$$ M_d = M_s $$

$$ kY - ji = \frac{M_o}{P} $$
    
Y, considerando que la tasa de interés nominal $(i)$, es igual a la tasa de interés real $(r)$: 

$$ kY - jr = \frac{M_o}{P} $$
    
Asimismo, para la curva LM, se necesita establecer la ecuación en función a r, por lo tanto se la despeja:

$$ kY - \frac{M_o}{P} = jr $$

$$ \frac{k}{j}Y - \frac{M_o}{Pj} = r $$
    
Por lo tanto, la función de la curva LM resulta ser:

$$ r = -\frac{1}{j}\frac{M_o}{P} + \frac{k}{j}Y $$

###### 2. ¿Cuál es el efecto de una disminución en la Masa Monetaria $∆M_s<0$? Explica usando la intuición y gráficos.

###### Intuición:

Teniendo en cuenta que la curva LM está definida por la función:

$$ r = -\frac{1}{j}\frac{M_o}{P} + \frac{k}{j}Y $$

Y sabiendo que $M_o$ es la masa monetaria $(M_s)$. Se puede intuir que, una disminución en la masa monetaria va a generar una contracción de la curva, generando un desplazamiento hacia la izquierda. Además, debido a la relación inversa que posee con la tasa de interés (observada en la función anterior), y al tratarse de un parámetro del intercepto, se puede notar un aumento en el mismo. Esto ocurre, porque ante una reducción de la oferta nominal de dinero, se incurre en un desequilibrio, debido a que, en este caso, no se perciben cambios en el ingreso (se mantiene). 

Por esa razón, en este caso, para regresar a un equilibrio en el mercado monetario, resulta necesario aumentar la tasa de interés.   

###### Gráfico:

```python
#--------------------------------------------------
    # Curva LM ORIGINAL

Y_size = 100

# Parámetros

k = 2
j = 1                
Ms = 700             
P  = 20               

Y = np.arange(Y_size)

# Ecuación

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i_Ms = i_LM( k, j, Ms, P, Y)

#--------------------------------------------------
    # NUEVA curva LM

#Nuevo parámetro

Ms = 200

# Nueva ecuación:

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i_LM_Ms = i_LM( k, j, Ms, P, Y)
```

```python
# Dimensiones del gráfico
y_max = np.max(i_Ms)
v = [0, Y_size, 0, y_max]   
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, i_Ms, label="LM_0", color = '#20B2AA')
ax.plot(Y, i_LM_Ms, label="LM_1", color = '#8470FF', linestyle = 'dashed')

# Texto agregado
plt.text(47, 76, '∆$M_s$', fontsize=12, color='black')
plt.text(47, 70, '←', fontsize=15, color='grey')

# Título y leyenda
ax.set(title = "Disminución en la Masa Monetaria $(M^s)$", xlabel=r'Y', ylabel=r'r')
ax.legend()


plt.show()
```

###### 3. ¿Cuál es el efecto de un aumento en k (∆k > 0)? Explica usando intuición y gráficos. 

###### Intuición: 

Un CAMBIO en la elasticidad del ingreso $(k)$, va a generar cambios en el empinamiento de la curva. Esto se debe a que es un parámetro que determina la pendiente de la curva, como se puede deducir al observar la función:

$$ r = -\frac{1}{j}\frac{M_o}{P} + \frac{k}{j}Y $$

Por lo tanto, en este caso, un aumento en la elasticidad del ingreso va a implicar un mayor empinamiento de la curva, debido a su relación directamente proporcional. 

Asimismo, como ya se estableció, esta elasticidad está relacionada al ingreso. Por lo tanto, si existe un aumento de $k$, por defecto, también existirá un aumento del ingreso $(Y)$, lo cual genera un desequilibrio en el mercado monetario. 

donde:

$$ Y > r $$

Por esa razón, se necesita también, que la tasa de interés aumente, de manera que se pueda volver al equilibrio. 

En ese sentido, ese aumento de la tasa y el ingreso, producido por el aumento de la elasticidad del ingreso, genera un mayor empinamiento de la curva LM. 

```python
#--------------------------------------------------
    # Curva LM ORIGINAL

Y_size = 100

# Parámetros

k = 2
j = 1                
Ms = 700             
P  = 20               

Y = np.arange(Y_size)

# Ecuación

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i_k = i_LM( k, j, Ms, P, Y)

#--------------------------------------------------
    # NUEVA curva LM

#Nuevo parámetro

k = 6

# Nueva ecuación:

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i_LM_k = i_LM( k, j, Ms, P, Y)
```

```python
# Dimensiones del gráfico
y_max = np.max(i_k)
v = [0, Y_size, 0, y_max]   
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, i_k, label="LM_0", color = '#CD919E')
ax.plot(Y, i_LM_k, label="LM_1", color = '#912CEE', linestyle = 'dashed')

# Texto agregado
plt.text(47, 120, '∆$k$', fontsize=12, color='black')
plt.text(47.5, 105, '←', fontsize=15, color='grey')

# Título y leyenda
ax.set(title = "Incremento de la elasticidad del Ingreso $(k)$", xlabel=r'Y', ylabel=r'r')
ax.legend()


plt.show()
```

Como se observa en el gráfico, el aumento de $k$ implicó un mayor empinamiento de la curva.
