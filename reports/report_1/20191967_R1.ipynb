1.Derive y explique la funci√≥n de demanda de Consumo. Replique el gr√°fico siguiente:

Las familias toman decisiones de gasto y ahorro. Sabemos que la suma de salarios de los trabajadores y de los beneficios de los propietarios del stock de capital es igual al Ingreso Agregado (Y). Parte de este ingreso (Y) se entrega al Estado como tributaci√≥n (T), y con lo que queda se decide consumir y ahorrar.
La tributaci√≥n total, es:

$$ùëá = ùë°ùëå$$

Donde ‚Äút‚Äù es la tasa de impuestos o de tributaci√≥n promedio. El ingreso neto de tributaci√≥n es el Ingreso Disponible, pues se est√° suponiendo que las transferencias (TR) son iguales a cero:

$$ùëå^ùëë =ùëå‚àíùëá$$
$$ùëå^ùëë =(1‚àíùë°)ùëå$$

Parte de este ingreso disponible se utiliza para los gastos en consumo:

$$ùê∂ = ùê∂_0 + ùëèùëå^ùëë$$

Donde ‚ÄúC0‚Äùes el nivel b√°sico de consumo, conocido como el consumo aut√≥nomo, y ‚Äúb‚Äù representa el cambio en el consumo por unidad de cambio en el ingreso disponible. Es conocida como la Propensi√≥n Marginal a Consumir, (PMgC). Esta propensi√≥n es mayor que cero y menor que uno, 0 < ùëè < 1.

```python

#Par√°metros

Y_size=100

b=0.5
Co=10

Yd=np.arange(Y_size)
Yd
```

```python
#Ecuaci√≥n

def C (Co, b, Yd):
    C = Co + b*Yd
    return C
C = C(Co, b, Yd)
C
```

```python
#Gr√°fico:

plt.plot(Yd, C, color='blue')
plt.title("Gr√°fico 4.2 - Funci√≥n de Demanda de Consumo")
plt.xlabel("C")
plt.ylabel("Yd")
plt.grid()

texto1 = plt.text(30,55, r'$PMgC=\frac{\Delta  C}{\Delta  Y}=b=0.5$', fontsize=15)
texto1 = plt.text(30,15, r'$\Delta  Y^d$', fontsize=15)
texto1 = plt.text(42,25, r'$\Delta  C$', fontsize=15)

texto2 = plt.text(20,19, r'$------$', fontsize=16)
texto2 = plt.text(39,30, r'$|$', fontsize=16)
texto2 = plt.text(39,21, r'$|$', fontsize=16)
texto2 = plt.text(39,26, r'$|$', fontsize=16)

```

2.Derive y explique la funci√≥n de Inversi√≥n. Replique el gr√°fico siguiente:

El gasto de inversi√≥n se puede representar con una funci√≥n que depende inversamente de la tasa de inter√©s real, ‚Äúr‚Äù, y donde ‚ÄúI0‚Äù es el componente aut√≥nomo que depende b√°sicamente de las expectativas de los inversionistas.

$$I = I_0 ‚àí hr$$

Donde:
- Io: Inversi√≥n aut√≥noma (inversi√≥n fija).
    Si sube la inversi√≥n aut√≥noma, sube la inversi√≥n general en el pa√≠s.
- r: Tasa de inter√©s real pasiva.
    Indica la tasa de inter√©s que recibe el banco al realizar pr√©stamos.
- h: Sensibilidad de los inversionistas ante la subida o bajada del inter√©s. Es fija para efectos del modelo.
    Se encuentra entre 0 y 1.


```python

#Par√°metros

Io=100
h=0.5
R_size=50

r=np.arange(R_size)
r
```

```python
#Ecuaci√≥n

def I (Io, h, r):
    I = Io - h*r
    return I
I = I(Io, h, r)
I
```

```python
#Gr√°fico

plt.plot(r, I, color="blue")
plt.title("I = Io - hr \n \n Gr√°fico 4.3 \n Funci√≥n de Demanda de Inversi√≥n")
plt.xlabel("r")
plt.ylabel("I")
plt.xticks([])
plt.yticks([])
```

```python
# 4.Encuentre el nivel de Ingreso de equilibrio Ye. 
#Replique el siguiente gr√°fico con todas descripciones encima de las rectas.

# Par√°metros

Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8 # b > m
m = 0.2
t = 0.3
r = 0.9

Y = np.arange(Y_size)

# Ecuaci√≥n de la curva del ingreso de equilibrio

def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)
```

```python
# Recta de 45¬∞

a = 2.5 

def L_45(a, Y):
    L_45 = a*Y
    return L_45

L_45 = L_45(a, Y)
```

```python
# Gr√°fico

# Dimensiones del gr√°fico
y_max = np.max(DA_IS_K)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, DA_IS_K, label = "DA", color = "#3D59AB") #Demanda agregada
ax.plot(Y, L_45, color = "#404040") #L√≠nea de 45¬∫

# Eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

# L√≠neas punteadas punto de equilibrio
plt.axvline(x=70.5,  ymin= 0, ymax= 0.69, linestyle = ":", color = "grey")
plt.axhline(y=176, xmin= 0, xmax= 0.7, linestyle = ":", color = "grey")

# Texto agregado
    # punto de equilibrio
plt.text(0, 180, '$DA^e$', fontsize = 11.5, color = 'black')
plt.text(72, 0, '$Y^e$', fontsize = 12, color = 'black')
plt.text(0, 152, '$Œ±_o$', fontsize = 15, color = 'black')
    # l√≠nea 45¬∫
plt.text(6, 4, '$45¬∞$', fontsize = 11.5, color = 'black')
plt.text(2.5, -3, '$‚óù$', fontsize = 30, color = '#404040')

# T√≠tulo y leyenda
ax.set(title="El ingreso de Equilibrio a Corto Plazo", xlabel= r'Y', ylabel= r'DA')
ax.legend() #mostrar leyenda

plt.show()
```
