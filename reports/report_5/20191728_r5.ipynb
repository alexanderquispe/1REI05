# A partir del sistema de ecuaciones que representan el modelo DA-OA, halle lo siguiente:

* Encuentre las ecuaciones de Ingreso , tasa de interés ,  de equilibrio (Escriba paso a paso la derivación de estas ecuaciones)

*Despejamos Y, para hallar (r^e):* 

$$ IS: Y=\dfrac{C_0+I_0+G_0-h*r}{1-b*(1-t)} $$

$$ LM: Y=\dfrac{M_0^s}{P_0*k}+\dfrac{j}{k}*r $$

$$ \dfrac{C_0+I_0+G_0-h*r}{1-b*(1-t)}=\dfrac{M_0^s}{P_0*k}+\dfrac{j}{k}*r $$

$$ \dfrac{C_0+I_0+G_0}{1-b*(1-t)}-\dfrac{h}{1-b*(1-t)}*r=\dfrac{M_0^s}{P_0*k}+\dfrac{j}{k}*r $$

$$ \dfrac{C_0+I_0+G_0}{1-b*(1-t)}-\dfrac{M_0^s}{P_0*k}=\dfrac{j}{k}*r+\dfrac{h}{1-b*(1-t)}*r $$

$$ \dfrac{\beta_0}{\beta_1}-\dfrac{M_0^s}{P_0*k}=\dfrac{j}{k}*r+\dfrac{h}{\beta_1}*r $$

$$ \dfrac{\beta_0*P_0*k-\beta_1*M_0^s}{\beta_1*P_0*k}=\dfrac{j*\beta_1+k*h}{k*\beta_1}*r $$

$$ \dfrac{\beta_0*P_0*k-\beta_1*M_0^s}{\beta_1*P_0*k}*\dfrac{k*\beta_1}{j*\beta_1+k*h}=r^e $$

*Despejamos r, para Hallar (Y^e):*

$$ IS: r=\dfrac{\beta_0}{h}-\dfrac{\beta_1}{h}*Y $$

$$ LM: r=-\dfrac{M_0^s}{P_0*j}+\dfrac{k}{j}*Y $$

$$ \dfrac{\beta_0}{h}-\dfrac{\beta_1}{h}*Y=-\dfrac{M_0^s}{P_0*j}+\dfrac{k}{j}*Y $$

$$ \dfrac{\beta_0}{h}+\dfrac{M_0^s}{P_0*j}=\dfrac{\beta_1}{h}*Y+\dfrac{k}{j}*Y $$

$$ \dfrac{\beta_0*P_0*j+M_0^s*h}{h*P_0*j}=\dfrac{\beta_1*j+h*k}{h*j}*Y $$

$$ \dfrac{\beta_0*P_0*j+M_0^s*h}{h*P_0*j}*\dfrac{h*j}{\beta_1*j+h*k}=Y^e $$

* **Grafique el equilibrio simultáneo en el modelo DA-OA**

$$ DA: P=\dfrac{h*M_0^s+j*\beta_0}{h}-Y*\dfrac{j*\beta_1+h*k}{h} $$

$$ OA: P=P^e+\theta*(Y-\bar Y) $$

```python
# OA
Y_size = 100 
Pe = 100
o = 3
Y_bar = 20

Y = np.arange(Y_size)   
#-----------------------
def r_IS(Pe, o, Y, Y_bar):
    r_IS = Pe+o*(Y-Y_bar)
    return r_IS

def Y_IS(Pe, Y_bar, o, P):
    Y_IS = Y_bar+P/o-Pe/o
    return Y_IS

r = r_IS(Pe, o, Y, Y_bar)
#-----------------------
# DA
Y_size = 100

h = 0.8
Ms = 200
j = 1
beta1 = 1.02
k = 2
beta0 = 127

Y = np.arange(Y_size)
#-----------------------
def i_LM(h, Ms, j, beta1, k, beta0, Y):
    i_LM = ((h*Ms+j*beta0)/h)-(Y*(j*beta1+h*k)/h)
    return i_LM

i = i_LM(h, Ms, j, beta1, k, beta0, Y)
#-----------------------
# No hay variación en la DA -> en equilibrio

beta0 = 127
def i_LM_Ms(h, Ms, j, beta1, k, beta0, Y):
    i_LM = ((h*Ms+j*beta0)/h)-(Y*(j*beta1+h*k)/h)
    return i_LM

i_Ms = i_LM_Ms(h, Ms, j, beta1, k, beta0, Y)
#-----------------------
# No hay variación en OA -> en equilibrio

Pe = 100

def r_IS_Go(Pe, o, Y, Y_bar):
    r_IS = Pe+o*(Y-Y_bar)
    return r_IS

def Y_IS_Go(Pe, Y_bar, o, P):
    Y_IS = Y_bar+P/o-Pe/o
    return Y_IS

r_Go = r_IS_Go(Pe, o, Y, Y_bar)
#-----------------------------
y_max = np.max(r)
x_max = Y_IS(Pe, Y_bar, o, P)

v = [0, x_max, 0, y_max]                        
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="DA-OA en equilibrio", xlabel= 'Y', ylabel= 'P')

ax.plot(Y, r, "k-", label = "OA", color = "olive")
ax.plot(Y, r_Go, "k-", label = "OA'", linestyle='dashed', color = "yellowgreen")
ax.plot(Y, i, "b-", label="DA", color = "firebrick")  
ax.plot(Y, i_Ms, "b-", label="DA'", linestyle='dashed', color = "lightcoral")  

plt.axhline(y=192.39, xmin= 0, xmax= 0.50, linestyle='dashed', color = 'lightgrey')
plt.axvline(x=50.8,  ymin= 0, ymax= 0.55, linestyle='dashed', color = 'lightgrey')


plt.plot(50.8, 192.39, marker="o", color="black")
plt.text(50,199, 'E', fontsize = 14, color = 'black')


ax.legend()
plt.show()
```
