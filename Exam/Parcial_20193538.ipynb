### 1. Defina el Multiplicador Keynesiano y dé un ejemplo de cómo influye este en el modelo Ingreso-Gasto Keynesiano.
El multiplicador Keynesiano es un cambio en la magnitud de los componentes autónomos de la demanda agregada que conforman el intercepto, genera un proceso multiplicador del ingreso hasta converger al nuevo ingreso y producto de equilibrio.

Así, el aumento multiplicado del ingreso es resultado de los efectos directos e indirectos ocasionados por el aumento de cualquiera de los componentes autónomos de la demanda agregada

Es representado con la formula de $\frac{1}{1-α_1}=\frac{1}{1-(b-m)(1-t)}$

**POLITICA MONETARIA**
- **Operacion de Mercado Abierto**

El Banco Central hace estas operaciones mediante la compra o venta de activos financieros o bonos a los bancos comerciales.
En las Políticas monetarias expansivas, el banco inyecta dinero a la economía, incrementando sus activos que aumenta la base y la oferta monetaria. En las políticas monetarias contractivas, se venden los bonos al mercado (bancos comerciales) y el banco central retira dinero de la economía, disminuyendo los depósitos de los bancos comerciales, reduciendo la base y oferta monetaria

- **Coeficiente Legal de encaje**

En la Política Monetaria Expansiva, el banco incrementa la cantidad de dinero que tienen los bancos para realizar préstamos mediante la disminución de la tasa de encaje, entonces los bancos comerciales tendrán menos dinero en formas de reservas en el balance del banco central, aumentando la oferta monetaria
En la Política Monetaria Contractiva, el Banco Central aumenta el coeficiente de encaje, los bancos comerciales deben tener una mayor proporción de depósitos en forma de reservas, lo que reduce el multiplicador bancario y, por tanto, origina una disminución de la oferta monetaria

- **Tasa de interes**

También conocida como la tasa de interés de referencia de la política monetaria
En la Política Monetaria Expansiva, el banco central reduce su tasa de interés de referencia, aumentando la cantidad de dinero prestada a los bancos comerciales lo que aumenta la base y oferta monetaria
En la Política Monetaria Contractiva, el banco central aumenta su tasa de interés de referencia, disminuyendo la cantidad de dinero prestada (el crédito) a los bancos comerciales, lo que reduce la base y oferta monetaria

### 1. Encuentre el Ingreso de Equilibrio e identifique el multiplicador keynesiano.
Tengamos en cuenta que:
- Demanda agregada:$DA=C+I+G+(X-M)$

Donde:
-	Consumo: $C=C_0+bY^d$
-	Inversión: $I=I_0-hr$
-	Gobierno: $G=G_0$
-	Exportaciones: $X=X_0$
-	Importaciones: $M=mY^d$

**Entonces el ingreso, al estar en equilibrio, tiene que ser igual a la demanda agregada (DA=Y)
Primero analizemos la demanda agregada, la cual se desdoblaria de la siguiente manera:**

$$DA = C_0 + b(1 − t)Y + I_0 − hr +G_0 + x_1Y^∗ + x_2ε− m_1(1 − t)Y + m2_ε$$


$$DA = C_0 + I_0 +G_0 + [b(1 − t)− m_1 (1 −t)]Y − hr + x_1Y^∗ + (x_2 + m_2)ε$$


$$DA = w −hr + x_1Y^∗ + (x_2 + m_2 )ε_0 − (x_2 + m_2 )ρr + (x_2 + m_2 )ρr^∗ + [(b− m_1)(1− t)]Y$$

**Donde:**

$$w = C_0 +I_0 + G_0$$


$$DA = [w + x_1Y^∗ + (x_2 + m_2 )ε0 + (x_2 + m_2 )ρr^∗ − (h + (x_2 + m_2 )ρ)r] + [(b− m_1)(1 − t)]Y$$

**Entonces si la Demanda Agregada es igual al Ingreso de Eequilibrio, las ecuaciones se igualan:**

$$Y = [w + x_1Y^∗ + (x_2 + m_2 )ε0 + (x_2 + m_2 )ρr^∗ − (h + (x_2 + m_2 )ρ)r] + [(b− m_1)(1 − t)]Y$$


$$Y = \frac{1} {1 − (b− m1)(1 −t)} [w + (x_2 + m_2 )ε_0 + (x_2 + m_2 )ρr ∗ + x_1Y^∗] − [ \frac{h + (x_2 + m_2 )ρ} {1 −(b− m1)(1− t)} ]r$$


**En esta ecuacion, se puede identificar el Multiplicador Keynesiano, el cual es**

$$\frac{1}{1-(b-m)(1-t)}$$
