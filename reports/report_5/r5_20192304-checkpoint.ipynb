```python
Una pol√≠tica fiscal con incremento de la tasa de impuesto, como una variable end√≥gnea, afecta de la siguiente manera

A el modelo de IS-LM

ùë°‚Üë=ùê∂ùëú‚Üì=ùê∑ùê¥‚Üì‚Üíùê∑ùê¥<ùëå=ùëå‚Üì‚Üíùê∑ùê¥=ùëå
 
ùëå‚Üì‚ÜíùëÄùëë‚ÜìùëÄùëë<ùëÄùë†‚Üíùëü‚Üì
 
Y en nuestro modelo DA-OA
ùëå‚Üì‚ÜíŒ∏(ùëå‚àíùëå¬Ø)‚Üì‚ÜíùëÉ‚Üì
 
Partiendo desde nuestro an√°lisis matem√°tico

Modelo IS-LM:
```

```python
Co, Io, Go, Xo, h, r, b, m, t, beta_0, beta_1  = symbols('Co Io Go Xo h r b m t beta_0, beta_1')

k, j, Ms, P, Y = symbols('k j Ms P Y')

beta_0 = (Co + Io + Go + Xo)
beta_1 = ( 1-(b-m)*(1-t) )

Y_eq = (k*beta_0)/(k*h + j*beta_1) - ( beta_1 / (k*h + j*beta_1) )*(Ms/P)
r_eq = (j*beta_0)/(k*h + j*beta_1) + ( h / (k*h + j*beta_1) )*(Ms/P)
```

```python
df_Y_eq_t = diff(Y_eq, t)
print("El Diferencial del Producto con respecto al diferencial de la tasa de inter√©s = ", df_Y_eq_t)

El Diferencial del Producto con respecto al diferencial de la tasa de inter√©s =  Ms*j*(b - m)*(-(1 - t)*(b - m) + 1)/(P*(h*k + j*(-(1 - t)*(b - m) + 1))**2) - Ms*(b - m)/(P*(h*k + j*(-(1 - t)*(b - m) + 1))) - j*k*(b - m)*(Co + Go + Io + Xo)/(h*k + j*(-(1 - t)*(b - m) + 1))**2
```

 
La pendiente es representada por el diferencial de $ ‚àÜt/‚àÜY $: $\frac{Ms*j*(b - m)*(-(1 - t)*(b - m) + 1)/(P*(h*k + j*(-(1 - t)*(b - m) + 1))**2) - Ms*(b - m)/(P*(h*k + j*(-(1 - t)*(b - m) + 1))) - j*k*(b - m)*(Co + Go + Io + Xo)}{(h*k + j*(-(1 - t)*(b - m) + 1))**2} $ y es negativo

```python
df_r_eq_t = diff(r_eq, t)
print("El Diferencial de la tasa de interes con respecto al diferencial de precios = ", df_r_eq_t)
El Diferencial de la tasa de interes con respecto al diferencial de precios =  -Ms*h*j*(b - m)/(P*(h*k + j*(-(1 - t)*(b - m) + 1))**2) - j**2*(b - m)*(Co + Go + Io + Xo)/(h*k + j*(-(1 - t)*(b - m) + 1))**2
```
