## Ejercicio 1

### 1.1 Explique cuáles son los intrumentos de política monetaria que puede utilizar el Banco Central.


El Banco Central es el agente que tiene control de la oferta de dinero nominal. Puede llevar a cabo dos tipos de políticas: Una política monetaria expansiva implica que la cantidad de dinero nominal aumente, mientras que una política monetaria contractiva implica que la cantidad de dinero nominal que circula en la economía disminuya.

### 1.2 Derive la oferta real de dinero y explique cada uno de sus componentes.

La oferta real de dinero se define como la oferta nominal, o masa monetaria, dividida por el nivel de precios.

$$Ms = Mo / P$$

La masa monetaria es la cantidad de dinero que el Banco Central coloca en circulación, y tiene una relación positiva con la oferta real. Por otro lado, el nivel de precios tiene una relación negativa con la oferta real.

### 1.3 Derive la demanda real de dinero. Explique qué papel cumplen los parametros "k" y "j"

La demanda real de dinero se define por dos partes. Primero, la referente a la demanda por transacción y precaución. Esta dependerá principalmente del producto y los ingresos que genera. Mientras más aumentan las ganancias, habrá más demanda para comprar más bienes. Por lo tanto, las variables que se utilizan son la elasticidad del ingreso (k) y el producto (Y). La variable (k) mide la sensibilidad de la demanda de dinero cuando se observa una variación en el producto (Y).

En segundo lugar, la parte referente a la especulación, donde la cantidad de dinero aumenta cuando la tasa de interés baja, pero esta se disminuye cuando la tasa de interés sube, debido a que las personas preferirán guardar su dinero en el banco cuando vean una tasa de interés alta, y viceversa. Por lo tanto, las variables que se utilizarán son la constante de elasticidad en la demanda de dinero (j) y la tasa de interés nominal (i). La variable (j) mide la sensibilidad de la demanda ante la variación de la tasa de interés nominal (i).

Por lo tanto, la ecuación que representa la demanda real sería de la siguiente manera:

$$Md = kY - ji$$

### 1.4 Asumiendo que no hay inflación podemos asumir que i = r. Escriba en terminos reales la ecuación de equilibrio en el mercado de dinero.

Si asumimos que la tasa de interés nominal es igual a la real, el equilibrio significará que la demanda real de dinero es igual a la oferta real.

$$Md = Ms$$

$$Mo = P(kY - jr)$$

### 1.5 Grafique el equilibrio en el mercado de dinero. 

```python
# Parameters
r_size = 100

k = 0.7
j = 0.4                
P  = 30 
Y = 46
MS_0 = 600

r = np.arange(r_size)

# Creamos la funcion de demanda 

def MD(k, j, P, r, Y):
    MD_eq = (k*Y - j*r)
    return MD_eq
MD_0 = MD(k, j, P, r, Y)

# Necesitamos crear la oferta de dinero.
MS = MS_0 / P
MS
```

```python
# Equilibrio en el mercado de dinero

fig, ax1 = plt.subplots(figsize=(10, 8))

# Título
ax1.set(title="Money Market Equilibrium", xlabel=r'M^s / P', ylabel=r'r')

# Demanda de dinero
ax1.plot(MD_0, label= '$L_0$', color = '#33C7FF')

# Oferta de dinero
ax1.axvline(x = MS,  ymin= 0, ymax= 1, color = "grey")
ax1.axhline(y=24.1, xmin= 0, xmax= 0.225, linestyle = ":", color = "black")
ax1.text(0, 24.1, "$r_0$", fontsize = 12, color = 'black')

ax1.yaxis.set_major_locator(plt.NullLocator())   
ax1.xaxis.set_major_locator(plt.NullLocator())

ax1.legend()

plt.show()
```

$$ kY-jr=\frac{M_0^s}{P} $$ 

Se deriva la ecuación de r en función a Y

$$ kY - \frac{M_0^s}{P} = jr $$ 

$$ \frac{kY}{j}-\frac{M_0^s}{Pj} = r $$

La tasa de interés que equilibra el mercado monetario se daría entonces por la ecuación:
$ r=-\frac{1}{j}\frac{M_0^s}{P_0}+\frac{k}{j}Y $

### 3.2 ¿Cuál es el efecto de una disminución en la Masa Monetaria  ΔMs< 0? Explica usando la intuición y gráficos.

Debido a que el valor de la masa monetaria se encuentra en el intercepto en r de la ecuación de la curva LM, con un símbolo negativo, si el valor de esta disminuye, el intercepto en r aumentará y por lo tanto la curva se desplazará hacia la izquierda. 

```python
 # Curva LM ORIGINAL

# Parámetros

Y_size = 100

k = 0.7
j = 0.4 
Ms = 40                
P  = 15               
            

Y = np.arange(Y_size)

# Ecuación

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)

#--------------------------------------------------
    # NUEVA curva LM

# Definir SOLO el parámetro cambiado
Ms = -50

# Generar la ecuación con el nuevo parámetro
def i_LM_Ms( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i_Ms = i_LM_Ms( k, j, Ms, P, Y)

# Dimensiones del gráfico
y_max = np.max(i)
v = [0, Y_size, 0, y_max]   
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, i, label="LM", color = 'black')
ax.plot(Y, i_Ms, label="LM_Ms", color = '#33FFCA', linestyle = 'dashed')

# Texto agregado
plt.text(45, 97, '∆$M^s$', fontsize=12, color='black')
plt.text(45, 103, '←', fontsize=15, color='black')

# Título y leyenda
ax.set(title = "Disminución en la Masa Monetaria $(M^s)$", xlabel=r'Y', ylabel=r'r')
ax.legend()


plt.show()
```

### 3.3 ¿Cuál es el efecto de un aumento en k Δk>0 ? Explica usando intuición y gráficos.

Debido a que el valor de k se encuentra dentro de la pendiente y afecta de manera positiva, un aumento en su valor implicará que la pendiente de la curva LM se empine hacia arriba.

```python
 # Curva LM ORIGINAL

# Parámetros

Y_size = 100

k = 0.7
j = 0.4 
Ms = 40                
P  = 15               
            

Y = np.arange(Y_size)

# Ecuación

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)

#--------------------------------------------------
    # NUEVA curva LM

# Definir SOLO el parámetro cambiado
k = 1

# Generar la ecuación con el nuevo parámetro
def i_LM_k( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i_Ms = i_LM_k( k, j, Ms, P, Y)

# Dimensiones del gráfico
y_max = np.max(i)
v = [0, Y_size, 0, y_max]   
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, i, label="LM", color = 'black')
ax.plot(Y, i_Ms, label="LM_k", color = '#FFC133', linestyle = 'dashed')

# Texto agregado
plt.text(42, 97, '∆$k$', fontsize=12, color='black')
plt.text(42, 103, '↑', fontsize=15, color='black')

# Título y leyenda
ax.set(title = "Incremento en la sensibilidad de la demanda de dinero ante variaciones del ingreso (k)", xlabel=r'Y', ylabel=r'r')
ax.legend()


plt.show()
```
