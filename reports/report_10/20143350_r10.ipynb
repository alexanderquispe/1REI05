# REPORTE DE LECTURA 

La pregunta de investigación es, ¿**qué factores causales determinan quién se convierte en inventor** (un individuo que produce una patente) **en los Estados Unidos**, teniedo en cuenta el rol de la habilidad inventiva y el entorno?

El artículo tiene dos hipótesis: una es que las diferencias se derivan de las diferencias heredadas en talentos o preferencias para seguir la innovación como carrera (*nature*); y la otra es que los niños de diferentes *backgrounds* crecen en diferentes entornos y, por ello, terminan siguiendo diferentes carreras, pues mientras unos se exponen a la innovación cuando son niños, los otros no lo hacen, y esto depende de variables como la clase socioeconómica, la raza y el género (*nurture*). 

La data que usa son los registros de patentes desde 1976 al presente y los registros tributarios (de *federal income tax*) de 1996 a 2012. Asimismo, usa los registros escolares del Distrito de Nueva York para obtener los resultados de los exámenes de los niños que atendieron escuelas públicas en dicha zona. Las tres bases de datos se relacionan usando el nombre del inventor, su ciudad y su sexo.    

Las variables usadas son *income*, *parent's income*, *geographic location*, *college attendance*, *test scores* y *summary statistics*.

Una fortaleza del artículo es que descarta la primera hipótesis, es decir, que las diferencias se basan en *nature*. Así, al analizar las diferencias entre los dos grupos (quiénes se volvieron inventores y quiénes no) en base al ingreso de los padres, la raza y etnicidad y el género; encuentra que estos no son factores causales que explican por sí mismos que alguien se convierta o no en inventor, pues las diferencias entre grupos tomando en cuenta solo estas variables no son significativas. 

Asimismo, en primer lugar, al usar regresiones discontinuas, los autores introducen la exposición a la innovación como variable de tratamiento $T$ y consideran como resultados potenciales si el inventor produce la misma clase de tecnología a la que fue expuesto ($Y_0$) o que no lo hace ($Y_1$). Para ello, determina umbrales $c$ en las variables *parents*, *parents' coworkers* y *neighborhoods*. Así, los autores pueden analizar los resultados en los dos grupos que separa el umbral $c$. De esa forma, aislan el efecto de la exposición. La idea detrás de esto es que las habilidades genéticas no generan diferencias en cuanto a la clase de tecnología a patentar en el futuro, pues estas no determinan la habilidad de patentar cierta clase específica de tecnología.

Luego de su análisis, los autores encuentran que quienes se vuelven inventores usualmente patentan la misma clase de tecnología (*chemicals*, *computers and communications*, *drugs and medical*, *electrical and electronic*, *mechanical*, *design and plant* y *other*) que la tecnología a la que fueron expuestos, sea la que produjeron sus padres, compañeros de trabajo de sus padres o la dominante en la zona en la que vivieron. Asimismo, en el caso de las mujeres, encuentran que estas son más proclives a innovar en una categoría particular de tecnología si hay mujeres innovando en esta.

Sin perjuicio de lo anterior, los autores son cautos al señalar que, si bien la exposición a innovaciones determina la clase de tecnología que patentará un inventor, esto no implica que quien está expuesto a innovaciones va a volverse o no un inventor. Por tanto, analizan si la exposición aumenta el nivel de innovación estudiando los efectos causales de vivir en un determinado vecindario en el nivel de innovación.

Otra vez, los autores usan regresiones discontinuas e introducen la mudanza a áreas con alto nivel de innovación como variable de tratamiento $T$ y consideran como resultados potenciales que el niño de vuelve inventor ($Y_0$) o que no lo hace ($Y_1$). Para ello, determina un umbral $c$ en referencia a la edad en la que los niños se mudaron. En ese sentido, 0 sería los niños que se mudaron en sus años tempranos a áreas con alto nivel de innovación y 1 los niños que lo hicieron luego. Así, otra vez, los autores pueden analizar los resultados en los dos grupos que separa el umbral $c$ y aislar el efecto de la exposición.

Luego de realizar los cálculos e interpretaciones, los autores llegan a la conclusión de que la exposición a innovación tiene impactos sustanciales tanto en el tipo de innovación que realizarán los niños como en si estos se vuelven  o no inventores.

El *paper* realiza una contribución a la ya amplia literatura que señala que el éxito futuro de los niños depende de su entorno, pues las investigaciones se han enfocado sobre todo cómo los barrios donde los niños viven y las escuelas que atienden afectan su *performance*, más no en la exposición (en este caso, a inventos). Asimismo, resuelve el debate de *nature* vs. *nurture* en el caso de los inventores, pues demuestra que la exposición es fundamental para que un niño (o una niña) devenga en inventor.

## Regresión discontinua 

### 1. Explique de manera muy breve y resumida en qué consiste una regresión discontinua.

La regresión discontinua sirve para explicar efectos causales cuando se producen "saltos artificiales". Para ello, se tiene una variable de tratamiento $T$ y los resultados potenciales (por ejemplo, $Y_1$ y $Y_2$). En ese sentido, $T$ es 0 cuando está por debajo del umbral y 1 cuando está por encima. $T$ es una función discontinua de una variable móvil observada $R$.

En comparar el resultado por encima y debajo de un umbral $c$ para identificar el efecto del tratamiento en el umbral. De esa forma, podemos observar $Y_1$ cuando $R>c$ y $Y_0$ cuando $R<c$. Entonces, $c$ permite ver cada una de las funciones, pero no ambas.

### 2. Detalle un ejemplo en la vida real en donde podría emplear una regresión discontinua.

Los promedios ponderados de los cursos en todas las facultades de la PUCP subieron considerablemente durante la pandemia, resaltando el caso de la Facultad de Ciencias e Ingeniería, que tiene los promedios más bajos de la universidad. La variable de tratamiento $T$ es la virtualización de los cursos. Los resultados potenciales son un promedio de notas mayor a 12 ($Y_1$) y un promedio de las notas menor a 12 ($Y_2$). El umbral $c$ es 12. De esa forma, podemos observar el efecto de $T$ por encima y debajo del umbral $c$.

### 3. Explique el diseño Sharp Desing y comente un ejemplo en la vida real en donde podría aplicarlo (sea lo más detallado posible).

El diseño *Sharp Design* no es sino una regresión discontinua en la que la posibilidad del tratamiento salta de 0 a 1 en el umbral $c$, lo que permite comparar los resultados por encima y por debajo del umbral, a fin de identificar el efecto del tratamiento en $c$.

Un ejemplo en la vida real en donde podría aplicarse el *Sharp Design* es

En ese sentido, $T$ sería la entrega del subsidio por parte del Gobierno peruano y el resultado que queremos analizar es si el empleo se mantuvo o no gracias al subsidio. El umbral $c$ sería la remuneración bruta de S/ 1500. El 0 correspondería a los trabajadores que perciben remuneraciones brutas de hasta S/ 1500 mensuales, mientras que el 1 correspondería a los trabajadores que perciben remuneraciones brutas mayores a esa cantidad.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns, warnings
import statsmodels.formula.api as smf
warnings.filterwarnings('ignore')
from causalgraphicalmodels import CausalGraphicalModel
from cProfile import label
%matplotlib inline
```

### Ejemplo 1

$$Y = 15 + 22D + 2.5X + \epsilon$$

#### Gráfico 

```python
dat = pd.DataFrame({'x': np.random.normal(50, 25, 1000)}) 
dat.loc[dat.x<0, 'x'] = 0 
dat = dat[dat.x<100]
dat['D'] = 0 

dat.loc[dat.x>50, 'D'] = 1

dat['y'] = 15 + (22*dat.D) + (2.5*dat.x) + np.random.normal(0, 20, dat.shape[0])

dat_0 = dat[dat.D == 0]

dat_1 = dat[dat.D == 1]
```

```python
treat_c = "#568A9B" 
n_treat = "#A10728" 

plt.scatter("x", "y", data = dat_0, color = n_treat, alpha= .7, s = 7, label = "Not treatment") 
plt.scatter("x", "y", data = dat_1, color = treat_c, alpha= .7, s = 7, label = "Treatment")
plt.axvline(50, color = "black", linestyle = "-.", label = r"$c_0$: Cutoff") 

plt.legend(); 
```

```python
def y_hat_lin(da_ta, x = "x", y = "y"): 
    yhat = smf.ols(f"{y} ~ {x}", data = da_ta).fit().predict()
    da_ta["yhat"] = yhat
    da_ta.sort_values('yhat', inplace = True)
    return da_ta

data_y0 = y_hat_lin(dat_0)
data_y1 = y_hat_lin(dat_1)
```

```python
plt.scatter("x", "y", data = dat_0, color = n_treat, alpha= .2, s = 7, label = "Not treatment")
plt.scatter("x", "y", data = dat_1, color = treat_c, alpha= .2, s = 7, label = "Treatment")
plt.axvline(50, color = "black", linestyle = "-.", label = r"$c_0$: Cutoff")

plt.plot("x", "yhat", data = data_y0, color = n_treat, lw = 3)
plt.plot("x", "yhat", data = data_y1, color = treat_c, lw = 3);
```

#### Jump effect

```python
eff = min(data_y1.yhat) - max(data_y0.yhat) 
eff = round(eff,  2)
print("Effect:", f"{eff}")
```

```python
plt.scatter("x", "y", data = dat_0, color = n_treat, alpha= .03, s = 7, label = "Not treatment")
plt.scatter("x", "y", data = dat_1, color = treat_c, alpha= .03, s = 7, label = "Treatment")
plt.axvline(50, color = "black", linestyle = "-.", label = r"$c_0$: Cutoff")

plt.plot("x", "yhat", data = data_y0, color = n_treat, lw = 3)
plt.plot("x", "yhat", data = data_y1, color = treat_c, lw = 3)
plt.axvline(50, color = "black", linestyle = "-.", label = r"$c_0$: Cutoff", lw = 1)
plt.annotate(f"{eff}", xy = (51, eff + 76))
plt.annotate("", xy = (51, max(data_y0.yhat)), xytext = (51, min(data_y1.yhat)), 
            arrowprops=dict(arrowstyle="<->",
                            connectionstyle="arc3"));
```
