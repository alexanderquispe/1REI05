### Who becomes an inventor in America? The importance of exposure to innovation -Bell, Chetty, Jaravel, Petkova y Van Reenen (2019). 

El texto parte de la premisa de que la innovación es una de las claves para lograr el crecimiento económico. Sin embargo, hay evidencia que muestra que existen brechas entre aquellos que se convierten en inventores y los que no (entendiendo como "inventor" a todo aquel que posea una patente). Tomando todo ello como punto de partida, Bell et al. (2019) buscan identificar qué factores influyen en que las personas se conviertan en inventores, enfocándose en los individuos y no en factores externos o sistémicos. Ello lo hacen entendiendo que, al analizar los factores llevan a que los propios individuos se conviertan o no en inventores, se pueden identificar nuevos enfoques para aumentar las tasas de innovación entre los grupos subrepresentados en innovación.

El documento avanza en la pregunta de investigación argumentando que uno de los factores críticos que determinan quién se convierte en inventor es la exposición de los individuos a la innovación durante su infancia (entendiendo que la ausencia de exposición a la innovación es un fuerte indicador del porqué los niños talentosos en familias de bajos ingresos, grupos minoritarios y mujeres tienen menor probabilidad de convertirse en inventores); dejando de lado a quienes podrían producir las innovaciones más importantes del futuro. Por ello, la principal contribución del texto hacia la literatura sobre innovación es que presenta nueva evidencia sobre los factores que determinan quién se convierte en inventor a través del seguimiento de la vida de los inventores en Estados Unidos desde el nacimiento hasta la edad adulta, a diferencia del resto de trabajos relacionados a innovación, que tratan sobre factores como incentivos financieros, las barreras de entrada y la educación. 

Con ello, los autores proponen que las políticas que aumentan el nivel de exposición de las personas a innovaciones tienen la capacidad de incrementar también la cantidad de inventos y descubrimientos que se harán. Estas políticas para aumentar la exposición a la innovación podrían ir desde brindar tutoría por parte de inventores hasta otorgar programas de pasantías en empresas locales; pero los autores no tratan de explicar qué políticas específicas se deben aplicar, sino que buscan servir como guía de cómo dichas políticas deberían ser orientadas (enfocadas principalmente en mujeres, grupos minoritarios y niños de familias de escasos recursos, pero que son sobresalientes en aspectos académicos como las matemáticas y las ciencias). Además, destacan que dichos programas pueden incrementar su impacto en gran medida si estos se adaptan a los antecedentes de las personas a las que son destinados.

Otra de las contribuciones del texto se realiza a la literatura sobre cómo las perspectivas de éxito de los niños están moldeadas por sus entornos, ya que los estudios que suelen realizarse se centran en la acumulación general de capital humano, mientras que este estudio se enfoca en el efecto que tiene el medio ambiente en el que se desarrollan las personas sobre las trayectorias profesionales que eligen y, por ende, en su desempeño como inventores. Además, explican que el incremento de la movilidad intergeneracional también podría impactar de manera positiva en el crecimiento económico del país ya que, al incrementar los ingresos de las personas de bajos ingresos y minorías, se disminuye la desigualdad, y se estimula el crecimiento económico mediante el aprovechamiento de talento en innovación. 

Quizás, la principal fortaleza de este texto está en la metodología utilizada, ya que usan datos anónimos sobre millones de inventores de registros de patentes vinculados a registros fiscales para mostrar que las posibilidades de que los niños se conviertan en inventores varían según sus características personales (como su raza, género y la clase socioeconómica de los padres), a pesar de que cuenten con un excelente desempeño académico (siendo los niños de altos ingresos los que tienen mayores probabilidades de convertirse en inventores, a comparación de aquellos con escasos ingresos), y según las diferencias del entorno en el que crecieron, más allá de que tengan o no habilidades para innovar. Así, también es una de las mayores fortalezas del estudio el hecho de que logra establecer la importancia del medio ambiente en la formación de futuros inventores al mostrar que la exposición a la innovación durante la infancia tiene importantes efectos causales en la propensión de los niños a inventar (los niños que viven en un área de alta innovación cuando son jóvenes tienen más probabilidades de convertirse en inventores que los que no), y en reflejar que también existe una diferencia de género en este aspecto (las niñas, un grupo subrepresentado en la innovación, son más propensas a inventar en una clase en particular si crecen en un ambiente con más mujeres que inventan).

Sin embargo, el estudio también cuenta con debilidades. La mayor de ellas es el hecho de que los resultados son difícilmente extrapolables a cualquier otro país, ya que los autores proponen una guía para orientar hacia dónde se deben dirigir las políticas relacionadas a una mayor exposición a la innovación en la principal potencia internacional. Ello quiere decir que, para que dicho estudio sea aplicable en otro país, este debe ser uno que cuente con los recursos suficientes como para realizar políticas enfocadas en la innovación. Sin embargo, una parte sustancial de los países del mundo siguen sin desarrollar su aparato de innovación por ser, principalmente, exportadores de materias primas, así que todavía no cuentan con datos que permitan desarrollar el estudio, y probablemente tampoco existan brechas por eliminar en cuestiones de innovación precisamente por el poco desarrollo en ese aspecto de muchos países; aunque sí podría servir como un aspecto a considerar en cuanto a los grupos que no se debería dejar de lado al momento de invertir en el futuro desarrollo de la innovación del país. Así que, en cuestiones de aplicabilidad, solo podría extrapolarse a países con experiencia en innovación, con recursos para desarrollar e implementar las políticas orientadas al aumento de la exposición, y bajo gobiernos liberales que se interesen por el bienestar económico y desarrollo tanto de su población como del país. 

Finalmente, para lo que los autores proponen, se pueden identificar algunos pasos a seguir en el artículo de Guzmán, Velasco y Martínez (2019), que trata de identificar aquellos factores que afectan la propensión de los investigadores académicos de México de ser inventores y su productividad a través de la propuesta de tres modelos econométricos. Para ello, su enfoque está orientado hacia el uso de micro datos de patentes durante un período de tiempo determinado, y de datos individuales e institucionales de individuos que ya son investigadores; obteniendo así resultados que se asocian a factores personales e institucionales, como la edad de los investigadores, el tamaño de la institución en la que estudiaron y la calidad o el tipo de programas de doctorado que realizaron. Además, podría ser de utilidad el artículo de Tahmooresnejad y Turkina (2022), cuyo objetivo es entender los factores que afectan la producción de innovación de mujeres inventoras. ### Regresión Discontinua

* **Explique en qué consiste una regresión discontinua.**

La regresión discontinua es un método cuasiexperimental que busca evaluar los efectos causales de algún tratamiento (como políticas o programas) que implican una discontinuidad en la implementación del mismo, por lo que la probabilidad de participación en él cambia de manera discontinua en un punto de corte de la variable de asignación (el umbral); la cual debe ser observable y el tratamiento debe depender de ella. Así, la regresión discontinua permitirá ver la relación causal de cambios en el estatus de tratamiento sobre la variable dependiente.

* **Detalle un ejemplo en la vida real en donde podría emplear una regresión discontinua.**

Un ejemplo en la vida real en donde podría emplear una regresión discontinua es para realizar un estudio sobre los efectos causales de cambios en la matriculación a una beca otorgada según el puntaje obtenido en un examen académico, ya que la unidad de análisis tiene un puntaje mínimo que se debe alcanzar para obtener la beca, es una situación que no puede ser manipulada ni predicha por los postulantes, y cuenta con un punto de corte que es el puntaje mínimo en el examen para la obtención de la beca. 


* **Explique el diseño Sharp Desing y comente un ejemplo en la vida real en donde podría aplicarlo (sea lo más detallado posible).**

El Sharp Design es un tipo de regresión ocurre cuando la variable de asignación determina completamente la participación en el grupo de tratamiento a partir del punto de corte, así que la probabilidad de formar parte del grupo de tratamiento cambia discontinuamente de 0 a 1 en el punto de corte que determina la discontinuidad (se asume que los individuos no tienen ningún control sobre la decisión de participación en el grupo de tratamiento, esta asignación no es manipulable).

Un ejemplo de ello podría ser el caso del beneficio al impuesto predial para adultos mayores, una política establecida por el gobierno peruano que, para otorgar el beneficio, implica una asignación a través de la edad de las personas adultas mayores, contando con la probabilidad de recibir el beneficio todas las personas adultas mayores de 60 años de edad. Así, es dicha edad el punto de corte a partir del cual la probabilidad de participación cambia de 0 a 1. En este contexto, y considerando que  el beneficio al impuesto predial para adultos mayores se implementa de forma simultánea en todo el territorio nacional a toda la población adulta mayor de 60 años de edad, el grupo de control sería el de aquellos no beneficiarios (las personas menores de 60 años); logrando así que las características del grupo de control sean similares a las del experimental, con la diferencia de que los del grupo de control no se beneficia de dicha política. A ello se le suma el hecho de que no es probable que los adultos mayores, pero menores a 60 años, pueda manipular la variable de asignación para recibir el beneficio (porque no pueden modificar su edad), así que da seguridad que solo los individuos del grupo de tratamiento, los que están a la derecha del punto de corte, sean los que realmente pertenecen ahí (los que tienen más de 60 años).

$$Ejemplo 1: Y= 15+22D+2.5X +ϵ$$

* Realice un gráfico: 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns, warnings
import statsmodels.formula.api as smf
warnings.filterwarnings('ignore')
from causalgraphicalmodels import CausalGraphicalModel
from cProfile import label
%matplotlib inline
```

```python
dat = pd.DataFrame({'x': np.random.normal(50, 25, 1000)})
dat.loc[dat.x<0, 'x'] = 0
dat = dat[dat.x<100]
dat['D'] = 0
# Cutoff
dat.loc[dat.x>50, 'D'] = 1
# Simulated Y
dat['y'] = 15 + 22*dat.D + 2.5 * dat.x + np.random.normal(0, 20, dat.shape[0])
# Not Treatment
dat_0 = dat[dat.D == 0]
# Treatment
dat_1 = dat[dat.D == 1]
```

```python
# Colors 
treat_c = "#568A9B" # Treatment
n_treat = "#A10728" # Not treatment
# Plot
plt.scatter("x", "y", data = dat_0, color = n_treat, alpha= .7, s = 7, label = "No tratamiento")
plt.scatter("x", "y", data = dat_1, color = treat_c, alpha= .7, s = 7, label = "Tratamiento")
plt.axvline(50, color = "black", linestyle = "-.", label = r"$c_0$: Corte")
plt.legend();
```

```python
def y_hat_lin(da_ta, x = "x", y = "y"):
    yhat = smf.ols(f"{y} ~ {x}", data = da_ta).fit().predict()
    da_ta["yhat"] = yhat
    da_ta.sort_values('yhat', inplace = True)
    return da_ta

data_y0 = y_hat_lin(dat_0)
data_y1 = y_hat_lin(dat_1)
```

```python
plt.scatter("x", "y", data = dat_0, color = n_treat, alpha= .2, s = 7, label = "No tratamiento")
plt.scatter("x", "y", data = dat_1, color = treat_c, alpha= .2, s = 7, label = "Tratamiento")
plt.axvline(50, color = "black", linestyle = "-.", label = r"$c_0$: Corte")

plt.plot("x", "yhat", data = data_y0, color = n_treat, lw = 3)
plt.plot("x", "yhat", data = data_y1, color = treat_c, lw = 3);
```

* Encontrar the Jump Effect y plotearlo: 

```python
eff = min(data_y1.yhat) - max(data_y0.yhat)
eff = round(eff,  2)
print("Effect:", f"{eff}")
```

```python
plt.scatter("x", "y", data = dat_0, color = n_treat, alpha= .03, s = 7, label = "No tratamiento")
plt.scatter("x", "y", data = dat_1, color = treat_c, alpha= .03, s = 7, label = "Tratamiento")
plt.axvline(50, color = "black", linestyle = "-.", label = r"$c_0$: Corte")

plt.plot("x", "yhat", data = data_y0, color = n_treat, lw = 3)
plt.plot("x", "yhat", data = data_y1, color = treat_c, lw = 3)
plt.axvline(50, color = "black", linestyle = "-.", label = r"$c_0$: Corte", lw = 1)
plt.annotate(f"{eff}", xy = (51, eff + 76))
plt.annotate("", xy = (51, max(data_y0.yhat)), xytext = (51, min(data_y1.yhat)), 
            arrowprops=dict(arrowstyle="<->",
                            connectionstyle="arc3"));
```
