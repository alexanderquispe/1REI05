Regresión Discontinua

a) Explique de manera muy breve y resumida en qué consiste una regresión discontinua.

La regresión discontinua es un diseño cuasi-experimental que permite realizar inferencia causal en ausencia de aleatorización. Se puede aplicar cuando la exposición de interés se asigna, al menos parcialmente, según el valor de una variable aleatoria continua si esta variable cae por encima o por debajo de un determinado valor umbral.

La idea de la discontinuidad de la regresión es comparar el resultado justo por encima y por debajo del umbral para identificar el efecto del tratamiento en el umbral. A esto se le llama un diseño de sharp RD o RD nítido, ya que la probabilidad de obtener el tratamiento salta de 0 a 1 en el umbral. Sin embargo, podríamos pensar en un diseño de fuzzy RD o RD borroso, donde la probabilidad también salta, pero es una manera menos dramática.

Por ejemplo, imagine que tiene una variable de tratamiento T y resultados Y0 y Y1 potenciales. El tratamiento T es una función discontinua de una variable observada R. El tratamiento es cero cuando esta por debajo de un umbral y uno en caso contrario.

Para estimar los impactos del alcohol en la muerte, podríamos usar el hecho de que la edad legal para beber impone una discontinuidad en la naturaleza.

En este contexto, podemos pensar en RDD como un ensayo aleatorio local. Para aquellos en el umbral, el tratamiento podría haber ido en cualquier dirección y, por casualidad, algunas personas cayeron por debajo del umbral y otras por encima. En nuestro ejemplo, en el mismo momento, algunas personas tienen poco más de 21 años y algunas personas tienen menos de 21. Lo que determina esto es si alguien nació algunos días después o no, lo cual es bastante aleatorio. Por esto el diseño de regresión discontinua(RDD) proporciona una historia causal muy convincente.

Este modelo nos está diciendo que la mortalidad aumenta en 7,6627 puntos con el consumo de alcohol. Puedes ver que es como si tuviéramos 2 modelos de regresión: uno para los que están por encima del umbral y uno por debajo de él. La discontinuidad de la regresión se basa en gran medida en las propiedades de extrapolación de la regresión lineal. Lo que puede suceder es que la regresión se centre demasiado en ajustar los otros puntos de datos a costa de un mal ajuste en el umbral.

Con la excepción del suicidio, parece que agregar el peso del grano hizo que el impacto negativo en el alcohol fuera mayor. Una vez más, si queremos minimizar la tasa de mortalidad, NO deberíamos recomendar bajar la edad legal para beber, ya que hay un claro impacto del alcohol en la muerte. EJEMPLO 1:


$$
    Y = 15 + 22*D + 2.5 * X + e
$$

```python
## Random data
dat = pd.DataFrame({'x': np.random.normal(50, 25, 1000)})
dat.loc[dat.x<0, 'x'] = 0
dat = dat[dat.x<100]
dat['D'] = 0
# Cutoff
dat.loc[dat.x>50, 'D'] = 1
# Simulated Y
dat['y'] = 15 + 22*dat.D + 2.5 * dat.x + np.random.normal(0, 20, dat.shape[0])
# Not Treatment
dat_0 = dat[dat.D == 0]
# Treatment
dat_1 = dat[dat.D == 1]
```

```python
# Colors 
treat_c = "#568A9B" # Treatment
n_treat = "#A10728" # Not treatment
# Plot
plt.scatter("x", "y", data = dat_0, color = n_treat, alpha= .7, s = 7, label = "Not treatment")
plt.scatter("x", "y", data = dat_1, color = treat_c, alpha= .7, s = 7, label = "Treatment")
plt.axvline(50, color = "black", linestyle = "-.", label = r"$c_0$: Cutoff")
plt.legend();
```

```python
def y_hat_lin(da_ta, x = "x", y = "y"):
    yhat = smf.ols(f"{y} ~ {x}", data = da_ta).fit().predict()
    da_ta["yhat"] = yhat
    da_ta.sort_values('yhat', inplace = True)
    return da_ta

data_y0 = y_hat_lin(dat_0)
data_y1 = y_hat_lin(dat_1)
```

```python
plt.scatter("x", "y", data = dat_0, color = n_treat, alpha= .2, s = 7, label = "Not treatment")
plt.scatter("x", "y", data = dat_1, color = treat_c, alpha= .2, s = 7, label = "Treatment")
plt.axvline(50, color = "black", linestyle = "-.", label = r"$c_0$: Cutoff")

plt.plot("x", "yhat", data = data_y0, color = n_treat, lw = 3)
plt.plot("x", "yhat", data = data_y1, color = treat_c, lw = 3);
```

```python
eff = min(data_y1.yhat) - max(data_y0.yhat)
eff = round(eff,  2)
print("Effect:", f"{eff}")
```

```python
plt.scatter("x", "y", data = dat_0, color = n_treat, alpha= .03, s = 7, label = "Not treatment")
plt.scatter("x", "y", data = dat_1, color = treat_c, alpha= .03, s = 7, label = "Treatment")
plt.axvline(50, color = "black", linestyle = "-.", label = r"$c_0$: Cutoff")

plt.plot("x", "yhat", data = data_y0, color = n_treat, lw = 3)
plt.plot("x", "yhat", data = data_y1, color = treat_c, lw = 3)
plt.axvline(50, color = "black", linestyle = "-.", label = r"$c_0$: Cutoff", lw = 1)
plt.annotate(f"{eff}", xy = (51, eff + 76))
plt.annotate("", xy = (51, max(data_y0.yhat)), xytext = (51, min(data_y1.yhat)), 
            arrowprops=dict(arrowstyle="<->",
                            connectionstyle="arc3"));
```
