## Regresion Discontinua
#### 1. Explique de manera muy breve y resumida en qué consiste una regresión discontinua.

La regresión discontinua es un diseño cuasi-experimental y este nos permite poder realizar inferencias causales en ausencia de la aleatorización. Se aplica cuando la exposición de interés se asigna, por lo menos en un ámbito parcial, en función del valor de una variable aleatoria continua, si es que esta se encuentra por encima o por debajo de un valor umbral determinado. En otras palabras analiza las causas de la intervención estudiada a través de la asignación de un valor de corte que se encuentre debajo o arriba del asignado por la intervención. 


#### 2. Detalle un ejemplo en la vida real en donde podría emplear una regresión discontinua.

## Aplicaciones
### CASO 1 :
$$ Y = 15 + 22D + 2.5 X + e $$

```python
## Random data
caso1 = pd.DataFrame({'x': np.random.normal(50, 25, 1000)})
caso1.loc[caso1.x<0, 'x'] = 0
caso1 = caso1[caso1.x<100]
caso1['D'] = 0
# Cutoff
caso1.loc[caso1.x>50, 'D'] = 1
# Simulated Y
caso1['y'] = 15 + 22*caso1.D + 2.5 * caso1.x + np.random.normal(0, 20, caso1.shape[0])
# Not Treatment
caso1_0 = caso1[caso1.D == 0]
# Treatment
caso1_1 = caso1[caso1.D == 1]
```

```python
# Colors 
treat_c = "#568A9B" # Treatment
n_treat = "#A10728" # Not treatment
# Plot
plt.scatter("x", "y", data = caso1_0, color = n_treat, alpha= .7, s = 7, label = "Not treatment")
plt.scatter("x", "y", data = caso1_1, color = treat_c, alpha= .7, s = 7, label = "Treatment")
plt.axvline(50, color = "black", linestyle = "-.", label = r"$c_0$: Cutoff")
plt.legend();
```

```python
def y_hat_lin(da_ta, x = "x", y = "y"):
    yhat = smf.ols(f"{y} ~ {x}", data = da_ta).fit().predict()
    da_ta["yhat"] = yhat
    da_ta.sort_values('yhat', inplace = True)
    return da_ta

caso1_y0 = y_hat_lin(caso1_0)
caso1_y1 = y_hat_lin(caso1_1)
```

```python
plt.scatter("x", "y", data = caso1_0, color = n_treat, alpha= .2, s = 7, label = "Not treatment")
plt.scatter("x", "y", data = caso1_1, color = treat_c, alpha= .2, s = 7, label = "Treatment")
plt.axvline(50, color = "black", linestyle = "-.", label = r"$c_0$: Cutoff")

plt.plot("x", "yhat", data = caso1_y0, color = n_treat, lw = 3)
plt.plot("x", "yhat", data = caso1_y1, color = treat_c, lw = 3);
```

### - Jump Effect 1

```python
eff = min(caso1_y1.yhat) - max(caso1_y0.yhat)
eff = round(eff,  2)
print("Effect:", f"{eff}")
```

```python
plt.scatter("x", "y", data = caso1_0, color = n_treat, alpha= .03, s = 7, label = "Not treatment")
plt.scatter("x", "y", data = caso1_1, color = treat_c, alpha= .03, s = 7, label = "Treatment")
plt.axvline(50, color = "black", linestyle = "-.", label = r"$c_0$: Cutoff")

plt.plot("x", "yhat", data = caso1_y0, color = n_treat, lw = 3)
plt.plot("x", "yhat", data = caso1_y1, color = treat_c, lw = 3)
plt.axvline(50, color = "black", linestyle = "-.", label = r"$c_0$: Cutoff", lw = 1)
plt.annotate(f"{eff}", xy = (51, eff + 76))
plt.annotate("", xy = (51, max(caso1_y0.yhat)), xytext = (51, min(caso1_y1.yhat)), 
            arrowprops=dict(arrowstyle="<->",
                            connectionstyle="arc3"));
```
