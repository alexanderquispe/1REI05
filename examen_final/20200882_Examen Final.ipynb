## 2. (3 puntos) Regression Discontinuity Design (RDD)

Supongamos que estamos estudiando la eficacia de un programa de becas creado por el gobierno Peruano sobre las notas de los alumnos al final del primer semestre en la universidad. PRONABEC ha decidido que entregará becas a aquellos postulantes que tengan un promedio superior durante la secundaria igual a 16 .

Identifique todos los elementos necesarios para implementar un RDD en este ejemplo.

En primer lugar, la variable donde se realizará el corte, que determinará si se forma parte del grupo de control o de tratamiento, que en este caso sería el promedio durante la secundaria. En segundo lugar, el corte o cutoff, la variable que determina a cada caso si forma parte de un determinado grupo, dependiendo de su posición respectiva al corte (superior o inferior), en este caso sería a partir del 16. En tercer lugar, se puede determinar el bandwidth apropiado con tal que la distribución no se presente muy escalonada. Finalmente, podemos considerar el supuesto de continuidad. Este implica que el punto de corte no se dé al mismo momento, o sea endógeno, a alguna otra intervención que pueda influir en los resultados. Por tanto, todas las características del resultado deben ser continuas, a excepción de la implicancia del tratamiento.

Explique el procedimiento para encontrar el efecto causal en este escenario. Puede utilizar ecuaciones.

La manera de determinar el efecto causal sería a través de una comparación del promedio de notas entre ambos grupos determinados por el cutoff, aquellos que forman parte de la beca vs aquellos que no. Se puede esperar que la regresión discontinua aquí sea de tipo fuzzy, por lo que la influencia de la beca sobre el promedio de notas no sería tan marcada. Para hayar el efecto causal podemos determinarlo como una proporción entre la diferencia reducida del promedio de los resultados alrededor del corte y de la diferencia reducida del promedio de la asignación del tratamiento alrededor del corte.
