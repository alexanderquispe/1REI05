# Solución
## Función de demanda de Consumo

La función demanda de consumo representa el ingreso se utiliza en los gastos de consumo. De manera que a mayor sea tu ingreso disponible, va a ser mayor tu consumo. Entonces, existe una relación entre el ingreso disponible y el consumo

##### C es Consumo variable
##### Co es Consumo autónomo
##### b es Propensión marginal a consumir -> variable -> b pertenece a ]0, 1[

Donde: 
$$ C = Co + bY^d $$

```python
# Parámetros

Y_size = 100 

Co = 45
b = 0.9
t = 0.4

Y = np.arange(Y_size)

# Ecuación de la curva del ingreso de equilibrio

def C(Co, b, t, Y):
    C = Co + b*(1-t)*Y
    return C

C = C(Co, b, t, Y)
```

```python
# Recta de 45°

a = 2 

def L45(a, Y):
    L45 = a*Y
    return L45

L45 = L45(a, Y)
```

```python
# Gráfico

# Dimensiones del gráfico
y_max = np.max(C)
fig, ax = plt.subplots(figsize=(10, 8))

custom_xlim = (0, 130)
custom_ylim = (0, 130)

plt.setp(ax, xlim=custom_xlim, ylim=custom_ylim)


# Curvas a graficar
ax.plot(Y, C, label = "Con", color = "#3D59AB") #Demanda agregada
ax.plot(L45, color = "#404040") #Línea de 45º

# Eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

# Líneas punteadas punto de equilibrio
plt.axvline(x=70.5,  ymin= 0, ymax= 0.69, linestyle = ":", color = "grey")
plt.axhline(y=176, xmin= 0, xmax= 0.7, linestyle = ":", color = "grey")

# Texto agregado
    # punto de equilibrio
plt.text(80, 95, '$PMgC = △C/△Y^d = b = 0.5$', fontsize = 13, color = 'black')
   # línea 45º
plt.text(6, 4, '$45°$', fontsize = 11.5, color = 'black')
plt.text(2.5, -3, '$◝$', fontsize = 30, color = '#404040')


# Título y leyenda
ax.set(title="Función de demanda de Consumo", xlabel= '$Y^d$', ylabel= 'C')
ax.legend() #mostrar leyenda

plt.show()
```

## Función de inversión

La función es:

La inversión fija nos permite comprender la forma en que qué medida afecta la inversión en la economía. Asimismo, la inversión fija tiene que ver con la tasa de interés. En  este sentido, si sube la inversión autónoma, que viene de las empresas, la inversión general también sube.
	I  Inversión fija
	Io  Inversión autónoma
	r  tasa de interés real -> real pasivo
	h  sensibilidad de los economistas ante la subida o bajada de la tasa de interés

Donde:
$$ I = I_0 - hr $$

```python
# Parámetros

Io = 60
h = 0.7

r = np.arange(start = 0.01, stop = 1, step =0.01)  
r

# Ecuación de la curva del ingreso de equilibrio

def I(Io, h, r):
    I = Io - h*r
    return I

I = I(Io, h, r)


# Gráfico

# Dimensiones del gráfico
y_max = np.max(r)
fig, ax = plt.subplots(figsize=(10, 8))

custom_xlim = (0.1, 1.25)
custom_ylim = (59.1, 60.2)

plt.setp(ax, xlim=custom_xlim, ylim=custom_ylim)


# Curvas a graficar
ax.plot(r, I, label = "Inversión", color = "#3D59AB") #Demanda agregada

# Eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

# Texto agregado
    # punto de equilibrio
plt.text(0.8, 39.5, '$I = Io - hr$', fontsize = 12.5, color = 'black')


# Título y leyenda
ax.set(title="Función de demanda de Inversión", xlabel= 'r', ylabel= 'I')
ax.legend() #mostrar leyenda

plt.show()
```
