1. Derive y explique la función de demanda de Consumo.

La función de demanda de Consumo $(C)$ está determinada por diversas variables, principalmente por el consumo  autónomo $(Co)$, la propensión marginal a consumir $(b)$ y el ingreso disponible $(Y^d)$.

$$ C = C_0 + bY^d $$

teniendo en cuenta que el valor de $b < 0$

Sin embargo, dicha función deriva de otras funciones previas. En primer lugar, se tiene la función de la Tributación total $(T)$, que es el producto de la tasa impositiva y el ingreso agregado de toda la economía.

$$ T = tY $$

donde:
- T: Tributación total del país
- t: tasa impositiva
- Y: Ingreso agregado de toda la economía

En segundo lugar, es necesario determinar el ingreso disponible de toda la economía $(Y^d)$, el cual resulta de la sustracción entre el ingreso agregado $(Y)$ y la tributación total $(T)$.

$$ Y^d = Y - T $$

donde: $$ T = tY $$

entonces: $$ Y^d = Y - tY $$

siendo la ecuación final: $$ Y^d = (1 - t)Y $$

entonces, se encuentra que, si la tasa de impositiva aumenta, el ingreso disponible disminuye, y viseversa.

Finalmente, en ese sentido, la función final de la demanda de Consumo resulta en la siguiente: $$ C = C_0 + b(1 - t)Y $$

teniendo en cuenta que: $$ C = C_0 + bY^d $$

donde $C_0$ es el consumo autónomo, que implica los gastos fijos en los que incurre cada familia. 

Entonces, se encuentra que, la función de Consumo va a depender de el Ingreso disponible, y por ende de la tasa impositiva, pues si esta aumenta, el ingreso disponible disminuye y por lo tanto, el consumo baja; mientras que, de bajar la tasa impositiva, el ingreso disponible aumenta y por ende, el consumo también.

#### Gráfico de la función de demanda de Consumo:

```python
#Parámetros

Yd = np.arange(start = 0, stop = 110, step = 10)   #ingreso agregado

Co = 10  #consumo autónomo 

b = 0.5             #propensión marginal a consumir
```

```python
#ecuación de la función de demanda de consumo
def C_1(Co, Yd, b):
    C_1 = Co + b*Yd
    return C_1

C = C_1(Co, Yd, b)
```

```python
#Dimensiones

y_max = np.max(C)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
plt.plot(C, color = "purple")

# Líneas punteadas punto de equilibrio
plt.axvline(x=2,  ymin= 0, ymax= 0.65, linestyle = ":", color = "grey")
plt.axhline(y=10, xmin= 0, xmax= 0.59, linestyle = ":", color = "grey")
plt.axhline(y=0, xmin= 2, xmax= 2.1, linestyle = ":", color = "grey")
plt.axhline(y=20, xmin= 0, xmax= 0.59, linestyle = ":", color = "grey")
plt.axvline(x=4,  ymin= 0, ymax= 0.65, linestyle = ":", color = "grey")
plt.axhline(y=30, xmin= 0, xmax= 0.594, linestyle = ":", color = "grey")
plt.axvline(x=6,  ymin= 0, ymax= 0.65, linestyle = ":", color = "grey")
plt.axhline(y=40, xmin= 0, xmax= 0.594, linestyle = ":", color = "grey")

#texto agregad0
plt.text(3.5, 50, "$PMgC = ∆C/∆Yd = b = 0.5$", fontsize = 14.5, color = "black")

#Título 
ax.set(title="Función de demanda de Consumo", xlabel= r"$Y^d$", ylabel= r"C")

plt.show()
```

2. Derive y explique la función de Inversión.  

La función de la inversión $(I)$ está determinada por la inversión autónoma del país $(I_0)$, la sensibilidad de los inversionistas frente a la tasa de interés $(h)$ y la tasa de interés real $(r)$.

$$ I = I_0 - hr $$

donde: $h < 0$ y $r < 0$, y h usualmente se mantiene fija.

además, en la función se presenta una sustracción porque existe una relación inversa entre la tasa de interés y la inversión autónoma. En ese sentido, se encuentra que, si la tasa de interés aumenta $(r)$, la inversión autónoma $(I_0)$ baja y por ende, también la inversión del Estado $(I)$; mientras que, si la tasa de interés baja, la inversión autónoma y la del Estado aumenta.   

#### Gráfico de la función de Inversión.

```python
# Parámetros

r = np.arange(start = 0, stop = 1, step = 0.1)

Io = 40
h = 0.7
```

```python
#Ecuación de la función de inversión

def I_1(Io, r, h):
    I_1 = Io - h*r
    return I_1

I = I_1(Io, r, h)
```

```python
# Dimensiones
y_max = np.max(I)
fig, ax = plt.subplots(figsize=(10, 8))

# Curva
ax.plot(I, label = "I", color = "orange") 

#eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

#Texto agregado
plt.text(7.5, 39.5, '$I = I_0 - hr$', fontsize = 19, color = 'black')

# Titulo
ax.set(title = "Función de demanda de Inversión", xlabel= 'r', ylabel= 'I')
ax.legend()

plt.show()
```

3. ¿Cuáles son los supuestos del modelo Ingreso-Gasto Keynesiano?

Este modelo tiene tres supuestos principales. En primer lugar, supone que el nivel de precios está fijo. En segundo lugar, supone que el nivel del producto se adapta a la demanda agregada (cada vez que existe un cambio en esta originados por sus componentes). Finalmente, en tercer lugar, se supone que la tasa de interés está determinada fuera del modelo. 


4. A partir de las siguientes ecuaciones

Función Consumo: $$ C = C_0 + bY^d $$

Función Inversión: $$ I = I_0 - hr $$

Gasto de Gobierno: $$ G = G_0 $$

Tributación: $$ T = tY $$

Exportaciones: $$ X = X_0 $$

Importaciones: $$ M = mY^d $$

Gasto o Demanda Agregada: $$ DA = C + I + G + X - M $$


#### - Encuentre el nivel de Ingreso de equilibrio $Y^e$

Sabiendo que $$ DA = C + I + G + X - M $$ 

y teniendo en cuenta las ecuaciones anteriores del modelo, se puede comenzar a reemplazar de la siguiente manera:

$$ DA = C_0 + bY^d + I_0 - hr + G_0 + X_0 - mY^d $$

donde: 
- Go: Gasto de gobierno autónomo
- Xo: ingresos por exportación
- m: propensión marginal a importar.

Además, sabiendo que $$ Y^d = (1 - t)Y $$

Se puede volver a reemplazar teniendo ahora: 

$$ DA = C_0 + b(1 - t)Y + I_0 - hr + G_0 + X_0 - m(1 - t)Y $$

Se factoriza y queda: 

$$ DA = (b - m)(1 - t)Y + C_0 + I_0 - hr + G_0 + X_0 $$

Ordenado: 

$$ DA =  (C_0 + I_0 + G_0 + X_0 - hr) + (b - m)(1 - t)Y $$

Resumido: 

$$ DA = α_0 + α_1Y $$

donde: 

$α_0 = C_0 + I_0 + G_0 + X_0 - hr$

$α_1 = (b - m)(1 - t)$

En ese sentido, sabiendo que en equilibrio el ingreso agregado $(Y)$ debe ser igual a la Demanda Agregada, se tiene que:

$$ DA = Y^e $$

entonces: 

$$ α_0 + α_1Y = Y $$

Por lo tanto, reemplazamos y se tiene: 

$$ Y = (C_0 + I_0 + G_0 + X_0 - hr) + (b - m)(1 - t)Y $$

Se factoriza:

$$ [1 - (b - m)(1 - t)]Y = C_0 + I_0 + G_0 + X_0 - hr $$

donde, finalmente: 

$$ Y^e = \frac{1}{1 - (b - m)(1 - t)} (C_0 + I_0 + G_0 + X_0 - hr) $$

además, se concluye que el multiplicador keynesiano $(k)$ es:

$$ k = \frac{1}{1 - (b - m)(1 - t)} $$

#### - Gráfico del Ingreso de equilibrio

```python
# Parámetros

Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3
r = 0.9

Y = np.arange(Y_size)

# Ecuación de la curva del ingreso de equilibrio

def DA_1(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_1 = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_1

DA = DA_1(Co, Io, Go, Xo, h, r, b, m, t, Y)
```

```python
# Recta de 45°

a = 2.5 

def L_45(a, Y):
    L_45 = a*Y
    return L_45

L_45 = L_45(a, Y)
```

- Analice una Política Fiscal Expansiva con aumento del Gasto del Gobierno $(∆G_0 > 0)$. El análisis debe ser intuitivo, matemático y gráfico.

###### - Análisis intuitivo: 
- Lo principal es tener en cuenta que el Producto se ajusta siempre a la demanda agregada. Entonces, teniendo en cuenta que se trata de una Política Fiscal Expansiva con un aumento del Gasto del Gobierno, se entiende que $(∆G_0 > 0)$ y que, por lo tanto: 

$$ G_0↑  →   G↑ $$

- Entonces, teniendo en cuenta la ecuación de Demanda Agregada: 

$$ DA=C+I+G+X-M $$

- Se puede intuir que, considerando que el Gasto de Gobierno (G) sube, la Demanda Agregada (DA) también lo hará, pues el Gasto de Gobierno es uno de los componentes que determina la misma, donde:

$$ DA ↑ = C + I + G↑ + X - M $$

- Por lo tanto, si: 
$$ G↑  →   DA↑ $$

- Entonces, como la Demanda Agregada aumenta, esta supera el nivel de producto, por lo que se tiene una situación de desequilibrio: $DA > Y$

- Sin embargo, como se estableció anteriormente, el nivel del producto siempre se ajusta al nivel de la demanda agregada. Por esa razón, teniendo en cuenta que la demanda agregada aumenta y que por eso se genera una situación de equilibrio, donde el nivel del producto es menor a la demanda agregada; entonces, el nivel del producto necesita aumentar también, para así regresar a la situación de equilibrio. 

- Entonces, se tiene que:

$$ G_0↑ → G↑ → DA↑ → Y↑ $$ 

$$ G_0↑ → Y↑ $$

- Y por ende, se concluye (intuitivamente) que: una política fiscal expansiva con aumento del gasto de gobierno, genera un aumento en la demanda agregada, y por ende, aumenta el nivel del producto, lo cual genera que la economía se expanda.

###### Análisis matemático:
- Se sabe que $(∆G_0 > 0)$, entonces ¿$∆Y$?

- En primer lugar, regresando a la función:
$$ Y = \frac{1}{1 - (b - m)(1 - t)} (C_0 + I_0 + G_0 + X_0 - hr) $$  
    podemos reemplazar al multiplicador keynesiano, considerando que este es: $k>0$
    
   Entonces: $$ Y = k(C_0 + I_0 + G_0 + X_0 - hr) $$
   
   Por lo tanto, aplicando el diferencial se tendría que:
   
   $$ ∆Y = k(∆C_0 + ∆I_0 + ∆G_0 + ∆X_0 - ∆hr) $$
   
   Pero, considerando que la política fiscal expansiva, en este caso, no considera cambios en $C_0, I_0, X_0, h$ y $r$, se puede concluir que sus diferenciales son iguales a 0: 
   $$ ∆C_0 = ∆I_0 = ∆X_0 = ∆h = ∆r = 0 $$
   
   Entonces: 
   $$ ∆Y = k(∆G_0) $$
   
   En ese sentido, sabiendo que $∆G_0>0$ y que $k>0$, se puede decir que se trata de la multiplicación de un número positivo (porque el gasto de gobierno aumenta y por ende, el diferencial es positivo), y otro número positivo (porque el multiplicador keynesiano siempre es mayor a 0 y por ende, positivo).
   
   donde: $$ ∆Y = (+)(+) $$
   
   por lo tanto: $$ ∆Y > 0 $$
   
   Entonces, matemáticamente también se concluye que al aplicar una Política Fiscal Expansiva con un aumento en el Gasto de Gobierno, se genera un aumento en el nivel del producto, y por lo tanto, se incentiva y avanza la economía.

- Analice una Política fiscal expansiva con una reducción de la Tasa de Tributación $(∆t < 0)$. El análisis debe ser intuitivo, matemático y gráfico.

$$ C = C_0 + bY^d $$

    donde: $$ Y^d = (1 - t)Y $$

    $$ t↓ → Y^d↑ → C↑ $$
    
    entonces, considerando la ecuación de Demanda Agregada:
    
    $$ DA=C+I+G+X-M $$
    
    Por lo tanto, considerando que el Consumo (C) aumenta, debido a la reducción de la tasa de tributación, entonces, se puede intuir que la Demanda Agregada (DA) también lo hará, pues el Consumo es uno de los componentes que determina la misma, donde:

    $$ DA ↑ = C↑ + I + G + X - M $$
    
    Entonces se puede decir que si la tasa de tributación se reduce, la Demanada Agregada aumenta: 
    
    $$ t↓ → DA ↑ $$
    
    Nuevamente, como en este caso la Demanda Agregada aumenta, entonces, supera al nivel del producto: $DA > Y$
    
    Sin embargo, como ya se estableció, el nivel del producto siempre se ajusta a la demanda agregada. Entonces, considerando que esta aumenta, debido a que la tasa de tributación disminuye y esta afecta al consumo, generando que este aumente, se puede decir que, para regresar a una situación de equilibrio, el nivel del producto debe ajustarse, y en este caso, debe aumentar. 
    
    $$ t↓ → Y^d↑ → C↑ → DA ↑ → Y↑ $$ 
    
    $$ t↓ → Y↑ $$
    
    Por ende, se concluye que: una política fiscal expansiva con reducción de la tasa de tributación, genera un aumento en la demanda agregada, y por ende, aumenta el nivel del producto, lo cual genera que la economía avance, se expande.

$$ DA = C + I + G + X - M $$
    
    Se puede concluir que, el aumentar el Gasto de Gobierno $(∆G>0)$, en un contexto en el que se sigue la Política Fiscal con Regla Contracíclica, tiene el propósito de recuperar la producción. Por lo tanto se puede decir que si el Gasto de gobierno aumenta, también lo hace el nivel de producción: $(∆Y>0)$  
    
    es decir: $$ G↑ → Y↑ $$

     $$ G↑ → Y↑ → DA↑ $$
    
    Por ende, se concluye que: una política fiscal expansiva con aumento del gasto de gobierno, siguiendo la regla contracíclica del gasto, genera una aumento en el nivel de producción, pues su objetivo es recuperar la misma, considerando que esta regla se aplica en contextos donde la economía está en recesión.

- Además en equilibrio: 
$$ DA = Y $$

- Entonces:
    
    $$ Y = C + I + G + X - M $$

     $$ Y = ∆C + ∆I + ∆G + ∆X - ∆M) $$

    $$ ∆C = ∆I = ∆X = ∆M = 0 $$
    
    Entonces queda: 
    $$ ∆Y = ∆G $$

    $$ ∆Y = (+) $$
    
    y por lo tanto:
    $$ ∆Y > 0 $$
    
    Entonces, matemáticamente, también se concluye que si el gasto de gobierno aumenta, el nivel de producción también lo hará, pues es el propósito principal de dicha acción (el incrementar el gasto de gobierno), según la regla contracíclica del gasto, de manera que, se estimule la recuperación de la producción.
