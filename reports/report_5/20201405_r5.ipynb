### C√ìDIGO

A partir del siguiente sistema de ecuaciones que representan el modelo DA-OA

Demanda agregada 

IS: [1-b(1-t)]Y = Co + Io + Go - hr

LM: Mo^s/Po = KY -jr

Oferta agregada 

P = P^e + ùõâ (Y - Y¬Ø) donde Y¬Ø = Yf

Encuentre las ecuaciines de ingreso (Ye) y tasa de inter√©s (r^e) de equilibrio 
Demanda agregada (DA)

La ecuaci√≥n IS es:

            r = ùõΩo/h - (ùõΩ1/h)Y
            
   donde: ùõΩo = Co + Io + Go + Xo
          ùõΩ1 = 1 - (b - m)(1 - t)
          
La ecuaci√≥n LM es: 

            r = -1/j * Mo^s/P + K/j*Y
            
    eliminamos "r" y despejamos P y obtenemos:
    
            ùõΩo/h - ùõΩ1/h*Y = -1/j * Mo^s/P + (K/j)Y
            ùõΩo/h - ùõΩ1/h*Y - (K/j)Y = -1/j * Mo^s/P
            P ( ùõΩo/h - ùõΩ1*Y/h - (K/j)Y) = Mo^s/j 
            P = h * Mo^s/ -j*ùõΩo + (j*ùõΩ1 + h*K)Y
            
    o tambi√©n podemos escribir la ecuaci√≥n de la DA en funci√≥n del nivel de ingresos (Y)
    
    KY/j + (ùõΩ1/h)*Y = ùõΩo/h + 1/j * Mo^s/P
    Y(K/j + ùõΩ1/h) = ùõΩo/h + 1/j * Mo^s/P
    Y = (j*ùõΩo/ j*ùõΩ1 + hK) + (h*Mo^s/j*ùõΩ1 + hK)1/P
    
    Ahora vamos a tener una nueva curva de oferta monetaria reemplazando 
    
    Mo^s/P ‚ü∂ Mo^s - P
    Mo^s - P = KY - jr
    
    reemplazamos r  y obtenemos la nueva ecuaci√≥n de DA que ya no es una hiperbola sino una recta 
    
    Mo^s - P = KY -j (ùõΩ0/h - ùõΩ1*Y/h)
    Mo^s - KY + j (ùõΩo/h - (ùõΩ1/h)Y) = P
    P = (h*Mo^s - hKY + j*ùõΩo - j*ùõΩ1Y)/h
    
    La ecuacion de demanda agregada es:
    P = (h*Mo^s + j*ùõΩo)/h - (j*ùõΩ1+hK/h)Y
    
Oferta agregada (OA)

    La curva de oferta agregada (OA) de corto plazo se puede representar con la siguiente ecuaci√≥n:
    
    P = P^e + ùúÉ (Y - Y¬Ø)
    
    Donde: P = nivel de precios
           P^e = precio esperado
           Y¬Ø = producto potencial
           
    Entonces para hablar la ecuaci√≥n de ingreso de equilibrio (Y) igualamos: 
    
    P^e + ùúÉ (Y - Y¬Ø) = (h*Mo^s + j*ùõΩo)/h - (j*ùõΩ1 + hK/h)Y
    
    Despejamos Y, hallamos el Y^e (Y de equilibrio)
    
    P^e + ùúÉ*Y - ùúÉ*Y¬Ø = (h*Mo^s + j*ùõΩo)/h - (j*ùõΩ1 + hK/h)Y
    ùúÉY + (j*ùõΩ1 + hK)/h*Y = (h*Mo^s + j*ùõΩo)/h + ùúÉ*Y¬Ø - P^e 
    Y(ùúÉ + (j*ùõΩ1 + h*K)/h) = (h*Mo^s + j*ùõΩo)/h + ùúÉ*Y¬Ø - P^e
    Y^eq = [1/ùúÉ + (j*ùõΩ1 + hK)/h][(h*Mo^s + j*ùõΩo)/h - P^e + ùúÉ*Y¬Ø]
    
    Hallamos P^eq reemplazando Y^eq en la ecuacion de oferta agregada (OA) 
    
    P = P^e + ùúÉ (Y-Y¬Ø)
    P^eq = P^e + ùúÉ[(1/ùúÉ + (j*ùõΩ1+hK)/h)((h*Mo^s + j*ùõΩo)/h - P^e + ùúÉ*Y¬Ø)] - Y¬Ø
    
    Para hallar r^eq reemplazamos P^eq en la ecuacion de la tasa de interes de equilibrio del modelo IS-LM 
    
    La tasa de interes de equilibrio del modelo IS-LM es: 
    
    r^eq = K*ùõΩo/ Kh + j*ùõΩ1 - (ùõΩ1/Kh+jùõΩ1)(Mo^s - P)
    
    La tasa de interes de equilibrio en DA-OA es:
    
    r^eq = (KùõΩo/Kh+jùõΩ1) - (ùõΩ1/Kh + j*ùõΩ1)(Mo^s - P^eq)
    
    r^eq = K*ùõΩo/(Kh + jùõΩ1) - (ùõΩ1/Kh + jùõΩ1)(Mo^s - P^e + ùúÉ[1/ùúÉ+(jùõΩ1+hK)/h)(h*Mo^s + j*ùõΩo/h - P^e + ùúÉ*Y¬Ø)] - Y¬Ø)
