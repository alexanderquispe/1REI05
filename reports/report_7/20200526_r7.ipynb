### Ploteo de Histogramas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns, warnings
import statsmodels.formula.api as smf
warnings.filterwarnings('ignore')
from causalgraphicalmodels import CausalGraphicalModel
from cProfile import label
%matplotlib inline
```

```python
data = pd.read_csv('https://github.com/matheusfacure/python-causality-handbook/raw/master/causal-inference-for-the-brave-and-true/data/online_classroom.csv')
print(data.shape)
data.head()
```

```python
# creado una columna extra para clasificar
data_2 = (data
 .assign(class_format = np.select(
     [data["format_ol"].astype(bool), data["format_blended"].astype(bool)],
     ["online", "blended"],
     default="face_to_face" #create a new variable
 ))) #group by the new variable (treatments)
```

```python
data_2
```

```python
# create dummy variables
data_3 = pd.get_dummies(data_2, columns = ["class_format"])
data_3
```

```python
blended = data_3.loc[data_3['class_format_blended'] == 1]
sns.displot(blended, x = "falsexam")

plt.title('Blended')
```

```python
face_to_face = data_3.loc[data_3['class_format_face_to_face'] == 1]
sns.displot(face_to_face, x = "falsexam")

plt.title('Face_to_face')
```

```python
online = data_3.loc[data_3['class_format_online'] == 1]
sns.displot(online, x = "falsexam")

plt.title('Online')
```

Interpretación: En base a los tres histogramas presentados se puede observar que el formato de clases online es aquel que tiene, en general, las notas más bajas. De hecho, en comparación a las otras dos modalidades, es la única donde se pueden observar notas de 0. Esto se puede deber a los nuevos problemas virtuales que fueron apareciendo con esta modalidad, pues antes no se había implementado a gran escala. En segundo lugar, se encuentra la modalidad "face to face" o presencial, en la que se observa un rango de notas en promedio más altas en comparación al resto de modalidades. En efecto, hay mayor cantidad de notas altas en el rango de 80-90 a diferencia de las otras. Finalmente, se encuentra el formato de "blended learning" o aprendizaje combinado, donde el promedio de las notas parece estar en el medio de las otras modalidades, pues si bien hay buena cantidad de notas de 80, no supera a las obtenidas en la modalidad presencial. De esta forma, puedo concluir que hay un efecto causal negativo de implementar una políitica llamada "online learning" sobre el rendimiento de los estudiantes.

### Difference-in-means estimator

```python
data_2.groupby(["class_format"]).mean() #get the exam's mean
```

```python
# Y(1)- Y(0) = ATE Average Treatment Effect - Causal 
# Y(1) -> online, Y(0) -> face_to_face
73.635263 - 78.547485
```

```python
data_4 = data_3.loc[(data_3['class_format_face_to_face'] == 1) | (data_3['class_format_online'] == 1)]
data_4
```

```python
data_5 = data_4.groupby(["class_format_face_to_face", 'class_format_online']).mean()
data_5
```

Interpretación: Con respecto al promedio de las clases presenciales se puede observar un valor obtenido de 78.54. Por otro lado, con los cursos "online" se obtiene un promedio con un valor de 73.63. Por ende, se puede obtener el ATT de una clase en línea, el cual vendría a ser -4.91. En otras palabras, esto significa que las clases en línea hacen que los estudiantes rindan aproximadamente 5 puntos menos que con las clases presenciales. Es decir, la política aplicada de irnos a la virtualidad nos ha traído como consecuencia un efecto negativo con respecto al rendimiento del estudiantado.
