### 2.1 Equilibrium in the IS-LM model:

Considering, on the one hand, the equation of the IS Curve:

$$ r = \frac{B_o}{h} - \frac{B_1}{h}Y $$

- Where $ B_0 = C_o + I_o + G_o + X_o $ and $ B_1 = 1 - (b - m)(1 - t) $

And, on the other hand, the equation of the LM Curve:

$$  r = -\frac{1}{j}\frac{Mo^s}{P_o} + \frac{k}{j}Y $$

Both equations can be equalized (substituted or reduced) to find the equilibrium level of Income $ (Y^e) $ and the equilibrium Interest Rate $ (r^e) $.

- To find the equilibrium level of Income $ (Y^e) $, the interest rates of each model are equalized $ (r) $:

$$ -\frac{1}{j}\frac{Mo^s}{P_o} + \frac{k}{j}Y = \frac{B_o}{h} - \frac{B_1}{h}Y $$

$$ Y^e = \frac{j B_o}{k h + j B_1} + (\frac{h}{k h + j B_1})\frac{Ms_o}{P_o} $$

- To find the equilibrium Interest Rate $ (r^e) $, the levels of Income $ (Y) $ of each model are equalized:

$$ r^e = \frac{kB_o}{kh + jB_1} - (\frac{B_1}{kh + jB_1})\frac{Ms_o}{P_o} $$

These two equations represent the equilibrium of the IS-LM model.

The equilibrium of the IS-LM model therefore is plotted as follows:

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import sympy as sy
```

```python
# IS Curve
Y_size = 100 
b = 0.4                 #propensión marginal a consumir
m = 0.2                 #propensión a importar
t = 0.9                   #tasa impositiva
Co = 20                 #consumo autónomo
Io = 10                  #inversión autónoma
Go = 50                  #gasto gobierno 
Xo = 2                  #exportación 
h =  1                  #constante de decisión a invertir

Y = np.arange(Y_size)
```

```python
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

def Y_IS(b, m, t, Co, Io, Go, Xo, h, r):
    Y_IS = (Co + Io + Go + Xo - h*r)/(1-(b-m)*(1-t))
    return Y_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
r
```

```python
# LM Curve
Y_size = 100

k = 15                # Transaction money demand
j = 10                # Speculation money demand
Ms = 9000             # Nominal money supply
P  = 20                # Price level

Y = np.arange(Y_size)
```

```python
def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)
i
```

```python
y_max = np.max(i)

fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="IS-LM Model", xlabel= r'Y', ylabel= r'r')

ax.plot(Y, r, "k-", label = "IS", color = "skyblue")
ax.plot(Y, i, "b-", label="LM", color = "orange")

ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())  

plt.axvline(x=51.1,  ymin= 0, ymax= 0.52, linestyle = ":", color = "black")
plt.axhline(y=32, xmin= 0, xmax= 0.52, linestyle = ":", color = "black")
plt.plot(51, 32, marker="o", color="black", label = "Eo")
plt.text(49,38, 'Eo', fontsize = 14, color = 'black')

ax.legend()
plt.show()
```

- *Intuition*: $ ↑∆G_o $
$$ G_o↑ → DA↑ → DA > Y → Y↑ $$
$$ Y↑ → M_d↑ → M_d > M_o → r↑ $$ 
    
    An increase in Public Spending $ (G_o) $ will mean an increase in Aggregate Demand $ (DA) $. So, Aggregate Demand will be greater than the Production Level $ (Y) $. Therefore, to return to equilibrium, the production level will increase.

     On the other hand, an increase in production level will cause the Money Demand $ (M^d) $ to increase in the money market. Thus, the demand for money will be greater than the Money Supply $ (M^o) $. For both to balance, the interest rate or the price of money $ (r) $ will increase.

     In conclusion, the increase in government gpending causes a growth in production level. However, when considering the effects on the money market, the increase in the interest rate will cause a decrease in investment.

--------------------------------------

$$ ∆r^e = \frac{k∆G_o}{kh + jB_1} $$

$$ ∆r^e = [\frac{k}{kh + jB_1}] ∆G_o > 0 $$

$$ r↓ → I_o↑ → DA↑ → DA > Y → Y↑ $$ 

   In the money market, an increase in the mass of money $(M^s)$ will cause the money supply to be greater than demand $(M^d)$. Therefore, for both to equalize, the interest rate $(r)$ must be reduced

     Since the interest rate also affects the IS curve, its reduction will encourage more investment $(I_o)$. As autonomous investment increases, aggregate demand $(DA)$ will increase. Then, the aggregate demand will be greater than production level $(Y)$. And, for both to return to equilibrium, the level of production must grow.
    
     In conclusion, an increase in the money supply will generate a lower interest rate which, by benefiting investors, will make GDP grow.

- *Intuition*: $ ↑∆M^o $ and $ ↓∆G_o $

$$ M^o↑ → M^o > M^d → r↓ $$
$$ r↓ → I_o↑ → DA↑ → DA > Y → Y↑ $$ 

$$ G_o↓ → DA↓ → DA < Y → Y↓ $$
$$ Y↓ → M^d↓ → M^d < M^o → r↓ $$

On the one hand, an expansionary monetary policy implies that the money supply $(M^o)$ increases. Therefore, the money supply will be greater than the demand $(M^d)$. And, to return to equilibrium, the interest rate $(r)$ or the price of money will be reduced.
    
This reduction in the interest rate will also cause investment $(I_o)$ and, consequently, aggregate demand $(DA)$ to grow. Then, since aggregate demand is greater than production level $(Y)$, the latter will increase to return to equilibrium.
    
On the other hand, a restrictive fiscal policy means that public spending $(G_o)$ is reduced. Thus, the aggregate demand $(DA)$ will also be reduced and will end up being less than production level $(Y)$. To return to equilibrium, the level of production must fall.
    
This will also have effects on the money market: given the reduction in the level of production, the money demand $(M^d)$ will decrease. When reduced, it will be less than the money supply $(M^o)$ and the interest rate (r) will have to fall for them to return to equilibrium.
    
In short, the combination of these fiscal and monetary policies supposes a reduction in the interest rate. However, there seems to be a contradiction regarding the production level: while in the IS model GDP will increase, in the LM model it will decrease. Therefore, as shown in the graph, the changes in both markets cause production level to remain constant.
    
So, in the face of an expansionary monetary policy and a restrictive fiscal policy, only a fall in the interest rate will be observed.

- *Math*:

Monetary Policy:

$$ ∆Y^e = (\frac{h}{kh + jB_1})\frac{1}{P_o}∆Mo^s > 0 $$
    
$$ ∆r^e = -(\frac{B_1}{kh + jB_1})\frac{1}{P_o}∆Mo^s < 0 $$

----------------------

Fiscal Policy:

$$ ∆Y^e = [\frac{j}{kh + jB_1}] ∆G_o < 0 $$

$$ ∆r^e = [\frac{k}{kh + jB_1}] ∆G_o < 0 $$

A reduction in the tax rate $(t)$ implies an increase in disposable income $(Y^d)$. This in turn poses two shocks: on the one hand, the increase in importance will imply a drop in net exports $(X-M)$. This fall will reduce aggregate demand $(DA)$, which will make it less than production level $(Y)$. To return to equilibrium, the production level will have to fall.

However, this shock is less significant than that of autonomous consumption $(C_o)$. The latter, by focusing on the income of the country's people, will mean an increase in aggregate demand $(DA)$. So, for aggregate demand and production level to return to equilibrium, the latter must increase.

Given an increase in the level of production, the money demand $(M^d)$ will also grow. Thus, the demand for money will be greater than the supply $(M^o)$. So, to return to equilibrium, the interest rate must increase.

In conclusion, an increase in the tax rate will cause that both production level and interest rate to grow.

## PART 2: Report

The paper’s objective is to present a macroeconomic model that reproduces what happened in Peru with the spread of COVID-19. In other words, the authors seek to simulate the macroeconomic effects of the start and end of the quarantine by focusing on two sectors: on the one hand, sector 1, which includes businesses such as hotels, restaurants or airlines, that was directly affected by the lockdown. On the other hand, sector 2 corresponds to the production of food and essential services. Although the latter was able to continue operating during quarantine, it was indirectly affected by the negative demand shock of sector 1.

To achieve their objective, an alternative macroeconomic model to the traditional IS-LM-DA-OA is used: the IS-MR-DA-OA model proposed by Waldo Mendoza in previous works. The importance of the second model lies in the fact that it emphasizes monetary policy and reflects the current practice of central banks. Nowadays, central banks do not control the money supply, as the IS-LM model assumes, but rather the interest rate. So, since the central bank controls the interest rate, the money supply plays a secondary role, since it has no effect on production.

While the introduction of this model and, above all, the equations used are somewhat difficult to understand, the conclusions presented at the end of each section of the paper show that the IS-MR-DA-OA model is not difficult to understand. For this reason, the article contributes not only by proposing an alternative model that is closer to reality, but also by applying it to explain a real situation of crucial importance, such as the macroeconomic effects of the pandemic. In this way, the use of this alternative model is consistent with Mendoza's aspiration that it could “be part of the heart of the teaching of General Equilibrium Macroeconomics in undergraduate”.

In line with this idea of getting closer to reality, the authors propose two types of exercises to analyze the effects of the quarantine on the Peruvian economy. Even though the first of them assumes that the fall in the potential product of autonomous consumption in sector 1 is permanent, the second one is similar to what happened in Peru:  a significant economic recovery to the point that the GDP has already exceeded the levels before the pandemic. Thus, a second strength of the paper is considering the transitory shocks of supply and demand, and not only the permanent ones.

The explanation for the economic recovery, according to the authors, is due to the credit policy applied in the 2020-2021 period that prevented the massive bankruptcy of companies. However, a weakness of the article is not delving deeper into the foundation and results of Reactiva Peru. Although it is presented as the “most powerful policy in Latin America”, its effects are not fully included in the designed macroeconomic model. Even when cataloging it as "the mechanical ventilator that saved thousands of human lives, the instrument that allowed the survival of thousands of companies, especially micro and small" there is no room for criticism and questioning of this policy.

It should be noted that Reactiva Peru, for the most part, benefited large companies more than small and micro ones. This was an important attempt by the State to help MYPES, but it did not reach the reach of many and, consequently, has caused informality to increase. Being informality a crucial issue in Peruvian economic matters that the article does not address either.

However, another strength of the paper is that it raises future lines of research. Specifically three: 1) following the alternative model, carry out a comparative statics exercise that considers public policies in economic matters and not only health (such as quarantine); 2) analyze the existence of agents with rational expectations and their implications in the steady state; and 3) develop a model that analyzes the effect of the duration of the quarantine on the potential product. Likewise, future research could also take up topics in a comparative manner: at the regional level, on the credit policies chosen and their results in Latin America and, at the domestic level, the effects of these on the different socioeconomic groups.
