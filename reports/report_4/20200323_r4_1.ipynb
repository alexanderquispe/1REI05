#### Encuentre las ecuaciones de Ingreso  y tasa de interes  de equilibrio(Escriba paso a paso la derivacion de estas ecuaciones).

Primero vamos a colocar las funciones de la curva IS:

$$ r = \frac{B_o}{h} - \frac{B_1}{h}Y $$

- Donde $ B_0 = C_o + I_o + G_o + X_o $ y $ B_1 = 1 - (b - m)(1 - t) $

y la curva LM:

$$  r = -\frac{1}{j}\frac{Mo^s}{P_o} + \frac{k}{j}Y $$

Para encontrar las ecuaciones de equilibrio, debemos entonces igualar ambas funciones, ya que ambas se igualan a r:

$$ -\frac{1}{j}\frac{Mo^s}{P_o} + \frac{k}{j}Y = \frac{B_o}{h} - \frac{B_1}{h}Y $$

Ahora podemos modificar las ecuaciones, y traer Y a la izquierda para la ecuación de ingreso de equilibrio:

$$ Y^e = \frac{j B_o}{k h + j B_1} + (\frac{h}{k h + j B_1})\frac{Ms_o}{P_o} $$

o r a la izquierda, para tener la ecuación de tasa de interés de equilibrio.

$$ r^e = \frac{kB_o}{kh + jB_1} - (\frac{B_1}{kh + jB_1})\frac{Ms_o}{P_o} $$

Estas dos ecuaciones representan el modelo IS-LM.

#### Grafique el equilibrio simultáneo en los mercados de bienes y de dinero.

```python
#--------------------------------------------------
    # Curva IS

# Parámetros

Y_size = 100 

Co = 35
Io = 40
Go = 50
Xo = 2
h = 0.8
b = 0.7
m = 0.5
t = 0.8

Y = np.arange(Y_size)


# Ecuación 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo)/h - ( ( 1-(b-m)*(1-t) ) / h)*Y  
    return r_IS

r_is = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)


#--------------------------------------------------
    # Curva LM 

# Parámetros

Y_size = 100

k = 2
j = 1                
Ms = 200             
P  = 20               

Y = np.arange(Y_size)

# Ecuación

def r_LM(k, j, Ms, P, Y):
    r_LM = - (1/j)*(Ms/P) + (k/j)*Y
    return r_LM

r_lm = r_LM( k, j, Ms, P, Y)
```

```python
# Gráfico del modelo IS-LM

# Dimensiones del gráfico
y_max = np.max(r_lm)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
# Curva IS
ax.plot(Y, r_is, label = "IS", color = "red") #IS
# Curva LM
ax.plot(Y, r_lm, label="LM", color = "blue")  #LM

# Eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

# Texto y figuras agregadas
# Graficar la linea horizontal - r
plt.axvline(x=53,  ymin= 0, ymax= 0.53, linestyle = ":", color = "black")
# Grafica la linea vertical - Y
plt.axhline(y=95, xmin= 0, xmax= 0.53, linestyle = ":", color = "black")

# Plotear los textos 
plt.text(49,100, '$E_0$', fontsize = 14, color = 'black')
plt.text(0,100, '$r_0$', fontsize = 12, color = 'black')
plt.text(53,-10, '$Y_0$', fontsize = 12, color = 'black')

plt.text(49,150, 'I', fontsize = 14, color = 'black')
plt.text(80,100, 'II', fontsize = 14, color = 'black')
plt.text(49,40, 'III', fontsize = 14, color = 'black')
plt.text(20,100, 'IV', fontsize = 14, color = 'black')

# Título, ejes y leyenda
ax.set(title="IS-LM Model", xlabel= r'Y', ylabel= r'r')
ax.legend()

plt.show()
```

##### Analice los efectos sobre las variables endógenas Y, r de una disminución del gasto fiscal. El análisis debe ser intuitivo, matemático y gráfico.

Intuición: 

- Mercado de Bienes
$$ Go↓ → DA↓ → Y>DA → Y↓ $$ 
- Mercado de dinero
$$ Y↓ → Md↓ → Ms>Md → r↓ $$ 

Si disminuye el gasto de gobierno, disminuye la demanda agregada, lo que hace que haya un exceso de oferta, y se deba bajar el producto. Esto genera que la demanda de dinero baje, y haya un exceso de oferta de dinero, por lo que la tasa de interés debe bajar.

```python
#Matemática

# nombrar variables como símbolos
Co, Io, Go, Xo, h, r, b, m, t, beta_0, beta_1  = symbols('Co Io Go Xo h r b m t beta_0, beta_1')

# nombrar variables como símbolos
k, j, Ms, P, Y = symbols('k j Ms P Y')

# Beta_0 y beta_1
beta_0 = (Co + Io + Go + Xo)
beta_1 = ( 1-(b-m)*(1-t) )

# Producto de equilibrio y la tasa de interes de equilibrio en el modelo IS-LM
r_eq = (k*beta_0)/(k*h + j*beta_1) - ( beta_1 / (k*h + j*beta_1) )*(Ms/P)
Y_eq = (j*beta_0)/(k*h + j*beta_1) + ( h / (k*h + j*beta_1) )*(Ms/P)
```

```python
df_Y_eq_Go = diff(Y_eq, Go)
print("El Diferencial del Producto con respecto al diferencial del gasto autonomo = ", df_Y_eq_Go)  
```

```python
df_r_eq_Go = diff(r_eq, Go)
print("El Diferencial de la tasa de interes con respecto al diferencial del gasto autonomo = ", df_r_eq_Go)  
```

Ambos diferenciales de Y y r son positivos cuando cambia el gasto autónomo. Sin embargo, en este caso, el gasto autónomo esta disminuyendo, por lo que la multiplicación de los diferenciales será uno positivo por uno negativo.

$$ ∆Y^eq = (+)(-) $$
$$ ∆r^eq = (+)(-) $$

Por lo tanto, tanto el producto como la tasa de interés van a disminuír.

```python
#1--------------------------------------------------
    # Curva IS ORIGINAL

# Parámetros

Y_size = 100 

Co = 35
Io = 40
Go = 80
Xo = 2
h = 0.8
b = 0.7
m = 0.5
t = 0.8

Y = np.arange(Y_size)


# Ecuación 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)


#2--------------------------------------------------
    # Curva LM ORIGINAL

# Parámetros

Y_size = 100

k = 2
j = 1                
Ms = 200             
P  = 20               

Y = np.arange(Y_size)

# Ecuación

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)
```

```python
#--------------------------------------------------
    # NUEVA curva IS: reducción Gasto de Gobierno (Go)
    
# Definir SOLO el parámetro cambiado
Go = 50

# Generar la ecuación con el nuevo parámetro
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r_G = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
```

```python
# Gráfico

# Dimensiones del gráfico
y_max = np.max(i)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, r, label = "IS_(G_0)", color = "C1") #IS_orginal
ax.plot(Y, r_G, label = "IS_(G_1)", color = "C1", linestyle = 'dashed') #IS_modificada

ax.plot(Y, i, label="LM", color = "C0")  #LM_original

# Texto y figuras agregadas
plt.axvline(x=51.5,  ymin= 0, ymax= 0.5, linestyle = ":", color = "grey")
plt.axhline(y=93, xmin= 0, xmax= 0.52, linestyle = ":", color = "grey")

plt.axvline(x=63,  ymin= 0, ymax= 0.6, linestyle = ":", color = "grey")
plt.axhline(y=115, xmin= 0, xmax= 0.62, linestyle = ":", color = "grey")
plt.text(65,117, '$E_0$', fontsize = 14, color = 'black')

plt.text(49,100, '$E_1$', fontsize = 14, color = 'black')
plt.text(-1,100, '$r_1$', fontsize = 12, color = 'black')
plt.text(60,-40, '$Y_0$', fontsize = 12, color = 'black')
plt.text(-1,120, '$r_0$', fontsize = 12, color = 'black')
plt.text(53,-40, '$Y_1$', fontsize = 12, color = 'black')

#plt.text(69, 115, '→', fontsize=15, color='grey')
#plt.text(69, 52, '←', fontsize=15, color='grey')

# Título, ejes y leyenda
ax.set(title="Politica Fiscal Contractiva", xlabel= r'Y', ylabel= r'r')
ax.legend()

plt.show()
```

### Puntos Extra Código

Primero vamos a colocar las nuevas funciones de la curva IS:

$$ i = (\frac{B_o + π}{h})  - (\frac{B_1 + π}{h}Y) $$

- Donde $ B_0 = C_o + I_o + G_o $ y $ B_1 = 1 - (b)(t) $

y la curva LM:

$$  i = -\frac{1}{j}\frac{Mo^s}{P_o} + \frac{k}{j}Y $$

Para encontrar las ecuaciones de equilibrio, debemos entonces igualar ambas funciones, ya que ambas se igualan a r:

$$ -\frac{1}{j}\frac{Mo^s}{P_o} + \frac{k}{j}Y = (\frac{B_o + π}{h})  - (\frac{B_1 + π}{h}Y) $$

Ahora podemos modificar las ecuaciones, y traer Y a la izquierda para la ecuación de ingreso de equilibrio:

$$ Y^e = \frac{j B_o + π}{k h + j B_1} + (\frac{h}{k h + j B_1})\frac{Ms_o}{P_o} $$

o i a la izquierda, para tener la ecuación de tasa de interés de equilibrio.

$$ i^e = \frac{kB_o + π}{kh + jB_1} - (\frac{B_1}{kh + jB_1})\frac{Ms_o}{P_o} $$

Estas dos ecuaciones representan el nuevo modelo IS-LM.

```python
#--------------------------------------------------
    # Curva IS

# Parámetros

Y_size = 100 

Co = 35
Io = 20
Go = 45
h = 0.8
b = 0.7
t = 0.8

Y = np.arange(Y_size)


# Ecuación 
def r_IS(b, t, Co, Io, Go, h, Y):
    r_IS = (Co + Io + Go)/h - ( ( 1-(b)*(t))/ h)*Y  
    return r_IS

r_is = r_IS(b, t, Co, Io, Go, h, Y)


#--------------------------------------------------
    # Curva LM 

# Parámetros

Y_size = 100

k = 2
j = 1                
Ms = 200             
P  = 20               

Y = np.arange(Y_size)

# Ecuación

def r_LM(k, j, Ms, P, Y):
    r_LM = - (1/j)*(Ms/P) + (k/j)*Y
    return r_LM

r_lm = r_LM( k, j, Ms, P, Y)
```

```python
# Gráfico del modelo IS-LM

# Dimensiones del gráfico
y_max = np.max(r_lm)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
# Curva IS
ax.plot(Y, r_is, label = "IS", color = "red") #IS
# Curva LM
ax.plot(Y, r_lm, label="LM", color = "blue")  #LM

# Eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

# Texto y figuras agregadas
# Graficar la linea horizontal - r
plt.axvline(x=53,  ymin= 0, ymax= 0.53, linestyle = ":", color = "black")
# Grafica la linea vertical - Y
plt.axhline(y=95, xmin= 0, xmax= 0.53, linestyle = ":", color = "black")

# Plotear los textos 
plt.text(49,100, '$E_0$', fontsize = 14, color = 'black')
plt.text(0,100, '$r_0$', fontsize = 12, color = 'black')
plt.text(53,-10, '$Y_0$', fontsize = 12, color = 'black')

plt.text(49,150, 'I', fontsize = 14, color = 'black')
plt.text(80,100, 'II', fontsize = 14, color = 'black')
plt.text(49,40, 'III', fontsize = 14, color = 'black')
plt.text(20,100, 'IV', fontsize = 14, color = 'black')

# Título, ejes y leyenda
ax.set(title="IS-LM Model", xlabel= r'Y', ylabel= r'r')
ax.legend()

plt.show()
```

##### Analice los efectos sobre las variables endógenas Y, r de una disminución de la inflación esperada. El análisis debe ser intuitivo, matemático y gráfico.

Intuición:

- Mercado de dinero
$$ π^e↓ → i↓ → Md↑ → Md>Ms → r↑ $$

- Mercado de bienes
$$ r↑ → I↓ → DA↓ → Y > DA → Y↓ $$

Si disminuye la inflación esperada, entones va a disminuír la tasa de interés nominal. Esto aumenta la demanda de dinero, y genera un exceso de demanda en el mercado monetario. Para volver al equilibrio, se necesita entonces un aumento de la tasa de interés real. Esto a su vez provoca que la inversión caiga, y con esto también caiga la DA. Para volver al equilibrio se tendrá que reducir el Producto.

```python
#--------------------------------------------------
    # Curva IS

# Parámetros

Y_size = 100 

Co = 35
Io = 40
Go = 20
h = 0.8
b = 0.7
t = 0.8

Y = np.arange(Y_size)


# Ecuación 
def r_IS(b, t, Co, Io, Go, h, Y):
    r_IS = (Co + Io + Go)/h - ( ( 1-(b)*(t))/ h)*Y  
    return r_IS

r = r_IS(b, t, Co, Io, Go, h, Y)


#--------------------------------------------------
    # Curva LM 

# Parámetros

Y_size = 100

k = 2
j = 1                
Ms = 500             
P  = 20               

Y = np.arange(Y_size)

# Ecuación

def r_LM(k, j, Ms, P, Y):
    r_LM = - (1/j)*(Ms/P) + (k/j)*Y
    return r_LM

i = r_LM( k, j, Ms, P, Y)
```

```python
# Definir SOLO el parámetro cambiado
Ms = 200

# Generar nueva curva LM con la variacion del Ms
def i_LM_Ms( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i_Ms = i_LM_Ms( k, j, Ms, P, Y)
```
