1. Defina el Multiplicador Keynesiano y dé un ejemplo de cómo influye este en el modelo Ingreso-Gasto Keynesiano.

El multiplicador keynesiano nos indica en qué magnitud una variación de un componente autónomo de la demanda agregada, como el consumo, gasto público o la inversión; repercute en el nivel de ingresos. Este multiplicador depende del valor de la propensión marginal a consumir en la medida en que cuando esté se aproxima a uno, mayor es el multiplicador. Para analizar como influye un multiplicador utilizaremos de ejemplo una variación en el consumo autónomo. 

Suponiendo que el resto de las variables se mantienen constantes, la ecuación nos indica que un aumento en el consumo autonomo ocasionara un aumento mayor en el ingreso. Suponiendo que la propensión marginal a consumir sea 0.3 y el consumo autonomo aumentará en 1000 soles, el ingreso se incrementaria en 1428.6 soles.

2. Grafique y explique cúando sucede un exceso y déficit de demanda en el modelo Ingreso-Gasto. Señale las áreas donde ocurre cada caso. Explique cómo se converge al equilibrio a partir de estos dos escenarios.

Un caso de exceso de demanda puede suceder cuando uno de los componentes autonomos se incrementa, por ejemplo, la inversión, lo cual genera que la curva de DA se desplace hacia arriba. En un primer momento, la demanda se incrementa en la misma magnitud del aumento de la inversión, generando así un exceso de demanda ya que se estaría por encima del anterior punto de convergencia (equilibrio) entre Y y DA en el punto 1. Sin embargo, el nuevo equilibrio se encuentra en el punto de intersección de la recta de 45º y la nueva curva de demanda, en el punto 2, ocasionando así que el ingreso Y se incremente en la medida establecida por el multiplicador. En el gráfico observamos que suceden incrementos consecutivos tanto de la demanda como del ingreso, esto se debe a que el incremento inciial de la demanda genera que las empresas incrementen su producción para satisfacer el nuevo nivel de demanda. Esto a su vez genera un incremento de la demanda y así sucesivamente hasta alcanzar el punto de convergencia.

El efecto opuesto sucede cuando un componente aútonomo de la DA disminuye, ya que esto ocasionaria un deficit de demanda.

3. ¿Cuáles son las herramientas de política fiscal y política monetaria? Dea un ejemplo para cada una de ellas dentro del contexto peruano. Coloque su fuente en caso sea necesario. (Solo necesita 1 para pol. Fiscal y 1 para Monetaria).

La política fiscal consiste en cualquier decisión por parte del gobierno sobre los impuestos o el gasto público. Un ejemplo de política fiscal aplicado en el Perú es la ley que establece el porcentaje de Igv, ya que esta debe ser establecida de tal modo que genere una base tributaria suficiente sin mermar en demasía el ingreso disponible de la población. Por otro lado, la política monetaria es aquella dirigida por el Banco Central que interviene en variables como la oferta monetaria, tipo de cambio y la tasa de interés. Por ejemplo, para el mes de octubre el BCR ha elevado la tasa de interés de referencia de 6,75% a 7% en vista del incremento del nivel de inflación. Esta decisión fue tomada para poder disminuir la demanda de dinero y su consecuente incremento en el nivel de precios.

4. Explique qué funciones del dinero respaldan el éxito de de Yape y Plin(Aplicativos moviles para intercambio de dinero).

La función del dinero que respalda el éxito de Yape es la de transacción. Entre las principales ventajas de estas aplicaciones se encuentra el hecho de que te permiten llevar un equivalente al efectivo en todo momento, el cual puede ser aceptado en la mayoría de locales comerciales pequeños, a diferencia de las tarjetas de crédito o débito. Asimismo, otra ventaja tanto para las personas naturales como para los negocios es que te permiten pagar con la cantidad exacta, facilitando así las transacciones. Además, gracias a su gran nivel de liquidez, cumple una importante función de reserva de valor, ya que el dinero en estas aplicaciones tiene gran poder adquisitivo o poder de compra al poder ser utilizado de manera rápida y eficaz para hacer transacciones medianas y cotidianas.

5. Cuáles son las conclusiones principales del paper de Dancurt "Inflation Targeting in Peru: The Reasons for the Success"

Tras el éxito del BCR en el período 2002-2013, a través de diferentes herramientas de política monetaria conseguía mantener la estabilidad macroeconómica. Por lo tanto, el artículo se pregunta cómo funcionaron los los principales instrumentos de política monetaria que el BCR utilizó en este período. De su investigación, llegó a las siguientes conclusiones. En primer lugar, desde la experencia de la crisis del 2008-2009, la estrategia que debe adoptar en respuesta a próximos shocks externos negativos es la que se usó en este momento, reducir la tasa de interés y vender moneda extranjera. Al contrario, ante shocks externos beneficiosos, el BCR debe aumentar la tasa de interés y comprar moneda extranjera. En segundo lugar, el banco debería limitar su endeudamiento externo con bancos comerciales, de lo contrario no se podrá desdolarizar el crédito bancario. Por tanto, la única herramienta efectiva es imponer requisitos de reserva más altos a la deuda externa de los bancos comerciales. Finalmente, el autor afirma que la caída del tipo de cambio real desde 2007 estuvo ligada al error en reducir la meta de inflación en 2007 y que una regla de intervención cambiaria simétrica permite evitar estas caídas del tipo de cambio real mientras se cumple la meta de inflación.

1. (2 puntos) Encuentre el Ingreso de Equilibrio $(Y^{eq})$ e identifique el multiplicador keynesiano.

#### Teniendo las ecuaciones:

Consumo:  $$C = c_0+bY^d$$
Inversión: $$I = I_0-hr$$
Gasto del gobierno: $$G = G_0$$
Exportaciones: $$X = x_1Y^*+x_2ε$$
Importaciones: $$M = m_1Y^d-m_2ε$$
Tributación: $$T = tY$$
Ingreso disponible: $$Y^d = Y-T$$
Demanda Agregada: $$DA = C+I+G+(X-M)$$
Equilibrio Ingreso-Gasto: $$Y = DA$$

Entonces: $$  DA = (c_0+bY^d) + (I_0-hr) + (G_0) + (x_1Y^*+x_2ε) - (m_1Y^d-m_2ε)$$

$$ DA = c_0+b(Y)(1-t) + I_0-hr + G_0 + x_1Y^*+x_2(ε_0-p(r-r^*)) - m_1(Y)(1-t)-m_2(ε_0-p(r-r^*)) $$

$$ DA = c_0 + I_0 + G_0 - hr + x_1Y^* + (x_2 + m_2)ε_0 - (x_2 + m_2)pr + (x_2 + m_2)pr^* + ((b-m_1)(1-t))Y$$

$$ DA =  [c_0 + I_0 + G_0 + x_1Y^* + (x_2 + m_2)ε_0 + (x_2 + m_2)pr^* - (h + (x_2 + m_2)p)r] + [(b-m_1)(1-t)]Y $$

Si $Y = DA$

$$ Y =  [c_0 + I_0 + G_0 + x_1Y^* + (x_2 + m_2)ε_0 + (x_2 + m_2)pr^* - (h + (x_2 + m_2)p)r] + [(b-m_1)(1-t)]Y $$

$$ Y = \frac{1}{1-(b-m_1)(1-t)}[c_0 + I_0 + G_0 + x_1Y^* + (x_2 + m_2)ε_0 + (x_2 + m_2)pr^*] - [\frac{h + (x_2 + m_2)p}{1-(b-m_1)(1-t)}]r$$

En esta ecuación de equilibrio, nuestro multiplicador sería $\frac{1}{1-(b-m_1)(1-t)}$

Además, la ecuación de equilibrio sería la siguiente:
$$ a_0 + a_1Y $$
Donde el intercepto es $a_0 = [c_0 + I_0 + G_0 + x_1Y^* + (x_2 + m_2)ε_0 + (x_2 + m_2)pr^* - (h + (x_2 + m_2)p)r]$

Y la pendiente es $a_1 = [(b-m_1)(1-t)]Y$
