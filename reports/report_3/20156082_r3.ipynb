## MERCADO MONETARIO

### Instrumentos de pol√≠tica monetaria que puede utilizar el Banco Central

Las pol√≠ticas monetarias hacen referencia a la cantidad de dinero circulando en la econom√≠a, cuyo principal objetivo es la estabilidad de los precios. El Banco Central es la autoridad monetaria y cabeza del sistema bancario. Como tal tiene
la capacidad de controlar o gestionar la oferta monetaria (pol√≠ticas monetaria extractivas o contractivas). Dichos movimientos son realizados mediante el uso de determinados instrumentos.
El primer instrumento de pol√≠tica monetaria que utiliza el Banco Central para aumentar o disminuir la oferta monetaria, es la Operaci√≥n de Mercado Abierto. El Banco Central hace estas operaciones mediante la compra o venta de activos financieros o bonos a los bancos comerciales. Aparte de las operaciones de mercado abierto, el banco central dispone como segundo instrumento el coeficiente legal de encaje. Por √∫ltimo, se encuentra la tasa de inter√©s de referencia de la pol√≠tica monetaria.

### Oferta real de dinero

Para expresar la oferta real de dinero (M^s) en el Mercado de Dinero se deriva la oferta nominal de dinero o masa monetaria (M^o) dividida entre el nivel general de precios de la econom√≠a (ùëÉ)

$$ M^s = \frac{M^o}{P} $$

Asimismo, se asume que el nivel de precios est√° dado para todo el an√°lisis de corto plazo.

### Demanda real de dinero

En general, la gente demanda dinero, como demanda cualquier otro bien o servicio, porque desea mantener cierto monto de riqueza l√≠quida. Hay varios motivos por los que una empresa o familias demandan dinero, por lo que los tipos de demanda de dinero son:
a. Transacci√≥n: se refiere a la necesidad de usar billetes para el intercambio
b. Precauci√≥n: cuando la demanda suele depender del nivel de ingresos
c. Especulaci√≥n: se refiere a jugar

La demanda de dinero por los motivos transacci√≥n y precauci√≥n dependen del ingreso, entonces:

$$ L_1 = kY $$

El motivo de especulaci√≥n depender√° negativa o inversamente de la tasa de inter√©s de los bonos:

$$ L_2 = -ji $$

Si asumimos que ambos tipos de demanda est√°n en t√©rminos reales, la funci√≥n de demanda real de dinero ser√°:

$$ L = L_1 + L_2 $$

$$ L = kY -ji $$

La magnitud del par√°metro ‚Äúk‚Äù indica la sensibilidad de la demanda de dinero ante variaciones del ingreso ‚ÄúY‚Äù, y la del par√°metro ‚Äúj‚Äù indica cu√°n sensible es la demanda de dinero ante las variaciones de la tasa de inter√©s nominal de los bonos, ‚Äúi‚Äù. 

### Ecuaci√≥n de equilibrio en el mercado de dinero

El corto plazo est√° caracterizado por el supuesto de un nivel de precios fijo y ex√≥geno. Por lo tanto, puede suponerse que la inflaci√≥n esperada es igual a cero, lo implica que no hay diferencia entre la tasa de inter√©s nominal y real, es decir, que son aproximadamente iguales: $$ i = r $$

Entonces la ecuaci√≥n del equilibrio en el mercado monetario es:
$$ \frac{M^s}{P} = kY - jr $$

#### Gr√°fico

```python
# Parameters
r_size = 100

k = 0.4
j = 0.2                
P  = 10 
Y = 43.7
MS_0 = 500

r = np.arange(r_size)

# Funcion de demanda 

def MD(k, j, P, r, Y):
    MD_eq = (k*Y - j*r)
    return MD_eq
MD_0 = MD(k, j, P, r, Y)


# Funci√≥n de la oferta de dinero.
MS = MS_0 / P
MS
```

```python
# Equilibrio en el mercado de dinero

# Creamos el seteo para la figura 
fig, ax1 = plt.subplots(figsize=(10, 8))

# Agregamos titulo t el nombre de las coordenadas
ax1.set(title="Equilibrio en el mercado de dinero", xlabel=r'M^s / P', ylabel=r'r')

# Ploteamos la demanda de dinero
ax1.plot(MD_0, label= '$L_0$', color = '#CD5C5C')

# Para plotear la oferta de dinero solo necesitamos crear una linea vertical
ax1.axvline(x = MS,  ymin= 0, ymax= 1, color = "black")

# Creamos las lineas puntadas para el equilibrio
ax1.axhline(y=7.5, xmin= 0, xmax= 0.5, linestyle = ":", color = "red")

# Agregamos texto
ax1.text(0, 7.5, "$r_0$", fontsize = 12, color = 'blue')

ax1.yaxis.set_major_locator(plt.NullLocator())   
ax1.xaxis.set_major_locator(plt.NullLocator())

ax1.legend()

plt.show()
```

## EFECTOS DENTRO DEL MERCADO DE DINERO

###  Disminuci√≥n en la demanda de dinero

Cuando disminuye el ingreso "Y", la demanda de dinero tambi√©n disminuye. Si el mercado estaba en equilibrio, esta disminuci√≥n del ingreso genera un d√©ficit de la demanda en el mercado. El equilibrio debe restablecerse con una depreciaci√≥n de la tasa de inter√©s. La curva de demanda de dinero se desplaza hacia abajo.


## CURVA LM

### Derivaci√≥n de la curva LM a partir del equilibrio en el mercado monetario

Se refiere a todos los puntos de equilibrio dentro del mercado monetario. El equilibrio es cuando la demanda es igual a la oferta:

$$ M^s = \frac{M_o^s}{P} $$
$$ M^d = kY -jr $$

$$ M^d = M^s $$ 

$$ kY -jr = \frac{M_o^s}{P} $$

$$ kY - \frac{M_o^s}{P} = jr $$

$$ \frac{kY}{j} - \frac{M_o^s}{Pj} = r $$

La tasa de inter√©s que equilibra el mercado monetario ser√° igual a:
$$ r = -\frac{1}{j}\frac{M_o^s}{P_o} + \frac{k}{j}Y $$

Son dos ecuaciones simult√°neas. Todos los par√°metros son positivos. La propensi√≥n marginal a consumir est√° entre cero y uno: ùüé < ùíÉ < 1

```python
# Parametros
Y_size = 100

k = 20
j = 20                
Ms = 210             
P  = 30               

Y = np.arange(Y_size)


# Ecuaci√≥n

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)
```

```python
# Gr√°fico de la curva LM

# Dimensiones del gr√°fico
y_max = np.max(i)
v = [0, Y_size, 0, y_max]   
fig, ax = plt.subplots(figsize=(10, 8))

# Eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

# Curvas a graficar
ax.plot(Y, i, label="LM", color = 'purple')

# T√≠tulo y leyenda
ax.set(title="LM SCHEDULE", xlabel=r'Y', ylabel=r'r')
ax.legend()


plt.show()
```

### Disminuci√≥n en la masa monetaria

Cambios en la oferta nominal de dinero tambi√©n producen cambios en la tasa de inter√©s para equilibrar el mercado monetario; por lo tanto, tambi√©n afecta la posici√≥n de la curva LM.

La disminuci√≥n de la oferta monetaria provoca aumentos de la tasa de inter√©s para los diferentes tipos de ingreso dados. En otras palabras, para un mismo nivel de ingreso, cuando disminuye la oferta de dinero, la tasa de inter√©s debe subir para que se restaure el equilibrio en el mercado monetario.

Intuici√≥n: Si disminuye la Masa Monetaria, la recta se desplaza a la izquierda y la tasa de inter√©s de equilibrio aumenta.

```python
#--------------------------------------------------
    # Curva LM ORIGINAL

# Par√°metros

Y_size = 100

k = 20
j = 20               
Ms = 700             
P  = 20               

Y = np.arange(Y_size)

# Ecuaci√≥n

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)

#--------------------------------------------------
    # NUEVA curva LM

# Definir SOLO el par√°metro cambiado
Ms = 200

# Generar la ecuaci√≥n con el nuevo par√°metro
def i_LM_Ms( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i_Ms = i_LM_Ms( k, j, Ms, P, Y)
```

```python
# Dimensiones del gr√°fico
y_max = np.max(i)
v = [0, Y_size, 0, y_max]   
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, i, label="LM", color = 'orange')
ax.plot(Y, i_Ms, label="LM_Ms", color = 'green', linestyle = 'dashed')

# Texto agregado
plt.text(47, 76, '‚àÜ$M^s$', fontsize=12, color='black')
plt.text(49, 70, '‚Üê', fontsize=15, color='grey')

# T√≠tulo y leyenda
ax.set(title = "Incremento en la Masa Monetaria $(M^s)$", xlabel=r'Y', ylabel=r'r')
ax.legend()


plt.show()
```
