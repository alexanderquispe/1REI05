# 1.1

* Curva IS:
$$ r=\dfrac{β_0}{h}-\dfrac{β_1}{h}*Y $$
  → Donde:
$$ β_0=C_0+I_0+G_0+X_0 $$

$$  β_1=1-(b-m)*(1-t) $$

$$  0<b<1 $$
* Curva LM:
$$ r=-\dfrac{1}{j}*\dfrac{(M_0^s)}{P_0} +\dfrac{k}{j}*Y $$

* → Se despeja r,para encontrar Y^e
$$ \dfrac{β_0}{h}-\dfrac{β_1}{h}*Y=-\dfrac{1}{j}*\dfrac{(M_0^s)}{P_0} +\dfrac{k}{j}*Y $$
$$ \dfrac{β_0}{h}+\dfrac{(M_0^s)}{(P_0*j)}=Y*(\dfrac{β_1}{h}+\dfrac{k}{j}) $$

$$ \dfrac{(β_0*P_0*j+h*M_0^s)}{(h*P_0*j)}=Y*\dfrac{(β_1*j+k*h)}{(h*j)} $$

$$ \dfrac{(β_0*P_0*j+h*M_0^s)}{P_0} =Y*(β_1*j+k*h) $$

$$ \dfrac{β_0*P_o*j+h*M_0^s}{P_0*(β_1*j+k*h) }=Y^e $$

$$ ∴Y^e=\dfrac{(j*β_0)}{(β_1*j+k*h)}+\dfrac{h}{(β_1*j+k*h)}*\dfrac{(M_0^s)}{P_0} $$

* → Ahora se despeja Y,para encontrar r^e

$$ \dfrac{β_0}{β_1} -\dfrac{(h*r)}{β_1} =\dfrac{(M_0^s)}{(P_0*k)}-\dfrac{(j*r)}{k} $$

$$ \dfrac{β_0}{β_1} -\dfrac{(M_0^s)}{(P_0*k)}=\dfrac{(h*r)}{β_1} -\dfrac{(j*r)}{k} $$

$$ β_0*k-\dfrac{(M_0^s*β_1)}{P_0} =h*k*r-j*β_1*r $$

$$ β_0*k-\dfrac{(M_0^s*β_1)}{P_0} =r*(h*k-j*β_1 ) $$

$$ ∴ r^e=\dfrac{(β_0*K)}{(h*k-j*β_1 )}-\dfrac{β_1}{(h*k-j*β_1 )}*\dfrac{(M_0^s)}{P_0}  $$

```python
import ipywidgets as widgets
import matplotlib.pyplot as plt
import numpy as np               

import sympy as sy
import pandas as pd
```

```python
# curva IS
Y_size = 100 
b = 0.6                
m = 0.1                 
t = 0.96                  
Co = 17                
Io = 11                 
Go = 44                   
Xo = 10                  
h =  1                  

Y = np.arange(Y_size)

def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t))/h)
    return r_IS

def Y_IS(b, m, t, Co, Io, Go, Xo, h, r):
    Y_IS = (Co + Io + Go + Xo - h*r)/(1-(b-m)*(1-t))
    return Y_IS

r_IS = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)

# LM Curva
Y_size = 100

K = 15.56                
J = 10.37              
Ms = 7000            
P  = 15               

Y = np.arange(Y_size)

def LM( K, J, Ms, P, Y):
    LM = ((-Ms/P)* 1/J + K/J*Y)
    return LM

LM = LM( K, J, Ms, P, Y)


##GRAFICOS
y_max = np.max(LM)

fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="Equilibrio Estable", xlabel= r'Y', ylabel= r'r')

ax.plot(Y, r_IS, "k-", label = "IS", color = "orange")
ax.plot(Y, LM, "b-", label="LM", color = "blue")

ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())  

plt.axvline(x=51.1,  ymin= 0, ymax= 0.52, linestyle = ":", color = "black")
plt.axhline(y=32, xmin= 0, xmax= 0.52, linestyle = ":", color = "black")
plt.plot(51, 32, marker="o", color="black", label = "Eo")
plt.text(49,38, 'Eo', fontsize = 14, color = 'black')

ax.legend()
plt.show()
```

```python
#Grafica LM

Y_size = 100          
k = 20               
j = 10                
Ms = 20000             
P  = 20
πe = 0.5

Y = np.arange(Y_size)  
def i_LM_0( k, j, Ms, P, πe, Y):
    i_LM = k*Y/j-Ms/(P*j)-πe
    return i_LM
i_0 = i_LM_0( k, j, Ms, P, πe, Y)

#Grafica IS
Y_size = 100
c = 20                 
b = 0.01                
m = 0.9            
T = 0.4                 
I_bar = 20             
G_bar = 55              
πe = 0.5             
h = 1                  
Y = np.arange(Y_size)  
def i_IS(c, b, m, T, I_bar, G_bar, X_bar, h, Y):
    i_IS = ((c+G_bar+I_bar)/h-Y*((1-b*(1-T)))/h-πe)
    return i_IS
def Y_IS(c, b, m, T, I_bar, G_bar, πe, h, i):
    Y_IS = ((c+I_bar+G_bar)/((1-b*(1-t)))-(h*πe)((1-b*(1-t)))+h/(1-b*(1-t))*i)
    return Y_IS

i = i_IS(c, b, m, T, I_bar, G_bar, πe, h, Y)


#Grafica LM2

Y_size = 100          
k = 20               
j = 10               
Ms = 20000             
P  = 15
πe = 0.5

Y = np.arange(Y_size)  
def i_LM2( k, j, Ms, P, πe, Y):
    i_LM2 = k*Y/j-Ms/(P*j)-πe
    return i_LM2
LM2 = i_LM2( k, j, Ms, P, πe, Y)


#modelo graficos
y_max = np.max(i_0)

v = [0, Y_size, 0, y_max]                       
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="Efecto de Aumento en los Precios", xlabel=r'Y', ylabel=r'r')
ax.plot(Y, i_0, "b-", label="LM", color="red")
ax.plot(Y, LM2, "b-", label="LM2", linestyle = ":", color="red")
ax.plot(Y, i, "r-",label="IS",color="purple")

plt.axvline(x=65.13,  ymin= 0, ymax= 0.68, linestyle = ":", color = "black") 
plt.axhline(y=29.76, xmin= 0, xmax= 0.65, linestyle = ":", color = "black")   
plt.plot(65.13,29.76, marker="o", color="black")
plt.text(64.13,34.76, 'E0', fontsize=12, color='black')

#Lineas LM2

plt.axvline(x=76.16,  ymin= 0, ymax= 0.65, linestyle = ":", color = "black") 
plt.axhline(y=18.69, xmin= 0, xmax= 0.76, linestyle = ":", color = "black")   
plt.plot(76.26,18.69, marker="o", color="black")
plt.text(76.16,20.69, 'E1', fontsize=12, color='black')

ax.legend()
plt.show()
```

* Intuición:
 
$$ ΔP_0<0 $$
 
$$ P_0↓→M^s↑→M^s>M^d→r↓ $$

$$ r↓→I↑→DA↑→DA>Y→Y↑ $$
 
 
* Matemática:
 
$$ Δr=\dfrac{\beta_0*k}{h*k+\beta_1*j}-\dfrac{M_0^s*\beta_1}{ΔP_0}-π^e $$
 
$$ Δr=-\dfrac{M_0^s*\beta_1}{ΔP_0} $$

$$ ΔY=\dfrac{\beta_0*P_0*j+h*M_0^s}{P_0*(\beta_1*j+h*k)} $$

$$ ΔY=\dfrac{h*M_0^s}{ΔP_0*(\beta_1*j+h*k)}>0 $$

* Intuición:

$$ Δπ^e<0 $$
 
$$ π^e↓→r↑→I↓→DA↓→DA<Y→Y↓ $$

* Matemática:


$$ Δr=\dfrac{\beta_0*k}{h*k+\beta_1*j}-\dfrac{M_0^s*\beta_1}{P_0}-Δπ^e $$

$$ Δr=-Δπ^e>0 $$

$$ ΔY=\dfrac{M_0^s}{P_0}+j*(r+π^e) $$

$$ ΔY=j*Δπ^e<0 $$
