Considerando el diferencial (∆) de la función de demanda de Consumo (C):

$ \frac{∆C}{∆Y^d} = \frac{(C_1-C_0)}{(Y^d_1 - Y^d_0)}  $
$ \frac{(40-30)}{(40-20)} = \frac{10}{20} = 0.5 $

La función de demanda de Consumo sería :

$ C = C_0 + bY^d $

En donde: 
    
 $C_0$ = Consumo autónomo (consumo fijo) 
        
 $Y^d$ = Ingreso disponible. 
    
 $b$ = Propensión marginal a consumir, ∈ ]0,1[

Se concluye que: 

Existe un  consumo autónimo, que es fijo, y un consumo variable, según el ingreso disponible, por lo que se puede decir que la función de consumo representa que todas las familias peruanas están ante una situación de doble forma de consumo.

Consideramos el diferencial de la función de la Demanda de Inversión:
$ ∆I = ∆I_0 - ∆hr $


* Los componentes autónomos se mantienen o no cambian: $ ∆I_0 = 0 $
$ ∆I = (-) $

La  Demanda de Inversión:
$ I = I_0 - hr $


* I = Inversión autónoma (inversión fija).
* r = Tasa de interés real (la tasa nominal neta de la inflación esperada).  
* h = Sensibilidad de los inversionistas ante la variación de la taza de interés.


```python
# Parámetros
r_size = 100
Io = 80
h = 0.4

r = np.arange(r_size)


def C_I(Io, h, r):
    C_I = Io - h*r
    return C_I

C_IS_I = C_I(Io, h, r)
```

```python
# Dimensiones del gráfico
y_max = np.max(C_IS_I)
fig, ax = plt.subplots(figsize=(14, 10))

# Curvas a graficar
ax.plot(r, C_IS_I, label = "I", color = "green") 

# Texto agregado
    # punto de equilibrio
plt.text(60, 80, '$I = Io - hr$', fontsize = 14, color = 'green')

# Título y leyenda
ax.set(title="Función de demanda de Inversión", xlabel= r'r', ylabel= r'I')
ax.legend() #mostrar leyenda

plt.show()
```

Si queremos calcular el "Ingreso de Equilibrio"(Y), en primer lugar, se debe hallar la ecuación de la "Demanda agregada" (DA): 

Datos: 
$$ C = C_0 + bY^d $$
$$ I = I_0 - hr $$
$$ G = G_0 $$
$$ T = tY $$
$$ X = X_0 $$
$$ M = mY^d $$
----------------------------
Suma de "Datos" que es igual a "DA":
$ DA = C + I + G + X -M $

Debemos reemplazar los datos previos  en "DA" 
- $ DA = C_0 + bY^d + I_0 -hr + G_0 + X_0 - mY^d $
- $ DA = C_0 + I_0 + G_0 + X_0 - hr + Y(b - m)(1 - t) $

En donde:
* Intercepto: $ α_0 = C_0 + I_0 + G_0 + X_0 - hr $ 
* Pendiente: $ α_1 = (b - m)(1 - t) $

Entonces, tenemos: $ DA = α_0 + α_1Y $

Y se reemplaza en la ecuación del "Ingreso de equilibrio" en donde  $Y = DA$
$$ Y = C_0 + bY^d + I_0 -hr + G_0 + X_0 - mY^d $$
$$ Y = \frac{1}{1 - (b - m)(1 - t)} (C_0 + I_0 + G_0 + X_0 - hr) $$

$$ k=\frac{1}{1 - (b - m)(1 - t)} $$

Otra forma: 
$$ Y = k (C_0 + I_0 + G_0 + X_0 - hr) $$

Si el gobierno decide aumentar su gasto (∆𝐺 > 0) para aumentar el empleo en la economía.

Asimismo, el aumento del gasto del gobierno significa que el intercepto de la función de la Demanda Agregada para se elevará para arriba. Se puede decir que: si el Gasto(Go) aumenta, la Demanda Agregada (DA) también lo hará. Por ende, el Ingreso de Equilibrio aumentará también. 

$$ ↑Go → ↑DA → DA > Y → ↑Y $$





- Matemáticamente: $∆G_0 > 0  →  ¿∆Y?$

$$ Y_e = \frac{1}{1 - (b - m)(1 - t)} (C_0 + I_0 + G_0 + X_0 - hr) $$

Si : $ k=\frac{1}{1 - (b - m)(1 - t)} $ , en donde $ k > 0 $ , tenemos lo siguiente:

$$ Y = k (C_0 + I_0 + G_0 + X_0 - hr) $$


$$ ∆Y = k (∆C_0 + ∆I_0 + ∆G_0 + ∆X_0 - ∆hr) $$


Como no hay cambios en $C_0$, $I_0$, $X_0$, $h$ ni $r$, entonces: 

$$∆C_0 = ∆I_0 = ∆X_0 = ∆h = ∆r = 0$$

$$ ∆Y = k (∆G_0) $$


Sabiendo que $∆G_0 > 0 $ y que $k > 0$, la multiplicación de un número positivo con otro positivo resultará también en  un resultado positivo:

$$ ∆Y = (+)(+) $$
$$ ∆Y > 0 $$

- Intuición: 

Si tomamos en cuenta el Consumo autónomo (Co) y las Importaciones (M), se generan estas dos contradicciones. 

 * Si la tasa de impuestos disminuye, el Ingreso de Equilibrio aumentará . Por endo, aumenta la demanda de consumo y la demanda agregada y el ingreso disponible
 $$ t↓ → Co↑ → DA↑ → DA > Y → Y↑ $$
 * Por otro lado,  aumentarían las importaciones, lo cual generaría una reducción  en la pendiente de la Demanda Agregada y el valor del multiplicador.Entonces, conlleva a la disminución del ingreso de equilibrio: 
 $$ t↓ → M↑ → DA↓ → DA < Y → Y↓ $$

- Matemáticamente: $∆t < 0  →  ¿∆Y?$


```python
Co, Io, Go, Xo, h, r, b, m, t = symbols('Co Io Go Xo h r b m t')

f = (Co + Io + Go + Xo - h*r)/(1-(b-m)*(1-t))


df_t = diff(f, t)
df_t #∆Y/∆t
```

Tomando en cuenta el diferencial de $∆t$:

 * $∆Y$ =  $ (m-b)(Co + Go + Io + Xo - hr) $
 * $∆t$ = $ (1-(1-t)(b-m)+1)^2 $
 
Significa que:
$$ \frac{∆Y}{∆t} = \frac{(m-b)(Co + Go + Io + Xo - hr)}{(1-(1-t)(b-m)+1)^2} $$

- Sabemos que b > m, entonces $(m-b) < 0$
- Los componentes autónomos no cambian: $∆C_0 = ∆I_0 = ∆X_0 = ∆h = ∆r = 0$
- Cualquier número elevado al cuadrado será positivo: $ (1-(1-t)(b-m)+1)^2 > 0 $

Entonces:

$$ \frac{∆Y}{∆t} = \frac{(-)}{(+)} $$

Dado que $∆t < 0$: La multiplicación de dos negativos, da como resultado uno positivo. Por tanto quedaría en que la división de un positivo y otro positivo resultará positivo también:

$$ \frac{∆Y}{(-)} = \frac{(-)}{(+)} $$

$$ ∆Y = \frac{(-)(-)}{(+)} $$

$$ ∆Y = \frac{(+)}{(+)} $$

$$ ∆Y > 0 $$
