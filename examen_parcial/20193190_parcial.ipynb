El Multiplicador Keynesiano representa en cunto se va a incrementar el producto ante un incremento de cualquiera de los componentes autnomos de la demanda agregada (que conforman el intercepto), ya que genera un proceso multiplicador del ingreso hasta converger al nuevo ingreso y producto de equilibrio.

Su impacto en el modelo Ingreso-Gasto Keynesiano se nota en que, si se produce un incremento en el gasto pblico, por ejemplo, el producto final tendr un incremento mayor al incremento original en la demanda; logrando as un aumento multiplicador del ingreso como consecuencia del efecto del Multiplicador Keynesiano en el aumento del gasto pblico.

En el punto B, la demanda agregada es menor que el nivel de produccin, as que hay un dficit de demanda o un exceso de produccin que da lugar a un aumento de los inventarios y a disminuciones de la produccin hasta llegar al punto de equilibrio.
