# 1.1:
Para calcular la "Curva IS"(IS), a partir de la condici√≥n de equilibrio Y^e = DA : 

Datos: 
$$ C = C_0 + bY^d $$
$$ I = I_0 - hr $$
$$ G = G_0 $$
$$ T = tY $$
$$ X = X_0 $$
$$ M = mY^d $$
----------------------------
Resultado de la suma de "Datos" que es igual a "DA":
$ DA = C + I + G + X -M $

Para encontrar el Y^e , le restaremos la "Tributaci√≥n Total" (T) a la ecuaci√≥n:

$ Y-T = Y-T = Y^dC + I + G + X -M -T $
* $ Y-T = Y^d $ #(dato)

Reemplazamos dato en: $ Y^d= C + I + G+X-M-T $ 

Despejando la inversi√≥n: $ Y^d - C - G - X + M + T = I $
* $ Y^d - C = Sector privado(Sp) $
* $ T - G = Sector gobierno(Sg) $
* $ M-X = sector externo o internacional(Se) $
* $ Sp + Sg+ Se = S $

Ordenando en la ecuaci√≥n:
$ (Y^d - C)+(T-G)+(M-X)= I $ 

* $ (Y^d - C)+(T-G)+(M-X) = Ahorro Total (S) $ 

Por tanto, tenemos que:

$ Sp + Sg+ Se = S $

$ S(Y) =I (r) $

Reemplazando en la ecuaci√≥n del Ahorro Total (S):
* $ Y^d =(1-t)y $ #(dato)

 $ [(Y^d - (C_O + bY^d)]+[t^y - G_O]+[mY^d-X_O]=I_O - hr $
 $ [1-(b-m)(1-t)]*Y - [C_O+G_O+X_O]=I_O - hr $
 
* $ [1-(b-m)(1-t)]*Y - [C_O+G_O+X_O] = S $ #(dato)
* $ I_O - hr =I $ #(dato)
 
Despejando "hr":

$ hr=I_0-[1-(b-m)(1-t)]*Y-(C_O+G_O+Y_O) $

$$ hr = (C_0 + G_0 + I_0 + X_0) - (1 - (b - m)(1 - t))Y $$

$$ r = \frac{1}{h}(C_0 + G_0 + I_0 + X_0) - \frac{1 - (b - m)(1 - t)}{h}Y $$

Simplificando:

$$ r = \frac{X_0}{h} - \frac{X_1}{h}Y $$

* Intercepto: $ X_0 = C_0 + G_0 + I_0 + X_0  $  
* Pendiente: $  X_1 = 1 - (b - m)(1 - t) $ 

# 1.2:

Partiendo de la ecuaci√≥n "r" ( vista en el punto 1.1):

$$ r = \frac{1}{h}(C_0 + G_0 + I_0 + X_0) - \frac{1 - (b - m)(1 - t)}{h}Y $$

Hallar: $ ‚àÜr/‚àÜY $

Considernado el diferencial de $‚àÜr$:

 * $‚àÜr$ =  $ \frac{1}{h}(C_0 + G_0 + I_0 + X_0) - \frac{1 - (b - m)(1 - t)}{h}* ‚àÜY $
 
 Dando forma: 
 
$ ‚àÜr/‚àÜY = - \frac{1 - (b - m)(1 - t)}{h} $ 
 

Entonces:

$$ \frac{‚àÜr}{‚àÜY} = (-)(+) $$

Dado que $‚àÜr < 0$,  la multiplicaci√≥n de unn negativo y un positivo, da como resultado un negativo. Por tanto quedar√≠a:

$$ \frac{‚àÜr}{‚àÜY} = (-) $$

$$ \frac{‚àÜr}{‚àÜY} < 0 $$


La pendiente de la Curva IS es mayor que cero y puede graficarse en el plano (Y,r) con una recta con pendiente negativa.

# Ejercicio 2: 


Ahorro total (ahorro privado + ahorro del gobierno + ahorro externo) = Inversi√≥n (I)

$ S=I_0-h*r $

$ S_p+S_g+S_e=I_0-h*r $

$(Y^d-C_0-b*Y^d)+(T-G_O)+(m*Y^d-X_0)=I_0-h*r$

$Y^d*(1-b+m)+T-G_0-C_0-X_0=I_0-h*r $

$Y^d=Y-T=(1-t)Y$

$Y*(1-t)*(1-b+m)+t*Y-G_0-C_0-X_0=I_0-h*r$

$Y*(t+(1-t)*(1-b+m))-G_0-C_0-X_0-I_0=-h*r$

$Y*(t+(1-t)-(1-t)*b+(1-t)*m)-G_0-C_0-X_0-I_0=-h*r$

$Y*(1-(1-t)*(m-b))-G_0-C_0-X_0-I_0=-h*r$

$G_0-C_0-X_0-I_0-Y*(1-(1-t)*(m-b))=h*r$

$\frac{G_0-C_0-X_0-I_0}{h}-\frac{(1-(1-t)*(m-b))}{h}*Y=r$

```python
# Curva IS
Y_size = 100 

Co = 15
Io = 18
Go = 28
Xo = 5
h = 0.8
b = 0.5
m = 0.3
t = 0.3

Y = np.arange(Y_size)


# Ecuaci√≥n 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = ((Co+Io+Go+Xo)/h-Y*((1-(1-t)*(b-m))/h))
    return r_IS

IS = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)


# Curva IS
   

fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="LA CURVA IS DE EQUILIBRIO EN EL MERCADO DE BIENES", xlabel=r'Y', ylabel=r'r')

ax.text(100, -20, '$IS$', fontsize = 12, color = 'green')

ax.plot(IS, "k-")
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())                                     
plt.show()
```

Los cambios en la propensi√≥n marginal a importar ‚Äúm‚Äù, tienen efectos en la pendiente de la curva IS parecidos a los provocados por los cambios en la tasa de tributaci√≥n ‚Äút‚Äù: cuando disminuye (aumenta) ‚Äúm‚Äù, la pendiente de la curva IS se disminuye (aumenta). Por el contrario, la disminuci√≥n de la propensi√≥n marginal a consumir, ‚Äúb‚Äù, disminuye el tama√±o del multiplicador, disminuye la pendiente de la DA y aumenta la pendiente de la curva IS.
