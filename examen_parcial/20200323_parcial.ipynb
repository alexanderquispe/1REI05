#### (2 puntos) Defina el Multiplicador Keynesiano y dé un ejemplo de cómo influye este en el modelo Ingreso-Gasto Keynesiano.

El multiplicador keynesiano "k" representa en cuanto se va a incrementar el producto ante cambios en los componentes autónomos de la demanda agregada. El multiplicador es un valor que toma en cuenta la propensión marginal a consumir, la tasa impositiva, y la propensión marginal a importar, para mostrar los efectos en el producto ante variaciones de los componentes autónomos (gasto del gobierno, exportaciones netas, consumo o inversión. Si cambia algún componente autónomo, k afecta ese cambio, y el ingreso de equilibrio cambia.

Un ejemplo de cómo influye en el modelo es cuando aumenta el Gasto del Gobierno, por ejemplo, cuando el gobierno decide construír escuelas y hospitales. Digamos que el Go aumenta de 1000 a 1500, y que nuestro multiplicador keynesiano es de 2. Por lo tanto, nuestro producto de equilibrio original que era de 2000 (1000 x 2) ahora va a aumentar, debido a que el Go aumentó a 1500. Por lo tanto, nuestro nuevo ingreso de equilibrio sería de 3000, ya que nuestro multiplicador keynesiano multiplica por 2 a la demanda agregada, que en este caso es solo los 1500 de Go.

En el modelo Ingreso-Gasto, un exceso de demanda se da cuando aumentan los componentes autónomos, es decir, el consumo, las exportaciones netas, las inversiones o el Gasto del Gobierno. En cambio, un déficit sucede cuando la demanda agregada disminuye, es decir, cuando el gasto que representan los compoentes autónomos baja. 

En el gráfico se muestra como el punto A representa un nivel donde la demanda agregada es mayor que el nivel de producción, generando un exceso de demanda. Mientras que el punto B representa un déficit de demanda, ya que la demanda agregada es menor que el producto. El punto E es el punto de equilibrio.

En ambos casos, el producto se debe adaptar a la demanda, lo que hace a Y la variable endógenas del modelo. Si hay un exceso de demanda, entonces la producción debe aumentar. Si hay un déficit de demanda, la producción debe disminuir. La oferta de producción se ajusta a los cambios de la demanda.

Las herramientas de política fiscal son 3:

- El Gasto de gobierno o también llamado gasto público: compras del Estado, remuneraciones de funcionarios y pago por compra de bienes y servicios.
- La tasa impositiva: porcentaje del ingreso que recolecta el gobierno de sus ciudadanos.
- Las transferencias (que en el modelo Ingreso Gasto son 0 por simplicidad)

Las herramientas de política monetaria son 3:

- Operación de Mercado Abierto: compra o venta de activos financieros o bonos a bancos comerciales por parte del Banco Central para aumentas o disminuír la oferta monetaria. Si compra bonos, entonces es una política expansiva, ya que aumenta la base monetaria y en consecuencia la oferta monetaria. Si vende bonos, entonces es una política contractiva, ya que se busca reducir la producción de la economía.

- Coeficiente legal de encaje: el BCR puede modificar la cantidad de dinero que tienen los bancos para realizar préstamos. Puede disminuír la tasa de encaje, lo que aumenta la posibilidad de creación de dinero bancario ya que aumenta el multiplicador de dinero bancario, y eso aumenta la oferta monetaria. En cambio, si aumenta la tasa de encaje, disminuye la oferta monetaria.

- Tasa de interés de referencia: desde 1990, el BCR puede modificar la cantidad de dinero prestado a los bancos comerciales. Reducir la tasa implica que se darán más préstamos y aumentará la oferta, mientras que aumentar la tasa de interés disminuye la oferta monetaria.

Vamos a encontrar el Ye en base a las ecuaciones del modelo Ingreso-Gasto en economía abierta.

Para llegar al equilibrio, debemos igualar DA = Y, es decir, la demanda agregada debe ser igual el ingreso agregado.

La función de demanda agregada es la siguiente:

$$ DA = C + I + G + X - M $$

Pero ahora debemos reemplazar las nuevas funciones de cada componente autónomo para el nuevo modelo de economía abierta:

$$ DA = C_0 + b(1 − t)Y + I_0 − hr + G_0 + x_1Y∗ + x_2ε − m1(1 − t)Y + m_2ε $$
$$ DA = C_0 + I_0 + G_0 + [b(1 − t) − m_1(1 − t)]Y − hr + x_1Y∗ + (x_2 + m_2)ε $$

Ahora podemos unir a Co, Io y Go en una sola variable "f"

$$ f = C_o + I_0 + G_0 $$

$$ DA = [f + x_1Y∗ + (x_2 + m_2)ε_0 + (x_2 + m_2)ρr∗ − (h + (x_2 + m_2)ρ)r] + [(b − m_1)(1 − t)]Y $$

Por lo tanto, este sería nuestra ecuación de la demanda agregada:

$$ DA = α_0 + α_1Y $$

Donde el intercepto es:

$$ α_0 = [C_0 + I_0 + G_0 + x_1Y∗ + (x_2 + m_2)ε_0 + (x_2 + m_2)ρr∗ − (h + (x_2 + m_2)ρ)r] $$

Y la pendiente es:

$$ α_1 = [(b − m_1)(1 − t)] $$

Para encontrar el equilibrio ahora, debemos igualar Y con DA:

$$ Y_e = f − [h + (x_2 + m_2)ρ]r + x_1Y∗ + (x_2 + m_2)ε_0 + (x_2 + m_2)ρr∗ + [(b − m_1)(1 − t)]Y $$

$$ Y_e = \frac{1}{1 − (b − m_1)(1 − t)} [f + (x_2 + m_2)ε_0 + (x_2 + m_2)ρr∗ + x_1Y∗] − \frac{h + (x_2 + m_2)ρ}{1 − (b − m_1)(1 − t) r} $$

En este caso, el multiplicador keynesiano es de:

$$ k = \frac{1}{1 − (b − m_1)(1 − t)} $$
