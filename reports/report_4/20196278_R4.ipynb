En primer lugar, mencoinamos las curva LM y IS como los dos mercados que estas en confluencia. Ambos mercados evaluan diferentes dimensiones, sin embargo, logran un punto de equilibiro al intersectarce. 
Las variables de gran importancia seran entonces "r, Y". Como taza y producto, en dónde los demás elemento influirán. 
1.En primer lugar: mencoinamos las ecuaciones
- $IS$ ------------→$r =(Co+Io+Xo)/h-((1-(b-m)*(1-t) ) / h)*Y$
- $LM$ ------------→$r = - (1/j)*(Ms/P) + (k/j)*Y$
2.Luego, evaluamos las ecuaciones. factorizamos todo para que quede en las dos ecuaciones la variable "r", a un costado. En ese estructura, igualamos ecuaciones:
- $(Co+Io+Xo)/h-((1-(b-m)*(1-t) ) / h)*Y=- (1/j)*(Ms/P) + (k/j)*Y$
3.En esta ecuación podemos realizar un factorización de la variable "Y".en una ecuación de equilibrio 
- $Ye= (j*(Co+Io+Go+Xo)/(h*k+j*(1-(b-m)*(1-t)))) + h/(h*k+j(1-(b-m)*(1-t)))*(Ms/P)$
4.así como se realizó el procedimeitnoe de factorización de "Y". Se realiza el procedimiento para "Y". El resultado serái el sigiente en otra ecuación de equilibiro
- $re= (k*(Co+Io+Go+Xo)/(h*k+j*(1-(b-m)*(1-t)))) + ((1-(b-m)*(1-t))/(h*k+j(1-(b-m)*(1-t)))*(Ms/P)$


$conslución$
- Para encontrar el punto de equilibrio, simplemente se necesitario encontrar tanto "Y" o "r", pero, depués, abría que reemplazar ese valor, en cualquiera de las ecuaciones de LM-IS, para encontrar la otra varible importante. 
Es decir, al encontrar en culaquiera de las ecuaciones de equilibrio, si encuentro "Y", entonces, los reemplazaré en cualquiera de las ecuacioens de LM-IS, para encontrar "r", y viceversa con "r" para enctrar posteriormente "Y". 


- PUNTO 2: graficar el punto de equilibrio

```python
#--------------------------------------------------
    # Curva IS

# Parámetros

Y_size = 100 

Co = 35
Io = 40
Go = 50
Xo = 2
h = 0.8
b = 0.4
m = 0.5
t = 0.8

Y = np.arange(Y_size)


# Ecuación 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo)/h - ( ( 1-(b-m)*(1-t) ) / h)*Y  
    return r_IS

r_is = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)


#--------------------------------------------------
    # Curva LM 

# Parámetros

Y_size = 100

k = 2
j = 1                
Ms = 200             
P  = 20               

Y = np.arange(Y_size)

# Ecuación LM

def r_LM(k, j, Ms, P, Y):
    r_LM = - (1/j)*(Ms/P) + (k/j)*Y
    return r_LM

r_lm = r_LM( k, j, Ms, P, Y)

# Gráfico del modelo IS-LM

# Dimensiones del gráfico
y_max = np.max(r_lm)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
# Curva IS
ax.plot(Y, r_is, label = "IS", color = "C1") #IS
# Curva LM
ax.plot(Y, r_lm, label="LM", color = "C0")  #LM

# Eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

# Texto y figuras agregadas
# Graficar la linea horizontal - r
plt.axvline(x=51.5,  ymin= 0, ymax= 0.52, linestyle = ":", color = "black")
# Grafica la linea vertical - Y
plt.axhline(y=93, xmin= 0, xmax= 0.52, linestyle = ":", color = "black")

# Plotear los textos 
plt.text(49,100, '$E_0$', fontsize = 14, color = 'black')
plt.text(0,100, '$r_0$', fontsize = 12, color = 'black')
plt.text(53,-10, '$Y_0$', fontsize = 12, color = 'black')

#Adicional texto para el grafico
plt.text(100,20, '$LM$', fontsize = 15, color = 'black')
plt.text(100,190, '$IS$', fontsize = 15, color = 'black')
plt.text(80,100, '$II$', fontsize = 20, color = 'black')
plt.text(15,100, '$IV$', fontsize = 20, color = 'black')
plt.text(48,145, '$I$', fontsize = 20, color = 'black')
plt.text(48,45, '$III$', fontsize = 20, color = 'black')

# Título, ejes y leyenda
ax.set(title="IS-LM Model", xlabel= r'Y', ylabel= r'r')
ax.legend()

plt.show()
```
