# 1. Explique cuales son los instrumentos de politica monetaria que puede utilizar el Banco Central

* El Banco Central de Reserva puede aplicar una política monetaria expansiva o contractiva mediante diferentes instrumentos como la oferta de dinero.
La política monetaria es expansiva cuando el Banco Central de Reserva incrementa la cantidad de dinero en el mercado y es contractiva cuando reduce la cantidad de dinero del mercado.
En relación a esto pueden existir los siguiente como instrumentos de política monetaria
1) La Oferta Monetaria, el BCR utiliza el instrumento Operación de Mercado Abierto mediante la cual compra o vende activos financieros o bonos a los bancos comerciales. Es expansiva cuando se compra y es contractiva cuando se vende.
2) Coeficiente Legal de encaje, mediante la cual en la política expansiva el BCR puede incrementar la cantidad de dinero que tienen los bancos para realizar préstamos pero esto conlleva a que los bancos deben tener menos dinero en forma de reservas. Y en una política contractiva el BCR aumenta el coeficiente de encaje, lo que conlleva que los bancos cuenten con mayor proporción de depósitos en forma de reservas.
3) La tasa de interés de referencia de la política monetaria, en una política monetaria expansiva el BCR reduce la tasa de interés de referencia y en una política monetaria contractiva el BCR aumenta la tasa de interés de referencia.     


# 2. Derive la oferta real del dinero y explique cada uno de sus componentes

* La ofreta nominal de dinero $ (M^S) $ es una variable exogena e instrumento de la politica monetaria. Para expresarla en términos reales o precios constantes, la oferta nominal del dinerso se dividde entre en nivel general de precios de la economia (P)
  
 $$ \frac{M^S}{P} = \frac{M_O^S}{P} $$
 
 Donde $M^S$ es la oferta de dinero $P$ son los precios y $M_O^S$ es la masa monetaria nominal.  


# 3. Derive la demanda real de dinero. Explique el papel que cumplen los parametros "k" y "j"

* La demanda de dinero por los motivos de transacción y precaución será esenta el equilibrio en el mercado de bienes

  $$ M^d = Ky - jr $$
  
  $$ jr = Ky - Md $$
  
  $$ r  = \frac{k}{j}y - \frac{M^d}{j} $$
  
  
    * Donde: k = Elasticidad del ingreso | Y =  Ingreso

    * Donde: i = tasa de interes nominal | j = elasticidad de sustitución del dinero en mano vs dinero en banco

* Si definimos que ambos tipos de demanda están en términos reales, la función de demanda del dinero $(M^d)$ ser:

  $$ M^d = L_1 + L_2 $$

  $$ M^d = k_Y - j_i  $$
  
  Donde: $L_1$ esta conformada por la demanda transacción y precuación y $L_2$ por la demanda de especulación.

* Por lo tanto se define lo siguiente: 

    * La magnitud del parámetro $"k"$ indica la sensibilidad de la demanda de dinero ante variaciones del ingreso $"Y"$
    * La magnitud del parámetro $"j"$ indica cuan sensible es de la demanda de dinero ante las variaciones de la tasa de
      interéses de los bonos $"i"$

# Derive paso a paso la curva LM matemáticamente (a partir del equilibrio en el Mercado Monetario) y grafique


   $$ M^d = KY - jr $$
   
   $$ M^S = \frac{M_O^S}{P} $$
   
   $$ M^S = M^d $$

   $$ \frac{M_O^S}{P} = KY - jr $$

   $$ Ky - \frac{M_O^S}{P} =  jr  $$

   $$ \frac{KY}{j} - \frac{M_O^S}{Pj} = r $$
   
   $$ r = -\frac{1}{j}\frac{M_O^S}{P_O} + \frac{K}{J}Y $$

```python
# Parameters
Y_size = 100

k = 10
j = 10                
Ms = 200             
P  = 20               

Y = np.arange(Y_size)


# Ecuación

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

LM = i_LM( k, j, Ms, P, Y)
LM
```

```python


y_max = np.max(LM)
v = [0, Y_size, 0, y_max]   
fig, ax = plt.subplots(figsize=(10, 8))


ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())


ax.plot(Y, LM, label=" CURVA LM", color = '#3D59AB')


ax.set(title="CURVA LM", xlabel=r'Y', ylabel=r'r')
ax.legend()


plt.show()
```
