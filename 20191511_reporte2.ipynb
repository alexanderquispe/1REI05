B)La Curva IS o el equilibrio Ahorro- Inversión
--

-Derive la ecuación de IS a partir de la igualdad Ahorro-Inversión y grafique la curva IS de equilibrio en el Mercado de Bienes

Entendemos nuestra fómula de equilibrio como 
            
            Y=DA
            
En donde 
            
            DA= C+I+G+M-X

Por lo tanto:

            Y=C+I+G+M-X
           
           
Además conocemos que 
-    T=ty
-    y(d)=Y-T
- C=Co+by(d)
- I=Io-hr
- G=Go
- X-M= Xo-my(d)

Ya que estamos analizando desde el análisis parte desde el tasa de interés; por lo que se hacen las siguientes operaciones

            Y-T=C+I+G+X-M-T
            
            y(d)=C+I+G+X-M-T
            
Pasamos todo al otro lado

            y(d)-C-G-X+M+T = I
            
            (y(d)-C) + (T-G) + (M-X) = I
            
            
De ahí encontramos tres subdivisiones de ahorro:

- El ahorro privado             (y(d)-C)

- El ahorro de Gobierno         (T-G)

- El ahorro Externo:            (M-X)


Entonces la suma de los tres tipos de ahorro (S) es la misma expresion de quilibro con respecto a la inversión: 

             S=I
             
Reemplazando econtramos la ecuación para la Curva IS en cuanto a el quilibrio Ahorro= Inversión

        (y(d)-C) + (T-G) + (M-X) = I
        
        (y(d)- Co -by(d)) + (ty-Go) + (my(d)-Xo)= Io - hr
        
        (1-(b-m)(1-t)Y - (Co + Go + Xo) = Io-hr
        
        
En donde:

- La función ahorro está representado por "(1-(b-m)(1-t)Y - (Co + Go + Xo)"
- La función inversión está representado por  "I = I0 − hr"
        
Siguiendo con la operación de la fórmula derivamos que:
 
                    hr= Io-[1-(b-m)(1-t)Y - (Co+Go+Xo)
                    hr= [Io+ (Co+Go+Xo)] - [1-(b-m)(1-t)]Y >0
                    r = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
                    
En donde nuestro intercepto está representado por:

            L1: Io+Co+Go+Xo
            
Y nuestra pendiente está representado por:

            L2: 1-(b-m)(1-t)
           

```python
# Parameters

Y_size = 100 

Co = 45
Io = 50
Go = 80
Xo = 2
h = 0.5
b = 0.8
m = 0.2
t = 0.3

Y = np.arange(Y_size)


def Ejer4(b, m, t, Co, Io, Go, Xo, h, Y):
    Ejer4 = (1/h)*(Co + Io + Go + Xo) - (1/h)*(1-(b-m)*(1-t))*Y
    return Ejer4

Ejercicio4 = Ejer4(b, m, t, Co, Io, Go, Xo, h, Y)
```

```python
#Graficamos
y_max = np.max(Ejercicio4)
fig, ax = plt.subplots(figsize=(10, 8))


ax.plot(Y, r, label = "IS", color = "#6C3483") 

ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

ax.set(title = "Curva de IS", xlabel= 'Y', ylabel= 'r')
ax.legend()

plt.show()
```

- Analice una Política Fiscal Expansiva con una caída de la Tasa de Impuestos . El análisis debe ser intuitivo y gráfico

Para el analisis intuitivo de un Política Fiscal Expansiva con una caída de la Tasa de Impuestos ,entendemos que existe dos cambios al modelo de equilibrio
En primer lugar, la caída de la tasa de impuestos tiene un efecto inverso en el consumo y en la demanda, en donde:
    
    t↓ = C↑ = DA↑
    
Entonces el equilibrio se pierde por el aumento de la demanda y Y tiene que subir para encontrar nuevamente el equilibrio:

     DA↑ = DA>Y = Y↑
     
Por otro lado, la caída de la tasa de impuesto tiene un efecto inverso con respecto a las importanciones y este, por consecuencia, afecta inversamente a la demanda, es por ello que vemos que:

     t↓ = M↑ = DA↓

Entonces el equilibrio se pierde por la disminución de la demanda y Y tiene que bajar para encontrar nuevamente el equilibrio

     DA↓ = DA<Y = Y↓
 
Eso significaría también que el intento de enquilibrio de ambos modelos de cambios son contradictorios


```python
Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.7

Y = np.arange(Y_size)


# Ecuación 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)


#--------------------------------------------------

t = 0.3


def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r_t = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
```

```python

y_max = np.max(r)
fig, ax = plt.subplots(figsize=(10, 8))


ax.plot(Y, r, label = "IS", color = "black") #IS orginal
ax.plot(Y, r_t, label = "IS_t", color = "#9863E4", linestyle = 'dashed') #Nueva IS


plt.text(47, 162, '∆t', fontsize=12, color='black')
plt.text(47, 158, '->', fontsize=15, color='grey')


ax.set(title = "Política Fiscal Expansiva con una caída de la Tasa de Impuestos $(t)$", xlabel= 'Y', ylabel= 'r')
ax.legend()

plt.show()
```
