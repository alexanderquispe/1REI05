#### 1. (2 puntos) Defina el Multiplicador Keynesiano y dé un ejemplo de cómo influye este en el modelo Ingreso-Gasto Keynesiano.

El modelo Ingreso-Gasto Keynesiano se caracteriza por dar un enfoque a la demanda agregada y su gran importancia en la economía.
El modelo establece 3 supuestos:
- el nivel de precios es fijo
- el producto se adapta a la DA
- la tasa de interés está determinada fuera del modelo
Se debe considerar además que este es un modelo a corto plazo.
En este sentido, cuando hay un cambio en la magnitud del consumo, la inversión, el gasto o las exportaciones, se genera un evento multiplicador del ingreso hasta ocasionar que la el nuevo ingreso y el producto de equilibrio puedan coincidir. De esta manera, el aumento multiplicado del ingreso es resultado de los cambios generados -por cualquier agente directo o indirecto- en el aumento de cualquiera de los elementos del intercepto de la DA.
Si se presenta que : 

$$DA = α_0 + α_1Y$$

Y, si la condición es:
$$Y = DA$$

Entonces:
$$(1 - α_1)Y = α_0$$

En este sentido, el multiplicador keynesiano es:
$$\frac{1}{1 - α_1} = \frac{1}{1-(b-m)(1-t)}$$

Por otro lado, cabe mencionar que el tamaño de este multiplicador dependerá de los parámetros "internos" b, m y t.
Partiendo de una economía cerrada y sin gobierno y del sector de bienes de inversión y de bienes de consumo, un ejemplo de como influye este multiplicador en el modelo se da ubicando el efecto multiplicador keynesiano: primero, cuando aumenta la demanda de inversión, hay un incremento en la producción e ingreso en el sector de bienes de inversión. Este incremento del ingreso ocasiona un aumento en la demanda de consumo y por eso, tambiñen hay un aumento en la producción e ingreso en el sector de bienes de consumo.

#### 2. (2 puntos) Grafique y explique cúando sucede un exceso y déficit de demanda en el modelo Ingreso-Gasto. Señale las áreas donde ocurre cada caso. Explique cómo se converge al equilibrio a partir de estos dos escenarios.

##### Graficando

```python
#Curva de ingreso --------------------------

#Parámetros
Y_size = 100 

Co = 35
Io = 40
Go = 40
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3
r = 0.9

Y = np.arange(Y_size)

    # Ecuación 
def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)

#--------------------------------------------------
# 45°

a = 2.5 

def L_45(a, Y):
    L_45 = a*Y
    return L_45

L_45 = L_45(a, Y)
```

```python
# Gráfico
y_max = np.max(DA_IS_K)
fig, ax = plt.subplots(figsize=(10, 8))


# Curvas a graficar
ax.plot(Y, DA_IS_K, label = "DA", color = "#EA047E") #curva ORIGINAL
ax.plot(Y, L_45, color = "#FF6D28") #línea de 45º

# Lineas punteadas
plt.axvline(x = 75, ymin= 0, ymax = 0.75, linestyle = ":", color = "grey")
plt.axvline(x = 56,  ymin= 0, ymax = 0.56, linestyle = ":", color = "grey")
plt.axvline(x = 20,  ymin= 0, ymax = 0.50, linestyle = ":", color = "grey")
plt.axhline(y = 139, xmin= 0, xmax = 0.55, linestyle = ":", color = "grey")

# Texto agregado
plt.text(0, 150, '$DA^e$', fontsize = 11.5, color = 'turquoise')
plt.text(55, 145, '$E$', fontsize = 11.5, color = 'turquoise')
plt.text(75, 150, '$B$', fontsize = 11.5, color = 'turquoise')
plt.text(20, 130, '$A$', fontsize = 11.5, color = 'turquoise')
plt.text(6, 4, '$45°$', fontsize = 11.5, color = 'turquoise')
plt.text(2.5, -3, '$◝$', fontsize = 30, color = 'turquoise')
plt.text(57, 0, '$Y^e$', fontsize = 12, color = 'turquoise')

# Título y leyenda
ax.set(title = "Exceso y déficit de demanda en el Modelo Ingreso-Gasto", xlabel = r'Y', ylabel = r'DA')
ax.legend()

plt.show()
```

Para poder explicar el exceso y déficit de la demanda en el modelo Ingreso-Gasto se tiene que observar "A", "E" (el equilibrio) y "B". 
- En "A": 
 $$DA > nivel de producción$$ 
Esto ocasiona que la producción se incremente hasta poder llegar al equilibrio.

- En "E":
Se presenta E como punto de equilibrio y referencia para la diferenciación entre A y B.

- En "B":
En B se puede apreciar un déficit de la demanda o, visto de otra manera, un aumento de la producción que ocasiona la disminución de la producción hasta llegar al equilibrio.

Si la Demanda Agregada es :
$$DA=C+I+G+X-M $$

Y tenemos que:
$C= C_o+bY^d$

$I= I_o-hr$

$G= G_o$

$X= x_1Y^*+x_2ε$

$M= m_1Y^d-m_2ε$

$T= tY$

$Y^d=Y-T$

$DA=C+I+G+(X-M)$

$Y=DA$

$ε=ε_0-p(r-r^*)$

Por lo tanto:

$Y= C_0+bY^d+I_0-hr+G_0+x_1Y^*+x_2ε-m_1Y^d+m_2ε$

$Y-(b-m_1)Y^d = C_0 + I_0 + G_0 + x_1Y^* +(x_2+m_2)ε-hr$

$Y-(b-m_1)(1-t)Y=C_0 + I_0 + G_0  + x_1Y^* + (x_2+m_2)(ε_0-pr+pr^*)-hr$

$Y[1-(b-m_1)(1-t)]=C_0 + I_0 + G_0 + x_1Y^* + (x_2+m_2)(ε_0+pr^*)-[(x_2+m_2)p+h]r$

$Y=\frac{1}{1-(b-m_1)(1-t)} [C_0 + I_0 + G_0 + x_1Y^* + (x_2+m_2)(ε_0+pr^*)]-\frac{1}{1-(b-m_1)(1-t)}[(x_2+m_2)p+h]r$

En este sentido, se reconoce que el multiplicador es: $\frac{1}{1-(b-m_1)(1-t)}$
