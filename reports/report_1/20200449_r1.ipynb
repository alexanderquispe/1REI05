Pregunta 1: Derive y explique la función de demanda de Consumo.

Formula de la demanda de consumo:
$$C= Co+bY^d$$

Formula de la tributación: 
$$T=tY$$
$$Y^d=Y-T$$

Factorizado:
$$Y^d=(1-t)Y$$

Con el Ingreso disponible se puede ver el consumo variable. Se obtiene de la multiplicación del ingreso disponible con l propensión marginal a consumir:
$$C=Co+bY^d$$


```python
# Parámetros

b = 0.5 
Co = 10

Y_size = 100 

Y = np.arange(Y_size)

# Ecuación de la curva del ingreso de equilibrio

def DA_A(Co, b):
    DA_A = (Co) + (b*Y)
    return DA_A

DA_A_A = DA_A(Co, b)

a = 2.5 

def L_45(a, Y):
    L_45 = a*Y
    return L_45

L_45 = L_45(a, Y)
```

```python
# GráfiO 
y_max = np.max(DA_A_A)
fig, ax = plt.subplots(figsize=(9, 7))

# Curvas a graficar
ax.plot(Y, DA_A_A, label = "C")

# Título y leyenda
ax.set(title="Función de demanda de consumo", xlabel= r'Y', ylabel= r'C')
ax.legend() #mostrar leyenda

plt.show()
```

Pregunta 2: Derive y explique la función de Inversión. Replique el gráfico siguiente:

La inversión autónoma depende principalmente de las expectativas de los inversionistas:
$$Io$$

r representa la tasa de interés (cuanto le cobran de interes a los inversionistas). Si el r esté más cerca al 1, influenciará de manera más negativa a la inversión en el país debido a que los inversionistas tendrán menos incentivos a invertir. Por ello se usa el signo (-) pues a mayor sea el r menor será la inversión.
$$-r$$

h es la constante de decisión a invertir (0-1)
$$h$$

La ecuación quedaría:
$$I=Io-hr$$ 

```python
# Parámetros
h =  0.7
Io = 40
r_size = 100 
r = np.arange(r_size)

# Ecuación de la curva del ingreso de equilibrio

def DA_B(Io, h):
    DA_B = (Io) - (h*r)
    return DA_B

DA_B_B = DA_B(Io, h)

#recta 45°

a = 2.5 

def L_45(a, r):
    L_45 = a*r
    return L_45

L_45 = L_45(a, r)
```

Pregunta 3: ¿Cuáles son los supuestos del modelo Ingreso-Gasto Keynesiano?

Las ecuaciones del modelo de Ingreso -Gasto y la función de Demanda Agregada: 
$$ DA = C + I + G + X - M $$
Obtenemos los datos necesarios para poder obtener la demanda de consumo, gasto publico, inversiones y exportaciones menos importacione
$$DA=Co+bY^d+Io-hr+Go+Xo-mY^d$$
Al factorizar
$$ DA = α_0 + α_1Y $$

En ese sentido, el supuesto del ingreso-Gasto Keynesiano se basa en el equilibrio
En equilibrio la Demanda Agregada (DA) debe ser igual al ingreso agregado Y donde se obtiene la siguiente formula de equilibrio de producción a corto plazo:
Dado que D=Y:
$$ Y = C_0 + bY^d + I_0 -hr + G_0 + X_0 - mY^d $$
A corto plazo se supone que los precios no son elasticos, sino fijos. MIentras que la oferta agregada es muy elastica en función a cada precio. En este sentido la oferta está en función de la demanda agregada.
El multiplicador keynesiano: El multiplicador keynesiano depende de las variables t, b y m, pues viene a representar su denominador. Además, un cambio de las variables autonomas de demanda agregada generan un cambio multiplicado en Y
