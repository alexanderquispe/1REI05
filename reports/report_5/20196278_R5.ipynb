Al evaluar el Ye de equilibrio, se debe considearar las ecuaciones de IS-LM y DA-OA En las que la variable endpogenas son 

-$ Y_e, r_e, P_e$
$$Encontrar "r_e"$$
Entendemos a la IS-LM desde las ecuaciones de equilibrio de 
- $r=β_0/h - (β_1)Y/h$  
- $r=- M_0/jP+ kY/j$

De ambas ecuaciones podemos depejar las variables Y de producción. Posteriormente igualariamos las ecuaciones en las que factorizaríamos "r" a un lado de la ecuación. Eso nos daría el "r" de equilibrio. 
- $r_e= ((kβ_0)/(kh+jβ_1)) + (M_0 β_1)/(kh+β_1)P_0$

$$Encontrar "Y_e"$$
A continuación, el Y_e de equilibrio simplemente se daría al igualar las dos ecuaciones DA- OA : demanda agregada y oferta agregada.
- $ DA= ((h*M_s + j*β_0)/h) - (Y*(j*β_1 + h*k)/h)$
- $ OA=  Pe + θ*(Y-_Y)$

De amabs ecuaciones al igualarlas 

- $((h*Ms + j*B0)/h) - (Y*(j*B1 + h*k)/h)= Pe + θ*(Y-_Y)$

despejamos a un ado de la ecuación al elemento Y,para encontrar el Y_e: finalmnete lo ordenamos.
- $Y_e = ( (1)/(θ + ( (j*β_1+h*k)/h) ) )*( ( (h*M_s+j*β_0)/h ) - Pe + θ* _Y )$


$$Encontrar "P_e"$$
Continuando, para encontrar el P_e de equilibrio consideramos en primer lugar, para motivos de graficar un recta simple de intercepto y pendiente: el elemento
- $M_0/P$

se reemplazara por

- $M_0 - P$

Ademas, para econtra el $P_e$ de equilibrio, primero se retoma la ecuación de la $Y_e$ de equilibrio de la función IS-LM. 
Con el cambio de $M_0-P$. Efectuamos la operación de todas las varaibles endógenas para hallar en el $Y_e$ de equilibrio. Una vez encontrado, se reemplaza en la ecuación de oferta agregada que se define de la siguiente manera
- $P=Pe+θ (Y-_Y)$
Se reemplaza el Y, con el $Y_e$ de qeuilibrio que se encontró previamente y se efectua para econtrar el $P_e$ de equilibrio
- $P_e=Pe + θ(Y_e-_Y)$
Notese que se reemplaza $Y$, por $Y_e$


```python
#1--------------------------
    # Demanda Agregada
    
# Parámetros

Y_size = 100 

Co = 35
Io = 40
Go = 50
Xo = 2
h = 0.8
b = 0.4
m = 0.5
t = 0.8

k = 2
j = 1                
Ms = 200             
P  = 8  

Y = np.arange(Y_size)


# Ecuación

B0 = Co + Io + Go + Xo
B1 = 1 - (b-m)*(1-t)

def P_AD(h, Ms, j, B0, B1, k, Y):
    P_AD = ((h*Ms + j*B0)/h) - (Y*(j*B1 + h*k)/h)
    return P_AD

P_AD = P_AD(h, Ms, j, B0, B1, k, Y)


#2--------------------------
    # Oferta Agregada
    
# Parámetros

Y_size = 100

Pe = 100 
θ = 3
_Y = 20   

Y = np.arange(Y_size)


# Ecuación

def P_AS(Pe, _Y, Y, θ):
    P_AS = Pe + θ*(Y-_Y)
    return P_AS

P_AS = P_AS(Pe, _Y, Y, θ)
# líneas punteadas autómaticas

# definir la función line_intersection
def line_intersection(line1, line2):
    xdiff = (line1[0][0] - line1[1][0], line2[0][0] - line2[1][0])
    ydiff = (line1[0][1] - line1[1][1], line2[0][1] - line2[1][1])

    def det(a, b):
        return a[0] * b[1] - a[1] * b[0]

    div = det(xdiff, ydiff)
    if div == 0:
       raise Exception('lines do not intersect')

    d = (det(*line1), det(*line2))
    x = det(d, xdiff) / div
    y = det(d, ydiff) / div
    return x, y

    # coordenadas de las curvas (x,y)
A = [P_AD[0], Y[0]] # DA, coordenada inicio
B = [P_AD[-1], Y[-1]] # DA, coordenada fin

C = [P_AS[0], Y[0]] # L_45, coordenada inicio
D = [P_AS[-1], Y[-1]] # L_45, coordenada fin

    # creación de intersección

intersec = line_intersection((A, B), (C, D))
intersec # (y,x)



    # Gráfico del modelo DA-OA

# Dimensiones del gráfico
y_max = np.max(P)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar 
ax.plot(Y, P_AD, label = "DA", color = "C4") #DA
ax.plot(Y, P_AS, label = "OA", color = "C8") #OA

# Líneas punteadas
plt.axhline(y=intersec[0], xmin= 0, xmax= 0.5, linestyle = ":", color = "grey")
plt.axvline(x=intersec[1],  ymin= 0, ymax= 0.49, linestyle = ":", color = "grey")

# Texto agregado
plt.text(0, 200, '$P_0$', fontsize = 12, color = 'black')
plt.text(53, 25, '$Y_0$', fontsize = 12, color = 'black')
plt.text(50, 202, '$E_0$', fontsize = 12, color = 'black')
plt.text(90, 70, '$DA$', fontsize = 16, color = 'black')
plt.text(90, 300, '$OA$', fontsize = 16, color = 'black')

# Eliminar valores de ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

# Título, ejes y leyenda
ax.set(title="DA-OA", xlabel= r'Y', ylabel= r'P')
ax.legend()

plt.show()
```

$ CASO 3$
 incremento de la tasa de impuestos. Δ t>0

$Intuición$



Se puede ver que tiene un correlato con el modelo IS. en la que la Taza "t" afecta a "r" por la expresión $r= (β_0 / h)- (β_1 * Y /h)$, genera que "r" se reduzca

 
$$ t↑ → r ↓$$

Como consecuencia, la taza "r" afecta a I de inversiones en tantao la expresión: I=Io - hr. Esto produce una subida en inversiones y por tanto en la DA demanda agregada
 
$$ r↓→DA↑$$

se genera un desvalance que el Y de producción de e corregir aumentadose, pero esto a su vez ocasiona que el "P" también finalmente cambie

$$ DA↑→Y↑→P↑$$

$Matemáticamente$

```python
# nombrar variables como símbolos de IS
Co, Io, Go, Xo, h, r, b, m, t, beta_0, beta_1  = symbols('Co Io Go Xo h r b m t beta_0, beta_1')

# nombrar variables como símbolos de LM 
k, j, Ms, P, Y = symbols('k j Ms P Y')

# nombrar variables como símbolos para curva de oferta agregada
Pe, _Y, Y, θ = symbols('Pe, _Y, Y, θ')

# Beta_0 y beta_1
beta_0 = (Co + Io + Go + Xo)
beta_1 = ( 1-(b-m)*(1-t) )

# Producto de equilibrio en el modelo DA-OA
Y_eq = ( (1)/(θ + ( (j*beta_1+h*k)/h) ) )*( ( (h*Ms+j*beta_0)/h ) - Pe + θ*_Y )

# Precio de equilibrio en el modelo DA-OA 
P_eq = Pe + θ*(Y_eq - _Y)

# Tasa de interés de equilibrio en el modelo DA-OA
r_eq = (j*beta_0)/(k*h + j*beta_1) + ( h / (k*h + j*beta_1) )*(Ms - P_eq)
#((h*Ms+j*beta_0)/h) - ((j*beta_1+h*r)/h)*((P-Pe-θ*_Y)/θ)
```

```python
# Y_eq con respecto a Ms
df_Y_eq_t = diff(Y_eq, t)
print("El Diferencial del Producto con respecto al diferencial de la taza t = ", df_Y_eq_t)

# P_eq con respecto a Ms
df_P_eq_t = diff(P_eq, t)
print("El Diferencial del nivel de precios con respecto a la taza t = ", df_P_eq_t)
```
