Entonces, el exceso de oferta se da cuando la tasa de inter√©s ‚Äúr‚Äù se mantiene constante, pero hay un mayor ingreso que el de equilibrio ‚ÄúYeq‚Äù. Por tanto, en la gr√°fica, ver√≠amos que los pares ordenados se encontrar√≠a a la derecha de la curva IS.

Adem√°s, el exceso de demanda se da cuando la tasa de inter√©s ‚Äúr‚Äù se mantiene constante, pero hay un menor ingreso que el de equilibrio ‚ÄúYeq‚Äù. Por tanto, en la gr√°fica, ver√≠amos que los pares ordenados se encontrar√≠a a la izquierda de la curva IS, y habr√≠a un menor ahorro.

```python
# Par√°metros

Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3

Y = np.arange(Y_size)


# Ecuaci√≥n 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (1/h)*(Co + Io + Go + Xo) - (1/h)*(1-(b-m)*(1-t))*Y
    return r_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
```

```python
# Gr√°fico de la curva IS

# Dimensiones del gr√°fico
y_max = np.max(r)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, r, label = "IS", color = "red") #Demanda agregada

# Eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

plt.text(98, 136, 'IS', fontsize = 12, color = 'black')
plt.text(70, 180, 'Exceso de Oferta', fontsize = 12, color = 'black')
plt.text(15, 150, 'Exceso de Demanda', fontsize = 12, color = 'black')
plt.text(30, 187, 'A', fontsize = 12, color = 'black')
plt.text(48, 187, 'B', fontsize = 12, color = 'black')
plt.text(12, 187, 'C', fontsize = 12, color = 'black')

plt.axhline(y = 185, xmin= 0, xmax = 0.49, linestyle = ":", color = "grey")

ax.axvline(x = 12, ymin= 0, ymax = 0.68, linestyle = ":", color = "grey")
ax.axvline(x = 30,  ymin= 0, ymax = 0.68, linestyle = ":", color = "grey")
ax.axvline(x = 48,  ymin= 0, ymax = 0.68, linestyle = ":", color = "grey")

# T√≠tulo, ejes y leyenda
ax.set(title = "Curva IS", xlabel= 'Y', ylabel= 'r')
ax.legend()


plt.show()
```

Respecto de la pol√≠tica monetaria:

Hay 3 posibles instrumentos de la pol√≠tica monetaria:
1) La Oferta Monetaria como instrumento de pol√≠tica

El primer instrumento de pol√≠tica monetaria que utiliza el Banco Central para aumentar o disminuir la oferta monetaria, es la Operaci√≥n de Mercado Abierto. El Banco Central hace estas operaciones mediante la compra o venta de activos financieros o bonos a los bancos comerciales. En esta, la oferta monetaria es una variable end√≥gena. 


2) El coeficiente legal de encaje como instrumento de pol√≠tica

El Banco Central puede incrementar la cantidad de dinero que tienen los bancos para realizar pr√©stamos mediante la disminuci√≥n de la tasa de encaje, porque ello implica que los bancos comerciales deben tener menos dinero en forma de reservas en el balance del Banco Central. En otras palabras, la disminuci√≥n de la tasa de encaje aumenta laposibilidad de creaci√≥n de dinero bancario porque aumenta el multiplicador de dinero bancario; y, este aumento del dinero bancario implica un aumento de la oferta monetaria.
Por otra parte, cuando el Banco Central aumenta el coeficiente de encaje, los bancos comerciales deben tener una mayor proporci√≥n de dep√≥sitos en forma de reservas, lo que reduce el multiplicador bancario y, por tanto, origina una disminuci√≥n de la oferta monetaria.

3) La tasa de inter√©s como instrumento de pol√≠tica

Desde los a√±os 1990 los Bancos Centrales optaron por un nuevo esquema de pol√≠tica monetaria. La tasa de inter√©s se convirti√≥ en instrumento de pol√≠tica y la oferta monetaria pas√≥ a ser una variable end√≥gena. La tasa de pol√≠tica es la tasa de inter√©s de referencia de la pol√≠tica monetaria.
El BCR, disminuyendo la tasa de inter√©s de referencia, aumenta la cantidad de dinero prestada a los bancos comerciales, que aumenta la pol√≠tica monetaria y con ello la oferta monetaria. Por otra parte, si aumenta la tasa de inter√©s de referencia, disminuye la cantidad de dinero prestada (el cr√©dito) a los bancos comerciales, lo que reduce la base monetaria y con ello la oferta monetaria.
