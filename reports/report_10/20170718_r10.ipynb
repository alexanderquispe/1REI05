The importance of exposure to innovation”, los autores analizan cómo es que la falta de exposición a la innovación puede ayudar a explicar por qué los niños talentosos de las familias de bajos ingresos, minorías y mujeres tienen menos probabilidades de convertirse en inventores. De esta manera, el problema principal es que existen brechas en la innovación, las cuales son explicadas por factores como la exposición a la innovación, definida como aquella que observa si estos investigadores han tenido padres que han sido investigadores, si sus colegas son investigadores o si han crecido en áreas con altas tasas de innovación o de invención; en este caso está definida como aquellas personas que tienen registradas las patentes. Por consiguiente, los autores construyen un conjunto de datos de panel que abarcan 1,2 millones de inventores (solicitantes o destinatarios de patentes) vinculando datos del universo de solicitudes y concesiones de patentes en los EE.UU entre 1996 y 2014 con declaraciones de impuestos federales.

En este sentido, la fortaleza del artículo, es que los autores utilizan tres análisis con diferentes metodologías. En el primer caso, respecto a las características del inventor al nacer, se realizan comparaciones de convertirse en inventor según características, como el ingreso, raza y seco; las cuales son condicionadas mediantes un algoritmo de emparejamiento por el ingreso y los resultados en test de matemáticas. En el segundo caso, respecto al entorno infantil y exposición a la innovación, el estudio realiza comparaciones y regresiones condicionadas a algunas características para determinar si estas explican el efecto de la brecha de innovación. En el tercer caso, respecto a los efectos del vecindario en el nivel de innovación, se realiza una regresión con efectos fijos: si el niño se vuelve inventor, la edad en la que se cambió de zona de bajo nivel de innovación a una zona con alto nivel de innovación, diferencia entre el ratio de patente en la zona de origen y destino, y el vector control (efectos fijos por edad de mudanza, año de nacimiento por cohorte, entre otros).

Por otro lado, una de las debilidades del artículo es que  los resultados no brindan evidencia concluyente sobre el papel del entorno porque los puntajes de las pruebas son una medida imperfecta de la capacidad; sino refleja la capacidad de un niño para innovar, especialmente a edades tempranas, la capacidad aún podría representar una parte sustancial de las brechas de innovación. Además, es importante considerar que el análisis de los autores no proporciona una guía sobre qué programas específicos son más efectivos, pero sí ofrece una guía sobre cómo deben enfocarse: los programas de exposición dirigidos a mujeres, minorías y niños de  familias de bajos ingresos que sobresalen en matemáticas y ciencias a edades  tempranas maximicen su impacto en la innovación. Además, se podría buscar otras variables relacionadas a la innovación, no solo las patentes ya que no toda innovación termina en patentes.

La contribución del artículo es que presenta nueva evidencia sobre los factores que determinan quién se convierte en inventor al rastrear las vidas de los inventores estadounidenses recientes desde el nacimiento hasta la edad adulta. Estos resultados presentan otro enfoque, la exposición a la innovación durante la infancia, como un factor determinante crítico. Por consiguiente, las políticas que aumentan la exposición, tienen la capacidad de aumentar en gran medida la innovación agregada ponderada por calidad. Además, los hallazgos de la investigación sugieren que las políticas diseñadas para aumentar la movilidad intergeneracional también pueden ser beneficiosas para aumentar el crecimiento económico. En otras palabras, si se atraen más niños pertenecientes a las minorías de bajos ingresos a la ciencia e innovación; se podría aumentar los ingresos y reducir la persistencia de la desigualdad entre generaciones, mientras que se estimula el crecimiento al aprovechar el talento actualmente subutilizado. 

Asimismo, en los resultados se observa que en las brechas de innovación por características de nacimiento; la relación entre ingresos e inventores crece de una manera exponencial, especialmente en los últimos percentiles de ingreso. Mientras que en el caso de las mujeres se ha ido reduciendo entre hombres y mujeres, pero a tasas muy bajas al 0.27 puntos porcentuales, a este ritmo en 118 años se llegaría a la paridad. En el caso del entorno infantil, se observa que cuando los niños tienen padres o compañeros inventores es más probable que se conviertan ellos mismos en inventores. Esto se debería a la transmisión de un capital humano específico al acceso de redes de contacto, que ayudan a los niños a perseguir un determinado campo o a la adquisición de información sobre determinadas carreras o efectos de modelo a seguir. En cuanto a los efectos de vecindario, se muestra que existen estas zonas donde hay alto índice de innovación, es más probable que el niño  se convierta en inventor en 22%,y si existe un movimiento de una zona de bajo nivel de innovación hacia una zona de alto nivel de innovación, el porcentaje se incrementará en un 37%. También se encontraron individuos que pudieron convertirse en inventores estrellas si estuvieran en igualdad de condiciones al nacer.

Finalmente, a partir de los hallazgos de este artículo, la línea de investigación en futuros trabajos podría replicar la metodología para el caso de países en desarrollo o se podría evaluar el impacto de los programas o políticas destinadas a incentivar la innovación y observar cuales son más efectivas en cerrar las brechas mencionadas. **Regresión Discontinua**

**a) Explique de manera muy breve y resumida en qué consiste una regresión discontinua.**

La regresión discontinua es un diseño cuasi-experimental que permite realizar inferencia causal en ausencia de aleatorización. Se puede aplicar cuando la exposición de interés se asigna, al menos parcialmente, según el valor de una variable aleatoria continua si esta variable cae por encima o por debajo de un determinado valor umbral.

La idea de la discontinuidad de la regresión es comparar el resultado justo por encima y por debajo del umbral para identificar el efecto del tratamiento en el umbral. A esto se le llama un diseño de **sharp RD** o RD nítido, ya que la probabilidad de obtener el tratamiento salta de 0 a 1 en el umbral. Sin embargo, podríamos pensar en un diseño de **fuzzy RD** o RD borroso, donde la probabilidad también salta, pero es una manera menos dramática.

Por ejemplo, imagine que tiene una variable de tratamiento T y resultados Y0 y Y1 potenciales. El tratamiento T es una función discontinua de una variable observada R. El tratamiento es cero cuando esta por debajo de un umbral y uno en caso contrario.



**b) Detalle un ejemplo en la vida real en donde podría emplear una regresión discontinua.**

Una pregunta de política muy relevante es cúal debería ser la edad mínima para beber. Si el consumo del alcohol aumenta mucho la tasa de mortalidad, deberíamos evitar bajar la edad mínima para beber. Para estimar los impactos del alcohol en la muerte, podríamos usar el hecho de que la edad legal para beber impone una discontinuidad en la naturaleza.

En este contexto, podemos pensar en RDD como un ensayo aleatorio local. Para 
aquellos en el umbral, el tratamiento podría haber ido en cualquier dirección y, por casualidad, algunas personas cayeron por debajo del umbral y otras por encima. En nuestro ejemplo, en el mismo momento, algunas personas tienen poco más de 21 años y algunas personas tienen menos de 21. Lo que determina esto es si alguien nació algunos días después o no, lo cual es bastante aleatorio. Por esto el diseño de regresión discontinua(RDD) proporciona una historia causal muy convincente.

Este modelo nos está diciendo que la mortalidad aumenta en 7,6627 puntos con el consumo de alcohol. Puedes ver que es como si tuviéramos 2 modelos de regresión: uno para los que están por encima del umbral y uno por debajo de él. La discontinuidad de la regresión se basa en gran medida en las propiedades de extrapolación de la regresión lineal. Lo que puede suceder es que la regresión se centre demasiado en ajustar los otros puntos de datos a costa de un mal ajuste en el umbral.

Con la excepción del suicidio, parece que agregar el peso del grano hizo que el impacto negativo en el alcohol fuera 
mayor. Una vez más, si queremos minimizar la tasa de mortalidad, NO deberíamos recomendar bajar la edad legal  para beber, ya que hay un claro impacto del alcohol en la muerte. Este caso simple cubre lo que sucede cuando el diseño de discontinuidad de regresión funciona perfectamente.

EJEMPLO 1:


$$
    Y = 15 + 22*D + 2.5 * X + e
$$

```python
## Random data
dat = pd.DataFrame({'x': np.random.normal(50, 25, 1000)})
dat.loc[dat.x<0, 'x'] = 0
dat = dat[dat.x<100]
dat['D'] = 0
# Cutoff
dat.loc[dat.x>50, 'D'] = 1
# Simulated Y
dat['y'] = 15 + 22*dat.D + 2.5 * dat.x + np.random.normal(0, 20, dat.shape[0])
# Not Treatment
dat_0 = dat[dat.D == 0]
# Treatment
dat_1 = dat[dat.D == 1]
```

```python
# Colors 
treat_c = "#568A9B" # Treatment
n_treat = "#A10728" # Not treatment
# Plot
plt.scatter("x", "y", data = dat_0, color = n_treat, alpha= .7, s = 7, label = "Not treatment")
plt.scatter("x", "y", data = dat_1, color = treat_c, alpha= .7, s = 7, label = "Treatment")
plt.axvline(50, color = "black", linestyle = "-.", label = r"$c_0$: Cutoff")
plt.legend();
```

```python
def y_hat_lin(da_ta, x = "x", y = "y"):
    yhat = smf.ols(f"{y} ~ {x}", data = da_ta).fit().predict()
    da_ta["yhat"] = yhat
    da_ta.sort_values('yhat', inplace = True)
    return da_ta

data_y0 = y_hat_lin(dat_0)
data_y1 = y_hat_lin(dat_1)
```

```python
plt.scatter("x", "y", data = dat_0, color = n_treat, alpha= .2, s = 7, label = "Not treatment")
plt.scatter("x", "y", data = dat_1, color = treat_c, alpha= .2, s = 7, label = "Treatment")
plt.axvline(50, color = "black", linestyle = "-.", label = r"$c_0$: Cutoff")

plt.plot("x", "yhat", data = data_y0, color = n_treat, lw = 3)
plt.plot("x", "yhat", data = data_y1, color = treat_c, lw = 3);
```

```python
eff = min(data_y1.yhat) - max(data_y0.yhat)
eff = round(eff,  2)
print("Effect:", f"{eff}")
```

```python
plt.scatter("x", "y", data = dat_0, color = n_treat, alpha= .03, s = 7, label = "Not treatment")
plt.scatter("x", "y", data = dat_1, color = treat_c, alpha= .03, s = 7, label = "Treatment")
plt.axvline(50, color = "black", linestyle = "-.", label = r"$c_0$: Cutoff")

plt.plot("x", "yhat", data = data_y0, color = n_treat, lw = 3)
plt.plot("x", "yhat", data = data_y1, color = treat_c, lw = 3)
plt.axvline(50, color = "black", linestyle = "-.", label = r"$c_0$: Cutoff", lw = 1)
plt.annotate(f"{eff}", xy = (51, eff + 76))
plt.annotate("", xy = (51, max(data_y0.yhat)), xytext = (51, min(data_y1.yhat)), 
            arrowprops=dict(arrowstyle="<->",
                            connectionstyle="arc3"));
```
