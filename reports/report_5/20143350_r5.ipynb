## A partir del siguiente sistema de ecuaciones que representan el modelo DA-OA:

- Encuentre las ecuaciones de Ingreso ($Y^e$) y tasa de interés ($r^e$) de equilibrio (escriba paso a paso la derivación de estas ecuaciones).

En primer lugar, debemos encontrar $DA$. La forma de su función se puede obtener de las ecuaciones de la $IS$ y $LM$ eliminando $r$ y despejando $P$. Para efectuar esa operación, se supondrá que $P$ no está dado.

$IS$:

$$r = \frac{B_0}{h} - \frac{B_1}{h}Y$$ 

$LM$:

$$r = -\frac{1}{j}\frac{M_0^s}{P} + \frac{k}{j}Y$$

Igualamos y despejamos $Y^e$:

$$-\frac{1}{j}\frac{M_0^s}{P} + \frac{k}{j}Y^e = \frac{B_0}{h} - \frac{B_1}{h}Y^e$$

$$\frac{k}{j}Y^e + \frac{B_1}{h}Y^e = \frac{B_0}{h}+\frac{1}{j}\frac{M_0^s}{P}$$

$$Y^e(\frac{k}{j} + \frac{B_1}{h}) = \frac{B_0}{h}+\frac{1}{j}\frac{M_0^s}{P}$$

$$Y^e(Pkh+jB_1P) = B_ojP+M_0^sh$$

$$Y^e = \frac{jB_0P+M_0^sh}{Pkh+B_1jP}$$

$$Y^e = \frac{jB_0P}{Pkh+B_1jP_0}+\frac{M_0^sh}{Pkh+B_1jP}$$

$$Y^e = \frac{P(jB_0)}{P(kh+B_1j)}+\frac{M_0^sh}{P(kh+B_1j)}$$

$$Y^e = \frac{jB_0}{jB_1+kh}+(\frac{hM_0^s}{jB_1+kh})\frac{1}{P}$$

Teniendo en cuenta que:

$$\frac{M_0^s}{P} = M_0^s-P$$

Despejamos $P$:

$$Y - \frac{jB_0}{kh+jB_1} = (\frac{h}{kh+jB_1})\frac{M_0^s}{P}$$

$$\frac{(kh+jB_1)Y-jB_0}{kh+jB_1} = (\frac{h}{kh+jB_1})\frac{M_0^s}{P}$$

$$[\frac{kh+jB_1}{h}][\frac{(kh+jB_1)Y-jB_0}{kh+jB_1}] = \frac{M_0^s}{P}$$

$$\frac{(kh+jB_1)Y-jB_0}{h} = \frac{M_0^s}{P}$$

$$\frac{(kh+jB_1)Y-jB_0}{h} = M_0^s-P$$

$$P = M_0^s - \frac{(kh+jB_1)Y-jB_0}{h} $$

$$P = \frac{hM_0^s}{h} - \frac{(kh+jB_1)Y-jB_0}{h} $$

$$P = \frac{hM_0^s+jB_0}{h} - \frac{jB_1+hk}{h}Y $$

Esta es la nueva curva de $DA$.

Ahora, para encontrar $Y^e_{DA-OA}$ igualamos las ecuaciones de la $DA$ y $OA$ que están en función a $P$:

$$\frac{hM_0^s+jB_0}{h} - \frac{jB_1+hk}{h}Y^e_{DA-OA} = P^e + θ(Y^e_{DA-OA} - \bar{Y})$$

$$\frac{hM_0^s+jB_0}{h} - \frac{jB_1+hk}{h}Y^e_{DA-OA} = P^e + θY^e_{DA-OA} - θ\bar{Y}$$

$$- \frac{jB_1+hk}{h}Y^e_{DA-OA} - θY^e_{DA-OA} = P^e - θ\bar{Y} - \frac{hM_0^s+jB_0}{h}$$

$$-Y^e_{DA-OA} (θ + \frac{jB_1+hk}{h}) = P^e - θ\bar{Y} - \frac{hM_0^s+jB_0}{h} $$

$$Y^e_{DA-OA} (θ + \frac{jB_1+hk}{h}) = - P^e + θ\bar{Y} + \frac{hM_0^s+jB_0}{h} $$

$$Y^e_{DA-OA} = [\frac {1}{(θ + \frac{jB_1+hk}{h})}][\frac{hM_0^s+jB_0}{h}- P^e + θ\bar{Y}]$$

Para encontrar $P^e_{DA-OA}$ reemplazamos $Y^e_{DA-OA}$ en la ecuación de $OA$:

$$P^e_{DA-OA} = P^e + θ(Y^e_{DA-OA} - \bar{Y})$$

$$P^e_{DA-OA} = P^e + θ( [ \frac{1}{(θ + \frac{jB_1 + hk}{h})} ][(\frac{h Mo^s + jB_o}{h} - P^e + θ\bar{Y})] - \bar{Y} )$$

Para encontrar $r^e$ reemplazamos $Y^e$ en la ecuación de $r^e$ del modelo $IS-LM$:

$$ r = \frac{B_0}{h} - \frac{B_1}{h}Y$$

$$ r^e = \frac{B_0}{h} - \frac{B_1}{h}Y^e$$

$$ r^e = \frac{B_0}{h} - \frac{B_1}{h} [\frac{jB_0}{kh+jB_1}+(\frac{h}{kh+jB_1})\frac{M_0^s}{P}]$$

$$ r^e = \frac{B_0}{h} - \frac{jB_0B_1}{h(kh+jB_1)} + \frac{B_1}{h}(\frac{h}{kh+jB_1})\frac{M_0^s}{P}$$

$$ r^e = \frac{B_0P(kh+jB_1)-jB_0B_1P+hB_1M_0^s}{hP(kh+jB_1)}$$

$$ r^e = \frac{khB_0P+jB_0B_1P-jB_0B_1P+hB_1M_0^s}{hP(kh+jB_1)}$$

$$ r^e = \frac{khB_0P+hB_1M_0^s}{hP(kh+jB_1)}$$

$$ r^e = \frac{khB_0P}{hP(kh+jB_1)}+\frac{hB_1M_0^s}{hP(kh+jB_1)}$$

$$r^e = \frac{kB_0}{kh + jB_1} - (\frac{B_1}{kh + jB_1})\frac{M^s_0}{P}$$

Teniendo en cuenta que:

$$\frac{M_0^s}{P} = M_0^s-P$$

Entonces:

$$r^e = \frac{kB_0}{kh + jB_1} - (\frac{B_1}{kh + jB_1})(M^s_0-P)$$

Para encontrar $r^e_{DA-OA}$ reemplazamos $P^e_{DA-OA}$ en la ecuación de $r^e$ del modelo $IS-LM$:

$$r^e_{DA-OA} = \frac{kB_0}{kh + jB_1} - (\frac{B_1}{kh + jB_1})(M^s_0-P^e_{DA-OA})$$

$$r^e_{DA-OA} = \frac{kB_0}{kh + jB_1} - (\frac{B_1}{kh + jB_1})(M^s_0-P^e + θ( [ \frac{1}{(θ + \frac{jB_1 + hk}{h})} ][(\frac{h Mo^s + jB_o}{h} - P^e + θ\bar{Y})] - \bar{Y} ))$$

- Grafique el equilibrio simultáneo en el modelo DA-OA.

```python
#Ecuación de demanda agregada

Y_size = 100 

Co = 35
Io = 40
Go = 50
Xo = 2
h = 0.8
b = 0.4
m = 0.5
t = 0.8

k = 2
j = 1                
Ms = 200             
P  = 8  

Y = np.arange(Y_size)

B0 = Co + Io + Go + Xo
B1 = 1 - (b-m)*(1-t)

def P_AD(h, Ms, j, B0, B1, k, Y):
    P_AD = ((h*Ms + j*B0)/h) - (Y*(j*B1 + h*k)/h)
    return P_AD

P_AD = P_AD(h, Ms, j, B0, B1, k, Y)


#Ecuación de la oferta agregada
    
Y_size = 100

Pe = 100 
θ = 3
_Y = 20   

Y = np.arange(Y_size)

def P_AS(Pe, _Y, Y, θ):
    P_AS = Pe + θ*(Y-_Y)
    return P_AS

P_AS = P_AS(Pe, _Y, Y, θ)
```

```python
#Punto de equilibrio

def line_intersection(line1, line2):
    xdiff = (line1[0][0] - line1[1][0], line2[0][0] - line2[1][0])
    ydiff = (line1[0][1] - line1[1][1], line2[0][1] - line2[1][1])

    def det(a, b):
        return a[0] * b[1] - a[1] * b[0]

    div = det(xdiff, ydiff)
    if div == 0:
       raise Exception('lines do not intersect')

    d = (det(*line1), det(*line2))
    x = det(d, xdiff) / div
    y = det(d, ydiff) / div
    return x, y

A = [P_AD[0], Y[0]] 
B = [P_AD[-1], Y[-1]] 

C = [P_AS[0], Y[0]] 
D = [P_AS[-1], Y[-1]]

intersec = line_intersection((A, B), (C, D))
intersec 
```

```python
#Gráfico del modelo DA-OA


y_max = np.max(P)
fig, ax = plt.subplots(figsize=(10, 8))

ax.plot(Y, P_AD, label = "$DA$", color = "red") 
ax.plot(Y, P_AS, label = "$OA$", color = "blue") 

plt.axhline(y=intersec[0], xmin= 0, xmax= 0.5, linestyle = "--", color = "grey")
plt.axvline(x=intersec[1],  ymin= 0, ymax= 0.49, linestyle = "--", color = "grey")

plt.text(0, 200, '$P^e$', fontsize = 12, color = 'black')
plt.text(53, 25, '$Y^e$', fontsize = 12, color = 'black')
plt.text(50, 202, '$E$', fontsize = 12, color = 'black')


ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

ax.set(title="Equilibrio simultáneo en el modelo $DA-OA$", xlabel= r'Y', ylabel= r'P')
ax.legend()
ax.xaxis.set_label_coords(1, -0.05)
ax.yaxis.set_label_coords(-0.05, 1)

plt.show()
```

**1. Análisis intuitivo**

Modelo IS-LM:

$$t↑ → DA↓ → DA < Y → Y↓$$

$$Y↓ → M^d↓ → M^d < M^s → r↓$$

Modelo DA-OA:

$$Y↓ → θ(Y-\bar{Y})↓ → P↓$$

**2. Análisis matemático**

```python
#Variables IS
Co, Io, Go, Xo, h, r, b, m, t, beta_0, beta_1  = symbols('Co Io Go Xo h r b m t beta_0, beta_1')

#Variables LM
k, j, Ms, P, Y = symbols('k j Ms P Y')

#Variables OA
Pe, _Y, Y, θ = symbols('Pe, _Y, Y, θ')

#Betas
beta_0 = (Co + Io + Go + Xo)
beta_1 = ( 1-(b-m)*(1-t) )

#Y de equilibrio DA-OA
Y_eq = ( (1)/(θ + ( (j*beta_1+h*k)/h) ) )*( ( (h*Ms+j*beta_0)/h ) - Pe + θ*_Y )

#P de equilibrio DA-OA 
P_eq = Pe + θ*(Y_eq - _Y)

#r de equilibrio DA-OA
r_eq = (j*beta_0)/(k*h + j*beta_1) + ( h / (k*h + j*beta_1) )*(Ms - P_eq)
```

```python
# Efecto del cambio de t sobre r en DA-OA
df_r_eq_t = diff(r_eq, t)
print("El diferencial de r con respecto al diferencial de t es:", df_r_eq_t)
print("\n")

# Efecto del cambio de t sobre P en DA-OA
df_P_eq_t = diff(P_eq, t)
print("El diferencial de P con respecto al diferencial de t es:", df_P_eq_t)
```
