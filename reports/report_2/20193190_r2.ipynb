* **¬øC√≥mo se deriva la curva IS a partir del equilibrio Y=DA?**

Dado que el equilibrio entre el Ingreso (Y) y la Demanda Agregada (DA) representa el equilibrio en el mercado de bienes, para derivar la curva IS a partir dicho equilibrio se deben hacer remplazos y algunas operaciones algebr√°icas, y as√≠ expresar la ecuaci√≥n en funci√≥n de la tasa de inter√©s (r). 
La ecuaci√≥n resultante indica que existen pares de valores ordenados de ingreso (ùëå) y de tasa de inter√©s (ùëü) que equilibran el mercado de bienes. Esta relaci√≥n inversa se puede **ilustrar gr√°ficamente** a partir del equilibro ùê∑ùê¥ = ùëå en la recta de 45¬∞ de la siguiente manera:

```python
#Demanda Agregada

# Par√°metros

Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3
r = 0.9

Y = np.arange(Y_size)

# Ecuaci√≥n de la curva del ingreso de equilibrio

def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)

#--------------------------------------------------
# Recta de 45¬∞

a = 2.5 

def L_45(a, Y):
    L_45 = a*Y
    return L_45

L_45 = L_45(a, Y)

#--------------------------------------------------
# Segunda curva de ingreso de equilibrio

    # Definir cualquier par√°metro aut√≥nomo
Go = 35

# Generar la ecuaci√≥n con el nuevo par√°metro
def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_G = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)
```

```python
#Curva IS

# Par√°metros

Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3

Y = np.arange(Y_size)


# Ecuaci√≥n 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
```

### b. La Curva IS o el equilibrio Ahorro- Inversi√≥n

* **Ecuaci√≥n de IS a partir de la igualdad Ahorro-Inversi√≥n:**


Para encontrar la ecuaci√≥n de IS partiendo de la igualdad Ahorro-Inversi√≥n, primero se debe considerar que la curva IS se deriva desde la igualdad entre el ingreso (Y) y la demanda agregada (DA): 

$$Y= C + I + G + X - M$$


Considerando que: 

$$C= Co + bY·µà$$

$$I= Io - hr$$

$$G= Go$$

$$T= tY$$

$$X= Xo$$

$$M= mY·µà$$


Para llegar al equilibrio Ahorro-Inversi√≥n, debemos restar la tributaci√≥n (T) de ambos miembros de la igualdad.

$$Y - T = C + I - T + G + X - M$$

$$Y·µà = C + I - T + G + X - M$$


Esta igualdad se puede reescribir de la siguiente forma:

$$(Y·µà - C) + (T - G) + (M - X) = I$$


Las tres partes de la derecha constituyen los tres componentes del ahorro total (S): ahorro privado (S‚Çö), ahorro del gobierno (Sùëî) y ahorro externo (S‚Çë):

$$S = S‚Çö + Sùëî + S‚Çë$$


Entonces, el ahorro total es igual a la inversi√≥n: 

$$S‚Çö + Sùëî + S‚Çë = I$$

$$ùëÜ(ùëå) = ùêº(ùëü)$$


Partiendo de esa igualdad Ahorro-Inversi√≥n, se debe remplazar para obtener que:

$$ùëÜùëù + ùëÜùëî + ùëÜùëí = ùêºo ‚àí ‚Ñéùëü$$

$$(ùëå·µà ‚àí ùê∂o ‚àí ùëèùëå·µà ) + (ùëá ‚àí ùê∫o) + (ùëöùëå·µà‚àí ùëão) = ùêºo ‚àí ‚Ñéùëü$$


Considerando las observaciones anteriores sobre los componentes de la condici√≥n de equilibrio (Y):

$$[1 ‚àí (ùëè ‚àí ùëö)(1 ‚àí ùë°)]ùëå ‚àí (ùê∂o + ùê∫o + ùëão) = ùêºo ‚àí ‚Ñéùëü$$


La curva IS se puede expresar con una ecuaci√≥n donde la tasa de inter√©s es una funci√≥n del ingreso:

$$‚Ñéùëü = (ùê∂o + ùêºo + ùê∫o + ùëão) ‚àí (1 ‚àí (ùëè ‚àí ùëö)(1 ‚àí ùë°))Y$$

                                                r =  1                            1 ‚àí (ùëè ‚àí ùëö)(1 ‚àí ùë°)
                                                   ‚Äî‚Äî‚Äî‚Äî‚Äî (ùê∂o + ùêºo + ùê∫o + ùëão) ‚àí  ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî Y
                                                     h                                  h 
                         
                         
Y puede simplificarse en:

                                                ùëü = ùõΩo     ùõΩ1
                                                   ‚Äî‚Äî‚Äî‚Äî - ‚Äî‚Äî‚Äî‚Äî Y
                                                    ‚Ñé      ‚Ñé
    

Donde Œ≤o = Co + Io + Go + Xo es el intercepto y Œ≤1 = 1 ‚àí (b ‚àí m)(1 ‚àí t) es la pendiente. 

*  **Gr√°fico de la curva IS de equilibrio en el Mercado de Bienes:** 

```python
# Par√°metros

Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3

Y = np.arange(Y_size)


# Ecuaci√≥n 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
```

```python
# Gr√°fico de la curva IS

# Dimensiones del gr√°fico
y_max = np.max(r)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, r, label = "IS", color = "#701b87") #Demanda agregada

# Eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

# T√≠tulo, ejes y leyenda
ax.set(title = "La Curva IS de Equilibrio en el Mercado de Bienes", xlabel= 'Y', ylabel= 'r')
ax.legend()

plt.show()
```

Una situaci√≥n similar ocurre con el punto "C", pero hacia el lado contrario, as√≠ que se puede identificar que a la izquierda de la IS, el desequilibrio del mercado de bienes es de **exceso de demanda**.
