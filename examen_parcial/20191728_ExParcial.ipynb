<h1>1).Defina el multiplicador keynesiano y dé un ejemplo de cómo influye este en el modelo Ingreso-Gasto Keynesiano<h1>

* El multiplicador keynesiano representa en cuánto se va incrementar el producto ante un incremento de cualquier componente autónomo de la demanda agregada (DA); es decir la parte del intercepto de la ecuación. 


* La influencia de este multiplicador se cuando un componente autónomo de la DA aumenta o disminuye; entonces esta variación se multiplica por el multiplicador keynesiano (el cual es positivo) para hacer un mayor efecto en la producción. 

<h1>2).	Grafique y explique cuándo sucede un exceso y déficit de demanda en el modelo Ingreso-Gasto. Señale las áreas donde ocurre cada caso. Explique cómo se converge al equilibrio a partir de estos dos escenarios. <h1>

* Los excesos de demanda son generados por un aumento en la DA, por lo que el nivel de producción es insuficiente; en este caso, para que se converja al equilibrio, los productores deciden aumentar la producción. Un déficit de demanda se da cuando la DA disminuye, por lo que hay un exceso de producción; en este caso, para que se converja al equilibrio, los productores deciden disminuir la producción.

```python
# Valores de Parametros
Y_size = 100 

Co = 5
Io = 18
Go = 28
Xo = 5
h = 0.8
b = 0.7
m = 0.3
t = 0.3
r = 8.2

Y = np.arange(Y_size)

# Ecuación de la curva del ingreso de equilibrio

def DA(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA= (Co+b*(1-t)*Y+Io-h*r+Go+Xo-m*(1-t)*Y)
    return DA

DA_0= DA(Co, Io, Go, Xo, h, r, b, m, t, Y)
```

```python
#--------------------------------------------------
# Recta de 45°

def L_45(Y):
    L_45 = Y
    return L_45

L_45 = L_45(Y)
```

```python
#--------------------------------------------------

# Parámetros

Y_size = 100 

Co = 15
Io = 18
Go = 28
Xo = 5
h = 0.8
b = 0.5
m = 0.3
t = 0.3

Y = np.arange(Y_size)


# Ecuación 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = ((Co+Io+Go+Xo)/h-Y*((1-(1-t)*(b-m))/h))
    return r_IS

IS_0 = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
```

$$DA= C_0+b(1-t)Y +I_0-hr+G_0+X_1Y^*+X_2*ε-m_1(1-t)Y+m_2*ε$$

$$W=C_0+I_0+G_0$$

$$DA=W+[b(1-t)-m_1(1-t)]Y-hr+X_1Y^*+(X_2+m_2)ε$$

$$DA=W-hr+X_1Y^*+(X_2+m_2)ε_0-(X_2+m_2)pr+(X_2+m_2)pr^*+[(b-m_1)(1-t)]Y$$

$$DA=[W+X_1Y^*+(X_2+m_2)ε_0+(X_2+m_2)pr^*-(h+(X_2+m_2)p)r]+[(b-m_1)(1-t)]Y$$

$$DA=α_0+α_1Y$$

$$DA=Y=α_0+α_1Y$$

$$Y(1-α_1)=α_0$$

$$Y=\dfrac{1}{1-α_1}*α_0$$

* Multiplicador Keynesiano

$$\dfrac{1}{1-(b-m_1)(1-t)}$$
