**1. Explique cuáles son los intrumentos de política monetaria que puede utilizar el Banco Central.**

El Banco Central es la autoridad monetaria y la cabeza del sistema bancario. En ese sentido, tiene la capacidad de controlar o gestionar la oferta monetaria. Para ello, puede hacer política monetaria expansiva (para, mediante el aumento de la oferta de dinero, reducir la tasa de interés y aumentar la demanda agregada) o contractiva (para, mediante la disminución de la oferta de dinero, aumentar la tasa de interés y disminuir la demanda agregada) haciendo uso de tres instrumentos: la *Operación de Mercado Abierto*, el *coeficiente legal de encaje* y la *tasa de interés*.

El primer instrumento, la *Operación de Mercado Abierto*, se refiere a la compra o venta activos financieros o bonos por parte del Banco Central a los bancos comerciales. 

La compra de bonos del mercado corresponde a una política monetaria expansiva, pues de esa forma inyecta dinero a la economía. El Banco Central registra un aumento de sus activos, lo que genera un incremento en el depósito de los bancos comerciales en el Banco Central. De esa forma, aumenta la base monetaria y, por ende, la oferta monetaria.

Por otro lado, la política monetaria contractiva se realiza mediante la venta de bonos al mercado (bancos comerciales). De esa forma, el Banco Central retira dinero de la economía. 

El segundo instrumento con el que cuenta el Banco Central es el *coeficiente legal de encaje*. El Banco central puede aumentar o disminuir la cantidad de dinero que tienen los bancos para realizar préstamos mediante el aumento o disminución de la tasa de encaje, pues ello implica que los bancos comerciales deben tener menos o más dinero en forma de reservas en el balance del Banco Central. La disminución de la tasa de encaje aumenta la posibilidad de creación de dinero bancario porque aumenta el multiplicador de dinero y el aumento de dinero bancario implica un aumento de la oferta monetaria. Por su parte, el aumento de la tasa de encaje reduce el multiplicador bancario y genera una disminución de la oferta monetaria.

Finalmente, la *tasa de interés* (de referencia de la política monetaria) es un instrumento usado por los Bancos Centrales desde 1990 gracias al cual la oferta monetaria pasó a ser una variable endógena. 

El Banco Central hace política monetaria expansiva reduciendo su tasa de interés de referencia para aumentar la cantidad de dinero prestada a los bancos comerciales, lo que aumenta la base monetaria y por tanto la oferta monetaria. Por otro lado, el Banco Central hace política monetaria aumentando su tasa de interés de referencia para disminuir el crédito a los bancos comerciales, lo que reduce la base monetaria y por tanto la oferta monetaria.

**2. Derive la oferta real de dinero y explique cada uno de sus componentes.**

La oferta nominal de dinero ($M^s$), que representa la cantidad de dinero que circula en la economía, parte del supuesto de que esta es una variable exógena e instrumento de política monetaria. 

Para expresar esta oferta en términos reales o a precios constantes, debe dividirse entre el nivel general de precios de la economía ($P$), pues este varía por la inflación:

$$\frac{M^s}{P}=\frac{M^s_0}{P}$$

**3. Derive la demanda real de dinero. Explique qué papel cumplen los parametros "$k$" y "$j$".**

Se tiene que:

- Demanda por motivos de transacción y precaución (depende del ingreso): $L_1=kY$ 

La magnitud de las transacciones de bienes y servicios tiene relación directa con $Y$. Asimismo, la precaución se refiere a la capacidad de pago de deudas, que depende también de $Y$. 

$k$ representa la sensibilidad o elasticidad de la demanda de dinero frente a las variaciones de $Y$. Ambos parámetros tienen una relación positiva y directa. En consecuencia, mientras mayor sea $k$, mayor será $Y$ y mayor será la demanda de dinero por motivos de transacción y precaución.

- Demanda por motivo de especulación (depende de la tasa de interés de los bonos): $L_2=-ji$

Se prefiere mantener liquidez en forma de dinero y no en forma de bonos cuando $i$ o el rendimiento de mercado de bonos se reduce. Por otro lado, pasa lo contrario si $i$ aumenta. Este tipo de demanda tiene relación inversa con la tasa de interés de los bonos.$j$ indica qué tan sensible es la demanda del dinero ante variaciones de la tasa de interés nominal de los bonos $i$.

- Función de demanda real de dinero: $L=L_1+L_2$

Si asumimos que ambos tipos de demanda están en términos reales, la función demanda real de dinero será $L=L_1+L_2$. Reemplazando:

$$L=kY-ji$$



**4. Asumiendo que no hay inflación podemos asumir que $i = r$. Escriba en terminos reales la eucacion de equilibrio en el mercado de dinero.**

En el corto plazo, se asume que el nivel de precios es fijo y exógeno (la inflación esperada es igual a 0), por lo que no hay diferencia entre la tasa de interés nominal y real. Cuando aumenta o disminuye el Ingreso o la oferta de dinero y genera un exceso de demanda u oferta monetaria, la tasa de interés es lo que equilibra el mercado monetario. Su ecuación es:

$$r=-\frac{1}{j}\frac{M^s_0}{P_o}+\frac{k}{j}Y$$

Es una función con pendiente positiva y puede graficarse en el plano ($Y$, $r$). La curva LM representa un *locus* de puntos ($Y$, $r$) que equilibran el mercado monetario.

**5. Grafique el equilibrio en el mercado de dinero.**

```python
r_size = 100

k = 0.5
j = 0.2                
P  = 10 
Y = 35
MS_0 = 500

r = np.arange(r_size)
```

```python
def MD(k, j, P, r, Y):
    MD_eq = (k*Y - j*r)
    return MD_eq
MD_0 = MD(k, j, P, r, Y)
```

**2. Explique y grafique qué sucede en el mercado de dinero si $\Delta  k < 0.$**

$k$ se encuentra en la función de demanda por motivos de transacción y precaución:

$$L_1 = kY$$

Ahora bien, la demanda de dinero por motivo de especulación es:

## Curva LM

**1. Derive paso a paso la curva $LM$ matemáticamente (a partir del equilibrio en el Mercado Monetario) y grafique.**

Sabemos que:

$$M^s = kY - jr$$

Y que:

$$M^s = \frac{M_o^s}{p}$$

El equilibrio se obtiene al igualar ambas ecuaciones:

$$kY - jr = \frac{M_o^s}{p}$$

$$ky - \frac{M_o^s}{p} = jr$$

$$\frac{kY}{j} - \frac{M_o^s}{pj} = r$$

$$r = - \frac{M_o^s}{pj} + \frac{kY}{j}$$

La ecuación que representa a la curva LM en función de la tasa de interés que equilibra el mercado monetario es:

$$r = - \frac{1}{j} \frac{M^o_s}{P_o} + \frac{k}{j}$$


```python
r_size = 100

k = 0.5
j = 0.2                
P  = 10 
Y = 35

r = np.arange(r_size)
```

```python
def Ms_MD(k, j, P, r, Y):
    Ms_MD = P*(k*Y - j*r)
    return Ms_MD

Ms_MD = Ms_MD(k, j, P, r, Y)
```

```python
Y1 = 45

def Ms_MD_Y1(k, j, P, r, Y1):
    Ms_MD = P*(k*Y1 - j*r)
    return Ms_MD

Ms_Y1 = Ms_MD_Y1(k, j, P, r, Y1)


Y2 = 25

def Ms_MD_Y2(k, j, P, r, Y2):
    Ms_MD = P*(k*Y2 - j*r)
    return Ms_MD

Ms_Y2 = Ms_MD_Y2(k, j, P, r, Y2)
```

```python
Y_size = 100

k = 0.5
j = 0.2                
P  = 10               
Ms = 30            

Y = np.arange(Y_size)
```

```python
def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)
```

**2. ¿Cuál es el efecto de una disminución en la Masa Monetaria $\Delta  Ms < 0.$? Explica usando la intuición y gráficos.**

Para que exista una disminución en la oferta real del dinero, el Banco Central debe haber aplicado una política monetaria contractiva (retirando el dinero en circulación mediante la vena de bonos o a través del aumento de la tasa de interés).

Además, se tiene que la tasa de interés aumenta porque la oferta de dinero ($M^s$) se mueve para la izquierda y con respecto a $M^d$. En otras palabras, el punto en el que demanda y oferta se encuentren en r, se dará más arriba (esto no implica que $M^d$ cambie.

```python
Y_size = 100

k = 2
j = 1                
Ms = 700            
P  = 20               

Y = np.arange(Y_size)
```

```python
def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)
```
