### 4.1 Equilbrio de Y^e


Considerando que $DA=Y$ 

En la que $DA = C_o+I_o+G_o+X_o-hr+Y(b-m)(1-t)$ ‚Üí $DA=ùõº^0+ùõº^1Y$

Entonces, tomando Y, la ecuasi√≥n de Ingreso de equilibrio $Y^e$:

$Y=C_o+I_o-hr+G_o+X_o+bY^d-mY^d$

$Y=\frac{1}{1-(b-m)(1-t)} (C_o+I_o+G_o+X_o-hr)$


### 4.2 Gr√°fico del Ingreso de Equilibrio a Corto Plazo

```python
Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8 # b > m
m = 0.2
t = 0.3
r = 0.9

Y = np.arange(Y_size)


def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)

a = 2.5 

def L_45(a, Y):
    L_45 = a*Y
    return L_45

L_45 = L_45(a, Y)
```

```python
y_max = np.max(DA_IS_K)
fig, ax = plt.subplots(figsize=(10, 8))


ax.plot(Y, DA_IS_K, label = "DA", color = "purple") #Demanda agregada
ax.plot(Y, L_45, color = "#404040") #L√≠nea de 45¬∫


# L√≠neas punteadas punto de equilibrio
plt.axvline(x=70.5,  ymin= 0, ymax= 0.69, linestyle = ":", color = "grey")
plt.axhline(y=176, xmin= 0, xmax= 0.7, linestyle = ":", color = "grey")
plt.axhline(y=146, xmin= 0, xmax= 0.8, linestyle = ":", color = "grey")

# Texto agregado
    # punto de equilibrio
plt.text(0, 180, '$DA^e$', fontsize = 11.5, color = 'black')
plt.text(72, 0, '$Y^e$', fontsize = 12, color = 'black')
plt.text(0, 152, '$Œ±_o$', fontsize = 15, color = 'black')
    # l√≠nea 45¬∫
plt.text(6, 4, '$45¬∞$', fontsize = 11.5, color = 'black')
plt.text(2.5, -3, '$‚óù$', fontsize = 30, color = '#404040')
    #Formula ùõº^0
plt.text(72, 137, '$Œ±_o=(C_o+I_o+G_o+X_o-hr)$', fontsize = 12, color = "grey")
    #Formula ùõº^1
plt.text(74, 167, '$Œ±_1=(b-m)(1-t)$', fontsize = 12, color = "grey")
    #Formula DA
plt.text(82, 193, '$DA=Œ±_o+Œ±_1Y$', fontsize = 11, color = "purple")
plt.text(82, 187, '$‚Üô$', fontsize = 11, color = "purple")

# T√≠tulo y leyenda
ax.set(title="El ingreso de Equilibrio a Corto Plazo", xlabel= r'Y', ylabel= r'DA')
ax.legend() #mostrar leyenda

plt.show()
```

### 5.1 Aumento del Gasto del Gobierno en una Pol√≠tica Fiscal Expansiva  

#### Manera instuitiva 
 
 Sabiendo que $ŒîG>0$
                 
$G‚Üë ‚Üí DA ‚Üë ‚Üí DA > Y ‚Üí Y‚Üë$
                     
Una pol√≠tica fiscal expansiva implica un incremento del ($G$). Ante esto, la demanda agregada ($DA$) a su vez sufrir√° un aumento, lo que implicar√° un desequilibrio con el nivel de producci√≥n ($Y$). Por lo que, para voler al equilibrio se debe aumentar el nivel de producci√≥n ($Y$).   
                            

#### Manera Gr√°fica

```python
# Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3
r = 0.9

Y = np.arange(Y_size)

    # Ecuaci√≥n 
def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)

# Definici√≥n del aumento del parametro del Gasto de Gobierno
Go = 90

def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_G = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)
```

```python
# Gr√°fico
y_max = np.max(DA_IS_K)
fig, ax = plt.subplots(figsize=(10, 8))

ax.set(title = "Aumento del Gasto del Gobierno $(G_0)$", xlabel = r'Y', ylabel = r'DA')

ax.plot(Y, DA_IS_K, label = "DA", color = "#3D59AB") #Curva sin el aumento del Gasto del Gobierno
ax.plot(Y, DA_G, label = "DA_G", color = "Purple", linestyle = 'dashed') #Curva ante el aumento del G
ax.plot(Y, L_45, color = "#404040") #l√≠nea de 45¬∫

plt.text(0, 180, '$DA_1$', fontsize = 11.5, color = 'purple')
plt.text(0, 135, '$DA^0$', fontsize = 11.5, color = 'black') 

plt.axvline(x = 70.5, ymin= 0, ymax = 0.69, linestyle = ":", color = "grey")
plt.axhline(y = 199, xmin= 0, xmax = 0.77, linestyle = ":", color = "grey")
plt.axvline(x = 80,  ymin= 0, ymax = 0.78, linestyle = ":", color = "grey")
plt.axhline(y = 175, xmin= 0, xmax = 0.69, linestyle = ":", color = "grey")

plt.text(76, 140, '$‚Üí$', fontsize = 18, color = 'grey')
plt.text(20, 160, '$‚Üë$', fontsize = 18, color = 'grey')
plt.text(6, 4, '$45¬∞$', fontsize = 11.5, color = 'black')
plt.text(2.5, -3, '$‚óù$', fontsize = 30, color = '#404040')

plt.show()
```

#### Manera Matem√°tica  

$ŒîG>0$ ‚Üí $ŒîY$

$Y=\frac{1}{1-(b-m)(1-t)} (C_o+ŒîI_o+ŒîG_o+ŒîX_o-hr)$

Puesto que $ŒîC_o=ŒîI_o=ŒîX_0=Œîr=0$ ‚Üí $ŒîG>0$

$ŒîY=\frac{1}{1-(b-m)(1-t)}(ŒîG)$

$ŒîY=k(ŒîG)$

Sabiendo que $ŒîG>0$ y que $k>0$, la multiplicaci√≥n de dos n√∫meros positivos: 

$ŒîY=(+)(+)$

$ŒîY>0$

#### Manera intuitiva 

Se ve un ingreso disponible ante la reducci√≥n de la tasa de impuestos, y por eso  aumenta la demanda de consumo y la demanda agregada, por lo que se aumenta el nivel de producci√≥n: 

$t‚Üì ‚Üí Co‚Üë ‚Üí DA ‚Üë ‚Üí DA > Y ‚Üí Y‚Üë$



#### Manera Gr√°fica

```python
Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3 #tasa de tributaci√≥n
r = 0.9

Y = np.arange(Y_size)

     
def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)

    # Definici√≥n de la reducci√≥n de la tasa de tributaci√≥n
t = 0.01

def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_t = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)
```
