# 2 - Regression Discontinuity Design (RDD)
Supongamos que estamos estudiando la eficacia de un programa de becas creado por el gobierno Peruano sobre las notas de los alumnos al final del primer semestre en la universidad. PRONABEC ha decidido que entregará becas a aquellos postulantes que tengan un promedio superior durante la secundaria igual a 16.

# 2.1 
Los elementos necesarios son $RV$ (running variable), la cual se encarga de determinar si el grupo de estudiantes que tiene un promedio inferior es igual a 16; es decir, pertenece al grupo de control, o, si el grupo de estudiantes que tiene un promedio superior durante la secundaria es igual a 16, de ser asi, pertenecería al grupo de tratamiento. Otro elemento imporante sería el $BondWidth$, el cual se encarga de determinar si la discontinuidad está lejos o cerca. También se debe considerar el elemento $CutOff$ (variable dentro del $RV$), la cual se encarga de determinar si pertece al grupo de control o al grupo de tratamiento, dependiendo de las obversaciones o el análisis que se encuentre. 

# 2.2
La causalidad en este escenario se puede determinar realizando la comparación del promedio de notas del grupo de control con el del grupo de tratamiento. Puesto que el promedio del grupo de tratamiento durante el primer semestre, se puede asegurar el efecto de causalidad es positivo.

# 2.3
El resultado de la política aplicada en cusnto al rendimiento de los estudiantes, se espera que sea positivo, puesto que la muestra analizada que se recopiló son de estudiantes con un alto promedio en la secundaria; por ende los estudiantes tienen altas probabilidades de obtener buenas notas en el primer semestre universitario, sin tomar en cuenta la beca. Se puede concluir que presenta una fuerte correlación mas no un efecto de causalidad.

# 3 - Difference in Difference (DiD)
Supongamos que estamos estudiando la eficacia de un programa de becas creado por el gobierno Peruano sobre las notas de los alumnos al final del primer semestre en la universidad. PRONABEC ha decidido que entregará becas a aquellos postulantes que tengan un promedio superior durante la secundaria igual a 16 .

# 3.1 
Para implementar un DiD, sería necesario, tener el grupo de tratamiento y el grupo de control. A su vez, el momento donde se existe un cambio en la política, los efectos que se tendría en el grupo de tratamiento. Se debe considerar que antes de que haya un cambio en la política, la recta del grupo de control debe estar con la misma dirección en la pendiente. Y, luego, del cambio en la política, se debe verificar que el grupo de tratamiento tenga un cambio en la dirección de la pendiente. 
# 3.2
Para encontrar el efecto causal en este escenario, se debe realizar una doble resta para determinar la causalidad. En otras palabras, se deber restar el grupo de tratamiento luego del cambio en la política  y el grupo de tratamiento antes del cambio en la política. El resultado obtenido restarlo con la resta de el grupo de control luego del cambio en la política y el grupo de control antes del cambio en la política.
# 3.3
Es predecible asumir que un incremento en la tasa de migración ocasione un declive en la tasa de empleos ya que se tendría un aumento en la demanda. Por lo tanto habría causalidad y correlación debido a que la situación es al azar, ya que tiene cualidades de un experimento natural.
