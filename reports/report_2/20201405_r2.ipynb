### a. El Modelo Ingreso-Gasto: la Curva IS

#####  Y = DA

#####  DA = C + I + G + X- M
#####  DA = Co + b * Y^d + Io - h*r + Go + Xo - m * Y^d
#####  DA = Co + b (1-t) + Io - h*r + Go + Xo - m (1-t) * Y 
#####  DA = Co + Io - h*r  + Go + Xo + (b-m) (1-t) Y
##### Y = Co + Io - h*r  + Go + Xo + (b-m) (1-t) Y
##### Y - (b-m) (1-t) Y = Co + Io - h*r  + Go + Xo
##### Y [1- (b-m) (1-t)] = Co + Io - h*r  + Go + Xo
##### Y = 1/[1- (b-m) (1-t)] (Co + Io + Go + Xo) - hr / [1- (b-m) (1-t)]

Entonces: 
##### r= 1/h (Co + Io + Go + Xo) - [1- (b-m) (1-t)]/h * Y

#### △r/△Y desde la función de equilibrio IS

```python
Co, Io, Go, Xo, h, r, b, m, t = symbols('Co Io Go Xo h r b m t')

f = ((Co + Io + Go + Xo)/(1-(b-m)*(1-t)))- ((h*r)/(1-(b-m)*(1-t)))


df_Y = diff(f, r)
df_Y #∆r/∆Y
```

### b. La Curva IS o el equilibrio Ahorro- Inversión

#####                                            Y = C + I + G + X- M
 
 Le restamos la cantidad de tributación que pagamos 

   #####   Y - T = C + I + G + X - M - T
   
Donde: 
   #####   C = Co + b* Y^d 
   #####   I = Io - h*r
   #####  G = Go
   #####  X = Xo
   #####  M = m * Y^d
   #####  Y^d = (1 - t) * Y
   #####  T = t * y
   
Entonces : 

#####   Y^d = C + I + G + X - M - T

#####   S = (Y^d - C) - (G - X) + (M + T) 

#####   (Y^d - C) - (G - X) + (M + T) = I

#####   S = I

#####   Y^d - Co - b * Y^d - Go - Xo + m * Y^d + t * Y = I 

#####   (1-t) Y - Co - b (1-t)*Y - Go - Xo + m (1 - t) + t * Y = Io - h*r

#####   Y [ (1 - t) - b (1 - t) + m(1- t ) + t ] - Co - Go - Xo = Io - h*r

#####   Y [ 1 - (1 - t) * (b - m) ] - Co - Go - Xo = Io - h*r 

#####   h*r = Io - Y [ 1 - (1 - t) * (b - m) ] + ( Co + Go + Xo )

#####  h*r = Io + Co + Go + Xo - [1 - (1 - t) * (b - m) ] * Y

#####  r = 1/h * (Io + Co + Go + Xo) - [1 - (1 - t) * (b - m)] * Y / h





### Gráfico
 

```python
# Parámetros

Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3

Y = np.arange(Y_size)


# Ecuación 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (1/h)*(Co + Io + Go + Xo) - (1/h)*(1-(b-m)*(1-t))*Y
    return r_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
```

```python
# Gráfico de la curva IS

# Dimensiones del gráfico
y_max = np.max(r)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, r, label = "IS", color = "#ff0080") #Demanda agregada

# Eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

# Título, ejes y leyenda
ax.set(title = "La Curva IS de Equilibrio en el Mercado de Bienes", xlabel= 'Y', ylabel= 'r')
ax.legend()

plt.show()
```
