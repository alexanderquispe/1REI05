1. Defina el Multiplicador Keynesiano y dé un ejemplo de cómo influye este en el modelo Ingreso-Gasto Keynesiano

2. Grafique y explique cúando sucede un exceso y déficit de demanda en el modelo Ingreso-Gasto. Señale las áreas donde ocurre cada caso. Explique cómo se converge al equilibrio a partir de estos dos escenarios.

- Cuando hay un excesos de demanda, las empresas deciden aumentar su producción. Además, como dato extra, en este modelo Cada vez que aumenta la producción, aumenta el empleo, disminuye la tasa de desempleo y aumenta el ingreso disponible de las familias.
- Cuando se da un deficit en la demanda aumentan los inventarios y las empresas deciden disminuir su producción y el empleo de trabajadores. 
- En ambos casos, la oferta de producción se ajusta a los cambios de la demanda.
- En área superior derecha se concetra el exceso de demanda y en la inferior izquierda el deficit. 

```python
Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.6
b = 0.7 # b > m
m = 0.3
t = 0.4
r = 0.8

Y = np.arange(Y_size)

# Ecuación de la curva del ingreso de equilibrio

def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)

#recta 45 grados

a = 2.5 

def L_45(a, Y):
    L_45 = a*Y
    return L_45

L_45 = L_45(a, Y)

# Dimensiones del gráfico
y_max = np.max(DA_IS_K)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(DA_IS_K, label = "DA", color = "#3D59AB") #Demanda agregada
ax.plot(L_45, color = "#404040") #Línea de 45º

# Texto agregado
    # punto de equilibrio
plt.text(0, 180, '$DA^e$', fontsize = 11.5, color = 'black')
plt.text(72, 0, '$Y^e$', fontsize = 12, color = 'black')
plt.text(0, 152, '$α_o$', fontsize = 15, color = 'black')

plt.text(70, 240, '$exceso_ demanda$', fontsize = 15, color = 'black')
plt.text(20, 22, '$deficit_ demanda$', fontsize = 15, color = 'black')
 
    # línea 45º
plt.text(6, 4, '$45°$', fontsize = 11.5, color = 'black')
plt.text(2.5, -3, '$◝$', fontsize = 30, color = '#404040')

# Título y leyenda
ax.set(title="El ingreso de Equilibrio a Corto Plazo", xlabel= r'Y', ylabel= r'DA')
ax.legend() #mostrar leyenda

plt.show()
```
