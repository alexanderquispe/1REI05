### A partir del siguiente sistema de ecuaciones que representan el modelo IS-LM

#### Encuentre las ecuaciones de Ingreso ($Y^e$) y tasa de interés ($r^e$) de equilibrio (Escriba paso a paso la derivación de estas ecuaciones)

Como el equilibrio es el punto en que IS y LM coinciden, se deve igualar ambas ecuaciones para hallar $Y^e$ y $r^e$...

$$ -\frac{1}{j}\frac{Mo^s}{P_o} + \frac{k}{j}Y = \frac{B_o}{h} - \frac{B_1}{h}Y $$

Para hallar $Y^e$...

$$\frac{k}{j}Y + \frac{B_1}{h}Y = \frac{B_o}{h} + \frac{1}{j}\frac{Mo^s}{P_o}$$

$$Y(\frac{k}{j} + \frac{B_1}{h}) = \frac{B_o}{h} + \frac{1}{j}\frac{Mo^s}{P_o}$$

$$Y = \frac{\frac{B_o}{h}}{\frac{k}{j} + \frac{B_1}{h}} + \frac{\frac{1}{j}}{\frac{k}{j} + \frac{B_1}{h}}\frac{Mo^s}{P_o}$$

$$ Y^e = \frac{j B_o}{k h + j B_1} + (\frac{h}{k h + j B_1})\frac{Ms_o}{P_o} $$

Para hallar $r^e$... (se nos explicó que se aplicaban derivadas...)

$$ r^e = \frac{kB_o}{kh + jB_1} - (\frac{B_1}{kh + jB_1})\frac{Ms_o}{P_o} $$

#### Grafique el equilibrio simultáneo en los mercados de bienes y de dinero

```python
#IS

TamañoY = 100 

Co = 30
Io = 50
Go = 40
Xo = 1
h = 0.7
b = 0.3
m = 0.6
t = 0.7

Y = np.arange(TamañoY)

def IS(b, m, t, Co, Io, Go, Xo, h, Y):
    IS = (Co + Io + Go + Xo)/h - ( ( 1-(b-m)*(1-t) ) / h)*Y  
    return IS

IS = IS(b, m, t, Co, Io, Go, Xo, h, Y)


# LM 

TamañoY = 100

k = 3
j = 2                
Ms = 250             
P  = 35               

Y = np.arange(TamañoY)

def LM(k, j, Ms, P, Y):
    LM = - (1/j)*(Ms/P) + (k/j)*Y
    return LM

LM = LM( k, j, Ms, P, Y)
```

```python
# Gráfico

y_max = np.max(LM)
fig, ax = plt.subplots(figsize=(12, 9))

ax.plot(Y, IS, label = "IS", color = "Black")
ax.plot(Y, LM, label="LM", color = "Black")

ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

plt.axvline(x=57.7,  ymin= 0, ymax= 0.49, linestyle = ":", color = "Grey")
plt.axhline(y=83.3, xmin= 0, xmax= 0.57, linestyle = ":", color = "Grey")

plt.text(57,90, '$E$', fontsize = 16, color = 'Grey')
plt.text(-10,83.3, '$r_0$', fontsize = 15, color = 'Grey')
plt.text(56.5,-20, '$Y_0$', fontsize = 15, color = 'Grey')

plt.text(95,150, '$LM$', fontsize = 17, color = 'Black')
plt.text(95,10, '$IS$', fontsize = 17, color = 'Black')

plt.text(90,81.5, '$II$', fontsize = 16, color = 'Black')
plt.text(22,81.5, '$IV$', fontsize = 16, color = 'Black')
plt.text(56.5,130, '$I$', fontsize = 16, color = 'Black')
plt.text(56.5,30, '$III$', fontsize = 16, color = 'Black')

ax.set(title="Equilibrio simultáneo en los mercados de bienes y de dinero", xlabel= r'Y', ylabel= r'r')

plt.show()
```

```python
#IS

TamañoY = 100 

Co = 30
Io = 50
Go = 40
Xo = 1
h = 0.7
b = 0.9
m = 0.4
t = 0.7

Y = np.arange(TamañoY)

def IS(b, m, t, Co, Io, Go, Xo, h, Y):
    IS = (Co + Io + Go + Xo)/h - ( ( 1-(b-m)*(1-t) ) / h)*Y  
    return IS

IS = IS(b, m, t, Co, Io, Go, Xo, h, Y)


# LM 

TamañoY = 100

k = 3
j = 2                
Ms = 250             
P  = 35               

Y = np.arange(TamañoY)

def LM(k, j, Ms, P, Y):
    LM = - (1/j)*(Ms/P) + (k/j)*Y
    return LM

LM = LM( k, j, Ms, P, Y)
```
