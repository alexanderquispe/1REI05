1. Explique cuáles son los instrumentos de política monetaria que puede utilizar el Banco Central.

De acuerdo con Jiménez (2020), hay 3 posibles instrumentos de política monetaria. El primero es la Oferta Monetaria como instrumento de política (Operación de Mercado Abierto), la cual consiste en la la compra o venta de activos financieros o bonos a los bancos comerciales. Durante una política monetaria expansiva, el Banco Central inyecta dinero en soles a la economía. Lo contrario sucedería en caso de una política monetaria contractiva, en la cual el Banco Central retira dinero en soles de su economía.

El segundo es el coeficiente legal de encaje como instrumento de política. Durante una política monetaria expansiva, el Banco central disminuye la tasa de encaje para aumentar la posibilidad de creación de dinero bancario, lo que imploca un aumento de la oferta monetaria. Por otra parte, durante una política monetaria contractiva, el Banco Central aumenta el coeficiente de encaje puesto que  los bancos comerciales deben tener mayores reservas, lo que reduce el multiplicador bancario y, como consecuencia, existe una disminución de la oferta monetaria.

El tercero y el último es la tasa de interés como instrumento de política (tasa de interés de referencia de la política monetaria), como consecuencia del nuevo esquema de política monetaria que prioriza la tasa de interés antes que la oferta monetaria. Durante una periodo de política monetaria expansiva, el Banco Central reduce su tasa de interés de referencia, lo que aumenta la base y oferta monetaria. Por el contrario, durante una política monetaria contractiva, se aumenta la tasa de interés, lo que disminuye la base y oferta monetaria.

4. Asumiendo que no hay inflación podemos asumir que i=r. Escriba en terminos reales la ecuación de equilibrio en el mercado de dinero.
El equilibrio en el Mercado de Dinero se deriva del equilibrio entre la Oferta de Dinero (Ms) y Demanda de Dinero (Md):

Ms=Md
Ms/P=kY-ji

Como a corto plazo el nivel de precios es fijo y exógeno, Sse asume que la inflación esperada es cero. Por lo tanto, no habría una gran diferencia entre la tasa de interés nominal (i) y la real (r).

Entonces: la ecuación que representa el equilibrio en el mercado de dinero es: 

Ms/P=kY-ji

5. Grafique el equilibrio en el mercado de dinero.

```python
#Parameters
r_size = 100

k = 0.5
j = 0.2                
P  = 10 
Y = 35
MS_0 = 500

r = np.arange(r_size)

#Creamos función de la demanda
def MD(k, j, P, r, Y):
    MD_eq = (k*Y - j*r)
    return MD_eq
MD_0 = MD(k, j, P, r, Y)

# Creamos la oferta de dinero.
MS = MS_0 / P
MS
```

```python
#Equilibrio en el mercado de dinero

# Creamos el seteo para la figura 
fig, ax1 = plt.subplots(figsize=(10, 8))

# Agregamos titulo t el nombre de las coordenadas
ax1.set(title="Money Market Equilibrium", xlabel=r'M^s / P', ylabel=r'r')

# Ploteamos la demanda de dinero
ax1.plot(MD_0, label= '$L_0$', color = '#008000')

# Para plotear la oferta de dinero solo necesitamos crear una linea vertical
ax1.axvline(x = MS,  ymin= 0, ymax= 1, color = "grey")

# Creamos las lineas puntadas para el equilibrio
ax1.axhline(y=7.5, xmin= 0, xmax= 0.5, linestyle = ":", color = "black")

# Agregamos texto
ax1.text(0, 7.5, "$r_0$", fontsize = 12, color = 'black')

ax1.yaxis.set_major_locator(plt.NullLocator())   
ax1.xaxis.set_major_locator(plt.NullLocator())

ax1.legend()

plt.show()
```

# Curva LM

1. Derive paso a paso la curva LM matemáticamente (a partir del equilibrio en el Mercado Monetario) y grafique.

```python
#CURVA LM
Ecuación: r= -[(1/j)(Mso/Po)+(k/j)Y]
```

```python
Y_size = 100

k = 10
j = 10                
Ms = 200             
P  = 20               

Y = np.arange(Y_size)


# Ecuación

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)
```

```python
#GRAFICO DE LA CRUVA LM
y_max = np.max(i)
v = [0, Y_size, 0, y_max]   
fig, ax = plt.subplots(figsize=(10, 8))

# Eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

# Curvas a graficar
ax.plot(Y, i, label="LM", color = '#3D59AB')

# Título y leyenda
ax.set(title="LM SCHEDULE", xlabel=r'Y', ylabel=r'r')
ax.legend()


plt.show()
```

2. ¿Cuál es el efecto de un disminución en k ? Explica usando intuición y gráficos.

```python
#--------------------------------------------------
    # Curva LM ORIGINAL

# Parámetros

Y_size = 100

k = 2
j = 1                
Ms = 200             
P  = 20               

Y = np.arange(Y_size)

# Ecuación

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)

#--------------------------------------------------
    # NUEVA curva LM

# Definir SOLO el parámetro cambiado
Ms = 50

# Generar la ecuación con el nuevo parámetro
def i_LM_Ms( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM
i_Ms = i_LM_Ms( k, j, Ms, P, Y)
```

```python
# Dimensiones del gráfico
y_max = np.max(i)
v = [0, Y_size, 0, y_max]   
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, i, label="LM", color = 'black')
ax.plot(Y, i_Ms, label="LM_Ms", color = '#3D59AB', linestyle = 'dashed')

# Texto agregado
plt.text(47, 76, '∆$M^s$', fontsize=12, color='black')
plt.text(49, 70, '→', fontsize=15, color='grey')

# Título y leyenda
ax.set(title = "Reducción en la Masa Monetaria $(M^s)$", xlabel=r'Y', ylabel=r'r')
ax.legend()


plt.show()
```

La disminución de la cantidad de dinero la curva LM se desplaza hacia la izquierda, pues hace subir la tasa de interés para que disminuya la demanda de dinero y exista mayor oferta en el mercado. En una política monetaria restrictiva, se reduce la inversión, la demanda agregada y el producto.
