## PART 1: Report

As indicated in the title, the article looks over the fiscal and monetary history of Peru from 1960 to 2017. It should be noted that the research focuses on two periods marked by mistrust in market allocations (the authoritarian regime of Velasco Alvarado, 1968 -1975, and the government of Alan García, 1985-1990) and in a period of stabilization (since Fujimori came to power in 1990).

To achieve its objective, to understand the Peruvian monetary policy and analyze the decisions adopted, a monetarist approach is used: the article not only explains the emergence of inflation and mistaken policies or "policy follies". But also highlights the importance that the BCRP autonomy and the absence of state interventionism, together with other variables (such as fiscal prudence and the presence of qualified officials) had and has for Peruvian macroeconomic stability.

On the one hand, the publication in Becker Friedman Institute and the University of Chicago is due to the evident monetarist tinge. The aforementioned importance of monetary policy and market opening are two components of the Fujimori government praised by the authors. Hence, they not only consider it a key economic stabilization period for the history of Peru, but also present the measures adopted and the political reforms as a kind of "recipe" in the face of an increase in inflation and a drop in productivity.

On the other hand, an important contribution is that it provides an analysis of monetary and fiscal policies in their specific contexts (national and international) and given the decisions of previous governments. Contrary to popular understanding, Martinelli and Vega go back to the first Belaunde government (and the widespread reformism of the time) to trace the origins of inflation. Although the excessive state intervention of the Velasquista government is identified as a cause of the fall in economic productivity and consequent increase in inflation, the foundations and the "good intentions" behind them are also recognized.

As noted with the latter, the document makes use of a historical methodology. Although quantitative tools and two accounting exercises are used to test the hypotheses, the historical review since 1960 allows to understand what was happening at the time and why it was happening. And although the accounting exercises use a slightly complex terminology, the sections dedicated to the successive governments since the middle of the 20th century not only allow them to be understood, but, despite the economic specificity of the article, they make it understandable before a public without too much economics knowledge.

Continuing with this more social than purely economic approach, Martinelli and Vega propose the idea of a "social learning" that was decisive in achieving economic stability in Peru. Given the chronic inflation and the disaster of the heterodox politics of Alan García, both the policymakers and, especially, the citizenry experienced a process of social learning: “The credibility of the policy regime change in the 1990s may be ultimately linked to the change in public opinion giving proper incentives to politicians, after the traumatic consequences of the hyperstagflation of 1987–1990".

Although this idea recalls the social science nature of economics and emphasizes the importance of individuals as economic agents, two weaknesses in it are identified:

First, in order to decipher the impact that social learning had on the change of direction of the Peruvian economy, the consequences of the two economic experiments of 1968 and 1985 had to be presented. Even if the history of the crisis is well known, an approximation to the social reality of the time, and not only to the macro indicators, would allow to understand the reason for the rejection of state intervention. In addition to this, the period of the Fujimori government should have been presented not only in terms of stabilization and growth, but also emphasizing the results (not so good or stable) in the short term.

Second, when speaking of social learning, it is implied that the citizenry, thanks to the traumas of the statist past and filled with inflation and crisis, ended up being the architect of the measures adopted during this last regime. However, history shows the opposite: Fujimori's election was due above all to popular discontent with the traditional political class. Though economic concerns and the mismanagement of previous governments played an important role, citizens did not vote for this outsider because of his economic plan. Moreover, in the first round Fujimori did not present clear ideas in this field. And in the second round a determining factor in his victory was precisely the defense of much less extreme measures than an economic shock or the complete liberalization of the economy, such as the ideas defended by Mario Vargas Llosa.

In this way, even if people do look to the past to make economic decisions (or policies with economic implications), the monetarist approach of the paper does not consider that individuals base their decisions, above all, on their current situation and the consequences in short term. Hence, a large part of the citizenry prioritized their concerns and fears for an even greater inflation that would come with liberalization, over the concern for the autonomy of the BCRP or certain state economic control.

Therefore, future research could focus on experiences where the adoption of new monetary and fiscal policies was in greater agreement with the social learning of citizens. Perhaps less extreme economic shocks in the region could initially have had popular support and not have too serious consequences in the short term. Likewise, although Martinelli and Vega only give some insights on the link between macroeconomic stability and the political difficulties that democracy entails, an interesting line of research could review the impact of internal political disputes (especially between the powers of the State) in the Peruvian economy and current macroeconomic policies.

## PART 2: IS Curve

```python
import matplotlib.pyplot as plt
import numpy as np               
import matplotlib.pyplot as plt
```

```python
#Parámetros
Y_size = 100 
b = 0.4                 #propensión marginal a consumir
m = 0.2                 #propensión a importar
t = 0.3                   #tasa impositiva
Co = 20                 #consumo autónomo
Io = 10                  #inversión autónoma
Go = 8                  #gasto gobierno 
Xo = 2                  #exportación 
h =  5                  #constante de decisión a invertir

Y = np.arange(Y_size)
```

```python
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

def Y_IS(b, m, t, Co, Io, Go, Xo, h, r):
    Y_IS = (Co + Io + Go + Xo - h*r)/(1-(b-m)*(1-t))
    return Y_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
r
```

### 2.1 Increase in Government Spending $(Go)$
Given an increase in Government Spending $ (Go) $, as this is an autonomous component and part of the intercept, aggregate demand $ (Y) $ will increase. Consequently, the IS curve will shift parallel to the right.

```python
Go = 15
def r_IS_G(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

def Y_IS_G(b, m, t, Co, Io, Go, Xo, h, r):
    Y_IS = (Co + Io + Go + Xo - h*r)/(1-(b-m)*(1-t))
    return Y_IS

r_G = r_IS_G(b, m, t, Co, Io, Go, Xo, h, Y)
r_G
```

```python
y_max = np.max(r)
x_max = Y_IS(b, m, t, Co, Io, Go, Xo, h, 0)

v = [0, x_max, 0, y_max]                        
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="IS SCHEDULE", xlabel=r'Y', ylabel=r'r')
ax.plot(Y, r, "k-",  scalex = t, scaley = t, label ='IS')
ax.plot(Y, r_G, "b-",  scalex = t, scaley = t, linestyle='dotted', color = "red", label = 'IS_G')
  
ax.legend()

plt.text(47, 0.07, '∆Go', fontsize=12, color='black')
plt.text(49, -0.4, '→', fontsize=15, color='grey')

plt.show()
```

### 2.2 Increase in Taxe Rate $(t)$:

A change in any of the slope components will cause a change in the angle of the curve. For example, if the tax rate $(t)$ is increased, the slope will move to the left.

```python
t = 0.9
def r_IS_t(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

def Y_IS_t(b, m, t, Co, Io, Go, Xo, h, r):
    Y_IS = (Co + Io + Go + Xo - h*r)/(1-(b-m)*(1-t))
    return Y_IS

r_t = r_IS_t(b, m, t, Co, Io, Go, Xo, h, Y)
r_t
```

```python
y_max = np.max(r)
x_max = Y_IS(b, m, t, Co, Io, Go, Xo, h, 0)

v = [0, x_max, 0, y_max]                        
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="IS SCHEDULE", xlabel=r'Y', ylabel=r'r')
ax.plot(Y, r, "k-", scalex = t, scaley = t,label ='IS')
ax.plot(Y, r_t, "b-",scalex = t, scaley = t, linestyle='dotted', color = "blue", label = 'IS_t')
   
ax.legend()

plt.text(51, -0.6, '∆t', fontsize=12, color='black')
plt.text(48, -1.2, '←', fontsize=15, color='grey')

plt.show()
```

### 2.3 Increase in Marginal Propensity to Consume $(b)$

Changes in the marginal propensity to consume $(b)$ have effects on the slope of the IS curve similar to those caused by the tax rate $(t)$. In this case, increasing $(b)$ increases the size of the multiplier, increases the slope, and reduces the slope of the IS curve.

```python
b = 0.6

def r_IS_b(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

def Y_IS_b(b, m, t, Co, Io, Go, Xo, h, r):
    Y_IS = (Co + Io + Go + Xo - h*r)/(1-(b-m)*(1-t))
    return Y_IS

r_b = r_IS_b(b, m, t, Co, Io, Go, Xo, h, Y)
r_b
```

```python
y_max = np.max(r)
x_max = Y_IS(b, m, t, Co, Io, Go, Xo, h, 0)

v = [0, x_max, 0, y_max]                        
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="IS SCHEDULE", xlabel=r'Y', ylabel=r'r')
ax.plot(Y, r, "k-", scalex = t, scaley = t, label = 'IS')
ax.plot(Y, r_b, "b-", scalex = t, scaley = t, linestyle='dotted', label="IS_b", color = "green")
  
ax.legend()

plt.text(43, 0.8, '∆b', fontsize=12, color='black')
plt.text(44.5, 0.3, '→', fontsize=15, color='grey')

plt.show()
```

### 3.1 Equilibrium in the Monetary Market

The equilibrium in the Money Market is derived from the equilibrium between Money Supply $ (M^o) $ and Money Demand $ (M^d) $: 

$$ M^o = M^d $$

Then the equilibrium equation in the money market, expressed in real terms, is:

$$ \frac{M^s}{P} = kY - ji $$

Where $ M^s $ is the nominal money supply, $ P $ the price level, $ k $ the transaction money demand, $ j $ the speculation money demand and $ i $ is the nominal interest rate. It must be noted that $ Y $ or the income is given.

Expected inflation can be assumed to be zero, since in the short run the price level is fixed and exogenous. Therefore, there would not be a big difference between the nominal $ (i) $ and the real interest rate $ (r) $.

The equation and the graph of equilibrium in the money market are:

$$ \frac{M^s}{P} = kY - jr $$

### 3.2 LM Curve

Following the last equilibrium equation, the money market interest rate $ (r) $ is equal to:

$$ jr = \frac{M^s_o}{P_o} - kY $$

$$ r = -\frac{1}{j} \frac{M^s_o}{P_o} + \frac{k}{j}Y $$


Where the slope, $ \frac{k}{j} $ , is positive and the intercept, $ \frac{M^s_o}{j P_o} $ , is negative.

The equilibrium LM curve in the Money Market is graphed as follows:

```python
# Parameters
Y_size = 100

k = 10                # Transaction money demand
j = 10                # Speculation money demand
Ms = 20000             # Nominal money supply
P  = 20                # Price level

Y = np.arange(Y_size)
```

```python
def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)
i
```

```python
y_max = np.max(i)

v = [0, Y_size, 0, y_max]                       
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="LM SCHEDULE", xlabel=r'Y', ylabel=r'r')
ax.plot(Y, i, "b-", label="LM", color = 'black')
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

ax.legend()
plt.show()
```
