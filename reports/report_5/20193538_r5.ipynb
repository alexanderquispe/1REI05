## A partir del siguiente sistema de ecuaciones que representan el modelo DA-OA
### - Encuentre las ecuaciones de Ingreso $(Y^e)$, tasa de interés $(r^e)$ , $(P^e)$ de equilibrio (Escriba paso a paso la derivación de estas ecuaciones).

Tomando en cuenta que la curva IS se representa mediante la ecuacion $r=\frac{B_o}{h}-\frac{B_1}{h}Y$, en donde $B_o=C_o+G_o+X_o+I_o$ y $B_1=1-(b-m)(1-t)$.

Ademas que la ecuacion LM es $r=-\frac{1}{j}{M_o^s}{P}+\frac{k}{j}Y$

**Entonces la ecuacion de DA es:**

$$P=\frac{hM_0^s+jB_o}{h}-\frac{jB_1+hk}{h}Y$$


**Y laa ecuacion de OA es:**

$$P=P^e+θ(Y-Ȳ)$$


Igualamos las ecuaciones y derivamos

*- Para hallar el $Y^E$*

$$\frac{hM_o^s+jB_o}{h}-\frac{jB_1+hk}{h}Y = P^e+θ(Y-Ȳ)$$

$$Y^E=[\frac{1}{(θ+\frac{jB_1+hk}{h})}]*[\frac{hM^S_o+jB_o}{h}-P^e+θȲ]$$



*- Para hallar el $P^E$*

$$P^E=P^e+θ(Y^e-Ȳ)$$

$$P^E=P^e+θ[\frac{1}{(θ+\frac{jB_1+hk}{h})}]*[\frac{hM^S_0+jB_0}{h}-P^e+θȲ]-Ȳ)$$



*- Para hallar $r^E$*

$$r^E=\frac{kB_o}{kh+jB_1}-(\frac{B_1}{kh+jB_1})*(M^s_o-P)$$

$$r^E=\frac{kB_o}{kh+jB_1}-(\frac{B_1}{kh+jB_1})*(M^s_o-P^E)$$

$$r^E=\frac{kB_o}{kh+jB_1}-(\frac{B_1}{kh+jB_1})*(M_s^o-P^e+θ(P^e+θ[\frac{1}{(θ+\frac{jB_1+hk}{h})}]*[\frac{hM^S_0+jB_0}{h}-P^e+θȲ]-Ȳ))$$


**Entonces los valores de equilibrio de las 3 variables serian**

- $Y^E=[\frac{1}{(θ+\frac{jB_1+hk}{h})}]*[\frac{hM^S_o+jB_o}{h}-P^e+θȲ]$


- $r^E=\frac{kB_o}{kh+jB_1}-(\frac{B_1}{kh+jB_1})*(M_s^o-P^e+θ(P^e+θ[\frac{1}{(θ+\frac{jB_1+hk}{h})}]*[\frac{hM^S_0+jB_0}{h}-P^e+θȲ]-Ȳ))$


- $P^E=P^e+θ[\frac{1}{(θ+\frac{jB_1+hk}{h})}]*[\frac{hM^S_0+jB_0}{h}-P^e+θȲ]-Ȳ)$

```python
# líneas punteadas autómaticas

    # definir la función line_intersection
def line_intersection(line1, line2):
    xdiff = (line1[0][0] - line1[1][0], line2[0][0] - line2[1][0])
    ydiff = (line1[0][1] - line1[1][1], line2[0][1] - line2[1][1])

    def det(a, b):
        return a[0] * b[1] - a[1] * b[0]

    div = det(xdiff, ydiff)
    if div == 0:
       raise Exception('lines do not intersect')

    d = (det(*line1), det(*line2))
    x = det(d, xdiff) / div
    y = det(d, ydiff) / div
    return x, y

    # coordenadas de las curvas (x,y)
A = [P_AD[0], Y[0]] # DA, coordenada inicio
B = [P_AD[-1], Y[-1]] # DA, coordenada fin

C = [P_AS[0], Y[0]] # L_45, coordenada inicio
D = [P_AS[-1], Y[-1]] # L_45, coordenada fin

    # creación de intersección

intersec = line_intersection((A, B), (C, D))
intersec # (y,x)
```

### Analice los efectos sobre las variables endógenas Y, P y r de un incremento de la tasa de impuestos $(▵t<0)$ . El análisis debe ser intuitivo, matemático y gráfico. En una figura, se debe que usar los ejes r e Y (modelo IS-LM), y en la otra, los ejes P, Y (modelo DA-OA).

**ANALISIS INTUITIVO**

**- Modelo IS-LM**

$$t↑→C↓→DA↓→DA<Y→↓$$
osea: $$t↑→DA↓→Y↓$$

.

$$Y↓→M_d↓→M_d<M^s→r↓$$
osea: $$Y↓→M_d↓→r↓$$


**- Modelo OA-DA**

$$Y↓→[P^e+θ(Y↓-Ȳ)]→[P^e+θ(Y-Ȳ)]↓→[P^e+θ(Y-Ȳ)]<P→P↓$$
osea: $$Y↓→Pe↓→P↓$$

.


**ANALISIS MATEMATICO**

**- Modelo IS-LM**
$$ΔY_e=\frac{ΔY_e}{Δt}>0 .... (-)+(-)=(-)<0$$

$$Δr=\frac{Δr}{Δt}>0 .... (-)+(-)=(-)<0$$

**- Modelo OA-DA**
$$ΔP=\frac{ΔP}{Δt}>0 .... (+)+(+)*(-)=(-)<0$$

**ANALISIS GRAFICO**
