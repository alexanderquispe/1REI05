### Instrumental Variables (IV)

Suponga que estamos interesados en estudiar la relación entre educación(X) y el nivel de ingresos(Y).

* Explique el procedimiento para encontrar el efecto causal en este escenario. Puede utilizar ecuaciones

U no es observado por el econometrista, lo que hace que el camino de la puerta trasera permanezca abierto. Se entiende que cuando cambie el nivel de educación X, también cambiará el nivel de ingresos Y por el efecto causal. Toda la correlación entre esas variables, y en esa situación, reflejará el efecto causal ya que X es un colisionador a lo largo del camino de la puerta trasera entre Z y Y.

### Regression Discontinuity Design (RDD)

Supongamos que estamos estudiando la eficacia de un programa de becas creado por el gobierno Peruano sobre las notas de los alumnos al final del primer semestre en la universidad. PRONABEC ha decidido que entregará becas solo a aquellos postulantes que tengan un promedio superior a 16 durante la secundaria.

* Indetifique todos los elementos necesarios para implementar un RDD en este ejemplo.

Los elementos de la RDD en este ejemplo son el puntaje mínimo del promedio en secundaria necesario para obtener la beca de PRONABEC (16 puntos); la regla de asignación objetiva, la cual no puede ser manipulada ni predicha por los postulantes, como lo es la que estipula que solo recibirán la beca aquellos cuyo promedio en secundaria supere los 16 puntos; el punto de corte o umbral, que es un promedio de 16 durante la secundaria; y la intervención, que se refiere al otorgamiento de la beca de PRONABEC para los seleccionados de entre los postulantes con el promedio mínimo. 


* Explique el procedimiento para encontrar el efecto causal en este escenario. Puede utilizar ecuaciones.

Para implementar un RDD en este ejemplo, se necesita X, una RunningVariable, que es una variable continua que asigna unidades al tratamiento. Esta asignación de unidades al tratamiento se basa en un puntaje de "límite", de modo que cualquier unidad con un puntaje superior al límite se coloca en el grupo de tratamiento y las unidades que están debajo no. En este caso, esa variable X sería el programa de becas de PRONABEC. 
También se necesita una variable de asignación puede afectar de forma independiente el resultado a través de la ruta de X a Y, e incluso puede estar relacionada con un conjunto de variables Z que determinan de forma independiente. Nótese por el momento que el estado de tratamiento de una unidad está determinado exclusivamente por la regla de asignación. El tratamiento no está determinado por Z. 
Además, el hecho de que no haya manipulación de la asignación al tratamiento alrededor del punto de corte permite que se le pueda atribuir causalidad a las becas: lo que determina quién recibe la beca y quién no es únicamente la regla de obtener un promedio mayor a 16 puntos, y no el esfuerzo de los estudiantes de manipular la regla de asignación y de mover selectivamente su puntaje por encima o por debajo del punto de corte.

* ¿El efecto de la politica sobre el rendimiento de los estudiantes sería positivo o negativo (solo necesitan usar su intuición)? Argumente si los resultados serian causales o solamente de correlación.

Al otorgar una beca a estudiantes que han demostrado tener un desempeño académico destacado o superior al de los demás (por encima de 16 puntos), se les está incentivando a que continuen con sus estudios y con tal nivel de desempeño académico, por lo que probablemente el efecto de la politica sobre el rendimiento de los estudiantes sería positivo. De esta manera, y por los requitos que cumple el método de asignación de las becas, los resultados de la ejecución de dicha política serían causales. 


### Difference in Difference (DiD)

Supongamos que estamos interesados en estudiar el impacto de la migración en la tasa de empleo en dos regiones del Perú en 2017-2018. Tenemos datos anuales sobre la tasa de empleo en Lima, que experimentó una afluencia significativa de inmigrantes en 2018, y Puno que no experimentó el ingreso de migrantes (2017 ninguna de las regiones experimentó afluencia de migrantes). Podemos usar un análisis DID para comparar el cambio en las tasas de empleo entre las dos regiones a lo largo del tiempo para ver si la migración tuvo algún efecto.

* Indetifique todos los elementos necesarios para implementar un DiD en este ejemplo.

Se necesita un grupo de tratamiento sobre el cual se tendrá una intervención (Lima, que experimentó una afluencia significativa de inmigrantes en 2018), por lo cual también se necesita un grupo de comparación que no tiene la intervención a través del tiempo (Puno, que no experimentó una afluencia de inmigrantes en el 2017 ni en el 2018).

* Explique el procedimiento para encontrar el efecto causal en este escenario. Puede utilizar ecuaciones.

Se debe aplicar una doble diferencia, es decir, comparar los cambios en el tiempo en la tasa de empleo por el impacto de la migración en Lima y en Puno. La condición necesaria para encontrar el efecto causal es contar con el supuesto de tendencias paralelas, con el que se asume que, en ausencia de la migración, los dos grupos hubiesen seguido la misma trayectoria en los resultados de interés (las tasas de empleo de Lima y Puno seguirían sin variar a lo largo del tiempo). 



* ¿El efecto de la migración sobre la tasa de empleo sería positivo o negativo (solo necesitan usar su intuición)? Argumente si los resultados serian causales o solamente de correlación. 

El efecto de la migración sobre la tasa de empleo sería negativo, ya que el incremento en la demanda laboral por el aumento de la cantidad de trabajadores migrantes no significa que habrán más empleos, sino que estos se ocuparán y la oferta laboral será menor. Así, se nota una relación causal entre el incremento de la migración y el impacto que tuvo en la tasa de empleo en Lima (donde sí llegaron más migrantes con el paso del tiempo) y en Puno (donde no incrementó la cantidad de migrantes con el tiempo).
