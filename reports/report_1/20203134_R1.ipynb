# 1.

La función de consumo, suma el consumo autónomo $Co$, es decir, el gasto mínimo y relativamente constante para la subsistencia de cada familia, que puede variar por el ingreso disponible de las familias. Esta variación es representada por la propensión marginal a consumir $b$, cuyo valor oscila entre el 0 al 1, multiplicado por el ingreso disponible $Yd$.

$C = C_0 + bY^d$

```python
import ipywidgets as widgets
import matplotlib.pyplot as plt
import numpy as np
import sympy as sy
from sympy import *
import pandas as pd
```

```python
# Parámetros

Yd_size = 70 

Co = 10
b = 0.5 # b > m

Yd = np.arange(Yd_size)

def C_K(Co, b, Yd):
    C_K = Co + (b*Yd)
    return C_K

C_IS_K = C_K(Co, b, Yd)
```

```python
# Gráfico

y_max = np.max(C_IS_K)
fig, ax = plt.subplots(figsize=(12, 10))

ax.plot(Yd, C_IS_K, label = "C=Co+bYd", color = "red") 

plt.text(57, 50, '$C$', fontsize = 12, color = 'black')

ax.set(title="Función de la demanda de Consumo", xlabel= r'Yd', ylabel= r'C')
ax.legend() 

custom_xlim = (0, 80)
custom_ylim = (0, 80)

plt.setp(ax, xlim=custom_xlim, ylim=custom_ylim)

plt.show()
```

# 2.

La fórmula de la Inversión es la inversión autónoma menos la multiplicación de la sensibilidad para invertir con la tasa de interés real:

$ I = I_0 - hr$



```python
# Parámetros

r_size = 100 


Io = 80
h = 0.7
r = 0.9

r = np.arange(r_size)

def I_K(Io, h, r):
    I_K = Io - (h)*(r)
    return I_K

I_IS_K = I_K(Io, h, r)
```

```python
# Gráfico

y_max = np.max(I_IS_K)
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(r, I_IS_K, label = "I=Io-hr", color = "red")

ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

plt.text(97, 20, '$I$', fontsize = 12, color = 'black')

ax.set(title="Función de demanda de Inversión", xlabel= r'r', ylabel= r'I')
ax.legend() 

custom_xlim = (0, 120)
custom_ylim = (-10, 120)

plt.setp(ax, xlim=custom_xlim, ylim=custom_ylim)

plt.show()
```

# 3.

Para el modelo Ingreso-Gasto Keynesiano, se consideran cuatro supuestos:

1. La cantidad o nivel de precios está fijo.
2. El nivel del producto se adapta a los cambios de la demanda agregada.
3. La tasa de interés siempre está determinada fuera del modelo (por el mercado monetario)
4. Es un modelo de corto plazo.

# 4. 

Para llegar al equilibrio, debemos igualar DA = Y, es decir, la demanda agregada debe ser igual el ingreso agregado

$ DA = C + I + G + X - M $

$ DA = Y $

$ Y = C + I + G + X - M $

Reemplazando con los elementos de la Demanga Agregada

$Y = C_0 + bY^d + I_0 -hr + G_0 + X_0 - mY^d$

Al derivar en función de $Y$, se obtiene, la $Y^e$, es decir el ingreso de equilibrio

$Y^e = \frac{1}{1 - (b - m)(1 - t)} (C_0 + I_0 + G_0 + X_0 - hr)$

```python
# Parámetros

Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8 # b > m
m = 0.2
t = 0.3
r = 0.9

Y = np.arange(Y_size)

# Ecuación de la curva del ingreso de equilibrio

def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)
```

```python
a = 2.5 

def L_45(a, Y):
    L_45 = a*Y
    return L_45

L_45 = L_45(a, Y)

y_max = np.max(DA_IS_K)
fig, ax = plt.subplots(figsize=(10, 10))

ax.plot(Y, DA_IS_K, label = "DA", color = "black") #Demanda agregada
ax.plot(Y, L_45, color = "red") #Línea de 45º

plt.axvline(x=70.5,  ymin= 0, ymax= 0.69, linestyle = ":", color = "grey")
plt.axhline(y=176, xmin= 0, xmax= 0.7, linestyle = ":", color = "grey")
plt.axhline(y=146, xmin= 0, xmax= 0.9, linestyle = ":", color = "grey")

plt.text(0, 180, '$DA^e$', fontsize = 11.5, color = 'black')
plt.text(72, 0, '$Y^e$', fontsize = 12, color = 'black')
plt.text(0, 152, '$α_o$', fontsize = 15, color = 'black')

plt.text(6, 4, '$45°$', fontsize = 11.5, color = 'black')
plt.text(2.5, -3, '$◝$', fontsize = 30, color = 'black')

plt.text(86, 200, '$DA = α_0 + α_1Y$', fontsize = 12, color = 'black')

ax.set(title="El ingreso de Equilibrio $Y$", xlabel= r'Y', ylabel= r'DA')
ax.legend()

plt.show()
```

# Política Fiscal Expansiva con aumento del Gasto de Gobierno:
    
- Intuitivo:
    
    ↑Go → ↑DA → DA > Y → ↑Y
    
    Si se aumenta el gasto del gobierno, este influirá en el aumento de la demanda agregada. Entonces, si la demanda agregada aumenta, va a ser mayor que el Ingreso Agregado. Por consecuencia, el Ingreso debe de aumentar para regresar al equilibrio en el mercado.


- Matemático:
    ∆G = ∆Y 

    ∆G/∆Y = 1

    ∆G ↑ = ∆Y  ↑

```python
#Gráfico: vamos a graficar la curva de equilibrio original, y luego 
#vamos a graficar la nueva curva, con un aumento del Gasto de gobierno.

#--------------------------------------------------
# Curva de ingreso de equilibrio ORIGINAL

    # Parámetros
Y_size = 100 

Co = 35
Io = 40
Go = 50
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3
r = 0.9

Y = np.arange(Y_size)

    # Ecuación 
def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)


#--------------------------------------------------
# NUEVA curva de ingreso de equilibrio con aumento del Gasto de gobierno

Go = 80

def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_G = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)
```

```python
# Gráfico
y_max = np.max(DA_IS_K)
fig, ax = plt.subplots(figsize=(12, 10))


# Curvas a graficar
ax.plot(Y, DA_IS_K, label = "DA", color = "blue") #curva ORIGINAL
ax.plot(Y, DA_G, label = "DA_G", color = "green", linestyle = 'dashed') #NUEVA curva
ax.plot(Y, L_45, color = "red") #línea de 45º

# Lineas punteadas
plt.axvline(x = 75, ymin= 0, ymax = 0.73, linestyle = ":", color = "grey")
plt.axhline(y = 187, xmin= 0, xmax = 0.73, linestyle = ":", color = "grey")
plt.axvline(x = 61,  ymin= 0, ymax = 0.61, linestyle = ":", color = "grey")
plt.axhline(y = 152, xmin= 0, xmax = 0.61, linestyle = ":", color = "grey")

# Texto agregado
plt.text(0, 133, '$DA^e$', fontsize = 11.5, color = 'black')
plt.text(0, 170, '$DA_G$', fontsize = 11.5, color = 'black')
plt.text(6, 4, '$45°$', fontsize = 11.5, color = 'black')
plt.text(2.5, -3, '$◝$', fontsize = 30, color = 'black')
plt.text(70, 0, '$Y_G$', fontsize = 12, color = 'black')
plt.text(57, 0, '$Y^e$', fontsize = 12, color = 'black')

# Título y leyenda
ax.set(title = "Aumento del Gasto del Gobierno $G_0$", xlabel = r'Y', ylabel = r'DA')
ax.legend()

plt.show()
```

# Política Fiscal expansiva con una reducción de la Tasa de Tributación:

- Intuitivo
           t↓ → Co↑ → DA↑ → DA > Y → Y↑
           t↓ → M↑ → DA↓ → DA < Y → Y↓
   
Al bajar la tasa impositiva, el consumo autónomo aumenta. Es así como, aumenta la demanda agregada, y por consecuencia, requiere que aumente el Ingreso Agregado para volver al equilibrio.

Otro resultado de bajar la tasa impositiva, es que genera que aumenten las importaciones. Esto reduce la demanda agregada, y genera un déficit de demanda, que solo regresará al equilibrio con una disminución del Ingreso.

Sin embargo, debido a que en la práctica "b" tiene un mayor efecto, ya que la cantidad de personas que consume es mayor a las que importa, el primer impacto descrito es el que predomina. 

- Matemático 
    


$\frac{∆Y}{ -∆t} = \frac{(-)}{(+)} $ 

$\frac{∆Y}{∆t} = \frac{(-)(-)}{(+)}$ 

$\frac{∆Y}{ -∆t} = 1 $

$ {∆Y↑}={ ∆t↓} $

```python
#Gráfico de la reducción de la tasa de tributación.

#--------------------------------------------------
# Curva de ingreso de equilibrio ORIGINAL

    # Parámetros
Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3 #tasa de tributación
r = 0.9

Y = np.arange(Y_size)

    # Ecuación 
def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)


#--------------------------------------------------
# NUEVA curva de ingreso de equilibrio

t = 0.01

# Generar la ecuación con el nuevo parámetros
def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_t = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)
```

```python
# Gráfico
y_max = np.max(DA_IS_K)
fig, ax = plt.subplots(figsize=(10, 8))


# Curvas a graficar
ax.plot(Y, DA_IS_K, label = "DA", color = "black") #curva ORIGINAL
ax.plot(Y, DA_t, label = "DA_t", color = "green", linestyle = 'dashed') #NUEVA curva
ax.plot(Y, L_45, color = "red") #línea de 45º

# Lineas punteadas
plt.axvline(x = 70.5, ymin= 0, ymax = 0.69, linestyle = ":", color = "grey")
plt.axhline(y = 176, xmin= 0, xmax = 0.7, linestyle = ":", color = "grey")
plt.axvline(x = 77,  ymin= 0, ymax = 0.75, linestyle = ":", color = "grey")
plt.axhline(y = 192, xmin= 0, xmax = 0.75, linestyle = ":", color = "grey")

# Texto agregado
plt.text(0, 180, '$DA^e$', fontsize = 11.5, color = 'black')
plt.text(0, 200, '$DA_t$', fontsize = 11.5, color = '#EE7600')
plt.text(6, 4, '$45°$', fontsize = 11.5, color = 'black')
plt.text(2.5, -3, '$◝$', fontsize = 30, color = '#404040')
plt.text(72, 0, '$Y^e$', fontsize = 12, color = 'black')
plt.text(80, 0, '$Y_t$', fontsize = 12, color = '#EE7600')


# Título y leyenda
ax.set(title = "Reducción de la Tasa de Tributación", xlabel = r'Y', ylabel = r'DA')
ax.legend()

plt.show()
```

# Equilibrio con regla contracíclica

Ya que el gasto del gobierno tiene un nuevo componente $"-gY"$,la ecuación de la demanda agregada, cambiaría de la siguiente manera:

$DA = C_0 + I_0 + G_0 + X_0 - hr + Y((b - m)(1 - t)-g)$

Consecuentemente, al tener en igualdad $DA=Y$, el ingreso de equilibrio estaría modelado por la siguiente ecuación

$Y = \frac{1}{1 - (b - m)(1 - t) + g} (C_0 + I_0 + G_0 + X_0 - hr)$

```python
#Gráfico: vamos a graficar la curva de equilibrio original, y luego 
#vamos a graficar la nueva curva, con el aumento del gasto de gobierno.

#--------------------------------------------------
# Curva de ingreso de equilibrio ORIGINAL

    # Parámetros
Y_size = 100 

Co = 35
Io = 40
Go = 50
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3
r = 0.9
g = 0.1

Y = np.arange(Y_size)

    # Ecuación 
def DA_IS_K(Co, Io, Go, Xo, h, r, b, m, t, g, Y):
    DA_IS_K = (Co + Io + Go + Xo - h*r) + (((b - m)*(1 - t)-g)*Y)
    return DA_IS_K

DA_IS_K = DA_IS_K(Co, Io, Go, Xo, h, r, b, m, t, g, Y)


#--------------------------------------------------
# NUEVA curva de ingreso de equilibrio con aumento del Gasto de gobierno

Go = 80

def DA_K(Co, Io, Go, Xo, h, r, b, m, t, g, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + (((b - m)*(1 - t)-g)*Y)
    return DA_K

DA_G = DA_K(Co, Io, Go, Xo, h, r, b, m, t, g, Y)
```

```python
# Gráfico
y_max = np.max(DA_IS_K)
fig, ax = plt.subplots(figsize=(10, 8))


# Curvas a graficar
ax.plot(Y, DA_IS_K, label = "DA", color = "blue") #curva ORIGINAL
ax.plot(Y, DA_G, label = "DA_G", color = "green", linestyle = 'dashed') #NUEVA curva
ax.plot(Y, L_45, color = "red") #línea de 45º

# Lineas punteadas
plt.axvline(x = 72, ymin= 0, ymax = 0.71, linestyle = ":", color = "grey")
plt.axhline(y = 180, xmin= 0, xmax = 0.71, linestyle = ":", color = "grey")
plt.axvline(x = 58,  ymin= 0, ymax = 0.57, linestyle = ":", color = "grey")
plt.axhline(y = 145, xmin= 0, xmax = 0.57, linestyle = ":", color = "grey")

# Texto agregado
plt.text(0, 133, '$DA^e$', fontsize = 11.5, color = 'black')
plt.text(0, 170, '$DA_G$', fontsize = 11.5, color = '#EE7600')
plt.text(6, 4, '$45°$', fontsize = 11.5, color = 'black')
plt.text(2.5, -3, '$◝$', fontsize = 30, color = '#404040')
plt.text(70, 0, '$Y_G$', fontsize = 12, color = '#EE7600')
plt.text(57, 0, '$Y^e$', fontsize = 12, color = 'black')


# Título y leyenda
ax.set(title = "Aumento del Gasto del Gobierno $(G_0)$", xlabel = r'Y', ylabel = r'DA')
ax.legend()

plt.show()
```

Como ha de observarse, el nuevo parámetro "g" aumenta el tamaño del denominador del multiplicador keynesiano. Al suceder esto, la cantidad resultante de la nueva división hará que se contraiga la pendiente de la recta de DA.

A diferencia del modelo de la pregunta 5, el valor de "k" es menor, ya que el denominador ha aumentado por la adición de "g". Por lo tanto, siendo "k" (denominador keynesiano) menor, el impacto de un aumento del gasto de gobierno se reducirá, a diferencia de cuando no existía el parámetro "g".
