### (1.5 puntos) Defina el Multiplicador Keynesiano y dé un ejemplo de cómo influye este en el modelo Ingreso-Gasto Keynesiano.

El Multiplicador Keynesiano representa en cuánto se va a incrementar el producto ante un incremento de cualquiera de los componentes autónomos de la demanda agregada (que conforman el intercepto), ya que genera un proceso multiplicador del ingreso hasta converger al nuevo ingreso y producto de equilibrio.

Su impacto en el modelo Ingreso-Gasto Keynesiano se nota en que, si se produce un incremento en el gasto público, por ejemplo, el producto final tendrá un incremento mayor al incremento original en la demanda; logrando así un aumento multiplicador del ingreso como consecuencia del efecto del Multiplicador Keynesiano en el aumento del gasto público.


### (2 puntos) Grafique y explique cúando sucede un exceso y déficit de demanda en el modelo Ingreso-Gasto. Señale las áreas donde ocurre cada caso. Explique cómo se converge al equilibrio a partir de estos dos escenarios.

Los excesos y déficits de demanda en el modelo Ingreso-Gasto suceden cuando hay ausencia de equilibrio, ya que puede que la Demanda Agregada sea mayor o igual que el ingreso o producto. 

Por un lado, los aumentos de la demanda agregada, que dan lugar a excesos de demanda, provocan la caída de los inventarios o existencias de las empresas, por lo que deben enfrentarse con aumentos de la producción (porque aumenta el empleo, disminuye la tasa de desempleo y aumenta el ingreso disponible de las familias, que distribuirán el incremento de su ingreso disponible entre consumo de bienes y servicios, y ahorro). 

Por otro lado, cuando se produce un déficit de demanda o exceso de producción, se aumentan los inventarios y las empresas deciden disminuir su producción y el empleo de trabajadores. En ambos casos, la oferta de producción se ajusta a los cambios en la demanda agregada.



En el gráfico se muestra el proceso de convergencia hacia el punto de equilibrio “E”, partiendo de un punto “A” (exceso de demanda) o del punto “B” (exceso de producción o déficit de demanda). 

En el punto “A”, la demanda agregada es mayor que el nivel de producción, así que hay exceso de demanda que provoca una disminución de los inventarios y aumentos de la producción hasta llegar al punto de equilibrio. 

En el punto “B”, la demanda agregada es menor que el nivel de producción, así que hay un déficit de demanda o un exceso de producción que da lugar a un aumento de los inventarios y a disminuciones de la producción hasta llegar al punto de equilibrio.

### (2 puntos) ¿Cuáles son las herramientas de política fiscal y política monetaria? Dea un ejemplo para cada una de ellas dentro del contexto peruano. Coloque su fuente. (Solo necesita 1 para pol. Fiscal y 1 para Monetaria).

Las herramientas de política fiscal son _el gasto público y la tasa impositiva_ , los cuales pueden ser utilizados por el gobierno para estabilizar o morigerar las fluctuaciones del producto mediante la política fiscal utilizando dichas herramientas. 

Un ejemplo aplicado al Perú de uno de los instrumentos de política fiscal se ve en el siguiente artículo: https://www.bloomberglinea.com/2022/10/14/maximo-torero-fao-peru-ya-no-es-el-pais-millonario-que-podia-distribuir-lo-que-queria/?outputType=amp en el que se explica que, por la fuerte caída del gasto público en el Perú, este ya no es un país con bajo riesgo, por lo que se deberían aplicar políticas fiscales más sólidas. 


Además, hay tres posibles instrumentos de política monetaria:

_1) La Oferta Monetaria:_ Suponiendo que la tasa de interés es una variable endógena, mientras la oferta monetaria es una variable exógena, uno de los instrumentos de política monetaria que utiliza el Banco Central para aumentar o disminuir la oferta monetaria es la Operación de Mercado Abierto; a través de la cual, el Banco Central compra o vende de activos financieros o bonos a los bancos comerciales para aumentar o disminuir la oferta monetaria.

_2) El coeficiente legal de encaje:_ El Banco Central puede incrementar o disminuir la cantidad de dinero que tienen los bancos para realizar préstamos mediante la disminución o aumento de la tasa de encaje, lo que implica que los bancos comerciales tendrán más o menos dinero en forma de reservas en el balance del Banco Central.

_3) La tasa de interés:_ Desde la década de 1990, los Bancos Centrales optaron por un nuevo esquema de política monetaria, a través del cual puede reducir o aumentar su tasa de interés de referencia (convirtiéndola en un instrumento de política y teniendo a la oferta monetaria como una variable endógena). De este modo, ante el aumento o disminución en la cantidad de dinero prestada a los bancos comerciales, la base monetaria y la oferta monetaria variarán.

Un ejemplo de la aplicación de la tasa de encaje como un instrumento monetario usado por el BCRP se nota en el siguiente artículo: https://elperuano.pe/noticia/65085-bcr-inyecta-us-181-millones-a-economia , en el que se explica que, debido a la reducción de las tasas de encaje en moneda extranjera por parte del BCRP, ingresarán alrededor de 181 millones de dólares al sistema financiero peruano.

* **Reserva de Valor:** el dinero es un medio para mantener riqueza. 

Al cumplir Yape y Plin con todas estas funciones del dinero, se entiende fácilmente su éxito en la economía actual del Perú. 

### (3 puntos) Cuáles son las conclusiones principales del paper de Dancurt "Inflation Targeting in Peru: The Reasons for the Success"

El paper de Dancurt "Inflation Targeting in Peru: The Reasons for the Success" está enfocado en mostrar cuáles fueron las principales acciones que emprendió el Banco Central de Reservas del Perú entre el 2002 y el 2013 para preservar la estabilidad macroeconómica en contextos externos favorables o no favorables; partiendo de un contexto de notable desempeño macroeconómico logrado por una serie de decisiones del Banco Central de Reserva del Perú, donde la inflación estaba en su nivel más bajo y el crecimiento del PIB en su nivel más alto en décadas, y no habían crisis bancarias. 

A partir de un análisis exahustivo, Dancurt concluye en que el BCRP realizó dos principales acciones que ayudaron a preservar la estabilidad macroeconómica del Perú (baja inflación, alto crecimiento del PIB o impidiendo recesiones severas) durante el período mencionado, en contextos externos favorables o desfavorables: primero, se implementó un sistema de metas de inflación con una tasa de interés de referencia como principal instrumento de política y, segundo, se acumularon suficientes reservas de divisas.

Con ello, Dancurt argumenta que, en primer lugar, el BCRP debe reducir la tasa de interés y vender divisas moneda en respuesta a choques externos adversos. En segundo lugar, el BCRP debe aumentar la tasa de interés y comprar divisas en respuesta a condiciones favorables. Además, explica que para prevenir crisis bancarias, el banco central debe limitar el endeudamiento externo de los bancos comerciales. Sin esto, desdolarizar el crédito bancario no es posible. Así, concluye que la única herramienta efectiva es imponer requisitos de reserva más altos para deuda externa de los bancos comerciales.
