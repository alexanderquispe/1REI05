**1. Explique cu√°les son los intrumentos de pol√≠tica monetaria que puede utilizar el Banco Central.**

Hay tres posibles instrumentos de pol√≠tica monetaria:

_1) La Oferta Monetaria como instrumento de pol√≠tica:_

Suponiendo que la tasa de inter√©s es una variable
end√≥gena, mientras la oferta monetaria es una variable ex√≥gena, uno de los instrumentos de pol√≠tica monetaria que utiliza el Banco Central para aumentar o disminuir la oferta monetaria es la Operaci√≥n de Mercado Abierto; que se realiza a trav√©s de la compra o venta de activos financieros o bonos a los bancos comerciales.

* Pol√≠tica Monetaria Expansiva:

Se realiza mediante la compra de bonos del mercado para inyectar dinero a la econom√≠a. As√≠, se aumenta la base monetaria y, por ende, tambi√©n incrementa a la oferta monetaria. 

* Pol√≠tica Monetaria Contractiva:

Se realiza mediante la venta de bonos al mercado (bancos comerciales) para retirar dinero de la econom√≠a. As√≠, reduce la base monetaria y, por ende, tambi√©n disminuye la oferta monetaria. 



_2) El coeficiente legal de encaje como instrumento de pol√≠tica:_

* Pol√≠tica Monetaria Expansiva:
    
La disminuci√≥n de la tasa de encaje aumenta la posibilidad de creaci√≥n de dinero bancario porque aumenta el multiplicador de dinero bancario; y, este aumento del dinero bancario implica un aumento de la oferta monetaria. 

* Pol√≠tica Monetaria Contractiva:

El aumento del coeficiente de encaje, lleva a que los bancos comerciales tengan una mayor proporci√≥n de dep√≥sitos en forma de reservas, lo que reduce el multiplicador bancario y, por tanto, origina una disminuci√≥n de la oferta monetaria.



_3) La tasa de inter√©s como instrumento de pol√≠tica:_

Desde los a√±os 1990 los Bancos Centrales optaron por un nuevo esquema de pol√≠tica
monetaria. La tasa de inter√©s se convirti√≥ en instrumento de pol√≠tica y la oferta monetaria
pas√≥ a ser una variable end√≥gena. 

Desde 1990, la tasa de pol√≠tica es la tasa de inter√©s de referencia de la pol√≠tica monetaria.

* Pol√≠tica Monetaria Expansiva:

El Banco Central hace pol√≠tica monetaria expansiva reduciendo su tasa de inter√©s de
referencia. De este modo aumenta la cantidad de dinero prestada a los bancos comerciales
lo que aumenta la base monetaria y con ello la oferta monetaria.

* Pol√≠tica Monetaria Contractiva:

El Banco Central hace pol√≠tica monetaria contractiva aumentando su tasa de inter√©s de
referencia. De este modo disminuye la cantidad de dinero prestada (el cr√©dito) a los
bancos comerciales, lo que reduce la base monetaria y con ello la oferta monetaria.

```python
# Par√°metros
r_size = 100

k = 0.5
j = 0.2                
P  = 10 
Y = 35
MS_0 = 500

r = np.arange(r_size)

# Necesitamos crear la funcion de demanda 

def MD(k, j, P, r, Y):
    MD_eq = (k*Y - j*r)
    return MD_eq
MD_0 = MD(k, j, P, r, Y)
# Necesitamos crear la oferta de dinero.
MS = MS_0 / P
MS
```

2. **¬øCu√°l es el efecto de una disminuci√≥n en la Masa Monetaria ŒîMs<0? Explica usando la intuici√≥n y gr√°ficos.**

    * An√°lisis intuitivo: 
    
$$Ms‚Üì ‚Üí Ms<Md ‚Üí i‚Üë ‚Üí r‚Üë$$

Al disminuir la masa monetaria, se produce un desvalance en el mercado ya que, como disminuye la oferta de dinero, la demanda de dinero existente ahora es mayor. As√≠, para volver al punto de equilibrio,y considerando que el precio del dinero es la tasa de inter√©s, la tasa de inter√©s tendr√° que subir. 


   * An√°lisis gr√°fico: 

```python
 # Curva LM ORIGINAL

# Par√°metros

Y_size = 100

k = 2
j = 1                
Ms = 600             
P  = 20               

Y = np.arange(Y_size)

# Ecuaci√≥n

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)

#--------------------------------------------------
    # NUEVA curva LM

# Definir SOLO el par√°metro cambiado
Ms = 100

# Generar la ecuaci√≥n con el nuevo par√°metro
def i_LM_Ms( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i_Ms = i_LM_Ms( k, j, Ms, P, Y)
```

```python
# Dimensiones del gr√°fico
y_max = np.max(i)
v = [0, Y_size, 0, y_max]   
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, i, label="LM", color = '#fcc660')
ax.plot(Y, i_Ms, label="LM_Ms", color = '#fc5603', linestyle = 'dashed')

# Texto agregado
plt.text(46, 76, '‚àÜ$M^s$', fontsize=12, color='black')
plt.text(46, 70, '‚Üê', fontsize=15, color='red')

# T√≠tulo y leyenda
ax.set(title = "Disminuci√≥n en la Masa Monetaria $(M^s)$", xlabel=r'Y', ylabel=r'r')
ax.legend()


plt.show()
```

3. **¬øCu√°l es el efecto de un aumento en k ŒîK>0? Explica usando intuici√≥n y gr√°ficos.**

    * An√°lisis intuitivo: 
    
$$k‚Üë ‚Üí Md‚Üë ‚Üí Md>Ms ‚Üí Ms‚Üë ‚Üí i‚Üë ‚Üí r‚Üë$$

Al aumentar k, aumenta tambi√©n la demanda de dinero, as√≠ que se produce un desvalance en el mercado ya que la demanda de dinero es mayor a la oferta. As√≠, considerando que el precio del dinero es la tasa de inter√©s, la tasa de inter√©s tendr√° que aumentar para volver al punto de equilibrio. 

        
* An√°lisis gr√°fico: 

```python
#--------------------------------------------------
    # Curva LM ORIGINAL

# Par√°metros

Y_size = 100

k = 2
j = 1                
Ms = 200             
P  = 20               

Y = np.arange(Y_size)

# Ecuaci√≥n

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)

#--------------------------------------------------
    # NUEVA curva LM
    
k = 3

# ecuaci√≥n con el nuevo par√°metro
def i_LM_k( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i_k = i_LM_k( k, j, Ms, P, Y)
```

```python
# Dimensiones del gr√°fico
y_max = np.max(i)
v = [0, Y_size, 0, y_max]   
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, i, label="LM", color = '#fa7d7d')
ax.plot(Y, i_k, label="LM_k", color = '#cc0808', linestyle = 'dashed')

# Texto agregado
plt.text(47, 116, '‚àÜ$k$', fontsize=12, color='black')
plt.text(47, 106, '‚Üê', fontsize=15, color='red')

# T√≠tulo y leyenda
ax.set(title = "Incremento en la sensibilidad de la demanda de dinero $(k)$", xlabel=r'Y', ylabel=r'r')
ax.legend()


plt.show()
```
