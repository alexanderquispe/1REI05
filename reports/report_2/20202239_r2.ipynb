#### a. El Modelo Ingreso-Gasto: la Curva IS

En el modelo Ingreso-Gasto el equilibrio entre el Ingreso y la DA representa el equilibrio en el mercado de bienes. 


##### - Utilizando las ecuaciones derive paso a paso la curva IS matemáticamente (a partir de la condición de equilibrio Y = DA).

    Primero, se parte de la igualdad entre el ingreso $(Y)$ y la demanda agregada $(DA)$. 
            
$$ Y = C + I + G + X - M $$

Luego, para llegar al equilibrio Ahorro-Inversión, se debe restar la Tributación $(T)$ en ambos miembros de la igualdad. 
     
$$ Y - T = C + I + G + X - M - T $$

$$ Y^d = C + I + G + X - M - T $$

Luego, se despeja la variable Inversión $(I)$:
     
$$ I = Y^d - C - G - X + M $$

Además, se puede agrupar: 

$$ (Y^d - C) + (T - G) + (M - X) = I $$

donde:
- $(Y^d - C)$ : es el ahorro privado
- $(T - G)$ : es el ahorro del gobierno
- $(M - X)$ : es el ahorro externo

Entonces, en base a las ecuaciones del modelo, se tiene que:

$$ [Y^d - (C_0 + bY^d)] + [tY - G_0] + [mY^d - X_0] = I_0 - hr $$

donde: 

$$ Y^d = (1 - t)Y $$

Por lo tanto, si se factoriza, se tiene que: 

$$ [1 - (b - m)(1 - t)]Y - (C_0 + G_0 + X_0) = I_0 - hr $$

En base a ello, se puede expresar la curva IS con una ecuación, en donde la tasa de interés $(r)$ se encuentra determinada en función del Ingreso $(Y)$:

$$ hr = (C_0 + G_0 + I_0 + X_0) - [1 - (b - m)(1 - t)]Y $$

$$ r = \frac{1}{h}(C_0 + G_0 + I_0 + X_0) - [\frac{1 - (b - m)(1 - t)}{h}]Y $$

Entonces, podemos simplificar y concluir que:

$$ r = \frac{θ_0}{h} - \frac{θ_1}{h}Y $$

donde:

- $ θ_0 = C_0 + G_0 + I_0 + X_0 $ es el intercepto 
- $ θ_1 = 1 - (b - m)(1 - t) $ es la pendiente

##### - Utilizando la función de equilibrio IS donde r está en función de Y, encuentre $∆r/∆Y$ que representa la pendiente y qué signo tiene?

En este caso, dado que la pendiente de la curva IS es mayor que cero y puede graficarse con una recta de pendiente negativa, se puede decir que:

$$ \frac{∆r}{∆Y}|_Is = - \frac{θ_1}{h} < 0 $$

En ese sentido, como se ve, se trata de una pendiente negativa, por lo tanto, su signo es negativo (-):

$$ \frac{∆r}{∆Y}|_Is = (-) $$

Entonces, como se sabe que, para mantener una situación de equilibrio, el precio debe ajustarse a la demanda de agregada, se tiene que un aumento en esta, generaría un aumento en el precio de equilibrio $(DA↑ → Y↑)$. En ese sentido, de esa manera se deriva la curva IS, pues dicha recta es esta, donde la curva representa el equilibrio en el mercado de bienes.

#### b. La curva IS o el equilibrio Ahorro-Inversión

###### - Derive la ecuación de IS a partir de la igualdad Ahorro-Inversión y grafique la curva IS de equilibrio en el Mercado de Bienes.

Partimos de la igual:

$$ S = I $$

y teniendo en cuenta que:

- $(Y^d - C)$ : es el ahorro privado $(S_p)$
- $(T - G)$ : es el ahorro del gobierno $(S_g)$
- $(M - X)$ : es el ahorro externo $(S_e)$     

Se puede decir que: 

$$ S_p + S_g + S_e = I $$

Entonces, reemplazamos:

$$ (Y^d - C) + (T - G) + (M - X) = I $$

y llegamos a la siguiente ecuación: 

$$ [Y^d - (C_0 + bY^d)] + [tY - G_0] + [mY^d - X_0] = I_0 - hr $$

Factorizando se llega a:

$$ [1 - (b - m)(1 - t)]Y - (C_0 + G_0 + X_0) = I_0 - hr $$

Y luego, se busca que la tasa de interés $(r)$ esté en función al ingreso: 

$$ hr = (C_0 + G_0 + I_0 + X_0) - [1 - (b - m)(1 - t)]Y $$

$$ r = \frac{1}{h}(C_0 + G_0 + I_0 + X_0) - [\frac{1 - (b - m)(1 - t)}{h}]Y $$

donde, de forma resumida, ya se estableció que: 

$$ r = \frac{θ_0}{h} - \frac{θ_1}{h}Y $$
