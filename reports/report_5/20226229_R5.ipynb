#  A partir del siguiente sistema de ecuaciones que representan el modelo DA-OA

* Demanda Agregada:
  
 $$ IS:[1-b(1-t)]Y = C_O + I_O + G_O - hr ...(1)$$
 
 $$ LM: \frac{M_O^S}{P_O} = kY - jr ...(2) $$
 
 * Oferta Agregada:
 
 $$ P_0 = P^e + Œ∏(Y- Y^-) ...(3) $$ donde $$ Y^- = Y_f $$ 

# Encuentre las ecuaciones de ingreso $(Y^e)$ , tasa de inter√©s $(r^e)$, $(P^e)$ de equilibrio (Escriba paso a paso la derivaci√≥n de estas ecuaciones)

* De la Ecuacion IS (3): 
  
   $$ hr = C_O + I_O + G_O - [1-b(1-t)]Y  $$

   $$ r = \frac{C_O + I_O + G_O}{h} - \frac{[1-b(1-t)]}{h}Y  ...(4)  $$
   
* Asumiendo que  $ B_0 = C_O + I_O + G_O $ y $B_1 = 1-b(1-t)$ ...(5)

* Remplazando (5) en (4) se obtiene la ecuacion lineal r(Y) 
   
  $$ r = \frac{B_0}{h} - \frac{B_1}{h}Y ...(6) $$

* De la Ecuacion LM (2):

  $$ jr = kY - \frac{M_O^S}{P_O}  $$
 
  $$ r = - \frac{1}{j}\frac{M_O^S}{P_O} + \frac{k}{j}Y ...(7) $$
  
* Igualando $ r $ ecuacion (6) y (7) para obtener P(Y) 

   $$ \frac{B_0}{h} - \frac{B_1}{h}Y  = -\frac{1}{j}\frac{M_O^S}{P_O} + \frac{k}{J}Y $$
   
   $$ P = \frac{hM_O^S}{-jB_0 + (jB_1 + hk)Y}  $$
   
 
 * Igualando $ r $ ecuacion (6) y (7) para obtener Y(P)
 
   $$ \frac{B_0}{h} - \frac{B_1}{h}Y  = -\frac{1}{j}\frac{M_O^S}{P_O} + \frac{k}{J}Y $$
   
   $$ Y = \frac{jB_0}{jB_1 + hk} + (\frac{hM_O^S}{jB_1 + hk}) \frac{1}{P}  $$
 
 * Considerando la nueva ecuacion de equilibrio en el mercado monetario donde hacemos el remplazo de 
 
* Para linealizar se modifica la oferta monetaria 

  $$ \frac{M_O^S}{P} -----> M_O^S - P $$
  
* Entonces remplazando 
   
  $$ M_O^S - P = kY - jr $$
 
 * Se remplaza (r) y se obtiene la ecaucion de la demanda agregada DA, que ahora es una recta no una hiperbola
 
   $$ P = \frac{hM_O^S + jB_O}{h} - \frac{jB_1 + hk}{h}Y $$
   
 * Donde  $ B_0 = C_O + I_O + G_O $ y $B_1 = 1-b(1-t)$

```python
# IS-LM

#1--------------------------------------------------
    # Curva IS ORIGINAL

# Par√°metros

Y_size = 100 

Co = 35
Io = 40
Go = 50
Xo = 2
h = 0.8
b = 0.4
m = 0.5
t = 0.8

Y = np.arange(Y_size)


# Ecuaci√≥n 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)

t = 3.6
r_t = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)

#2--------------------------------------------------
    # Curva LM ORIGINAL

# Par√°metros

Y_size = 100

k = 2
j = 1                
Ms = 200             
P  = 8           

Y = np.arange(Y_size)

# Ecuaci√≥n

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)

#DA-OA

#1--------------------------
# Demanda Agregada ORGINAL
    
# Par√°metros

Y_size = 100 

Co = 35
Io = 40
Go = 50
Xo = 2
h = 0.8
b = 0.4
m = 0.5
t = 0.8

k = 2
j = 1                
Ms = 200             
P  = 8  

Y = np.arange(Y_size)


# Ecuaci√≥n

B0 = Co + Io + Go + Xo
B1 = 1 - (b-m)*(1-t)

def P_AD(h, Ms, j, B0, B1, k, Y):
    P_AD = ((h*Ms + j*B0)/h) - (Y*(j*B1 + h*k)/h)
    return P_AD

P_AD = P_AD(h, Ms, j, B0, B1, k, Y)

#--------------------------------------------------
    # NUEVA Demanda Agregada

# Definir SOLO el par√°metro cambiado

t = 3.6
B1 = 1 - (b-m)*(1-t)
# Generar la ecuaci√≥n con el nuevo par√°metro

def P_AD_Ms(h, Ms, j, B0, B1, k, Y):
    P_AD = ((h*Ms + j*B0)/h) - (Y*(j*B1 + h*k)/h)
    return P_AD

P_AD_t= P_AD_Ms(h, Ms, j, B0, B1, k, Y)


#2--------------------------
    # Oferta Agregada ORIGINAL
    
# Par√°metros

Y_size = 100

Pe = 70
Œ∏ = 3
_Y = 20  

Y = np.arange(Y_size)

# Ecuaci√≥n

def P_AS(Pe, _Y, Y, Œ∏):
    P_AS = Pe + Œ∏*(Y-_Y)
    return P_AS

P_AS = P_AS(Pe, _Y, Y, Œ∏)

# Dos gr√°ficos en un solo cuadro
fig, (ax1, ax2) = plt.subplots(2, figsize=(8, 16)) 

#---------------------------------
# Gr√°fico 1: IS-LM
    
ax1.plot(Y, r, label = "IS(t = 0.8)", color = "C1") #IS
ax1.plot(Y, r_t, label = "IS(t = 1.6)", color = "C0", linestyle ='dashed') #IS
ax1.plot(Y, i, label="LM", color = "C0")  #LM

ax1.text(60, 85, '‚àÜ$M_s$', fontsize=12, color='black')
ax1.text(60, 90, '‚Üê', fontsize=15, color='grey')
ax1.set(title="Efectos de unincremento en la tasa de impuesto", xlabel= r'Y', ylabel= r'r')
ax1.legend()


#---------------------------------
    # Gr√°fico 2: DA-OA

ax2.plot(Y, P_AD, label = "AD(t = 0.8)", color = "C4") #DA
ax2.plot(Y, P_AD_t, label = "AD(t = 1.6)", color = "C5", linestyle = 'dashed') #DA_Ms
ax2.plot(Y, P_AS, label = "AS", color = "C8") #OA

ax2.axvline(x=67.5,  ymin= 0, ymax= 1, linestyle = ":", color = "grey")
ax2.axvline(x=56,  ymin= 0, ymax= 1, linestyle = ":", color = "grey")
ax2.axhline(y=213,  xmin= 0, xmax= 0.67, linestyle = ":", color = "grey")
ax2.axhline(y=175,  xmin= 0, xmax= 0.56, linestyle = ":", color = "grey")

ax2.text(60, 100, '‚àÜ$M_s$', fontsize=12, color='black')
ax2.text(60, 130, '‚Üë', fontsize=15, color='grey')

ax2.set(xlabel= r'Y', ylabel= r'P')
ax2.legend()

plt.show
```
