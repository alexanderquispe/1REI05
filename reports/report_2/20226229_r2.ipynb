# a. El Modelo Ingreso-Gasto: la Curva IS 

* Utilizando las ecuaciones derive paso a paso la curva IS matem√°ticamente (a partir de la condici√≥n de equilibrio Y = DA).
  En el modelo Ingreso-Gasto el equilibrio entre el Ingreso y la DA representa el equilibrio en el mercado de bienes
  
  $$ ùëå = (ùê∂_0 + ùêº_0 + ùê∫_0 + ùëã_0 ‚àí ‚Ñéùëü) + (ùëè ‚àí ùëö)(1 ‚àí ùë°)ùëå $$
  $$ [1 ‚àí (ùëè ‚àí ùëö)(1 ‚àí ùë°)]ùëå = ùê∂_0 + ùêº_0 + ùê∫_0 + ùëã_0 ‚àí ‚Ñéùëü $$
  $$ Y = \frac{1}{ùê∂_0 + ùêº_0 + ùê∫_0 + ùëã_0 ‚àí ‚Ñéùëü)}  $$

* Y en funcion de r $$ Y = f(r) $$ 

  $$ Y = \frac{1}{(1 ‚àí (ùëè ‚àí ùëö)(1 ‚àí ùë°))} (C_0 + I_0 + G_0 + X_0) - \frac{h}{1 ‚àí (ùëè ‚àí ùëö)(1 ‚àí ùë°)} r $$   

* Utilizando la funci√≥n de equilibrio IS donde r est√° en funci√≥n de Y, encuentre  Ar / Ay 
  que representa la pendiente y qu√© signo tiene?
  
  La igualdad entre el ahorro y la inversion representa la curva IS
  De la pregunta (1) se deriva que r en funcion de y $$ r=ùëì(Y)$$
  
  $$ r = \frac{1}{h} (C_0 + I_0 + G_0 + X_0) - \frac{1 ‚àí (ùëè ‚àí ùëö)(1 ‚àí ùë°)}{h} Y $$   
  
  La ecuacion lineal $$ r = b + m * y $$
  
  $$ r = \frac{B_0}{h} - \frac{B_1}{h} Y $$
  
  Derivando r en funcion a Y  $$ \frac{‚àÜr}{‚àÜy} = 0 - \frac{B_1}{h} $$
   
  # La pendiente tiene signo negativo

* Lea la secci√≥n 4.4 del material de ense√±anza y explique c√≥mo se deriva la curva IS a partir del equilibrio $$ Y = DA $$  y grafique:

El equilibrio de  la curva IS representa una situaci√≥n en la que todo lo que se produce en la econom√≠a (Y) 
es consumido por todos los agentes de la misma (sector p√∫blico, sector privado y sector exterior), o lo que es lo mismo, 
el ahorro iguala a la inversi√≥n.

```python
# Recta de 45¬∞
a = 2.5 
def L_45(a, Y):
    L_45 = a*Y
    return L_45
L_45 = L_45(a, Y)

#--------------------------------------------------
# Curva de ingreso de equilibrio ORIGINAL

# Par√°metros
Y_size = 100 

Co = 25
Io = 30
Go = 60
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3 #tasa de tributaci√≥n
r = 0.9

Y = np.arange(Y_size)

    # Ecuaci√≥n 
def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)

#--------------------------------------------------
# NUEVA curva de ingreso de equilibrio

# Definir SOLO el par√°metro cambiado
Co = 35
Io = 40
Go = 70
DA_G = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)
```

## IS CURVE

```python
# Curves to plot
# Parameters

Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3

Y = np.arange(Y_size)


# Equation 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (1/h)*(Co + Io + Go + Xo) - (1/h)*(1-(b-m)*(1-t))*Y
    return r_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
```

```python
# Plot of the derivation of the IS curve from equality (DA = Y)

    # Two graphs in one frame
fig, (ax1, ax2) = plt.subplots(2, figsize=(8, 16)) 

#---------------------------------
    # Graph 1: Short-Term Equilibrium Income
    
ax1.yaxis.set_major_locator(plt.NullLocator())   
ax1.xaxis.set_major_locator(plt.NullLocator())

ax1.plot(Y, DA_IS_K, label = "DA_0", color = "C0", linestyle = "-") 
ax1.plot(Y, DA_G, label = "DA_1", color = "C0", linestyle = ":") 
ax1.plot(Y, L_45, color = "#404040") 

ax1.axvline(x = 70.5, ymin= 0, ymax = 0.69, linestyle = ":", color = "grey")
ax1.axvline(x = 54,  ymin= 0, ymax = 0.54, linestyle = ":", color = "grey")

ax1.text(6, 4, '$45¬∞$', fontsize = 11.5, color = 'black')
ax1.text(2.5, -3, '$‚óù$', fontsize = 30, color = 'black')
ax1.text(72, 0, '$Y_1$', fontsize = 12, color = 'black')
ax1.text(56, 0, '$Y_0$', fontsize = 12, color = 'black')
ax1.text(67, 185, 'E_1', fontsize = 12, color = 'black')
ax1.text(50, 142, 'E_0', fontsize = 12, color = 'black')

ax1.set(title = "Derivacion de la Curva IS", xlabel = r'Y', ylabel = r'DA')
ax1.legend()

#---------------------------------
    # Graph 2: IS Curve

ax2.yaxis.set_major_locator(plt.NullLocator())   
ax2.xaxis.set_major_locator(plt.NullLocator())

ax2.plot(Y, r, label = "IS", color = "C1") 

ax2.axvline(x = 70.5, ymin= 0, ymax = 1, linestyle = ":", color = "grey")
ax2.axvline(x = 54,  ymin= 0, ymax = 1, linestyle = ":", color = "grey")
plt.axhline(y = 151.5, xmin= 0, xmax = 0.7, linestyle = ":", color = "grey")
plt.axhline(y = 165, xmin= 0, xmax = 0.55, linestyle = ":", color = "grey")

ax2.text(72, 128, '$Y_1$', fontsize = 12, color = 'black')
ax2.text(56, 128, '$Y_0$', fontsize = 12, color = 'black')
ax2.text(1, 153, '$r_1$', fontsize = 12, color = 'black')
ax2.text(1, 167, '$r_0$', fontsize = 12, color = 'black')
ax2.text(72, 152, 'E_1', fontsize = 12, color = 'black')
ax2.text(55, 166, 'E_0', fontsize = 12, color = 'black')

ax2.set(title = " IS", xlabel= 'Y', ylabel= 'r')
ax2.legend()

plt.show()
```

## b. La Curva IS o el equilibrio Ahorro Inversi√≥n

* Derive la ecuacion de IS  a partir de la igualdad Ahorro inversion y grafique la curva IS de equilibrio del Mercado de Bienes

* El equilibrio entre el ingreso y la demanda agregada es equivalente al equilibrio entre el ahorro total y la inversi√≥n 
   $$ DA= Y $$
   $$ ùëå = ùê∂  + ùêº  + ùê∫  + ùëã - M $$
  
  Restando la tributaci√≥n T en ambos lado
  $$ ùëå - T = ùê∂  + ùêº - T  + ùê∫  + ùëã - M $$
  $$ ùëå^d = ùê∂  + ùêº - T  + ùê∫  + ùëã - M $$
  
  Esta igualdad se puede reescribir de la siguiente manera:
  $$ (ùëå^d - ùê∂) + ( T  - ùê∫ ) + (M - ùëã ) = I $$
  
  El lado derecho constituye el ahorro total que es igual a la inversi√≥n
 
  $$ S_p + S_g + S_e = ùêº_O - hr $$
  
  $$ S = ùêº_O - hr $$
  
  Esta igualdad entre el ahorro y la inversi√≥n representa precisamente la curva IS
  
  $$ S(Y) = ùêº(r) $$
 
  Haciendo reemplazos se obtiene que:
  
  $$ (ùëå^d - C_O - bùëå^d) + (T -G_O) + (mùëå^d - X_O)  = ùêº_O - hr $$

  Mediante operaciones algebr√°icas se obtiene
  
   $$ [1 -( b - m) (1 - t)] ùëå - ( C_O + G_O + X_O) = ùêº_O - hr $$
   
   $$ r = \frac{1}{h} (C_0 + I_0 + G_0 + X_0) - \frac{1 ‚àí (ùëè ‚àí ùëö)(1 ‚àí ùë°)}{h} Y $$   
  
  La ecuaci√≥n lineal $$ r = b + m * y $$
  
  $$ r = \frac{B_0}{h} - \frac{B_1}{h} Y $$

```python
# Graph of the IS curve


# Dimensions
y_max = np.max(r)
fig, ax = plt.subplots(figsize=(10, 8))

# Curves to plot
ax.plot(Y, r, label = "IS", color = "C1") #Demanda agregada

# Remove the quantities from the axes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())
ax.text(90, 140, 'IS', fontsize = 12, color = 'black')


# Title, axes adn legend
ax.set(title = "La Curva IS de equilibrio en el Mercado de Bienes", xlabel= 'Y', ylabel= 'r')
ax.legend()

ax.legend()
plt.show()
```

## c. Desequilibrio en el mercado de bienes

Grafique las √°reas de exceso de oferta o exceso de demanda en el mercado de bienes. Explique por que estos puntos est√°n fuera de los puntos de equilibrio en el mercado de bienes.

Cada vez que los mercados experimentan una inestabilidad como por ejemplo se producen precios desequilibrados, excedentes y escasez,  las fuerzas del mercado conducir√°n nuevamente a precios de equilibrio.
En el gr√°fico existe un exceso de oferta cuando el precio real del producto est√° por encima del precio del mercado, dando lugar a un aumento de la oferta, y asu vez conllevar√° a que disminuya la demanda, lo que dar√° como resultado que el precio real del producto baje.
Si sucede lo contrario, es decir, el precio real est√° por debajo del precio de mercado, habr√° menos oferta y a la vez una mayor demanda, de esta forma, el producto se agotar√≠a en stock y el precio subir√≠a.


```python
# Curves to plot
# Parameters

Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3
Y = np.arange(Y_size)
# Equation 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (1/h)*(Co + Io + Go + Xo) - (1/h)*(1-(b-m)*(1-t))*Y
    return r_IS
r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)

# Dimensions
y_max = np.max(r)
fig, ax = plt.subplots(figsize=(10, 8))

# Curves to plot
ax.plot(Y, r, label = "IS", color = "C1") #Demanda agregada

# Remove the quantities from the axes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())
ax.text(0, 180, '$r_A$', fontsize = 12, color = 'black')
ax.text(22, 180, 'C', fontsize = 12, color = 'black')
ax.text(37, 180, 'A', fontsize = 12, color = 'black')
ax.text(52, 180, 'B', fontsize = 12, color = 'black')
ax.text(22, 127, '$Y_C$', fontsize = 12, color = 'black')
ax.text(37, 127, '$Y_A$', fontsize = 12, color = 'black')
ax.text(52, 127, '$Y_B$', fontsize = 12, color = 'black')
#/// Lineas horizontales | Lineas verticales
plt.axvline(x = 22,  ymin= 0, ymax= 0.69, linestyle = ":", color = "grey")
plt.axvline(x = 37,  ymin= 0, ymax= 0.69, linestyle = ":", color = "grey")
plt.axvline(x = 52,  ymin= 0, ymax= 0.69, linestyle = ":", color = "grey")
plt.axhline(y = 180, xmin= 0, xmax = 0.7, linestyle = ":", color = "grey")
#// Texto
plt.text(24, 160, 'Exceso de', fontsize = 11.5, color = '#3D59AB')
plt.text(24, 155, 'demanda', fontsize = 11.5, color = '#3D59AB')
#// Texto
plt.text(65, 190, 'Exceso de', fontsize = 11.5, color = '#3D59AB')
plt.text(65, 185, 'Oferta', fontsize = 11.5, color = '#3D59AB')

# Title, axes adn legend
ax.set(title = "Equilibrio y Desequilibrio en el Mercado de Bienes", xlabel= 'Y', ylabel= 'r')
ax.legend()

ax.legend()
plt.show()
```

## d. Movimiento de la curva IS

```python
#-------------------------------------------------- 
# ORIGINAL IS Curve
# Parameters
Y_size = 100 
Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3
Y = np.arange(Y_size)

# Equation 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)

#--------------------------------------------------
    # New IS Curve

# Contraccion en el gasto del gobierno $(G_0)$
Go = 60
r_G = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)

# Politica fiscal expansiva caida tasa de impuesto $(‚Üìt)
Go = 70
t = -0.1
r_t = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)

# Politica fiscal expansiva caida tasa de impuesto $(‚Üìt)
Go = 70
b = 0.4
r_b = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
```
