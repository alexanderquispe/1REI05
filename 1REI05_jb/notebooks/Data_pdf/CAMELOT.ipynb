```python
# Importamos las librerias
import camelot
import pandas as pd

# Read the PDF file and extract tables
tablas = camelot.read_pdf("semidefinitivos.pdf", pages='all')

# Concatenate all tables into a single DataFrame
df = pd.concat([tabla.df for tabla in tablas])

# Print the concatenated DataFrame
print(df)
```

```python
df
```

```python
# Crearemos las columnas para nuestra data
df1= df.columns= ["REGION","DRE_UGEL","GRUPO","DNI","APELLIDO_PATERNO"
,"APELLIDO_MATERNO","NOMBRES","ID_PLAZA"]
```

```python
df
```

```python
# Como ha iterado hoja por hoja, ha jalado las cabeceras en cada hoja, procedemos a localizarlas y eliminarlas
df["REGION"] = df["REGION"].str.strip()
errors = df.loc[df["REGION"]=="REGIÓN"]
errors_one= df.loc[df["ID_PLAZA"]=="CÓDIGO DE PLAZA"]
errors_list = errors.index.tolist()
```

```python
Cargo_Especialistas = df.drop(df.index[errors_list])
```

```python
Cargo_Especialistas
```

```python
Especialistas_clean = Cargo_Especialistas.drop([1],axis=0)
Especialistas_clean
```
