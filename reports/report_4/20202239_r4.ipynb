### Modelo IS-LM

###### - Encuentre las ecuaciones de Ingreso $(Y^e)$ y tasa de interés $(r^e)$ de equilibrio. Escriba paso a paso la derivación de estas ecuaciones. 

En este caso, se debe tener en cuenta, en primer lugar, la función de la curva IS:

$$ r = \frac{β_0}{h} - \frac{β_1}{h}Y $$

donde:

- $β_0 = C_0 + I_0 + G_0 + X_0$
- $β_1 = 1 - (b - m)(1 - t)$

Asimismo, en segundo lugar, la función de la curva LM:

$$ r = -\frac{1}{j}\frac{M_o}{P} + \frac{k}{j}Y $$

En ese sentido, para poder hallar la ecuación de nivel de producción de equilibrio $(Y^e)$ es necesario igualar ambas ecuaciones:

$$ \frac{β_0}{h} - \frac{β_1}{h}Y = -\frac{1}{j}\frac{M_o}{P} + \frac{k}{j}Y $$

Luego se debe despejar $Y$:

$$ j(β_0) - j(β_1)Y = h(k)Y - h\frac{M_o}{P} $$

$$ j(β_0) + h\frac{M_o}{P} = (kh + jβ_1)Y $$


$$ Y^e = \frac{jβ_0}{kh + jβ_1} + \frac{h}{kh + jβ_1}\frac{M_o}{P} $$

Por otro lado, para hallar la tasa de interés de equilibrio, primero se deben despejar las funciones de la curva IS y la curva LM, en función de Y:

   - Curva IS:
   
   $$ r = \frac{β_0}{h} - \frac{β_1}{h}Y $$
   
   $$ rh = β_0 - β_1Y $$
   
   $$ β_1Y = β_0 - rh $$
   
   $$ Y = \frac{β_0 - rh}{β_1} $$
   
   - Curva LM:
   
   $$ r = -\frac{1}{j}\frac{M_o}{P} + \frac{k}{j}Y $$
   
   $$ rj + \frac{M_o}{P} = kY $$
   
   $$ \frac{rj}{k} + \frac{1}{k}\frac{M_o}{P} = Y $$
   
Luego, se igualan ambas ecuaciones y se despeja $r$:

$$ \frac{β_0 - rh}{β_1} = \frac{rj}{k} + \frac{1}{k}\frac{M_o}{P} $$ 

$$ kβ_0 - k(rh) = β_1(rj) + β_1\frac{M_o}{P} $$

$$ kβ_0 - β_1\frac{M_o}{P} = β_1(rj) + k(rh) $$

$$ kβ_0 - β_1\frac{M_o}{P} = r(β_1j + kh) $$


$$ r^e = \frac{kβ_0}{kh + jβ_1} - \frac{β_1}{kh + jβ_1}\frac{M_o}{P} $$ 

###### - Grafique el equilibrio simultáneo en los mercados de bienes y de dinero. 

```python
#--------------------------------------------------

#Equilibrio simultáneo en ambos mercados 

#--------------------------------------------------
    # Curva IS

# Parámetros

Y_size = 100 

Co = 35
Io = 40
Go = 50
Xo = 2
h = 0.8
b = 0.4
m = 0.5
t = 0.8

Y = np.arange(Y_size)


# Ecuación 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo)/h - ( ( 1-(b-m)*(1-t) ) / h)*Y  
    return r_IS

r_is = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)


#--------------------------------------------------
    # Curva LM 

# Parámetros

Y_size = 100

k = 2
j = 1                
Ms = 200             
P  = 20

Y = np.arange(Y_size)

# Ecuación

def r_LM(k, j, Ms, P, Y):
    r_LM = - (1/j)*(Ms/P) + (k/j)*Y
    return r_LM

r_lm = r_LM( k, j, Ms, P, Y)
```

```python
# Gráfico del modelo IS-LM

# Dimensiones del gráfico
y_max = np.max(r_lm)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
# Curva IS
ax.plot(Y, r_is, label = "IS", color = "#9932CC") #IS
# Curva LM
ax.plot(Y, r_lm, label="LM", color = "#7CFC00")  #LM

# Eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

# Texto y figuras agregadas
# Graficar la linea horizontal - r
plt.axvline(x=51.5,  ymin= 0, ymax= 0.52, linestyle = ":", color = "black")
# Grafica la linea vertical - Y
plt.axhline(y=93, xmin= 0, xmax= 0.52, linestyle = ":", color = "black")

# Plotear los textos 
plt.text(49,100, '$E_0$', fontsize = 14, color = 'black')
plt.text(0,100, '$r_0$', fontsize = 12, color = 'black')
plt.text(53,-10, '$Y_0$', fontsize = 12, color = 'black')

# Título, ejes y leyenda
ax.set(title="Modelo IS-LM", xlabel= r'Y', ylabel= r'r')
ax.legend()

plt.show()
```

### Estática comparativa
###### Analice los efectos sobre las variables endógenas Y, r de una disminución del gasto fiscal $(∆G_0 < 0)$. El análisis debe ser intuitivo, matemático y gráfico.

###### - Análisis intuitivo:

- En el mercado de bienes, si se da una política fiscal contractiva, con una disminución del gasto fiscal, se genera una caída de la Demanda agregada:

    $$ G_0↓ → DA↓ $$

    Esta reducción en la Demanda Agregada genera un desequilibrio en el mercado de bienes, donde $DA = Y$, pues al disminuir, resulta menor al nivel de producción:
    
    $$ DA↓ → DA<Y $$
    
    Por lo tanto, para regresar al equilibrio el nivel de producción debe ajustarse al nivel de la demanda agregada, es decir, reducirse. Entonces, se puede decir que una disminución del gasto fiscal implica una disminución del nivel de producción:

$$ DA↓ → Y↓ $$

$$ G_0↓ → Y↓ $$

- Continuando con el mercado de dinero, una disminución en el nivel de producción o de ingreso, va a generar una caída en la demanda de dinero, debido a su relación directamente proporcional, según $M_d = kY - jr$. Entonces:

    $$ Y↓ → M_d↓ $$
    
    Esta contracción de la demanda de dinero, va a generar un desequilibrio en el mercado de dinero, donde $M_d = M_s$, pues al disminuir la demanda, la oferta resulta mayor a la nueva demanda:
    
    $$ M_d↓ → M_d<M_s $$
    
    Entonces, para regresar al equilibrio, como no se plantea ningún cambio en la oferta de dinero, se supone que esta se mantiene. Por esa razón, es necesario que ante la caída de la demanda de dinero, la tasa de interés se reduzca. De esa manera, se regresa a un equilibrio, pues las personas van a buscar tener nuevamente su dinero en la "mano" antes que en los bancos, lo cual aumenta la circulación del dinero en la economía:
    
    $$ M_d↓ → r↓ $$
    
    $$ Y↓ → r↓ $$

###### - Análisis matemático:

###### Para el nivel de producción de equilibrio

Teniendo en cuenta:

$$ Y^e = \frac{j*β_0}{kh + jβ_1} + \frac{h}{kh + jβ_1}\frac{M_o}{P} $$

y sabiendo que:

- $β_0 = C_0 + I_0 + G_0 + X_0$
- $β_1 = 1 - (b - m)(1 - t)$

Se puede concluir que, en la segunda parte de la función: $(\frac{h}{kh + jβ_1}\frac{M_o}{P})$ no hay cambios, pues se trata de una política fiscal contractiva. Entonces, al no percibirse cambios, su diferencial es igual a 0. Por lo tanto, nos quedamos con la primera parte de la ecuación: $(\frac{j*β_0}{kh + jβ_1})$, pero identificamos que la disminución del gasto fiscal solo afecta a la variable $β_0$, pues es uno de sus parámetros, entonces lo separamos:

$$ ∆Y^e = ∆(\frac{j}{kh + jβ_1})∆β_0 $$

Además, como $G_0$ es el único parámetro que cambia en β_0, entonces, podemos reemplazar a este último por el primero:

$$ ∆Y^e = ∆(\frac{j}{kh + jβ_1})∆G_0 $$

Y como se sabe que $∆G_0<0$, se tiene:

$$ ∆Y^e = (+)(-) $$

Por lo tanto, se concluye que el diferencial de $Y^e$ es negativo y, por ende, menor a 0:

$$ ∆Y^e = (-) $$

$$ ∆Y^e<0 $$

###### Para la tasa de interés de equilibrio:

Teniendo en cuenta:

$$ r^e = \frac{kβ_0}{kh + jβ_1} - \frac{β_1}{kh + jβ_1}\frac{M_o}{P} $$

y sabiendo que:

- $β_0 = C_0 + I_0 + G_0 + X_0$
- $β_1 = 1 - (b - m)(1 - t)$

Se puede concluir que en la segunda parte de la función: $(\frac{β_1}{kh + jβ_1}\frac{M_o}{P})$ no hay cambios. Entonces se iguala a 0. Por lo tanto, nos quedamos con la primera parte de la ecuación: $(\frac{kβ_0}{kh + jβ_1})$, pero como la disminución del gasto fiscal solo afecta a la variable $β_0$, al ser uno de sus parámetros, entonces lo separamos:

$$ ∆r^e = ∆(\frac{k}{kh + jβ_1})∆β_0 $$

Asimismo, reemplazamos $β_0$ por $G_0$, pues el único componente que cambia:

$$ ∆r^e = ∆(\frac{k}{kh + jβ_1})∆G_0 $$

Y como el gasto fiscal disminuye $∆G_0<0$, se tiene:

$$ ∆r^e = (+)(-) $$

Entonces, se concluye que el diferencial de $r^e$ es negativo y, por lo tanto, menor a 0:

$$ ∆r^e = (-) $$

$$ ∆r^e<0 $$

###### Analice los efectos sobre las variables endógenas Y, r de una disminución de la masa monetaria $(∆M^s<0)$. El análisis deber ser intuitivo, matemático y gráfico. 

###### - Análisis Intuitivo:

- En el mercado de dinero, aplicar una política monetaria contractiva, que implica la disminución de la masa monetaria, como esta, implica una reducción en la oferta de real de dinero. Por lo tanto, se genera un desequilibrio, considerando que en el mercado de dinero, el equilibrio está determinado por $M^s = M^d$. Por lo tanto, al disminuir la oferta de dinero, la demanda resulta mayor, por lo que, para que regresar al equilibrio, se necesita que la tasa de interes aumente. Esto debido a la relación inversamente proporcional que tiene la tasa de interés y la masa monetaria $(r = -\frac{1}{j}\frac{M_o}{P} + \frac{k}{j}Y)$. Entonces, se puede concluir lo siguiente:

$$ M_s↓ → M^o↓ → M^o < M^d → r↑ $$

$$ M_s↓ → r↑ $$

- Entonces, una reducción de la tasa de interés, genera un cambio en el mercado de bienes. En este caso, afecta a la Inversión $(I)$, teniendo en cuenta que esta es:

    $$ I = I_0 - hr $$
    
    En ese sentido, considerando su relación inversa, se puede concluir, que un aumento de la tasa de interés, genera una disminución de la inversión. Además, esta última, al ser un parámetro de la Demanda Agregada, genera una disminución en la misma. Entonces, se produce un desequilibrio en el mercado de bienes, donde la demanda agregada debe ser igual al nivel de producción $(DA = Y)$. En este caso, la disminución de la demanda, implica que el nivel de producción es mayor. Por lo tanto, para regresar al nivel de equilibrio, el nivel de producción debe ajustarse y disminuir. Entonces, podemos decir:
    
    $$ r↑ → I↓ → DA↓ → DA < Y → Y↓ $$
    
    $$ r↑ → Y↓ $$
    
    De esa manera, se concluye que una disminución de la masa monetaria, implica un aumento de la tasa de interés de equilibrio, y esto a su vez, genera una reducción del nivel de producción. 

###### - Análisis matemático:
###### Para el nivel de producción de equilibrio

Teniendo en cuenta:

$$ Y^e = \frac{j*β_0}{kh + jβ_1} + \frac{h}{kh + jβ_1}\frac{M_o}{P} $$

Se puede concluir que, al aplicar una política monetaria contractiva $(M_o<0)$, en la primera parte de la función: $(\frac{j*β_0}{kh + jβ_1})$ no se perciben cambios. Entonces, su diferencia es igual a 0. Por lo tanto, nos quedamos con la segunda parte de la ecuación: $\frac{h}{kh + jβ_1}\frac{M_o}{P}$, y de esa manera, se tiene:

$$ ∆Y^e = ∆(\frac{h}{kh + jβ_1}) ∆(\frac{M_o}{P}) $$


Y como se sabe que $∆M^s<0$, se tiene:

$$ ∆Y^e = (+)\frac{(-)}{(+)} $$

$$ ∆Y^e = (+)(-) $$

Por lo tanto, se concluye que el diferencial de $Y^e$ es negativo y, por ende, menor a 0:

$$ ∆Y^e = (-) $$

$$ ∆Y^e<0 $$

###### Para la tasa de interés de equilibrio:

Teniendo en cuenta:

$$ r^e = \frac{kβ_0}{kh + jβ_1} - \frac{β_1}{kh + jβ_1}\frac{M_o}{P} $$

Se puede concluir que los cambios no se perciben en la primera parte de la función: $(\frac{kβ_0}{kh + jβ_1})$. Entonces, al no haber ningún cambio, su diferencia se iguala a 0. Por lo tanto, nos quedamos con la segunda parte de la ecuación: $(- \frac{β_1}{kh + jβ_1}\frac{M_o}{P})$, pues es donde se ecuentra la oferta de dinero. Entonces tenemos:

$$ ∆r^e = ∆(- \frac{β_1}{kh + jβ_1}\frac{M_o}{P}) $$

Y como la masa monetaria disminuye $∆M^s<0$, se tiene:

$$ ∆r^e = (-)(+)\frac{(-)}{(+)} $$

$$ ∆r^e = (-)(+)(-) $$

$$ ∆r^e = (-)(-) $$

Entonces, se concluye que el diferencial de $r^e$ es positivo y, por lo tanto, mayor a 0:

$$ ∆r^e = (+) $$

$$ ∆r^e>0 $$

###### Analice los efectos sobre las variables endógenas Y, r de un incremento de la tasa de impuestos (∆t > 0). El análisis debe ser intuitivo, matemático y gráfico.

###### - Análisis intuitivo:

    $$ DA = (C_0 + I_0 + G_0 + X_0 - hr) + (b - m)(1 - t)Y $$
    
    Se puede intuir que, debido a la relación inversa que existe entre la tasa impositiva, un incremento de esta generaría una caía en la demanda agregada. Esto a su vez, generaría un desequilibrio en el mercado de bienes, donde se debe mantener la igualdad: $DA = Y$. En ese sentido, al disminuir la demanda agregada, se tiene un exceso de producción, pues este resulta mayor frente a la nueva demanda. Por lo tanto, para volver al equilibrio, el nivel de producción debe ajustarse a la demanda y disminuir. Entonces, se intuye que:
    
    $$ t↑ → DA↓ → DA < Y → Y↓ $$
    
    $$ t↑ → Y↓ $$
    
- En el mercado de dinero, también se perciben cambios, producto de la contracción del nivel de producción. En ese sentido, teniendo en cuenta: $M_d = kY - jr$, se puede inferir que una reducción en el nivel de producción, va a generar una contracción de la demanda de dinero, debido a su relación directamente proporcional:

    $$ Y↓ → M_d↓ $$
    
    Esto, genera un desquilibrio en el mercado de dinero, donde se procura: $M_d = M^s$. En este caso, como se plantea un cambio en la oferta de dinero, se presume que esta se mantiene, por lo que, una reducción de la demanda, implica un desequilibrio, donde la oferta resulta mayor a la demanda que existe en el mercado de dinero:
    
    $$ M_d < M^s $$
    
    Por lo tanto, para regresar al equilibrio, considerando que no hay cambios en la oferta de dinero, es necesario reducir la tasa de interés, de esa manera, como se explicó anteriormente, las personas van a demandar tener más dinero en las "manos", cubriendo la oferta que ya existe:

### A partir del siguiente sistema de ecuaciones que representan el modelo IS-LM



#### Hay que recordar que la tasa de interés real es igual a la diferencia entre la tasa de interés nominal y la tasa de inflación esparada $(r = i - π^e)$. Esta es la conocida definición de Irving Fishes. De aquí se infiere que la tasa de interés nominal es igual a: $i = r + π^e$

##### - Encuentre las ecuaciones de Ingreso $(Y^e)$ y tasa de interés $(r^e)$ de equilibrio. Escriba paso a paso la derivación de estas ecuaciones. 

En este caso, se debe tener en cuenta, en primer lugar, la función de la curva IS:

$$ r = \frac{β_0}{h} - \frac{β_1}{h}Y $$

donde:

- $β_0 = C_0 + I_0 + G_0 + X_0$
- $β_1 = 1 - (b - m)(1 - t)$

Asimismo, en segundo lugar, se debe hallar la curva LM, ahora teniendo en cuenta $M_d = kY - ji$:

- En ese sentido, sabe que la Curva LM se deriva del equilibrio del mercado monetario, en función de la tasa de interés. Entonces:

$$ M_s = M_d $$

$$ \frac{M_o}{P} = kY - ji $$

$$ \frac{M_o}{P} = kY - j(r + π^e) $$

$$ j(r + π^e) = kY - \frac{M_o}{P} $$

$$ r + π^e = \frac{kY}{j} - \frac{1}{j}\frac{M_o}{P} $$

- Por lo tanto, la función de la curva LM es:

$$ r = - \frac{1}{j}\frac{M_o}{P} + \frac{kY}{j} - π^e $$

En ese sentido, para poder hallar la ecuación de nivel de producción de equilibrio $(Y^e)$ es necesario igualar ambas ecuaciones:

$$ \frac{β_0}{h} - \frac{β_1}{h}Y = - \frac{1}{j}\frac{M_o}{P} + \frac{kY}{j} - π^e $$

Luego se debe despejar $Y$:

$$ β_0 - β_1Y = - h(\frac{1}{j}\frac{M_o}{P}) + h(\frac{kY}{j}) - hπ^e $$

$$ β_0 - β_1Y + hπ^e = - h(\frac{1}{j}\frac{M_o}{P}) + h(\frac{kY}{j}) $$

$$ j(β_0) - j(β_1)Y + jhπ^e = - h\frac{M_o}{P} + (kh)Y $$

$$ j(β_0) + jhπ^e + h\frac{M_o}{P} = (kh + jβ_1)Y $$

$$ Y^e = \frac{jβ_0}{kh + jβ_1} + \frac{h}{kh + jβ_1}\frac{M_o}{P} + \frac{jhπ^e}{kh + jβ_1} $$

Por otro lado, para hallar la tasa de interés de equilibrio, primero se deben despejar las funciones de la curva IS y la curva LM, en función de Y:

   - Curva IS:
   
   $$ r = \frac{β_0}{h} - \frac{β_1}{h}Y $$
   
   $$ rh = β_0 - β_1Y $$
   
   $$ β_1Y = β_0 - rh $$
   
   $$ Y = \frac{β_0 - rh}{β_1} $$
   
   - Curva LM:
   
   $$ r = - \frac{1}{j}\frac{M_o}{P} + \frac{kY}{j} - π^e $$
   
   $$ r + π^e = - \frac{1}{j}\frac{M_o}{P} + \frac{kY}{j} $$
   
   $$ rj + π^ej + \frac{M_o}{P} = kY $$
     
   $$ Y = \frac{rj}{k} + \frac{π^ej}{k} + \frac{1}{k}\frac{M_o}{P} $$
   
Luego, se igualan ambas ecuaciones y se despeja $r$:

$$ \frac{β_0 - rh}{β_1} = \frac{rj}{k} + \frac{π^ej}{k} + \frac{1}{k}\frac{M_o}{P} $$ 


$$ kβ_0 - k(rh) = β_1(rj) + β_1(π^ej) + β_1\frac{M_o}{P} $$


$$ kβ_0 - β_1π^ej - β_1\frac{M_o}{P} = β_1(rj) + k(rh) $$


$$ kβ_0 - β_1π^ej - β_1\frac{M_o}{P} = r(jβ_1 + kh) $$


$$ r^e = \frac{kβ_0}{kh + jβ_1} - \frac{β_1π^ej}{kh + jβ_1} - \frac{β_1}{kh + jβ_1}\frac{M_o}{P} $$ 

##### - Grafique el equilibrio simultáneo en los mercados de bienes y de dinero. 

```python
#--------------------------------------------------

#Equilibrio simultáneo en ambos mercados 

#--------------------------------------------------
    # Curva IS

# Parámetros

Y_size = 100 

Co = 35
Io = 40
Go = 50
Xo = 2
h = 0.8
b = 0.4
m = 0.5
t = 0.8

Y = np.arange(Y_size)


# Ecuación 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo)/h - ( ( 1-(b-m)*(1-t) ) / h)*Y  
    return r_IS

r_is_i = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)


#--------------------------------------------------
    # Curva LM 

# Parámetros

Y_size = 100

k = 2
j = 1                
Ms = 200             
P  = 20
π = 0.3

Y = np.arange(Y_size)

# Ecuación

def r_LM(k, j, Ms, P, Y, π):
    r_LM = - (1/j)*(Ms/P) + (k/j)*Y - π
    return r_LM

r_lm_i = r_LM( k, j, Ms, P, Y, π)
```

    $$ P_o ↓ → M_s ↑ $$
    
    Entonces, se genera un desequilibrio en el mercado, pues al no percibirse cambios en la demanda de dinero, se produce un exceso de oferta:
    
    $$ M_s > M_d $$
    
    Por lo tanto, para regresar al equilibrio se vuelve necesario bajar la tasa de interés, de esa manera, la gente demanda de tener más dinero en la "mano", así se vuelve al equilibrio: 
    
    $$ M_s ↑ → r ↓ $$
    
    
- En mercado de bienes, en ese sentido, también se perciben cambios. En este caso, la disminución de la tasa de interés, provoca un aumento en la inversión, debido a su relación inversa $(I = I_o - hr)$: 

    $$ r ↓ → I ↑ $$
    
    Por lo tanto, ese aumento en la inversión, genera un aumento en la demanda agregada $(DA = C + I + G + X - M)$:
    
    $$ I ↑ → DA ↑ $$

    $$ DA > Y $$
    
    Por lo tanto, para regresar al equilibrio es necesario que el nivel de producción se ajuste a la demanda, lo cual implica que también aumente. 
    
    $$ DA ↑ → Y ↑ $$

    Por lo tanto, esa reducción de la inversión, genera una contracción de la demanda agregada $(DA = C + I + G + X - M)$:
    
    $$ I ↓ → DA ↓ $$
    
    Esto produce un desquilibrio en el mercado de bienes, en el que se procura que la demanda agregada sea igual al nivel de producción. En este caso, al disminuir la demanda agregada, el nivel de producción resulta excesivo, pues resulta mayor a la demanda agregada:
    
    $$ DA < Y $$
    
    Por lo tanto, para regresar al equilibrio es necesario que el nivel de producción se ajuste a la demanda, lo cual implica que también se contraiga. 
    
    $$ DA ↓ → Y ↓ $$

###### - Análisis Matemático:

###### Para la tasa de interés de equilibrio:
- Teniendo en cuenta: 

    $$ r^e = \frac{kβ_0}{kh + jβ_1} - \frac{β_1πj}{kh + jβ_1} - \frac{β_1}{kh + jβ_1}\frac{M_o}{P} $$ 
    
    
- Y sabiendo que no se perciben cambios las partes: $(\frac{kβ_0}{kh + jβ_1})$ y $(\frac{β_1}{kh + jβ_1}\frac{M_o}{P})$, podemos decir que sus diferenciales son iguales a 0. Por eso, nos podemos quedar solo con la parte en la que se encuentra la inflación:

    $$ ∆r^e = - ∆(\frac{β_1πj}{kh + jβ_1}) $$


- Además, se sabe que no se perciben cambios en los parámetros de: $β_1$, $k$, $h$ y $j$. Por lo tanto, sus diferenciales pueden ser igualados a 0:
    
    $$ ∆β_1 = ∆k = ∆h = ∆j = 0 $$
    
    
- Entonces, se tiene que el diferencial de la tasa de interés es:

    $$ ∆r^e = - ∆π $$
    
    
- Además, sabemos que $∆π < 0$, por lo que su diferencial es negativo:

    $$ ∆r^e = - (-) $$
         
    
- Entonces, se concluye que el diferencial de la tasa de interés es positiva, por lo tanto, aumenta y es mayor a 0:

    $$ ∆r^e = (+) $$
    
    $$ ∆r^e > 0 $$

###### Para el nivel de producción de equilibrio:

- Teniendo en cuenta:

$$ Y^e = \frac{jβ_0}{kh + jβ_1} + \frac{h}{kh + jβ_1}\frac{M_o}{P} + \frac{jhπ}{kh + jβ_1} $$


- Y sabiendo que no se perciben cambios en las dos primeras partes: $(\frac{jβ_0}{kh + jβ_1})$ y $(\frac{h}{kh + jβ_1}\frac{M_o}{P})$, se puede decir que sus diferenciales son iguales a cero. Por eso, podemos quedarnos con la tercera parte de la función, que es la que contiene a la inflación:

$$ ∆Y^e = ∆(\frac{jhπ}{kh + jβ_1}) $$


- Además, como no se perciben cambios en los parámetros de $β_1$, $k$, $h$ y $j$, podemos igualar sus diferenciales a 0:

$$ ∆β_1 = ∆k = ∆h = ∆j = 0 $$


- Entonces, se tiene que el diferencial del nivel de producción es:

    $$ ∆Y^e = ∆π $$
    
    
- Asimismo, sabemos que $∆π < 0$, por lo que, su diferencial es negativo:

    $$ ∆Y^e = (-) $$  
    
    
- Entonces, se concluye que el diferencial del nivel de produción es negativo, por lo tanto, disminuye y es menor a 0:

    $$ ∆Y^e < 0 $$

```python
#1--------------------------------------------------
    # Curva IS ORIGINAL

# Parámetros
Y_size = 100 

Co = 35
Io = 40
Go = 50
Xo = 2
h = 0.8
b = 0.4
m = 0.5
t = 0.8

Y = np.arange(Y_size)


# Ecuación 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r_π = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)


#2--------------------------------------------------
    # Curva LM ORIGINAL

# Parámetros

Y_size = 100

k = 2
j = 1                
Ms = 700             
P  = 20
π = 20

Y = np.arange(Y_size)

# Ecuación

def i_LM( k, j, Ms, P, Y, π):
    i_LM = (-Ms/P)/j + k/j*Y - π
    return i_LM

i_π_0 = i_LM( k, j, Ms, P, Y, π)

#-------------------------------------------------
    # Curva LM NUEVA
    
#Parámetro que cambia: Inflación esperada (π)

π = 0.2

# Nueva curva LM con la variacion del Ms
def i_LM_π( k, j, Ms, P, Y, π):
    i_LM = (-Ms/P)/j + k/j*Y - π
    return i_LM

i_π_1 = i_LM_π( k, j, Ms, P, Y, π)
```
