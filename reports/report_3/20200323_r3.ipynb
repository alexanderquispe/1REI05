### Lectura

La investigación de Mendoza, Macinlla y Velarde tiene como objetivo analizar y replicar, mediante un modelo macroeconómico, los efectos producidos en la economía nacional durante el período de la cuarentena en la pandemia por el COVID 19 en el año 2020, que produjo una de las caídas más duras del PBI en LATAM y sobre todo en Perú. Los investigadores proponen entonces un modelo de 2 sectores para reproducir lo sucedido durante la pandemia, y a su vez los efectos de su fin, que poco a poco llevaron a reestabilizar la producción.

Una de las principales fortalezas de esta investigación es la interrelación que se realiza entre ambos sectores del mercado de bienes. Mientras el sector 1 de servicios se ve afectado, esta situación tiene consecuencias en el sector 2, de bienes indispensables. Esto genera un dinamismo, y demuestra que la economía no cuenta con diversos sectores que se interrelacionan y afectan mutuamente. Asimismo, es ventajoso que el modelo se pruebe con diferentes situaciones, tanto choques permanentes com transitorios, ya que esto pone a prueba las hipótesis de los investigadores en diferentes casos. Finalmente, otro punto a favor de la investigación es que reconoce que este no es el último paso para investigar este tema, y reconoce que hace falta llevar este tipo de análisis por otras rutas, como por ejemplo, tomando en cuenta las expectativas racionales de los actores, lo que le daría un aspecto más social a la investigación.

Por otro lado, a pesar de las ventajas de este modelo, también podemos identificar algunas debilidades. En primer lugar, tal y como los autores lo especifican, este modelo solo incluye 2 sectores keynesianos del mercado de bienes. Es decir, el mercado monetario no afecta a la producción, y la única conexión entre ambos mercados es la tasa de interés, que ya es fija. Esto limita la investigación, ya que en la realidad las decisiones del banco central sobre la oferta monetaria tienen un efecto en la producción y la demanda. Asimismo, los factores internacionales no son tomados en cuenta al momento del análisis. Durante la pandemia, las importaciones también se vieron afectadas, pero esto no se toma en cuenta dentro del modelo, que no incluye a las exportaciones netas dentro de su fórmula de demanda agregada.

El aporte de esta investigación es vasto. El modelo resulta interesante, ya que combina 2 sectores del mercado de bienes: uno directamente afectado por la pandemia, el de servicios; y otro indirectamente afectado, el de bienes indispensables. Pero lo particular de esta investigación es que combina ambos sectores, y evalúa los efectos que cada uno de ellos tienen sobre el otro. Mediante la simulación de los efectos de la pandemia en estos sectores, que generó la mayor contracción económica desde la guerra con Chile, podemos analizar su efecto en la producción, incluso en diferentes sitaciones, como choques permamentes de oferta y demanda, y luego con choques transitorios. Tomar en cuenta los efectos indirectos de la cuarentena es crucial, ya que muchas otras investigaciones se centran en ver los efectos directos de la disminución de la producción en el PBI. En cambio, esta investigación reproduce de manera más orgánica el funcionamiento de una economía nacional interdependiente.

Asimismo, podemos pensar en algunos avances que se podrían realizar en investigaciones sobre este tema. ### Pregunta 1

#### Explique cuáles son los intrumentos de política monetaria que puede utilizar el Banco Central.

El Banco Central tiene 3 instrumentos de política monetaria, que pueden ser tanto expansiva o contractiva. El primero es la Operación de Mercado Abierto, donde el Banco compra o vende activos financieros o bonos a los bancos comerciales. Si compra bonos del mercado, entonces es una política expansiva, ya que aumenta la base monetaria y en consecuencia la oferta monetaria. Si vende bonos, entonces es una política contractiva.

El segundo instrumento es el coeficiente legal de encaje, donde el Banco Central puede incrementar o disminuír la cantidad de dinero que tienen los bancos para realizar préstamos. Si decide disminuír la tasa de encaje, entonces aumenta la posibilidad de creación de dinero bancario ya que aumenta el multiplicador de dinero bancario, lo que aumenta la oferta monetaria. Al contrario, si aumenta la tasa de encaje, se reduce el multiplicador, y disminuye la oferta monetaria.

Finalmente, el tercer instrumento es la tasa de interés de referencia de la política monetaria, que modifica la cantidad de dinero prestado a los bancos comerciales. Si se reduce la tasa, los bancos podrán obtener prestado máws dinero, aumentando la oferta monetaria. Si aumenta la tasa de interés, entonces se prestará menos dinero, y por lo tanto disminuirá la oferta monetaria.

#### Derive la oferta real de dinero y explique cada uno de sus componentes.

La oferta real de dinero se forma con 2 componentes:

La oferta nominal de dinero: es una variable exógena del modelo, y es un instrumento de política monetaria. Representa la masa de dinero que está en la economía actual.

Precios: nivel de precios, que es un valor dado debido a que el análisis es a corto plazo.

Para encontrar el valor real del dinero, que es el valor de dinero que verdaderamente hace circular el BCR, entonces debemos dividir la oferta nominal de dinero entre los precios.

$$ M_s = \frac{M^s_0}{P} $$

#### Derive la demanda real de dinero. Explique qué papel cumplen los parametros "k" y "j"

La demanda real de dinero explica las razones por las que se demanda dinero en una economía.

Existen dos razones para demandar dinero: transacción y precaución. Ambas condiciones dependen del ingreso Y y del parámetro k, que representa la elasticidad del ingreso, es decir, la sensibilidad de la demanda de dinero ante variaciones en Y. Por lo tanto, la multiplicación de estos dos componentes nos da una primera ecuación:

$$ L_1 = kY $$

Ahora también debemos tomar en cuenta dónde se guarda el dinero de las personas, por medio de la especulación, que representa la decisión de guardar el dinero en el banco o en el bolsillo. Esto se define por la multiplicación de i, que es la tasa de interés nominal, con j negativo, que representa la elasticidad de sustitución, de tener el dinero en el bolsillo o en el banco. Si aumenta j, entonces la gente pondrá más dinero en el banco, y habrá menos dinero en su bolsillo. Ahora tenemos una segunda ecuación:

$$ L_2 = -ji $$

La demanda de dinero se consigue entonces obteniendo la cantidad de dinero en el bolsillo de la gente. Es decir, debemos hacer la siguiente ecuación:

$$ M_d = L_1 - L_2 $$

$$ M_d = kY - ji $$



#### Asumiendo que no hay inflación podemos asumir que i=r. Escriba en terminos reales la eucacion de equilibrio en el mercado de dinero.

El equilibrio en el mercado de dinero se alcanza con el equilibrio entre la oferta de dinero y demanda de dinero:

$$ M^s = M^d $$

$$ \frac{M^s_0}{P} = kY - ji $$

Pero si se supone que no hay inflación, entonces, la tasa de interés nominal y real no debería variar tanto. Por lo tanto podemos decir que la ecuación de equilibrio del mercado de dinero en términos reales es:

$$ \frac{M^s_0}{P} = kY - jr $$


#### Grafique el equilibrio en el mercado de dinero.

```python
# Parameters
r_size = 100

k = 0.5
j = 0.2                
P  = 10 
Y = 35
MS_0 = 500

r = np.arange(r_size)

# Necesitamos crear la funcion de demanda 

def MD(k, j, P, r, Y):
    MD_eq = (k*Y - j*r)
    return MD_eq
MD_0 = MD(k, j, P, r, Y)
# Necesitamos crear la oferta de dinero.
MS = MS_0 / P
MS
```

```python

fig, ax1 = plt.subplots(figsize=(10, 8))

ax1.set(title="Equilibrio en el mercado de Dinero", xlabel=r'M^s / P', ylabel=r'r')

ax1.plot(MD_0, label= '$L_0$', color = '#CD5C5C')

ax1.axvline(x = MS,  ymin= 0, ymax= 1, color = "grey")

ax1.axhline(y=7.5, xmin= 0, xmax= 0.47, linestyle = ":", color = "black")

ax1.text(0, 8, "$r^e_0$", fontsize = 12, color = 'black')

ax1.text(100, -1, "$L(Y_0)$", fontsize = 12, color = 'black')

ax1.text(50, -1, "$M^s_0 / P_0$", fontsize = 12, color = 'black')

ax1.yaxis.set_major_locator(plt.NullLocator())   
ax1.xaxis.set_major_locator(plt.NullLocator())

custom_xlim = (-10, 120)
custom_ylim = (-10, 20)

plt.setp(ax1, xlim=custom_xlim, ylim=custom_ylim)

ax1.legend()

plt.show()
```

#### Explique y grafique qué sucede en el mercado de dinero si el diferencial de k es menor a 0

K representa la elasticidad del ingreso. Por lo tanto, si k disminuye, entonces la cantidad de dinero que las personas guardan en su bolsillo va a disminuír, y va a haber más dinero en los bancos. ### Pregunta 3

#### Derive paso a paso la curva LM matemáticamente (a partir del equilibrio en el Mercado Monetario) y grafique.

Primero tenemos el equilibrio del mercado monetario.

$$ \frac{M^s_0}{P} = kY - jr $$

Ahora, podemos seguir derivando esta fórmula para ponerla en función de r, que sería la tasa de interés que equilibra el mercado monetario.

$$ kY - \frac{M^s_0}{P} = jr $$

$$ \frac{kY}{j} - \frac{M^s_0}{P} = r $$

LM = $$ r = -\frac{M^s_0}{Pj} + \frac{k}{j}Y $$

Esta sería la ecuación de la curva LM.

```python
#1----------------------Equilibrio mercado monetario

    # Parameters
r_size = 100

k = 0.5
j = 0.2                
P  = 10 
Y = 35

r = np.arange(r_size)


    # Ecuación
def Ms_MD(k, j, P, r, Y):
    Ms_MD = P*(k*Y - j*r)
    return Ms_MD

Ms_MD = Ms_MD(k, j, P, r, Y)


    # Nuevos valores de Y
Y1 = 45

def Ms_MD_Y1(k, j, P, r, Y1):
    Ms_MD = P*(k*Y1 - j*r)
    return Ms_MD

Ms_Y1 = Ms_MD_Y1(k, j, P, r, Y1)


Y2 = 25

def Ms_MD_Y2(k, j, P, r, Y2):
    Ms_MD = P*(k*Y2 - j*r)
    return Ms_MD

Ms_Y2 = Ms_MD_Y2(k, j, P, r, Y2)

#2----------------------Curva LM

    # Parameters
Y_size = 100

k = 0.5
j = 0.2                
P  = 10               
Ms = 30            

Y = np.arange(Y_size)


# Ecuación

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)
```

#### ¿Cuál es el efecto de una disminución en la Masa Monetaria si el diferencial de Ms es menor a 0? Explica usando la intuición y gráficos.

Si disminuye la masa monetaria, entonces se genera un exceso de demanda en el mercado, lo que mueve la curva LM hacia la izquierda. Esto se debe a que el Banco Central está realizando una política contractiva, disminuyendo la cantidad de dinero en la economía. Pero al no haber disminuído la demanda de la población, se da un exceso de esta. Por lo tanto, para regresar al equilibrio, debe de aumentar la tasa de interés, para que disminuya la demanda de dinero.

$$ M^s↓ → M^s_0↓ → M^s < M^d → r↑ $$

```python
Y_size = 100

k = 2
j = 1                
Ms = 600             
P  = 20               

Y = np.arange(Y_size)

# Ecuación

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)

#--------------------------------------------------
    # NUEVA curva LM

# Definir SOLO el parámetro cambiado
Ms = 100

# Generar la ecuación con el nuevo parámetro
def i_LM_Ms( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i_Ms = i_LM_Ms( k, j, Ms, P, Y)
```

```python
# Dimensiones del gráfico
y_max = np.max(i)
v = [0, Y_size, 0, y_max]   
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, i, label="LM", color = 'black')
ax.plot(Y, i_Ms, label="LM_Ms", color = '#3D59AB', linestyle = 'dashed')

# Texto agregado
plt.text(47, 76, '∆$M^s$', fontsize=12, color='black')
plt.text(47, 70, '←', fontsize=15, color='grey')

# Título y leyenda
ax.set(title = "Disminución en la Masa Monetaria $(M^s)$", xlabel=r'Y', ylabel=r'r')
ax.legend()


plt.show()
```

#### ¿Cuál es el efecto de un aumento en k. Explica usando intuición y gráficos.

Si aumenta k, entonces la pendiente de la curva LM va a aumentar. Esto se debe a que si k aumenta, la sensibilidad de los ingresos de las personas para tenerlos en su bolsillo es más alta. Esto genera que se requiera un aumento de la tasa de interés para regresar al equilibrio, para que así las personas decidan poner más de sus ingresos en los bancos.

$$ k↑ → M^d↑ → M^s < M^d → r↑ $$


```python
#--------------------------------------------------
    # Curva LM ORIGINAL

# Parámetros

Y_size = 100

k = 0.5
j = 1                
Ms = 200             
P  = 20               

Y = np.arange(Y_size)

# Ecuación

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)

#--------------------------------------------------
    # NUEVA curva LM

# Definir SOLO el parámetro cambiado
k = 0.9

# Generar la ecuación con el nuevo parámetro
def i_LM_k( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i_k = i_LM_k( k, j, Ms, P, Y)
```

```python
# Dimensiones del gráfico
y_max = np.max(i)
v = [0, Y_size, 0, y_max]   
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, i, label="LM", color = 'black')
ax.plot(Y, i_k, label="LM_k", color = '#3D59AB', linestyle = 'dashed')

# Texto agregado
plt.text(80, 45, '∆$k$', fontsize=12, color='black')
plt.text(80, 50, '↑', fontsize=15, color='grey')

# Título y leyenda
ax.set(title = "Incremento en k", xlabel=r'Y', ylabel=r'r')
ax.legend()


plt.show()
```
